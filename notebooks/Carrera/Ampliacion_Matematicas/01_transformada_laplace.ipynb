## 1. Definición de la Transformada de Laplace

La **transformada de Laplace** convierte una función del tiempo $f(t)$ en una función de la variable compleja $s$, facilitando la resolución de ecuaciones diferenciales lineales al transformarlas en ecuaciones algebraicas.

$$\boxed{\mathcal{L}\{f(t)\} = F(s) = \int_0^{\infty} f(t)\,e^{-st}\,dt, \qquad s \in \mathbb{C}}$$

**Condiciones de existencia:** $f(t)$ debe ser de orden exponencial, es decir, $|f(t)| \leq M e^{\alpha t}$ para $t$ suficientemente grande. Entonces $F(s)$ existe para $\text{Re}(s) > \alpha$.

### 1.1 Transformadas fundamentales

**Ejemplo 1:** $f(t) = 1$ (función constante)

$$\mathcal{L}\{1\} = \int_0^{\infty} e^{-st}\,dt = \left[-\frac{e^{-st}}{s}\right]_0^{\infty} = \frac{1}{s}, \qquad s > 0$$

$$\boxed{\mathcal{L}\{1\} = \frac{1}{s}}$$

**Ejemplo 2:** $f(t) = t^n$

Aplicando integración por partes $n$ veces (o usando la función Gamma):

$$\boxed{\mathcal{L}\{t^n\} = \frac{n!}{s^{n+1}}, \qquad n = 0, 1, 2, \ldots}$$

Casos particulares: $\mathcal{L}\{t\} = \dfrac{1}{s^2}$, $\quad \mathcal{L}\{t^2\} = \dfrac{2}{s^3}$, $\quad \mathcal{L}\{t^3\} = \dfrac{6}{s^4}$

**Ejemplo 3:** $f(t) = e^{at}$

$$\mathcal{L}\{e^{at}\} = \int_0^{\infty} e^{at}e^{-st}\,dt = \int_0^{\infty} e^{-(s-a)t}\,dt = \frac{1}{s-a}, \qquad s > a$$

$$\boxed{\mathcal{L}\{e^{at}\} = \frac{1}{s-a}}$$

**Ejemplo 4:** $f(t) = \sin(\omega t)$

Usando la identidad de Euler $\sin(\omega t) = \dfrac{e^{j\omega t} - e^{-j\omega t}}{2j}$ o integrando por partes dos veces:

$$\boxed{\mathcal{L}\{\sin(\omega t)\} = \frac{\omega}{s^2 + \omega^2}}$$

**Ejemplo 5:** $f(t) = \cos(\omega t)$

$$\boxed{\mathcal{L}\{\cos(\omega t)\} = \frac{s}{s^2 + \omega^2}}$$

## 3. Función escalón de Heaviside y función pulso

### 3.1 Función escalón de Heaviside

La **función escalón unitario** (o de Heaviside) se define como:

$$\boxed{u(t-a) = \begin{cases} 0 & \text{si } t < a \\ 1 & \text{si } t \geq a \end{cases}}$$

Su transformada de Laplace es:

$$\boxed{\mathcal{L}\{u(t-a)\} = \frac{e^{-as}}{s}, \qquad a \geq 0}$$

### 3.2 Función pulso rectangular

Un **pulso** de amplitud $A$ entre $t = a$ y $t = b$ se construye como:

$$p(t) = A\left[u(t-a) - u(t-b)\right]$$

$$\mathcal{L}\{p(t)\} = A\left(\frac{e^{-as} - e^{-bs}}{s}\right)$$

**Utilidad:** permite modelar señales que se activan y desactivan en instantes concretos, muy común en circuitos y sistemas de control.

```python
# Gráficas de la función escalón de Heaviside y función pulso
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

t = np.linspace(-1, 6, 1000)

# --- Escalón u(t) ---
u_t = np.where(t >= 0, 1.0, 0.0)
axes[0].plot(t, u_t, color=COLOR_PRINCIPAL, lw=2.5)
axes[0].plot(0, 1, 'o', color=COLOR_PRINCIPAL, ms=8, zorder=5)
axes[0].plot(0, 0, 'o', color=COLOR_PRINCIPAL, ms=8, mfc='white', zorder=5)
axes[0].set_title(r'Escalón unitario $u(t)$', fontweight='bold')
axes[0].set_xlabel(r'$t$')
axes[0].set_ylabel(r'$u(t)$')
axes[0].set_ylim(-0.2, 1.5)
axes[0].axhline(y=0, color='gray', lw=0.5)
axes[0].axvline(x=0, color='gray', lw=0.5)
axes[0].grid(True, alpha=0.3)

# --- Escalón desplazado u(t-2) ---
a = 2
u_ta = np.where(t >= a, 1.0, 0.0)
axes[1].plot(t, u_ta, color=COLOR_RECTA, lw=2.5)
axes[1].plot(a, 1, 'o', color=COLOR_RECTA, ms=8, zorder=5)
axes[1].plot(a, 0, 'o', color=COLOR_RECTA, ms=8, mfc='white', zorder=5)
axes[1].set_title(r'Escalón desplazado $u(t-2)$', fontweight='bold')
axes[1].set_xlabel(r'$t$')
axes[1].set_ylabel(r'$u(t-2)$')
axes[1].set_ylim(-0.2, 1.5)
axes[1].axhline(y=0, color='gray', lw=0.5)
axes[1].axvline(x=0, color='gray', lw=0.5)
axes[1].grid(True, alpha=0.3)

# --- Función pulso ---
a_p, b_p, A_p = 1, 4, 2
pulso = A_p * (np.where(t >= a_p, 1.0, 0.0) - np.where(t >= b_p, 1.0, 0.0))
axes[2].plot(t, pulso, color=COLOR_PUNTO, lw=2.5)
axes[2].fill_between(t, 0, pulso, alpha=0.2, color=COLOR_PUNTO)
axes[2].plot(a_p, A_p, 'o', color=COLOR_PUNTO, ms=8, zorder=5)
axes[2].plot(b_p, 0, 'o', color=COLOR_PUNTO, ms=8, zorder=5)
axes[2].set_title(r'Pulso $2[u(t-1) - u(t-4)]$', fontweight='bold')
axes[2].set_xlabel(r'$t$')
axes[2].set_ylabel(r'$p(t)$')
axes[2].set_ylim(-0.5, 3)
axes[2].axhline(y=0, color='gray', lw=0.5)
axes[2].axvline(x=0, color='gray', lw=0.5)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Propiedades de la Transformada de Laplace

Estas propiedades permiten calcular transformadas complejas a partir de las básicas de la tabla.

### 4.1 Linealidad

$$\boxed{\mathcal{L}\{\alpha f(t) + \beta g(t)\} = \alpha F(s) + \beta G(s)}$$

**Ejemplo:** $\mathcal{L}\{3t^2 + 5e^{-2t}\} = 3 \cdot \dfrac{2}{s^3} + 5 \cdot \dfrac{1}{s+2} = \dfrac{6}{s^3} + \dfrac{5}{s+2}$

### 4.2 Desplazamiento en frecuencia (*First Shifting Theorem*)

Si $\mathcal{L}\{f(t)\} = F(s)$, entonces:

$$\boxed{\mathcal{L}\{e^{at}f(t)\} = F(s-a)}$$

**Ejemplo:** $\mathcal{L}\{e^{-3t}\cos(2t)\}$

Sabemos que $\mathcal{L}\{\cos(2t)\} = \dfrac{s}{s^2+4}$. Sustituyendo $s \to s+3$:

$$\mathcal{L}\{e^{-3t}\cos(2t)\} = \frac{s+3}{(s+3)^2+4}$$

### 4.3 Desplazamiento en el tiempo (*Second Shifting Theorem*)

$$\boxed{\mathcal{L}\{f(t-a)\,u(t-a)\} = e^{-as}F(s), \qquad a \geq 0}$$

**Uso:** cuando una función "se enciende" en $t = a$.

**Ejemplo:** $\mathcal{L}\{(t-2)^3\,u(t-2)\}$

Con $f(t) = t^3$ y $a = 2$: $F(s) = \dfrac{6}{s^4}$, luego:

$$\mathcal{L}\{(t-2)^3\,u(t-2)\} = \frac{6\,e^{-2s}}{s^4}$$

### 4.4 Transformada de la derivada

Esta es la propiedad **clave** para resolver EDOs:

$$\boxed{\mathcal{L}\{f'(t)\} = s\,F(s) - f(0)}$$

$$\boxed{\mathcal{L}\{f''(t)\} = s^2 F(s) - s\,f(0) - f'(0)}$$

En general, para la derivada $n$-ésima:

$$\mathcal{L}\{f^{(n)}(t)\} = s^n F(s) - s^{n-1}f(0) - s^{n-2}f'(0) - \cdots - f^{(n-1)}(0)$$

**Truco para el examen:** cada derivada "sube" una potencia de $s$ y añade una condición inicial con signo negativo.

### 4.5 Transformada de la integral

$$\boxed{\mathcal{L}\left\{\int_0^t f(\tau)\,d\tau\right\} = \frac{F(s)}{s}}$$

**Ejemplo:** $\mathcal{L}\left\{\int_0^t e^{2\tau}\,d\tau\right\} = \dfrac{1}{s} \cdot \dfrac{1}{s-2} = \dfrac{1}{s(s-2)}$

### 4.6 Transformada de funciones periódicas

Si $f(t)$ es periódica con periodo $T > 0$, es decir, $f(t+T) = f(t)$:

$$\boxed{\mathcal{L}\{f(t)\} = \frac{1}{1 - e^{-sT}} \int_0^T f(t)\,e^{-st}\,dt}$$

**Utilidad:** solo necesitamos integrar en un periodo.

### 4.7 Resumen de propiedades

| Propiedad | $f(t)$ | $F(s)$ |
|:---|:---|:---|
| Linealidad | $\alpha f + \beta g$ | $\alpha F + \beta G$ |
| Despl. frecuencia | $e^{at}f(t)$ | $F(s-a)$ |
| Despl. tiempo | $f(t-a)\,u(t-a)$ | $e^{-as}F(s)$ |
| Derivada 1ª | $f'(t)$ | $sF(s) - f(0)$ |
| Derivada 2ª | $f''(t)$ | $s^2F(s) - sf(0) - f'(0)$ |
| Integral | $\int_0^t f(\tau)\,d\tau$ | $\dfrac{F(s)}{s}$ |
| Mult. por $t$ | $t\,f(t)$ | $-F'(s)$ |
| Div. por $t$ | $\dfrac{f(t)}{t}$ | $\int_s^{\infty}F(\sigma)\,d\sigma$ |
| Periódica | $f(t+T)=f(t)$ | $\dfrac{1}{1-e^{-sT}}\int_0^T f(t)e^{-st}dt$ |

```python
# Tabla visual: funciones f(t) y sus transformadas F(s)
fig, axes = plt.subplots(4, 2, figsize=(14, 16))

t = np.linspace(0, 5, 500)
s = np.linspace(0.1, 5, 500)

# Pares (f(t), F(s), nombre_f, nombre_F)
funciones = [
    (np.ones_like(t), 1/s, r'$f(t)=1$', r'$F(s)=1/s$'),
    (t, 1/s**2, r'$f(t)=t$', r'$F(s)=1/s^2$'),
    (np.exp(-t), 1/(s+1), r'$f(t)=e^{-t}$', r'$F(s)=1/(s+1)$'),
    (np.sin(2*t), 2/(s**2+4), r'$f(t)=\sin(2t)$', r'$F(s)=2/(s^2+4)$'),
]

colores = [COLOR_PRINCIPAL, COLOR_RECTA, COLOR_PUNTO, COLOR_AUX1]

for i, (ft, Fs, nombre_f, nombre_F) in enumerate(funciones):
    # Columna izquierda: f(t)
    axes[i, 0].plot(t, ft, color=colores[i], lw=2.5)
    axes[i, 0].set_title(nombre_f, fontsize=13, fontweight='bold')
    axes[i, 0].set_xlabel(r'$t$')
    axes[i, 0].set_ylabel(r'$f(t)$')
    axes[i, 0].grid(True, alpha=0.3)
    axes[i, 0].axhline(y=0, color='gray', lw=0.5)
    
    # Columna derecha: F(s)
    axes[i, 1].plot(s, Fs, color=colores[i], lw=2.5, ls='--')
    axes[i, 1].set_title(nombre_F, fontsize=13, fontweight='bold')
    axes[i, 1].set_xlabel(r'$s$')
    axes[i, 1].set_ylabel(r'$F(s)$')
    axes[i, 1].grid(True, alpha=0.3)
    axes[i, 1].axhline(y=0, color='gray', lw=0.5)

# Etiquetas de columna
axes[0, 0].annotate(r'Dominio del tiempo', xy=(0.5, 1.25),
                     xycoords='axes fraction', fontsize=14,
                     fontweight='bold', ha='center', color=COLOR_PRINCIPAL)
axes[0, 1].annotate(r'Dominio de Laplace', xy=(0.5, 1.25),
                     xycoords='axes fraction', fontsize=14,
                     fontweight='bold', ha='center', color=COLOR_RECTA)

plt.tight_layout()
plt.show()
```

## 5. Transformada inversa de Laplace

La **transformada inversa** recupera $f(t)$ a partir de $F(s)$:

$$\mathcal{L}^{-1}\{F(s)\} = f(t)$$

En la práctica, se usa la **descomposición en fracciones simples** para expresar $F(s)$ como suma de términos que aparecen en la tabla.

### 5.1 Raíces reales distintas

Si $F(s) = \dfrac{P(s)}{(s-a)(s-b)}$ con $a \neq b$:

$$F(s) = \frac{A}{s-a} + \frac{B}{s-b}$$

Los coeficientes se calculan por el **método de Heaviside** (cubrir):

$$A = \left.(s-a)\,F(s)\right|_{s=a}, \qquad B = \left.(s-b)\,F(s)\right|_{s=b}$$

**Ejemplo:** $F(s) = \dfrac{5s+3}{(s+1)(s+2)}$

$$A = \left.\frac{5s+3}{s+2}\right|_{s=-1} = \frac{-5+3}{-1+2} = \frac{-2}{1} = -2$$

$$B = \left.\frac{5s+3}{s+1}\right|_{s=-2} = \frac{-10+3}{-2+1} = \frac{-7}{-1} = 7$$

$$F(s) = \frac{-2}{s+1} + \frac{7}{s+2} \implies f(t) = -2e^{-t} + 7e^{-2t}$$

### 5.2 Raíces reales repetidas

Si $F(s) = \dfrac{P(s)}{(s-a)^n}$, la descomposición incluye:

$$F(s) = \frac{A_1}{s-a} + \frac{A_2}{(s-a)^2} + \cdots + \frac{A_n}{(s-a)^n}$$

**Ejemplo:** $F(s) = \dfrac{2s+5}{(s+1)^2(s+3)}$

$$F(s) = \frac{A}{s+1} + \frac{B}{(s+1)^2} + \frac{C}{s+3}$$

Calculamos:
- $B = \left.(s+1)^2 F(s)\right|_{s=-1} = \left.\frac{2s+5}{s+3}\right|_{s=-1} = \frac{3}{2}$
- $C = \left.(s+3)F(s)\right|_{s=-3} = \left.\frac{2s+5}{(s+1)^2}\right|_{s=-3} = \frac{-1}{4}$
- $A$: igualando coeficientes o sustituyendo un valor cómodo, por ejemplo $s = 0$:

$$\frac{5}{1 \cdot 3} = \frac{A}{1} + \frac{3/2}{1} + \frac{-1/4}{3} \implies \frac{5}{3} = A + \frac{3}{2} - \frac{1}{12} \implies A = \frac{5}{3} - \frac{17}{12} = \frac{1}{4}$$

$$f(t) = \frac{1}{4}e^{-t} + \frac{3}{2}\,t\,e^{-t} - \frac{1}{4}e^{-3t}$$

### 5.3 Raíces complejas conjugadas

Si el denominador tiene factores cuadráticos irreducibles $s^2 + bs + c$ con $b^2 - 4c < 0$:

$$\frac{As + B}{s^2 + bs + c}$$

Se completa el cuadrado en el denominador para obtener la forma $(s-\alpha)^2 + \omega^2$ y luego se identifica con senos/cosenos amortiguados.

**Ejemplo:** $F(s) = \dfrac{3s+7}{s^2+4s+13}$

Completando cuadrado: $s^2+4s+13 = (s+2)^2 + 9 = (s+2)^2 + 3^2$

$$F(s) = \frac{3s+7}{(s+2)^2+9} = \frac{3(s+2)+1}{(s+2)^2+9} = 3\cdot\frac{s+2}{(s+2)^2+9} + \frac{1}{3}\cdot\frac{3}{(s+2)^2+9}$$

$$\boxed{f(t) = 3\,e^{-2t}\cos(3t) + \frac{1}{3}\,e^{-2t}\sin(3t)}$$

**Truco:** siempre reescribir el numerador como $\alpha(s+a) + \beta$ donde $a$ es la parte real de las raíces.

## 6. Aplicación a EDOs de segundo orden

### 6.1 Método general

Para resolver una EDO lineal con coeficientes constantes y condiciones iniciales:

**Paso 1:** Aplicar $\mathcal{L}$ a ambos lados de la ecuación, usando las propiedades de la derivada.

**Paso 2:** Sustituir las condiciones iniciales $y(0)$, $y'(0)$.

**Paso 3:** Despejar $Y(s) = \mathcal{L}\{y(t)\}$.

**Paso 4:** Descomponer $Y(s)$ en fracciones simples.

**Paso 5:** Aplicar $\mathcal{L}^{-1}$ término a término para obtener $y(t)$.

### 6.2 Ejemplo resuelto: $y'' + 5y' + 6y = 2e^{-t}$

**Datos:** $y(0) = 1$, $y'(0) = 0$

**Paso 1:** Aplicar $\mathcal{L}$:

$$\mathcal{L}\{y''\} + 5\mathcal{L}\{y'\} + 6\mathcal{L}\{y\} = 2\mathcal{L}\{e^{-t}\}$$

$$[s^2Y(s) - sy(0) - y'(0)] + 5[sY(s) - y(0)] + 6Y(s) = \frac{2}{s+1}$$

**Paso 2:** Sustituir $y(0) = 1$, $y'(0) = 0$:

$$s^2Y - s - 0 + 5sY - 5 + 6Y = \frac{2}{s+1}$$

$$(s^2+5s+6)Y(s) = \frac{2}{s+1} + s + 5$$

**Paso 3:** Despejar $Y(s)$. Notemos que $s^2+5s+6 = (s+2)(s+3)$:

$$Y(s) = \frac{2}{(s+1)(s+2)(s+3)} + \frac{s+5}{(s+2)(s+3)}$$

**Paso 4:** Fracciones simples del primer término:

$$\frac{2}{(s+1)(s+2)(s+3)} = \frac{A}{s+1} + \frac{B}{s+2} + \frac{C}{s+3}$$

- $A = \left.\dfrac{2}{(s+2)(s+3)}\right|_{s=-1} = \dfrac{2}{1 \cdot 2} = 1$
- $B = \left.\dfrac{2}{(s+1)(s+3)}\right|_{s=-2} = \dfrac{2}{(-1)(1)} = -2$
- $C = \left.\dfrac{2}{(s+1)(s+2)}\right|_{s=-3} = \dfrac{2}{(-2)(-1)} = 1$

Segundo término:

$$\frac{s+5}{(s+2)(s+3)} = \frac{D}{s+2} + \frac{E}{s+3}$$

- $D = \left.\dfrac{s+5}{s+3}\right|_{s=-2} = \dfrac{3}{1} = 3$
- $E = \left.\dfrac{s+5}{s+2}\right|_{s=-3} = \dfrac{2}{-1} = -2$

**Total:**

$$Y(s) = \frac{1}{s+1} + \frac{-2+3}{s+2} + \frac{1-2}{s+3} = \frac{1}{s+1} + \frac{1}{s+2} - \frac{1}{s+3}$$

**Paso 5:** Antitransformar:

$$\boxed{y(t) = e^{-t} + e^{-2t} - e^{-3t}}$$

**Verificación:** $y(0) = 1 + 1 - 1 = 1$ $\checkmark$, $\;y'(0) = -1 - 2 + 3 = 0$ $\checkmark$

```python
# Solución de la EDO: y'' + 5y' + 6y = 2e^(-t), y(0)=1, y'(0)=0
t = np.linspace(0, 6, 500)

# Solución analítica
y = np.exp(-t) + np.exp(-2*t) - np.exp(-3*t)
dy = -np.exp(-t) - 2*np.exp(-2*t) + 3*np.exp(-3*t)
forzamiento = 2*np.exp(-t)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Gráfica principal: y(t) y y'(t) ---
axes[0].plot(t, y, color=COLOR_PRINCIPAL, lw=2.5, label=r"$y(t) = e^{-t} + e^{-2t} - e^{-3t}$")
axes[0].plot(t, dy, color=COLOR_RECTA, lw=2, ls='--', label=r"$y'(t)$")
axes[0].plot(0, 1, 'o', color=COLOR_PUNTO, ms=12, zorder=5, label=r'$y(0)=1$')
axes[0].plot(0, 0, 's', color=COLOR_AUX1, ms=10, zorder=5, label=r"$y'(0)=0$")
axes[0].axhline(y=0, color='gray', lw=0.5)
axes[0].set_xlabel(r'$t$')
axes[0].set_ylabel(r'$y$')
axes[0].set_title(r"Solución: $y'' + 5y' + 6y = 2e^{-t}$", fontweight='bold')
axes[0].legend(fontsize=10, loc='upper right')
axes[0].grid(True, alpha=0.3)

# --- Componentes de la solución ---
axes[1].plot(t, np.exp(-t), color=COLOR_PRINCIPAL, lw=2, label=r'$e^{-t}$')
axes[1].plot(t, np.exp(-2*t), color=COLOR_RECTA, lw=2, label=r'$e^{-2t}$')
axes[1].plot(t, -np.exp(-3*t), color=COLOR_AUX1, lw=2, label=r'$-e^{-3t}$')
axes[1].plot(t, y, color='black', lw=2.5, ls='--', label=r'$y(t)$ (suma)')
axes[1].axhline(y=0, color='gray', lw=0.5)
axes[1].set_xlabel(r'$t$')
axes[1].set_ylabel(r'Amplitud')
axes[1].set_title('Descomposición de la solución', fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Convolución y Teorema de Convolución

### 7.1 Definición de convolución

La **convolución** de dos funciones $f$ y $g$ se define como:

$$\boxed{(f * g)(t) = \int_0^t f(\tau)\,g(t-\tau)\,d\tau}$$

**Propiedades:**
- **Conmutativa:** $f * g = g * f$
- **Asociativa:** $(f * g) * h = f * (g * h)$
- **Distributiva:** $f * (g + h) = f * g + f * h$

### 7.2 Teorema de convolución

$$\boxed{\mathcal{L}\{f * g\} = F(s) \cdot G(s)}$$

Equivalentemente:

$$\mathcal{L}^{-1}\{F(s) \cdot G(s)\} = (f * g)(t) = \int_0^t f(\tau)\,g(t-\tau)\,d\tau$$

**Utilidad:** cuando $F(s) \cdot G(s)$ es difícil de descomponer en fracciones simples, pero $F(s)$ y $G(s)$ por separado son fáciles de antitransformar.

### 7.3 Ejemplo resuelto

Calcular $\mathcal{L}^{-1}\left\{\dfrac{1}{s(s+1)}\right\}$ usando convolución.

Identificamos: $F(s) = \dfrac{1}{s} \implies f(t) = 1$, $\quad G(s) = \dfrac{1}{s+1} \implies g(t) = e^{-t}$

$$(f * g)(t) = \int_0^t 1 \cdot e^{-\tau}\,d\tau = \left[-e^{-\tau}\right]_0^t = 1 - e^{-t}$$

$$\boxed{\mathcal{L}^{-1}\left\{\frac{1}{s(s+1)}\right\} = 1 - e^{-t}}$$

**Verificación por fracciones simples:** $\dfrac{1}{s(s+1)} = \dfrac{1}{s} - \dfrac{1}{s+1} \implies f(t) = 1 - e^{-t}$ $\checkmark$

```python
# Ejemplo visual de convolución: f(t)=1 con g(t)=e^(-t)
t = np.linspace(0, 6, 500)

f = np.ones_like(t)
g = np.exp(-t)
conv = 1 - np.exp(-t)  # resultado analítico

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# f(t)
axes[0].plot(t, f, color=COLOR_PRINCIPAL, lw=2.5)
axes[0].fill_between(t, 0, f, alpha=0.15, color=COLOR_PRINCIPAL)
axes[0].set_title(r'$f(t) = 1$', fontweight='bold')
axes[0].set_xlabel(r'$t$')
axes[0].set_ylabel(r'$f(t)$')
axes[0].set_ylim(-0.2, 1.5)
axes[0].grid(True, alpha=0.3)

# g(t)
axes[1].plot(t, g, color=COLOR_RECTA, lw=2.5)
axes[1].fill_between(t, 0, g, alpha=0.15, color=COLOR_RECTA)
axes[1].set_title(r'$g(t) = e^{-t}$', fontweight='bold')
axes[1].set_xlabel(r'$t$')
axes[1].set_ylabel(r'$g(t)$')
axes[1].set_ylim(-0.2, 1.5)
axes[1].grid(True, alpha=0.3)

# Convolución
axes[2].plot(t, conv, color=COLOR_PUNTO, lw=2.5)
axes[2].fill_between(t, 0, conv, alpha=0.15, color=COLOR_PUNTO)
axes[2].set_title(r'$(f * g)(t) = 1 - e^{-t}$', fontweight='bold')
axes[2].set_xlabel(r'$t$')
axes[2].set_ylabel(r'$(f*g)(t)$')
axes[2].set_ylim(-0.2, 1.5)
axes[2].axhline(y=1, color='gray', ls='--', lw=1, alpha=0.5)
axes[2].annotate(r'Asíntota $y=1$', xy=(4.5, 1.05), fontsize=10, color='gray')
axes[2].grid(True, alpha=0.3)

plt.suptitle(r'Teorema de convolución: $\mathcal{L}^{-1}\{F(s)\cdot G(s)\} = f * g$',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 8. Ejercicios resueltos del boletín

### Ejercicio 1: Transformada directa con propiedades

**Calcular** $\mathcal{L}\{t^2 e^{3t} + 4\sin(2t)\}$

**Solución:**

Por linealidad:

$$\mathcal{L}\{t^2 e^{3t}\} + 4\,\mathcal{L}\{\sin(2t)\}$$

Primer término: usamos la tabla $\mathcal{L}\{t^n e^{at}\} = \dfrac{n!}{(s-a)^{n+1}}$ con $n=2$, $a=3$:

$$\mathcal{L}\{t^2 e^{3t}\} = \frac{2!}{(s-3)^3} = \frac{2}{(s-3)^3}$$

Segundo término:

$$4\,\mathcal{L}\{\sin(2t)\} = 4 \cdot \frac{2}{s^2+4} = \frac{8}{s^2+4}$$

$$\boxed{\mathcal{L}\{t^2 e^{3t} + 4\sin(2t)\} = \frac{2}{(s-3)^3} + \frac{8}{s^2+4}}$$

### Ejercicio 2: Transformada inversa con raíces distintas

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{3s-1}{s^2-3s+2}\right\}$

**Solución:**

Factorizamos el denominador: $s^2-3s+2 = (s-1)(s-2)$

Fracciones simples:

$$\frac{3s-1}{(s-1)(s-2)} = \frac{A}{s-1} + \frac{B}{s-2}$$

- $A = \left.\dfrac{3s-1}{s-2}\right|_{s=1} = \dfrac{2}{-1} = -2$

- $B = \left.\dfrac{3s-1}{s-1}\right|_{s=2} = \dfrac{5}{1} = 5$

$$\boxed{f(t) = -2e^{t} + 5e^{2t}}$$

### Ejercicio 3: Transformada inversa con raíces complejas

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{2s+10}{s^2+6s+25}\right\}$

**Solución:**

Completamos cuadrado: $s^2+6s+25 = (s+3)^2 + 16 = (s+3)^2 + 4^2$

Reescribimos el numerador en función de $(s+3)$:

$$2s+10 = 2(s+3) + 4$$

$$F(s) = 2 \cdot \frac{s+3}{(s+3)^2+16} + \frac{4}{(s+3)^2+16} = 2 \cdot \frac{s+3}{(s+3)^2+16} + 1 \cdot \frac{4}{(s+3)^2+16}$$

$$\boxed{f(t) = 2\,e^{-3t}\cos(4t) + e^{-3t}\sin(4t)}$$

### Ejercicio 4: EDO de primer orden

**Resolver** $y' + 3y = e^{-t}$, con $y(0) = 2$.

**Solución:**

**Paso 1:** Aplicar $\mathcal{L}$:

$$sY(s) - y(0) + 3Y(s) = \frac{1}{s+1}$$

$$(s+3)Y(s) = \frac{1}{s+1} + 2$$

$$Y(s) = \frac{1}{(s+1)(s+3)} + \frac{2}{s+3}$$

**Paso 2:** Fracciones simples del primer término:

$$\frac{1}{(s+1)(s+3)} = \frac{A}{s+1} + \frac{B}{s+3}$$

- $A = \left.\dfrac{1}{s+3}\right|_{s=-1} = \dfrac{1}{2}$
- $B = \left.\dfrac{1}{s+1}\right|_{s=-3} = -\dfrac{1}{2}$

$$Y(s) = \frac{1/2}{s+1} + \frac{-1/2 + 2}{s+3} = \frac{1/2}{s+1} + \frac{3/2}{s+3}$$

$$\boxed{y(t) = \frac{1}{2}e^{-t} + \frac{3}{2}e^{-3t}}$$

**Verificación:** $y(0) = \frac{1}{2} + \frac{3}{2} = 2$ $\checkmark$

### Ejercicio 5: EDO de segundo orden con raíces complejas

**Resolver** $y'' + 4y = 0$, con $y(0) = 1$, $y'(0) = 6$.

**Solución:**

**Paso 1:** Aplicar $\mathcal{L}$:

$$s^2Y(s) - sy(0) - y'(0) + 4Y(s) = 0$$

$$(s^2+4)Y(s) = s + 6$$

$$Y(s) = \frac{s}{s^2+4} + \frac{6}{s^2+4} = \frac{s}{s^2+2^2} + 3 \cdot \frac{2}{s^2+2^2}$$

**Paso 2:** Antitransformar directamente de la tabla:

$$\boxed{y(t) = \cos(2t) + 3\sin(2t)}$$

**Verificación:** $y(0) = 1 + 0 = 1$ $\checkmark$, $\; y'(0) = 0 + 6 = 6$ $\checkmark$

### Ejercicio 6: Transformada con función escalón

**Calcular** $\mathcal{L}\{(t-3)^2\,u(t-3)\}$

**Solución:**

Usamos el segundo teorema de desplazamiento: $\mathcal{L}\{f(t-a)\,u(t-a)\} = e^{-as}F(s)$

Con $f(t) = t^2$ y $a = 3$: $F(s) = \dfrac{2}{s^3}$

$$\boxed{\mathcal{L}\{(t-3)^2\,u(t-3)\} = \frac{2\,e^{-3s}}{s^3}}$$

### Ejercicio 7: Transformada inversa con raíces repetidas

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{s+3}{(s+1)^2}\right\}$

**Solución:**

$$\frac{s+3}{(s+1)^2} = \frac{(s+1)+2}{(s+1)^2} = \frac{1}{s+1} + \frac{2}{(s+1)^2}$$

De la tabla: $\mathcal{L}^{-1}\left\{\dfrac{1}{s+1}\right\} = e^{-t}$ y $\mathcal{L}^{-1}\left\{\dfrac{1}{(s+1)^2}\right\} = t\,e^{-t}$

$$\boxed{f(t) = e^{-t} + 2t\,e^{-t} = (1+2t)e^{-t}}$$

### 9.1 Tipo 1: Transformada directa básica

Se aplica directamente la tabla de transformadas.

$$\boxed{\mathcal{L}\{t^n e^{at}\} = \frac{n!}{(s-a)^{n+1}}}$$

#### Ejercicio resuelto: Tipo 1

**Calcular** $\mathcal{L}\{5t^3 e^{-2t}\}$

**Datos:** $n = 3$, $a = -2$

$$\mathcal{L}\{5t^3 e^{-2t}\} = 5 \cdot \frac{3!}{(s+2)^4} = 5 \cdot \frac{6}{(s+2)^4}$$

$$\boxed{= \frac{30}{(s+2)^4}}$$

### 9.2 Tipo 2: Transformada con linealidad

Se descompone la función en sumandos y se transforma cada uno por separado.

$$\boxed{\mathcal{L}\{\alpha f + \beta g\} = \alpha F(s) + \beta G(s)}$$

#### Ejercicio resuelto: Tipo 2

**Calcular** $\mathcal{L}\{3\cosh(5t) - 7t^2 + e^{4t}\}$

$$= 3 \cdot \frac{s}{s^2 - 25} - 7 \cdot \frac{2}{s^3} + \frac{1}{s-4}$$

$$\boxed{= \frac{3s}{s^2-25} - \frac{14}{s^3} + \frac{1}{s-4}}$$

### 9.3 Tipo 3: Desplazamiento en frecuencia

Cuando $f(t)$ va multiplicada por $e^{at}$, se sustituye $s$ por $s-a$ en la transformada.

$$\boxed{\mathcal{L}\{e^{at}f(t)\} = F(s-a)}$$

#### Ejercicio resuelto: Tipo 3

**Calcular** $\mathcal{L}\{e^{2t}\sin(3t)\}$

**Paso 1:** $\mathcal{L}\{\sin(3t)\} = \dfrac{3}{s^2+9}$

**Paso 2:** Sustituir $s \to s-2$:

$$\boxed{\mathcal{L}\{e^{2t}\sin(3t)\} = \frac{3}{(s-2)^2+9}}$$

### 9.4 Tipo 4: Desplazamiento en tiempo (Heaviside)

Para funciones que "se encienden" en $t = a$.

$$\boxed{\mathcal{L}\{f(t-a)\,u(t-a)\} = e^{-as}F(s)}$$

#### Ejercicio resuelto: Tipo 4

**Calcular** $\mathcal{L}\{\sin(2(t-\pi))\,u(t-\pi)\}$

**Paso 1:** Identificar $f(t) = \sin(2t)$, $a = \pi$

**Paso 2:** $F(s) = \dfrac{2}{s^2+4}$

$$\boxed{\mathcal{L}\{\sin(2(t-\pi))\,u(t-\pi)\} = \frac{2\,e^{-\pi s}}{s^2+4}}$$

### 9.5 Tipo 5: Inversa con raíces reales distintas

Se factoriza el denominador y se aplica el **método de Heaviside** (cubrir).

$$\boxed{\frac{P(s)}{(s-a)(s-b)} = \frac{A}{s-a} + \frac{B}{s-b}, \quad A = \left.\frac{P(s)}{s-b}\right|_{s=a}}$$

#### Ejercicio resuelto: Tipo 5

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{7s+2}{s^2+s-6}\right\}$

**Paso 1:** $s^2+s-6 = (s+3)(s-2)$

**Paso 2:** $\dfrac{7s+2}{(s+3)(s-2)} = \dfrac{A}{s+3} + \dfrac{B}{s-2}$

- $A = \left.\dfrac{7s+2}{s-2}\right|_{s=-3} = \dfrac{-19}{-5} = \dfrac{19}{5}$

- $B = \left.\dfrac{7s+2}{s+3}\right|_{s=2} = \dfrac{16}{5}$

$$\boxed{f(t) = \frac{19}{5}e^{-3t} + \frac{16}{5}e^{2t}}$$

### 9.6 Tipo 6: Inversa con raíces repetidas

La descomposición incluye potencias crecientes del factor repetido.

$$\boxed{\frac{P(s)}{(s-a)^n} = \frac{A_1}{s-a} + \frac{A_2}{(s-a)^2} + \cdots + \frac{A_n}{(s-a)^n}}$$

#### Ejercicio resuelto: Tipo 6

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{3s^2+1}{s(s-1)^2}\right\}$

$$\frac{3s^2+1}{s(s-1)^2} = \frac{A}{s} + \frac{B}{s-1} + \frac{C}{(s-1)^2}$$

- $A = \left.\dfrac{3s^2+1}{(s-1)^2}\right|_{s=0} = \dfrac{1}{1} = 1$
- $C = \left.\dfrac{3s^2+1}{s}\right|_{s=1} = \dfrac{4}{1} = 4$
- $B$: sustituir $s=2$: $\dfrac{13}{2} = \dfrac{1}{2} + \dfrac{B}{1} + \dfrac{4}{1} \implies B = \dfrac{13}{2} - \dfrac{1}{2} - 4 = 2$

$$\boxed{f(t) = 1 + 2e^{t} + 4t\,e^{t}}$$

### 9.7 Tipo 7: Inversa con raíces complejas

Se completa el cuadrado y se reescribe el numerador.

$$\boxed{\frac{\alpha(s-a) + \beta\omega}{(s-a)^2+\omega^2} \implies \alpha\,e^{at}\cos(\omega t) + \beta\,e^{at}\sin(\omega t)}$$

#### Ejercicio resuelto: Tipo 7

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{s+5}{s^2+2s+10}\right\}$

**Paso 1:** Completar cuadrado: $s^2+2s+10 = (s+1)^2+9 = (s+1)^2+3^2$

**Paso 2:** Reescribir numerador: $s+5 = (s+1) + 4$

$$\frac{(s+1)+4}{(s+1)^2+9} = \frac{s+1}{(s+1)^2+9} + \frac{4}{3} \cdot \frac{3}{(s+1)^2+9}$$

$$\boxed{f(t) = e^{-t}\cos(3t) + \frac{4}{3}\,e^{-t}\sin(3t)}$$

### 9.8 Tipo 8: EDO de primer orden

$$\boxed{y' + ay = g(t) \implies (s+a)Y(s) = G(s) + y(0)}$$

#### Ejercicio resuelto: Tipo 8

**Resolver** $y' - 2y = 4$, con $y(0) = 1$.

$$sY - 1 - 2Y = \frac{4}{s} \implies (s-2)Y = \frac{4}{s} + 1 = \frac{s+4}{s}$$

$$Y(s) = \frac{s+4}{s(s-2)} = \frac{A}{s} + \frac{B}{s-2}$$

- $A = \left.\dfrac{s+4}{s-2}\right|_{s=0} = \dfrac{4}{-2} = -2$
- $B = \left.\dfrac{s+4}{s}\right|_{s=2} = \dfrac{6}{2} = 3$

$$\boxed{y(t) = -2 + 3e^{2t}}$$

**Verificación:** $y(0) = -2 + 3 = 1$ $\checkmark$

### 9.9 Tipo 9: EDO de segundo orden

$$\boxed{y'' + ay' + by = g(t) \implies (s^2+as+b)Y = G(s) + (s+a)y(0) + y'(0)}$$

#### Ejercicio resuelto: Tipo 9

**Resolver** $y'' + 2y' + 5y = 0$, con $y(0) = 2$, $y'(0) = -4$.

$$(s^2+2s+5)Y(s) = 2s + 4 + (-4) = 2s$$

$$Y(s) = \frac{2s}{s^2+2s+5} = \frac{2s}{(s+1)^2+4}$$

Reescribimos: $2s = 2(s+1) - 2$

$$Y(s) = 2\cdot\frac{s+1}{(s+1)^2+4} - \frac{2}{(s+1)^2+4} = 2\cdot\frac{s+1}{(s+1)^2+4} - 1\cdot\frac{2}{(s+1)^2+4}$$

$$\boxed{y(t) = 2\,e^{-t}\cos(2t) - e^{-t}\sin(2t)}$$

**Verificación:** $y(0) = 2 \cdot 1 - 0 = 2$ $\checkmark$

### 9.10 Tipo 10: Convolución

$$\boxed{\mathcal{L}^{-1}\{F(s)\cdot G(s)\} = \int_0^t f(\tau)\,g(t-\tau)\,d\tau}$$

#### Ejercicio resuelto: Tipo 10

**Calcular** $\mathcal{L}^{-1}\left\{\dfrac{1}{(s+1)(s+2)}\right\}$ usando convolución.

Identificamos: $F(s) = \dfrac{1}{s+1} \implies f(t) = e^{-t}$, $\quad G(s) = \dfrac{1}{s+2} \implies g(t) = e^{-2t}$

$$(f*g)(t) = \int_0^t e^{-\tau}\cdot e^{-2(t-\tau)}\,d\tau = e^{-2t}\int_0^t e^{\tau}\,d\tau = e^{-2t}\left[e^{\tau}\right]_0^t = e^{-2t}(e^t - 1)$$

$$\boxed{f(t) = e^{-t} - e^{-2t}}$$

### 9.11 Tipo 11: Función periódica

$$\boxed{\mathcal{L}\{f(t)\} = \frac{1}{1-e^{-sT}}\int_0^T f(t)\,e^{-st}\,dt}$$

#### Ejercicio resuelto: Tipo 11

**Calcular** la transformada de la onda cuadrada: $f(t) = 1$ para $0 \leq t < 1$, $f(t) = -1$ para $1 \leq t < 2$, con periodo $T = 2$.

$$\int_0^2 f(t)\,e^{-st}\,dt = \int_0^1 e^{-st}\,dt + \int_1^2 (-1)e^{-st}\,dt$$

$$= \frac{1-e^{-s}}{s} - \frac{e^{-s} - e^{-2s}}{s} = \frac{1 - 2e^{-s} + e^{-2s}}{s} = \frac{(1-e^{-s})^2}{s}$$

$$\boxed{F(s) = \frac{(1-e^{-s})^2}{s(1-e^{-2s})} = \frac{1-e^{-s}}{s(1+e^{-s})}}$$

donde usamos $1 - e^{-2s} = (1-e^{-s})(1+e^{-s})$.

### 9.12 Tipo 12: Multiplicación por $t$ (derivada de $F(s)$)

$$\boxed{\mathcal{L}\{t\,f(t)\} = -F'(s)}$$

#### Ejercicio resuelto: Tipo 12

**Calcular** $\mathcal{L}\{t\cos(3t)\}$

**Paso 1:** $F(s) = \mathcal{L}\{\cos(3t)\} = \dfrac{s}{s^2+9}$

**Paso 2:** Derivar:

$$F'(s) = \frac{(s^2+9) - s \cdot 2s}{(s^2+9)^2} = \frac{9-s^2}{(s^2+9)^2}$$

**Paso 3:** Aplicar el signo negativo:

$$\boxed{\mathcal{L}\{t\cos(3t)\} = -F'(s) = \frac{s^2-9}{(s^2+9)^2}}$$
