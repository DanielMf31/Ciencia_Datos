## Formulario

### Modelo atomico de Bohr

$$\boxed{E_n = -\left(\frac{Z}{n}\right)^2 E_0, \quad E_0 = 13.6 \text{ eV}}$$

$$\boxed{r_n = \frac{n^2}{Z} a_0, \quad a_0 = 0.529 \text{ A}}$$

### Energia del foton emitido/absorbido

$$\boxed{E_\gamma = h\nu = \frac{hc}{\lambda}, \quad \lambda \approx \frac{12400}{E_{nf} - E_{ni}} \text{ (A)}}$$

### Concentracion intrinseca de portadores

$$\boxed{n_i^2 = N_c \cdot N_v \cdot \exp\left(-\frac{E_g}{kT}\right)}$$

### Ley de accion de masas

$$\boxed{n \cdot p = n_i^2}$$

### Semiconductor tipo N (dopado con donores)

$$\boxed{n \approx N_D, \quad p = \frac{n_i^2}{N_D}}$$

### Semiconductor tipo P (dopado con aceptores)

$$\boxed{p \approx N_A, \quad n = \frac{n_i^2}{N_A}}$$

### Distribucion de Fermi-Dirac

$$\boxed{f(E) = \frac{1}{1 + \exp\left(\frac{E - E_F}{kT}\right)}}$$

### Nivel de Fermi intrinseco

$$\boxed{E_{Fi} \approx \frac{E_c + E_v}{2}}$$

### Constantes fundamentales

| Constante | Valor |
|-----------|-------|
| Carga elemental $e$ | $1.6 \times 10^{-19}$ C |
| Constante de Boltzmann $k$ | $8.62 \times 10^{-5}$ eV/K |

## 1. Modelo atomico: de Rutherford a la mecanica cuantica

### 1.1 Modelo de Rutherford (1911)

El modelo planetario propone que los electrones orbitan alrededor del nucleo, pero tiene un
problema fundamental: un electron acelerado (en movimiento circular) **emite radiacion**
y perderia energia continuamente, colapsando sobre el nucleo.

### 1.2 Modelo de Bohr (1913)

Bohr resuelve el problema imponiendo **postulados de cuantizacion**:

1. Los electrones solo pueden ocupar **orbitas estacionarias** con momento angular cuantizado: $L = n\hbar$
2. Solo se emite/absorbe radiacion al **transitar entre orbitas**: $E_\gamma = E_{n_f} - E_{n_i}$
3. Las orbitas permitidas tienen **energias discretas**: $E_n = -13.6/n^2$ eV (para el hidrogeno)

### 1.3 Modelo de Schrodinger (1926)

La mecanica cuantica completa reemplaza las orbitas por **orbitales** (funciones de onda $\psi$).
La probabilidad de encontrar al electron en una posicion se obtiene con $|\psi|^2$.
Cada estado se describe con 4 numeros cuanticos: $n, l, m_l, m_s$.

```python
# Niveles de energia del atomo de hidrogeno (modelo de Bohr)
fig, ax = plt.subplots(figsize=(10, 7))

n_levels = np.arange(1, 7)
E_n = -13.6 / n_levels**2

for n, E in zip(n_levels, E_n):
    ax.hlines(E, 0.2, 0.8, colors=COLOR_PRINCIPAL, linewidth=2.5)
    ax.text(0.85, E, f'n = {n}\n$E_{n}$ = {E:.2f} eV',
            va='center', fontsize=11, color=COLOR_FIJO)

# Transiciones de la serie de Balmer (a n=2)
transiciones = [(3, 2), (4, 2), (5, 2)]
colores_trans = [COLOR_ROJO, COLOR_NARANJA, COLOR_MORADO]
nombres = ['H-alpha (656 nm)', 'H-beta (486 nm)', 'H-gamma (434 nm)']

for (ni, nf), color, nombre in zip(transiciones, colores_trans, nombres):
    E_i = -13.6 / ni**2
    E_f = -13.6 / nf**2
    ax.annotate('', xy=(0.5, E_f), xytext=(0.5, E_i),
                arrowprops=dict(arrowstyle='->', color=color, lw=2))
    ax.text(0.52, (E_i + E_f)/2, nombre, fontsize=9, color=color)

ax.set_xlim(0, 1.2)
ax.set_ylim(-15, 1)
ax.set_ylabel('Energia (eV)', fontsize=13)
ax.set_title('Niveles de energia del atomo de hidrogeno', fontsize=14, fontweight='bold')
ax.set_xticks([])
ax.axhline(0, color=COLOR_FIJO, linestyle='--', alpha=0.5, label='Ionizacion (E = 0)')
ax.legend(fontsize=11)
ax.grid(False)
plt.tight_layout()
plt.show()
```

## 2. Bandas de energia en solidos cristalinos

Cuando los atomos se unen formando un cristal, los niveles de energia discretos se ensanchan
en **bandas** debido a la interaccion entre atomos vecinos.

### Conceptos clave

- **Banda de valencia (BV):** Ultima banda completamente llena a 0 K. Los electrones aqui
  estan ligados a los enlaces covalentes.
- **Banda de conduccion (BC):** Primera banda vacia (o parcialmente llena) por encima de la BV.
  Los electrones aqui son **libres** para conducir corriente.
- **Banda prohibida ($E_g$):** Gap de energia entre BV y BC. No existen estados permitidos.

### Clasificacion de materiales segun $E_g$

| Material | $E_g$ | Ejemplo |
|----------|--------|---------|
| **Conductor** | 0 eV (bandas solapadas) | Cu, Al, Au |
| **Semiconductor** | 0.1 - 4 eV | Si (1.12), Ge (0.66), GaAs (1.42) |
| **Aislante** | > 4 eV | SiO$_2$ (9), Diamante (5.5) |

```python
# Diagrama de bandas: conductor, semiconductor, aislante
fig, axes = plt.subplots(1, 3, figsize=(14, 6))

materiales = ['Conductor (Cu)', 'Semiconductor (Si)', 'Aislante (SiO$_2$)']
gaps = [0, 1.12, 9.0]

for ax, mat, Eg in zip(axes, materiales, gaps):
    # Banda de valencia
    bv = mpatches.FancyBboxPatch((0.15, 0.5), 0.7, 2.0,
                                  boxstyle='round,pad=0.05',
                                  facecolor=COLOR_CLARO, edgecolor=COLOR_PRINCIPAL, lw=2)
    ax.add_patch(bv)
    ax.text(0.5, 1.5, 'Banda de\nValencia', ha='center', va='center',
            fontsize=11, fontweight='bold', color=COLOR_PRINCIPAL)
    # Banda de conduccion
    bc_y = 2.5 + Eg * 0.3 if Eg > 0 else 2.3
    bc = mpatches.FancyBboxPatch((0.15, bc_y), 0.7, 2.0,
                                  boxstyle='round,pad=0.05',
                                  facecolor=to_rgba(COLOR_ROJO, 0.15),
                                  edgecolor=COLOR_ROJO, lw=2)
    ax.add_patch(bc)
    ax.text(0.5, bc_y + 1.0, 'Banda de\nConduccion', ha='center', va='center',
            fontsize=11, fontweight='bold', color=COLOR_ROJO)
    # Gap
    if Eg > 0:
        gap_mid = (2.5 + bc_y) / 2
        ax.annotate('', xy=(0.92, 2.5), xytext=(0.92, bc_y),
                    arrowprops=dict(arrowstyle='<->', color=COLOR_FIJO, lw=1.5))
        ax.text(0.97, gap_mid, f'$E_g$ = {Eg} eV', fontsize=10,
                va='center', color=COLOR_FIJO)
    else:
        ax.text(0.5, 2.65, 'Solapamiento', ha='center', fontsize=10,
                color=COLOR_NARANJA, fontstyle='italic')
    ax.set_xlim(0, 1.3)
    ax.set_ylim(0, 6.5)
    ax.set_title(mat, fontsize=13, fontweight='bold')
    ax.axis('off')

plt.suptitle('Diagrama de bandas de energia', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 3. Semiconductores intrinsecos

Un semiconductor **intrinseco** es un cristal puro, sin impurezas. A 0 K se comporta como
aislante, pero a temperatura ambiente los electrones adquieren energia termica suficiente
para saltar de la BV a la BC.

### Generacion de pares electron-hueco

Cada electron que salta a la BC deja un **hueco** en la BV. En un SC intrinseco:

$$n = p = n_i$$

donde $n_i$ es la **concentracion intrinseca**, que depende fuertemente de la temperatura:

$$n_i^2 = N_c \cdot N_v \cdot \exp\left(-\frac{E_g}{kT}\right)$$

### Valores de $n_i$ a 300 K

| Material | $E_g$ (eV) | $n_i$ (cm$^{-3}$) |
|----------|-----------|-------------------|
| Germanio (Ge) | 0.66 | $2.4 \times 10^{13}$ |
| Silicio (Si) | 1.12 | $1.5 \times 10^{10}$ |
| Arseniuro de Galio (GaAs) | 1.42 | $1.8 \times 10^{6}$ |

> **Nota:** A mayor $E_g$, menor $n_i$. El GaAs tiene muchisimos menos portadores
> intrinsecos que el Ge, lo que lo hace mejor para alta temperatura.

```python
# Concentracion intrinseca ni vs temperatura para Si, Ge y GaAs
k_eV = 8.62e-5  # eV/K

# Parametros: (Eg en eV a 300K, Nc*Nv producto aprox)
materiales_sc = {
    'Si':   {'Eg': 1.12, 'NcNv': 2.86e19 * 1.04e19, 'color': COLOR_PRINCIPAL},
    'Ge':   {'Eg': 0.66, 'NcNv': 1.04e19 * 6.0e18,  'color': COLOR_ROJO},
    'GaAs': {'Eg': 1.42, 'NcNv': 4.7e17 * 7.0e18,    'color': COLOR_PUNTO}
}

T = np.linspace(200, 600, 300)

fig, ax = plt.subplots(figsize=(10, 7))

for nombre, params in materiales_sc.items():
    ni = np.sqrt(params['NcNv'] * np.exp(-params['Eg'] / (k_eV * T)))
    ax.semilogy(T, ni, label=f"{nombre} ($E_g$ = {params['Eg']} eV)",
                color=params['color'], linewidth=2.5)

# Marcar T = 300 K
ax.axvline(300, color=COLOR_FIJO, linestyle='--', alpha=0.6, label='T = 300 K')

ax.set_xlabel('Temperatura (K)', fontsize=13)
ax.set_ylabel('$n_i$ (cm$^{-3}$)', fontsize=13)
ax.set_title('Concentracion intrinseca vs Temperatura', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.set_xlim(200, 600)
ax.set_ylim(1e4, 1e20)
plt.tight_layout()
plt.show()
```

## 4. Semiconductores extrinsecos (dopaje)

El **dopaje** consiste en introducir impurezas controladas para modificar la concentracion
de portadores.

### 4.1 Tipo N (donores)

Se introduce un atomo del grupo V (P, As, Sb) que tiene **5 electrones de valencia**.
Cuatro forman enlaces covalentes con el Si, y el quinto queda debilmente ligado.

- El atomo donor crea un **nivel de energia** $E_D$ justo debajo de $E_c$ ($\sim 0.045$ eV para P en Si)
- A temperatura ambiente, el electron extra se ioniza facilmente: $n \approx N_D$
- Los electrones son **portadores mayoritarios**, los huecos son minoritarios

$$n \approx N_D, \quad p = \frac{n_i^2}{N_D}$$

### 4.2 Tipo P (aceptores)

Se introduce un atomo del grupo III (B, Al, Ga) que tiene **3 electrones de valencia**.
Le falta un electron para completar los 4 enlaces, creando un hueco.

- El atomo aceptor crea un **nivel de energia** $E_A$ justo encima de $E_v$ ($\sim 0.045$ eV para B en Si)
- A temperatura ambiente, acepta un electron de la BV: $p \approx N_A$
- Los huecos son **portadores mayoritarios**, los electrones son minoritarios

$$p \approx N_A, \quad n = \frac{n_i^2}{N_A}$$

```python
# Diagramas de bandas para semiconductores tipo N y tipo P
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# --- Tipo N ---
ax = axes[0]
# Banda de conduccion
ax.hlines(4.0, 0.1, 0.9, colors=COLOR_ROJO, linewidth=3, label='$E_c$')
ax.text(0.92, 4.0, '$E_c$', fontsize=12, va='center', color=COLOR_ROJO)
# Banda de valencia
ax.hlines(1.0, 0.1, 0.9, colors=COLOR_PRINCIPAL, linewidth=3, label='$E_v$')
ax.text(0.92, 1.0, '$E_v$', fontsize=12, va='center', color=COLOR_PRINCIPAL)
# Nivel donor
ax.hlines(3.7, 0.25, 0.75, colors=COLOR_PUNTO, linewidth=2, linestyle='--')
ax.text(0.77, 3.7, '$E_D$', fontsize=12, va='center', color=COLOR_PUNTO)
# Nivel Fermi
ax.hlines(3.4, 0.1, 0.9, colors=COLOR_NARANJA, linewidth=2, linestyle='-.')
ax.text(0.92, 3.4, '$E_F$', fontsize=12, va='center', color=COLOR_NARANJA)
# Nivel Fermi intrinseco
ax.hlines(2.5, 0.1, 0.9, colors=COLOR_FIJO, linewidth=1.5, linestyle=':')
ax.text(0.92, 2.5, '$E_{Fi}$', fontsize=11, va='center', color=COLOR_FIJO)
# Electrones (portadores mayoritarios)
for x in [0.25, 0.4, 0.5, 0.6, 0.75]:
    ax.plot(x, 4.15, 'o', color=COLOR_ROJO, markersize=10)
ax.text(0.5, 4.45, 'Electrones (mayoritarios)', ha='center', fontsize=10, color=COLOR_ROJO)
# Hueco (minoritario)
ax.plot(0.5, 0.85, 'o', color=COLOR_PRINCIPAL, markersize=8, fillstyle='none', mew=2)
ax.text(0.5, 0.55, 'Huecos (minoritarios)', ha='center', fontsize=10, color=COLOR_PRINCIPAL)
ax.set_title('Semiconductor Tipo N', fontsize=14, fontweight='bold')
ax.set_xlim(0, 1.15)
ax.set_ylim(0, 5)
ax.set_ylabel('Energia (eV) - esquematico', fontsize=12)
ax.axis('off')

# --- Tipo P ---
ax = axes[1]
# Banda de conduccion
ax.hlines(4.0, 0.1, 0.9, colors=COLOR_ROJO, linewidth=3)
ax.text(0.92, 4.0, '$E_c$', fontsize=12, va='center', color=COLOR_ROJO)
# Banda de valencia
ax.hlines(1.0, 0.1, 0.9, colors=COLOR_PRINCIPAL, linewidth=3)
ax.text(0.92, 1.0, '$E_v$', fontsize=12, va='center', color=COLOR_PRINCIPAL)
# Nivel aceptor
ax.hlines(1.3, 0.25, 0.75, colors=COLOR_MORADO, linewidth=2, linestyle='--')
ax.text(0.77, 1.3, '$E_A$', fontsize=12, va='center', color=COLOR_MORADO)
# Nivel Fermi
ax.hlines(1.6, 0.1, 0.9, colors=COLOR_NARANJA, linewidth=2, linestyle='-.')
ax.text(0.92, 1.6, '$E_F$', fontsize=12, va='center', color=COLOR_NARANJA)
# Nivel Fermi intrinseco
ax.hlines(2.5, 0.1, 0.9, colors=COLOR_FIJO, linewidth=1.5, linestyle=':')
ax.text(0.92, 2.5, '$E_{Fi}$', fontsize=11, va='center', color=COLOR_FIJO)
# Huecos (mayoritarios)
for x in [0.25, 0.4, 0.5, 0.6, 0.75]:
    ax.plot(x, 0.85, 'o', color=COLOR_PRINCIPAL, markersize=10, fillstyle='none', mew=2)
ax.text(0.5, 0.55, 'Huecos (mayoritarios)', ha='center', fontsize=10, color=COLOR_PRINCIPAL)
# Electron (minoritario)
ax.plot(0.5, 4.15, 'o', color=COLOR_ROJO, markersize=8)
ax.text(0.5, 4.45, 'Electrones (minoritarios)', ha='center', fontsize=10, color=COLOR_ROJO)
ax.set_title('Semiconductor Tipo P', fontsize=14, fontweight='bold')
ax.set_xlim(0, 1.15)
ax.set_ylim(0, 5)
ax.axis('off')

plt.suptitle('Diagramas de bandas con dopaje', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 5. Estadistica de Fermi-Dirac

La funcion de distribucion de Fermi-Dirac da la **probabilidad** de que un estado de
energia $E$ este ocupado por un electron:

$$f(E) = \frac{1}{1 + \exp\left(\frac{E - E_F}{kT}\right)}$$

### Propiedades fundamentales

1. **En $E = E_F$:** $f(E_F) = 0.5$ (50% de probabilidad de ocupacion)
2. **Para $E \gg E_F$:** $f(E) \to 0$ (estados vacios)
3. **Para $E \ll E_F$:** $f(E) \to 1$ (estados llenos)
4. **A $T = 0$ K:** La funcion es un escalon: $f = 1$ si $E < E_F$, $f = 0$ si $E > E_F$
5. **Al aumentar T:** La transicion se suaviza, mas electrones tienen energia para
   saltar por encima de $E_F$

### Interpretacion para semiconductores

- La probabilidad de encontrar un **electron** en la BC es $f(E)$
- La probabilidad de encontrar un **hueco** en la BV es $1 - f(E)$
- El nivel de Fermi $E_F$ actua como referencia energetica del sistema

```python
# Distribucion de Fermi-Dirac para diferentes temperaturas
E_F = 0  # Referencia: nivel de Fermi en 0 eV
k_eV = 8.62e-5  # eV/K

E = np.linspace(-0.5, 0.5, 500)

temperaturas = [50, 150, 300, 500]
colores_T = [COLOR_PRINCIPAL, COLOR_PUNTO, COLOR_NARANJA, COLOR_ROJO]

fig, ax = plt.subplots(figsize=(10, 7))

for T_val, color in zip(temperaturas, colores_T):
    kT = k_eV * T_val
    f_E = 1.0 / (1.0 + np.exp((E - E_F) / kT))
    ax.plot(E, f_E, label=f'T = {T_val} K (kT = {kT*1000:.1f} meV)',
            color=color, linewidth=2.5)

# T = 0 K (escalon ideal)
ax.step([-0.5, 0, 0, 0.5], [1, 1, 0, 0], color=COLOR_FIJO,
        linewidth=2, linestyle='--', label='T = 0 K (escalon)')

ax.axhline(0.5, color=COLOR_FIJO, linestyle=':', alpha=0.5)
ax.axvline(0, color=COLOR_FIJO, linestyle=':', alpha=0.5)
ax.text(0.02, 0.52, '$f(E_F) = 0.5$', fontsize=11, color=COLOR_FIJO)

ax.set_xlabel('$E - E_F$ (eV)', fontsize=13)
ax.set_ylabel('$f(E)$', fontsize=13)
ax.set_title('Distribucion de Fermi-Dirac', fontsize=14, fontweight='bold')
ax.legend(fontsize=11, loc='upper right')
ax.set_ylim(-0.05, 1.1)
plt.tight_layout()
plt.show()
```

## 6. Concentracion de portadores en equilibrio

La concentracion de electrones y huecos en equilibrio termico se calcula combinando la
densidad de estados $g(E)$ con la funcion de Fermi-Dirac $f(E)$:

### En la banda de conduccion (electrones)

$$n = N_c \exp\left(-\frac{E_c - E_F}{kT}\right)$$

donde $N_c = 2\left(\frac{2\pi m_n^* kT}{h^2}\right)^{3/2}$ es la densidad efectiva de estados en la BC.

### En la banda de valencia (huecos)

$$p = N_v \exp\left(-\frac{E_F - E_v}{kT}\right)$$

donde $N_v = 2\left(\frac{2\pi m_p^* kT}{h^2}\right)^{3/2}$ es la densidad efectiva de estados en la BV.

### Ley de accion de masas

Multiplicando $n \cdot p$:

$$n \cdot p = N_c N_v \exp\left(-\frac{E_g}{kT}\right) = n_i^2$$

Esta relacion es **independiente del nivel de Fermi** y se cumple siempre en equilibrio,
tanto para semiconductores intrinsecos como extrinsecos.

```python
# Concentracion de portadores vs dopaje (Si a 300 K)
ni_Si = 1.5e10  # cm^-3 a 300 K

N_D = np.logspace(12, 18, 200)  # rango de dopaje donores
n_N = N_D.copy()  # mayoritarios tipo N
p_N = ni_Si**2 / N_D  # minoritarios tipo N

fig, ax = plt.subplots(figsize=(10, 7))

ax.loglog(N_D, n_N, label='$n$ (electrones, mayoritarios)',
          color=COLOR_ROJO, linewidth=2.5)
ax.loglog(N_D, p_N, label='$p$ (huecos, minoritarios)',
          color=COLOR_PRINCIPAL, linewidth=2.5)
ax.axhline(ni_Si, color=COLOR_PUNTO, linestyle='--', linewidth=2,
           label=f'$n_i$ = {ni_Si:.1e} cm$^{{-3}}$')

ax.fill_between(N_D, ni_Si, n_N, alpha=0.1, color=COLOR_ROJO)
ax.fill_between(N_D, p_N, ni_Si, alpha=0.1, color=COLOR_PRINCIPAL)

ax.set_xlabel('Concentracion de donores $N_D$ (cm$^{-3}$)', fontsize=13)
ax.set_ylabel('Concentracion de portadores (cm$^{-3}$)', fontsize=13)
ax.set_title('Portadores en Si tipo N a 300 K', fontsize=14, fontweight='bold')
ax.legend(fontsize=12)
ax.set_xlim(1e12, 1e18)
ax.set_ylim(1e2, 1e19)
plt.tight_layout()
plt.show()
```

### Ejercicio 1: Niveles de energia del atomo de hidrogeno

**Enunciado:** Calcular la energia de los 4 primeros niveles del atomo de hidrogeno
segun el modelo de Bohr. Determinar la longitud de onda del foton emitido en la
transicion $n = 3 \to n = 2$ (linea H-alpha de la serie de Balmer).

```python
# Ejercicio 1: Niveles de energia del hidrogeno
print('=' * 60)
print('EJERCICIO 1: Niveles de energia de Bohr (Hidrogeno)')
print('=' * 60)

# Datos
E_0 = 13.6  # eV
Z = 1       # Hidrogeno

# Paso 1: Calcular energias para n = 1, 2, 3, 4
print('\nPaso 1: Energias de cada nivel')
print('-' * 40)
for n in range(1, 5):
    E_n = -(Z/n)**2 * E_0
    print(f'  n = {n}: E_{n} = -({Z}/{n})^2 * {E_0} = {E_n:.4f} eV')

# Paso 2: Transicion n=3 -> n=2
print('\nPaso 2: Transicion n = 3 -> n = 2')
print('-' * 40)
E_3 = -(Z/3)**2 * E_0
E_2 = -(Z/2)**2 * E_0
delta_E = E_3 - E_2  # negativo (emision)
print(f'  E_3 = {E_3:.4f} eV')
print(f'  E_2 = {E_2:.4f} eV')
print(f'  Delta E = E_3 - E_2 = {delta_E:.4f} eV')
print(f'  Energia del foton: E_gamma = |Delta E| = {abs(delta_E):.4f} eV')

# Paso 3: Longitud de onda
print('\nPaso 3: Longitud de onda')
print('-' * 40)
lambda_A = 12400 / abs(delta_E)  # en Angstrom
lambda_nm = lambda_A / 10        # en nm
print(f'  lambda = 12400 / {abs(delta_E):.4f} = {lambda_A:.1f} A = {lambda_nm:.1f} nm')
print(f'\n  >>> RESULTADO: lambda(H-alpha) = {lambda_nm:.1f} nm <<<')
```

### Ejercicio 2: Concentracion intrinseca del Silicio

**Enunciado:** Calcular la concentracion intrinseca de portadores $n_i$ del Silicio
a $T = 300$ K, sabiendo que $N_c = 2.86 \times 10^{19}$ cm$^{-3}$,
$N_v = 1.04 \times 10^{19}$ cm$^{-3}$ y $E_g = 1.12$ eV.

```python
# Ejercicio 2: Concentracion intrinseca del Si a 300 K
print('=' * 60)
print('EJERCICIO 2: Concentracion intrinseca ni del Silicio')
print('=' * 60)

# Datos
Nc = 2.86e19   # cm^-3
Nv = 1.04e19   # cm^-3
Eg = 1.12      # eV
T = 300        # K
k = 8.62e-5    # eV/K

# Paso 1: Calcular kT
print('\nPaso 1: Energia termica')
print('-' * 40)
kT = k * T
print(f'  kT = {k:.2e} * {T} = {kT:.4f} eV = {kT*1000:.2f} meV')

# Paso 2: Calcular el exponencial
print('\nPaso 2: Factor exponencial')
print('-' * 40)
exponente = -Eg / kT
factor_exp = np.exp(exponente)
print(f'  -Eg/(kT) = -{Eg}/({kT:.4f}) = {exponente:.2f}')
print(f'  exp(-Eg/(kT)) = exp({exponente:.2f}) = {factor_exp:.4e}')

# Paso 3: Calcular ni^2 y ni
print('\nPaso 3: Concentracion intrinseca')
print('-' * 40)
ni_cuad = Nc * Nv * factor_exp
ni = np.sqrt(ni_cuad)
print(f'  ni^2 = Nc * Nv * exp(-Eg/(kT))')
print(f'  ni^2 = {Nc:.2e} * {Nv:.2e} * {factor_exp:.4e}')
print(f'  ni^2 = {ni_cuad:.4e} cm^-6')
print(f'  ni = sqrt(ni^2) = {ni:.4e} cm^-3')
print(f'\n  >>> RESULTADO: ni(Si, 300K) = {ni:.2e} cm^-3 <<<')
print(f'  (Valor de referencia: 1.5e10 cm^-3)')
```

### Ejercicio 3: Semiconductor tipo N

**Enunciado:** Una muestra de Silicio se dopa con $N_D = 10^{16}$ cm$^{-3}$ atomos de fosforo.
A $T = 300$ K, calcular:

a) La concentracion de electrones $n$ y huecos $p$

b) La posicion del nivel de Fermi respecto al nivel de Fermi intrinseco $E_F - E_{Fi}$

```python
# Ejercicio 3: Si tipo N con ND = 1e16 cm^-3
print('=' * 60)
print('EJERCICIO 3: Semiconductor tipo N')
print('=' * 60)

# Datos
ND = 1e16      # cm^-3
ni = 1.5e10    # cm^-3 (Si a 300 K)
T = 300        # K
k = 8.62e-5    # eV/K
kT = k * T     # eV

print(f'\nDatos: ND = {ND:.0e} cm^-3, ni = {ni:.1e} cm^-3, T = {T} K')

# Paso 1: Verificar dopaje >> ni (regimen extrinseco)
print('\nPaso 1: Verificar regimen extrinseco')
print('-' * 40)
ratio = ND / ni
print(f'  ND / ni = {ND:.0e} / {ni:.1e} = {ratio:.2e}')
print(f'  Como ND >> ni, estamos en regimen extrinseco')

# Paso 2: Concentracion de electrones
print('\nPaso 2: Concentracion de electrones (mayoritarios)')
print('-' * 40)
n = ND
print(f'  n = ND = {n:.2e} cm^-3')

# Paso 3: Concentracion de huecos (ley de accion de masas)
print('\nPaso 3: Concentracion de huecos (minoritarios)')
print('-' * 40)
p = ni**2 / n
print(f'  p = ni^2 / n = ({ni:.1e})^2 / {n:.0e}')
print(f'  p = {ni**2:.2e} / {n:.0e}')
print(f'  p = {p:.2e} cm^-3')

# Paso 4: Nivel de Fermi
print('\nPaso 4: Posicion del nivel de Fermi')
print('-' * 40)
delta_EF = kT * np.log(n / ni)
print(f'  EF - EFi = kT * ln(n/ni)')
print(f'  EF - EFi = {kT:.4f} * ln({n:.0e}/{ni:.1e})')
print(f'  EF - EFi = {kT:.4f} * {np.log(n/ni):.4f}')
print(f'  EF - EFi = {delta_EF:.4f} eV')

print(f'\n  >>> RESULTADOS <<<')
print(f'  n = {n:.2e} cm^-3')
print(f'  p = {p:.2e} cm^-3')
print(f'  EF - EFi = +{delta_EF:.4f} eV (EF por encima de EFi)')
```

### Ejercicio 4: Semiconductor tipo P

**Enunciado:** Una muestra de Silicio se dopa con $N_A = 10^{17}$ cm$^{-3}$ atomos de boro.
A $T = 300$ K, calcular:

a) La concentracion de huecos $p$ y electrones $n$

b) La posicion del nivel de Fermi respecto al nivel de Fermi intrinseco $E_F - E_{Fi}$

c) Verificar que se cumple la ley de accion de masas $n \cdot p = n_i^2$

```python
# Ejercicio 4: Si tipo P con NA = 1e17 cm^-3
print('=' * 60)
print('EJERCICIO 4: Semiconductor tipo P')
print('=' * 60)

# Datos
NA = 1e17      # cm^-3
ni = 1.5e10    # cm^-3 (Si a 300 K)
T = 300        # K
k = 8.62e-5    # eV/K
kT = k * T     # eV

print(f'\nDatos: NA = {NA:.0e} cm^-3, ni = {ni:.1e} cm^-3, T = {T} K')

# Paso 1: Verificar regimen extrinseco
print('\nPaso 1: Verificar regimen extrinseco')
print('-' * 40)
ratio = NA / ni
print(f'  NA / ni = {NA:.0e} / {ni:.1e} = {ratio:.2e}')
print(f'  Como NA >> ni, estamos en regimen extrinseco')

# Paso 2: Concentracion de huecos (mayoritarios)
print('\nPaso 2: Concentracion de huecos (mayoritarios)')
print('-' * 40)
p = NA
print(f'  p = NA = {p:.2e} cm^-3')

# Paso 3: Concentracion de electrones (minoritarios)
print('\nPaso 3: Concentracion de electrones (minoritarios)')
print('-' * 40)
n = ni**2 / p
print(f'  n = ni^2 / p = ({ni:.1e})^2 / {p:.0e}')
print(f'  n = {ni**2:.2e} / {p:.0e}')
print(f'  n = {n:.2e} cm^-3')

# Paso 4: Nivel de Fermi
print('\nPaso 4: Posicion del nivel de Fermi')
print('-' * 40)
delta_EF = kT * np.log(n / ni)  # Sera negativo (EF por debajo de EFi)
print(f'  EF - EFi = kT * ln(n/ni)')
print(f'  EF - EFi = {kT:.4f} * ln({n:.2e}/{ni:.1e})')
print(f'  EF - EFi = {kT:.4f} * ({np.log(n/ni):.4f})')
print(f'  EF - EFi = {delta_EF:.4f} eV')

# Paso 5: Verificacion ley de accion de masas
print('\nPaso 5: Verificacion ley de accion de masas')
print('-' * 40)
np_prod = n * p
ni_cuad = ni**2
print(f'  n * p = {n:.2e} * {p:.2e} = {np_prod:.4e} cm^-6')
print(f'  ni^2 = ({ni:.1e})^2 = {ni_cuad:.4e} cm^-6')
print(f'  Verificacion: n*p = ni^2? {np.isclose(np_prod, ni_cuad, rtol=1e-6)}')

print(f'\n  >>> RESULTADOS <<<')
print(f'  p = {p:.2e} cm^-3')
print(f'  n = {n:.2e} cm^-3')
print(f'  EF - EFi = {delta_EF:.4f} eV (EF por debajo de EFi)')
print(f'  n*p = ni^2 VERIFICADO')
```
