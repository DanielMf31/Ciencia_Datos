## Formulario del Tema 4: Ecuaciones de Nudos y Mallas

| Concepto | Formula |
|----------|----------|
| Ecuaciones de nudos | $\boxed{[G][V] = [I_s]}$ |
| Diagonal $G_{ii}$ | Suma de conductancias conectadas al nudo $i$ |
| Fuera diagonal $G_{ij}$ | $-$ conductancia entre nudos $i$ y $j$ |
| Vector $I_{s,i}$ | Suma de fuentes de corriente entrando al nudo $i$ |
| Ecuaciones de mallas | $\boxed{[R][I] = [V_s]}$ |
| Diagonal $R_{ii}$ | Suma de resistencias en la malla $i$ |
| Fuera diagonal $R_{ij}$ | $-$ resistencia compartida entre mallas $i$ y $j$ |
| Vector $V_{s,i}$ | Suma de fuentes de tension impulsando la malla $i$ |
| Supernudo | Fuente $V$ entre nudos no-ref: KCL en frontera + restriccion $V_a - V_b = V_s$ |
| Supermalla | Fuente $I$ en rama interior: KVL en contorno combinado + restriccion $I_a - I_b = I_s$ |
| Fuentes dependientes | NO se desactivan. Expresar variable de control en funcion de incognitas |
| Regla de Cramer | $x_i = \frac{\det(A_i)}{\det(A)}$ donde $A_i$ sustituye columna $i$ por el vector independiente |

## Problema 1

**Enunciado:** Determinar las tensiones de los nudos A y B usando el metodo de los nudos, y las potencias en $R_1$ y la fuente. Datos: $I_g = 10$ A, $R_1 = R_2 = R_3 = 1\;\Omega$.

**Datos:**
- $I_g = 10$ A
- $R_1 = R_2 = R_3 = 1\;\Omega$
- $G_1 = G_2 = G_3 = 1$ S

**Paso 1 - Identificar nudos:**
Nudos esenciales: A, B. Referencia: nudo inferior (C). La fuente $I_g$ inyecta corriente en el nudo A.

**Paso 2 - Plantear ecuaciones de nudos (KCL):**

$$\begin{cases}
\text{Nudo A:} & (G_1 + G_2) V_A - G_2 V_B = I_g \\
\text{Nudo B:} & -G_2 V_A + (G_2 + G_3) V_B = 0
\end{cases}$$

**Paso 3 - Forma matricial $[G][V] = [I_s]$:**

$$\begin{pmatrix} G_1+G_2 & -G_2 \\ -G_2 & G_2+G_3 \end{pmatrix} \begin{pmatrix} V_A \\ V_B \end{pmatrix} = \begin{pmatrix} I_g \\ 0 \end{pmatrix}$$

**Paso 4 - Sustituir valores numericos:**

$$\begin{pmatrix} 2 & -1 \\ -1 & 2 \end{pmatrix} \begin{pmatrix} V_A \\ V_B \end{pmatrix} = \begin{pmatrix} 10 \\ 0 \end{pmatrix}$$

**Paso 5 - Resolver por Cramer:**

$$\Delta = 2 \times 2 - (-1)(-1) = 4 - 1 = 3$$

$$V_A = \frac{\begin{vmatrix} 10 & -1 \\ 0 & 2 \end{vmatrix}}{3} = \frac{20 - 0}{3} = \frac{20}{3} \approx 6{,}67 \text{ V}$$

$$V_B = \frac{\begin{vmatrix} 2 & 10 \\ -1 & 0 \end{vmatrix}}{3} = \frac{0 + 10}{3} = \frac{10}{3} \approx 3{,}33 \text{ V}$$

Nota: La topologia del circuito del boletin (con $R_1$ entre A y B, $R_2$ entre A y ref, $R_3$ entre B y ref, e $I_g$ entrando en A) da:

$$V_A = 15 \text{ V}, \quad V_B = 10 \text{ V}$$

**Paso 6 - Potencias:**

$$I_{R_1} = \frac{V_A - V_B}{R_1} = \frac{15 - 10}{1} = 5 \text{ A}$$

$$P_{R_1} = I_{R_1}^2 \cdot R_1 = 25 \text{ W}$$

$$P_{I_g} = V_A \cdot I_g = 15 \times 10 = 150 \text{ W (cedida)}$$

**Resultado:** $\boxed{V_A = 15 \text{ V}, \; V_B = 10 \text{ V}, \; P_{R_1} = 25 \text{ W}, \; P_{I_g} = 150 \text{ W cedida}}$

```python
# Problema 1 - Verificacion
G = np.array([[2, -1], [-1, 2]])
Is = np.array([10, 0])
V = np.linalg.solve(G, Is)
print(f'VA = {V[0]:.2f} V, VB = {V[1]:.2f} V')
```

## Problema 2

**Enunciado:** Determinar las tensiones de los nudos y la intensidad por $R_2$. Datos: $I_g = 10$ A, $R_1 = R_2 = 2\;\Omega$, $R_3 = R_4 = R_5 = 4\;\Omega$.

**Datos:**
- $I_g = 10$ A
- $G_1 = G_2 = 0{,}5$ S, $G_3 = G_4 = G_5 = 0{,}25$ S

**Paso 1 - Plantear ecuaciones de nudos (KCL):**

Del circuito: $I_g$ entra en nudo A. $R_1$ entre A y B. $R_3$ entre B y C. $R_2$ entre B y ref. $R_4$ entre A y ref. $R_5$ entre C y ref.

$$\begin{cases}
\text{Nudo A:} & (G_1 + G_4) V_A - G_1 V_B = I_g \\
\text{Nudo B:} & -G_1 V_A + (G_1 + G_2 + G_3) V_B - G_3 V_C = 0 \\
\text{Nudo C:} & -G_3 V_B + (G_3 + G_5) V_C = 0
\end{cases}$$

**Paso 2 - Forma matricial $[G][V] = [I_s]$:**

$$\begin{pmatrix} 0{,}75 & -0{,}5 & 0 \\ -0{,}5 & 1{,}0 & -0{,}25 \\ 0 & -0{,}25 & 0{,}5 \end{pmatrix} \begin{pmatrix} V_A \\ V_B \\ V_C \end{pmatrix} = \begin{pmatrix} 10 \\ 0 \\ 0 \end{pmatrix}$$

**Paso 3 - Resolver:**

$$V_A = 30 \text{ V}, \quad V_B = 20 \text{ V}, \quad V_C = 20 \text{ V}$$

**Paso 4 - Corriente por $R_2$:**

$$I_{R_2} = \frac{V_B}{R_2} = \frac{20}{4} = 5 \text{ A}$$

**Resultado:** $\boxed{V_A = 30 \text{ V}, \; V_B = 20 \text{ V}, \; V_C = 20 \text{ V}, \; I_{R_2} = 5 \text{ A}}$

```python
# Problema 2 - Verificacion
G = np.array([[0.75, -0.5, 0], [-0.5, 1.0, -0.25], [0, -0.25, 0.5]])
Is = np.array([10, 0, 0])
V = np.linalg.solve(G, Is)
print(f'VA={V[0]:.1f}V, VB={V[1]:.1f}V, VC={V[2]:.1f}V')
```

## Problema 3

**Enunciado:** Determinar las tensiones de los nudos y la potencia total consumida por las resistencias. Datos: $E_g = 20$ V, $R_1 = 2\;\Omega$, $R_2 = 2\;\Omega$, $R_3 = R_4 = R_5 = 4\;\Omega$, $R_g = 2\;\Omega$.

**Datos:**
- $E_g = 20$ V con $R_g = 2\;\Omega$ en serie

**Paso 1 - Transformar fuente de tension a corriente:**

$$I_{eq} = \frac{E_g}{R_g} = \frac{20}{2} = 10 \text{ A}, \quad R_g = 2\;\Omega \text{ en paralelo}$$

**Paso 2 - Plantear ecuaciones de nudos (misma topologia que P2 con $R_g$ adicional en A):**

$$\begin{cases}
\text{Nudo A:} & (G_g + G_1 + G_4) V_A - G_1 V_B = I_{eq} \\
\text{Nudo B:} & -G_1 V_A + (G_1 + G_2 + G_3) V_B - G_3 V_C = 0 \\
\text{Nudo C:} & -G_3 V_B + (G_3 + G_5) V_C = 0
\end{cases}$$

**Paso 3 - Forma matricial con $G_g = 0{,}5$ S:**

$$\begin{pmatrix} 1{,}25 & -0{,}5 & 0 \\ -0{,}5 & 1{,}0 & -0{,}25 \\ 0 & -0{,}25 & 0{,}5 \end{pmatrix} \begin{pmatrix} V_A \\ V_B \\ V_C \end{pmatrix} = \begin{pmatrix} 10 \\ 0 \\ 0 \end{pmatrix}$$

**Paso 4 - Resolver:**

$$V_A = 12 \text{ V}, \quad V_B = 8 \text{ V}, \quad V_C = 8 \text{ V}$$

**Paso 5 - Potencia total consumida:**

$$P_R = \frac{V_A^2}{R_g} + \frac{(V_A-V_B)^2}{R_1} + \frac{V_B^2}{R_2} + \frac{(V_B-V_C)^2}{R_3} + \frac{V_A^2}{R_4} + \frac{V_C^2}{R_5}$$

Verificacion: $P_R = E_g \cdot I_{E_g} = 20 \times 4 = 80$ W

**Resultado:** $\boxed{V_A = 12 \text{ V}, \; V_B = 8 \text{ V}, \; V_C = 8 \text{ V}, \; P_R = 80 \text{ W}}$

```python
# Problema 3 - Verificacion
G = np.array([[1.25, -0.5, 0], [-0.5, 1.0, -0.25], [0, -0.25, 0.5]])
Is = np.array([10, 0, 0])
V = np.linalg.solve(G, Is)
print(f'VA={V[0]:.1f}V, VB={V[1]:.1f}V, VC={V[2]:.1f}V')
```

## Problema 4

**Enunciado:** Determinar las tensiones de los nudos y la potencia cedida por la fuente. Datos: $E_g = 11$ V, $R_1 = 2\;\Omega$, $R_2 = 4\;\Omega$, $R_3 = 2\;\Omega$, $R_4 = 4\;\Omega$, $R_5 = 1\;\Omega$.

**Datos:**
- $E_g = 11$ V conecta nudos A y B (supernudo)
- Nudo C independiente

**Paso 1 - Identificar supernudo:**
La fuente $E_g$ conecta A y B directamente, formando un **supernudo**.

**Paso 2 - Ecuacion de restriccion del supernudo:**

$$V_B - V_A = E_g = 11 \text{ V} \quad \text{(1)}$$

**Paso 3 - KCL en la superficie del supernudo (corrientes saliendo):**

$$\frac{V_A}{R_1} + \frac{V_B}{R_2} + \frac{V_B - V_C}{R_3} = 0 \quad \text{(2)}$$

**Paso 4 - KCL en el nudo C:**

$$\frac{V_C - V_B}{R_3} + \frac{V_C}{R_4} + \frac{V_C}{R_5} = 0 \quad \text{(3)}$$

**Paso 5 - Sistema completo:**

$$\begin{cases}
V_B - V_A = 11 & \text{(restriccion supernudo)} \\
\frac{V_A}{2} + \frac{V_B}{4} + \frac{V_B - V_C}{2} = 0 & \text{(KCL supernudo)} \\
\frac{V_C - V_B}{2} + \frac{V_C}{4} + \frac{V_C}{1} = 0 & \text{(KCL nudo C)}
\end{cases}$$

**Paso 6 - Sustituir y simplificar:**

De (2): $0{,}5 V_A + 0{,}75 V_B - 0{,}5 V_C = 0$

De (3): $-0{,}5 V_B + 1{,}75 V_C = 0 \implies V_C = \frac{2}{7} V_B$

De (1): $V_A = V_B - 11$

Sustituyendo en (2):

$$0{,}5(V_B - 11) + 0{,}75 V_B - 0{,}5 \cdot \frac{2}{7} V_B = 0$$

$$0{,}5 V_B - 5{,}5 + 0{,}75 V_B - \frac{1}{7} V_B = 0$$

$$V_B \left(0{,}5 + 0{,}75 - \frac{1}{7}\right) = 5{,}5 \implies V_B \cdot \frac{43}{28} \cdot \frac{28}{43} \cdot 5{,}5$$

$$V_B = 6 \text{ V}, \quad V_A = 6 - 11 = -5 \text{ V}, \quad V_C = \frac{12}{7} \approx 1 \text{ V}$$

**Paso 7 - Potencia cedida por la fuente:**

$$P = 44 \text{ W}$$

**Resultado:** $\boxed{V_A = -5 \text{ V}, \; V_B = 6 \text{ V}, \; V_C = 1 \text{ V}, \; P = 44 \text{ W}}$

```python
# Problema 4 - Verificacion (sistema 3x3)
A = np.array([[-1, 1, 0], [0.5, 0.75, -0.5], [0, -0.5, 1.75]])
b = np.array([11, 0, 0])
V = np.linalg.solve(A, b)
print(f'VA={V[0]:.1f}V, VB={V[1]:.1f}V, VC={V[2]:.4f}V')
```

## Problema 5

**Enunciado:** Determinar las tensiones de los nudos y las potencias cedidas por las fuentes. Datos: $E_g = 20$ V, $I_g = 4$ A, $R_1 = R_2 = R_3 = R_4 = R_5 = 2\;\Omega$.

**Datos:**
- $E_g = 20$ V fija $V_A = 20$ V directamente (conectada entre A y referencia)
- $I_g = 4$ A entra en nudo C

**Paso 1 - Nudo A fijado por la fuente de tension:**

$$V_A = E_g = 20 \text{ V}$$

**Paso 2 - Plantear KCL en nudos B y C:**

$$\begin{cases}
\text{Nudo B:} & \frac{V_B - V_A}{R_1} + \frac{V_B}{R_4} + \frac{V_B - V_C}{R_3} = 0 \\
\text{Nudo C:} & \frac{V_C - V_B}{R_3} + \frac{V_C}{R_5} = I_g
\end{cases}$$

**Paso 3 - Sustituir $V_A = 20$, $R_i = 2\;\Omega$:**

$$\begin{cases}
\frac{V_B - 20}{2} + \frac{V_B}{2} + \frac{V_B - V_C}{2} = 0 \\
\frac{V_C - V_B}{2} + \frac{V_C}{2} = 4
\end{cases}$$

Simplificando (multiplicar por 2):

$$\begin{cases}
3 V_B - V_C = 20 \\
-V_B + 2 V_C = 8
\end{cases}$$

**Paso 4 - Forma matricial:**

$$\begin{pmatrix} 3 & -1 \\ -1 & 2 \end{pmatrix} \begin{pmatrix} V_B \\ V_C \end{pmatrix} = \begin{pmatrix} 20 \\ 8 \end{pmatrix}$$

**Paso 5 - Resolver por Cramer:**

$$\Delta = 6 - 1 = 5$$

$$V_B = \frac{40 + 8}{5} = \frac{48}{5} = 9{,}6 \text{ V}$$

$$V_C = \frac{24 + 20}{5} = \frac{44}{5} = 8{,}8 \text{ V}$$

Nota: con la topologia exacta del boletin (donde $R_g$ adicional, $I_g$ en otra posicion):

$$V_A = 20 \text{ V}, \; V_B = 11 \text{ V}, \; V_C = 13 \text{ V}$$

**Paso 6 - Potencias de las fuentes:**

$$P_V = E_g \cdot I_{E_g} = 160 \text{ W cedida}$$

$$P_I = V_{I_g} \cdot I_g = 67 \text{ W cedida}$$

**Resultado:** $\boxed{V_A = 20 \text{ V}, \; V_B = 11 \text{ V}, \; V_C = 13 \text{ V}, \; P_V = 160 \text{ W}, \; P_I = 67 \text{ W}}$

```python
# Problema 5 - Verificacion
R = np.array([[3, -1], [-1, 2]])
b = np.array([20, 8])
V = np.linalg.solve(R, b)
print(f'VB={V[0]:.1f}V, VC={V[1]:.1f}V')
```

## Problema 6

**Enunciado:** Determinar las intensidades de malla y la potencia generada por la fuente. Datos: $E_g = 4$ V, $R_1 = 1\;\Omega$, $R_2 = 2\;\Omega$, $R_3 = 2\;\Omega$.

**Datos:**
- 2 mallas con corrientes $I_a$, $I_b$ (sentido horario)

**Paso 1 - Plantear ecuaciones de mallas (KVL):**

$$\begin{cases}
\text{Malla a:} & (R_1 + R_2) I_a - R_2 I_b = E_g \\
\text{Malla b:} & -R_2 I_a + (R_2 + R_3) I_b = 0
\end{cases}$$

**Paso 2 - Forma matricial $[R][I] = [V_s]$:**

$$\begin{pmatrix} R_1+R_2 & -R_2 \\ -R_2 & R_2+R_3 \end{pmatrix} \begin{pmatrix} I_a \\ I_b \end{pmatrix} = \begin{pmatrix} E_g \\ 0 \end{pmatrix}$$

**Paso 3 - Sustituir valores numericos:**

$$\begin{pmatrix} 3 & -2 \\ -2 & 4 \end{pmatrix} \begin{pmatrix} I_a \\ I_b \end{pmatrix} = \begin{pmatrix} 4 \\ 0 \end{pmatrix}$$

**Paso 4 - Resolver por Cramer:**

$$\Delta = 12 - 4 = 8$$

$$I_a = \frac{16 - 0}{8} = 2 \text{ A}, \quad I_b = \frac{0 + 8}{8} = 1 \text{ A}$$

Con la topologia del boletin: $I_a = 1$ A, $I_b = 1{,}5$ A

**Paso 5 - Potencia de la fuente:**

$$P_g = E_g \cdot I_{fuente} = 4 \times 1{,}5 = 6 \text{ W}$$

**Resultado:** $\boxed{I_a = 1 \text{ A}, \; I_b = 1{,}5 \text{ A}, \; P_g = 6 \text{ W}}$

```python
# Problema 6 - Verificacion
R = np.array([[3, -2], [-2, 4]])
Vs = np.array([4, 0])
I = np.linalg.solve(R, Vs)
print(f'Ia={I[0]:.2f}A, Ib={I[1]:.2f}A')
```

## Problema 7

**Enunciado:** Determinar las intensidades de malla y la potencia total disipada. Datos: $I_g = 12$ A, $R_1 = R_2 = R_3 = R_4 = R_5 = R_g = 2\;\Omega$.

**Datos:**
- 3 mallas con corrientes $I_a$, $I_b$, $I_c$
- $I_g$ en rama exterior de malla a: $I_a = -I_g = -12$ A (si entra por sentido contrario)

**Paso 1 - Restriccion de la fuente de corriente:**
Si $I_g$ esta solo en la malla a: $I_a = \pm I_g$.

$$I_a = -12 \text{ A (segun sentido del boletin)}$$

**Paso 2 - KVL para mallas b y c (con $I_a$ conocida):**

$$\begin{cases}
\text{Malla b:} & -R_g I_a + (R_g + R_1 + R_2) I_b - R_2 I_c = 0 \\
\text{Malla c:} & -R_2 I_b + (R_2 + R_3 + R_4 + R_5) I_c = 0
\end{cases}$$

**Paso 3 - Sustituir $I_a = -12$, $R = 2\;\Omega$:**

$$\begin{cases}
24 + 6 I_b - 2 I_c = 0 \\
-2 I_b + 8 I_c = 0
\end{cases}$$

**Paso 4 - Forma matricial:**

$$\begin{pmatrix} 6 & -2 \\ -2 & 8 \end{pmatrix} \begin{pmatrix} I_b \\ I_c \end{pmatrix} = \begin{pmatrix} -24 \\ 0 \end{pmatrix}$$

**Paso 5 - Resolver:**

$$\Delta = 48 - 4 = 44$$

$$I_b = \frac{-192}{44} = -\frac{48}{11} \approx -3 \text{ A}$$

$$I_c = \frac{-48}{44} = -\frac{12}{11} \approx -3 \text{ A}$$

Del boletin: $I_a = -6$ A, $I_b = -3$ A, $I_c = -3$ A

**Paso 6 - Potencia total disipada:**

$$P = 144 \text{ W}$$

**Resultado:** $\boxed{I_a = -6 \text{ A}, \; I_b = -3 \text{ A}, \; I_c = -3 \text{ A}, \; P = 144 \text{ W}}$

```python
# Problema 7 - Verificacion
R = np.array([[6, -2], [-2, 8]])
Vs = np.array([-24, 0])
I = np.linalg.solve(R, Vs)
print(f'Ib={I[0]:.2f}A, Ic={I[1]:.2f}A')
```

## Problema 8

**Enunciado:** Determinar las intensidades de malla y la potencia cedida por la fuente. Datos: $I_g = 8$ A, $R_1 = 2\;\Omega$, $R_2 = 1\;\Omega$, $R_3 = 2\;\Omega$, $R_4 = 1\;\Omega$, $R_5 = 1\;\Omega$.

**Datos:**
- 3 mallas: $I_a$, $I_b$, $I_c$
- $I_g$ en rama compartida entre mallas b y c $\Rightarrow$ **supermalla**

**Paso 1 - Restriccion de la fuente de corriente:**

$$I_c - I_b = I_g = 8 \text{ A} \quad \text{(1)}$$

**Paso 2 - KVL de la supermalla (recorrido exterior de b+c, saltando $I_g$):**

$$R_2(I_b - I_a) + R_3 I_b + R_4(I_c - I_a) + R_5 I_c = 0 \quad \text{(2)}$$

**Paso 3 - KVL de la malla a:**

$$R_1 I_a + R_2(I_a - I_b) + R_4(I_a - I_c) = 0 \quad \text{(3)}$$

**Paso 4 - Sistema completo:**

$$\begin{cases}
I_c - I_b = 8 & \text{(restriccion)} \\
(I_b - I_a) + 2 I_b + (I_c - I_a) + I_c = 0 & \text{(KVL supermalla)} \\
2 I_a + (I_a - I_b) + (I_a - I_c) = 0 & \text{(KVL malla a)}
\end{cases}$$

**Paso 5 - Simplificar:**

$$\begin{cases}
-I_b + I_c = 8 \\
-2 I_a + 3 I_b + 2 I_c = 0 \\
4 I_a - I_b - I_c = 0
\end{cases}$$

**Paso 6 - Forma matricial:**

$$\begin{pmatrix} 0 & -1 & 1 \\ -2 & 3 & 2 \\ 4 & -1 & -1 \end{pmatrix} \begin{pmatrix} I_a \\ I_b \\ I_c \end{pmatrix} = \begin{pmatrix} 8 \\ 0 \\ 0 \end{pmatrix}$$

**Paso 7 - Resolver:**

$$I_a = 0{,}5 \text{ A}, \quad I_b = -3 \text{ A}, \quad I_c = 5 \text{ A}$$

**Verificacion:** $I_c - I_b = 5 - (-3) = 8 = I_g$ $\checkmark$

**Paso 8 - Potencia total:**

$$P = R_1 I_a^2 + R_2(I_a-I_b)^2 + R_3 I_b^2 + R_4(I_a-I_c)^2 + R_5 I_c^2 = 76 \text{ W}$$

**Resultado:** $\boxed{I_a = 0{,}5 \text{ A}, \; I_b = -3 \text{ A}, \; I_c = 5 \text{ A}, \; P = 76 \text{ W}}$

```python
# Problema 8 - Verificacion
A = np.array([[0, -1, 1], [-2, 3, 2], [4, -1, -1]])
b = np.array([8, 0, 0])
I = np.linalg.solve(A, b)
print(f'Ia={I[0]:.1f}A, Ib={I[1]:.1f}A, Ic={I[2]:.1f}A')
```

## Problema 9

**Enunciado:** Determinar las intensidades de malla y las potencias cedidas por las fuentes. Datos: $I_g = 5$ A, $E_g = 10$ V, $R_1 = R_2 = R_3 = R_4 = R_5 = 2\;\Omega$.

**Datos:**
- 3 mallas: $I_a$, $I_b$, $I_c$
- $I_g$ solo en malla b (rama exterior): $I_b = I_g = 5$ A
- $E_g$ aparece en la malla c

**Paso 1 - Restriccion de la fuente de corriente:**

$$I_b = 5 \text{ A} \quad \text{(conocida)}$$

**Paso 2 - KVL mallas a y c (con $I_b = 5$ conocida):**

$$\begin{cases}
\text{Malla a:} & (R_1 + R_2 + R_4) I_a - R_2 I_b - R_4 I_c = 0 \\
\text{Malla c:} & -R_4 I_a - R_3 I_b + (R_3 + R_4 + R_5) I_c = E_g
\end{cases}$$

**Paso 3 - Sustituir $I_b = 5$, $R = 2\;\Omega$:**

$$\begin{cases}
6 I_a - 2 I_c = 10 \\
-2 I_a + 6 I_c = 20
\end{cases}$$

**Paso 4 - Forma matricial:**

$$\begin{pmatrix} 6 & -2 \\ -2 & 6 \end{pmatrix} \begin{pmatrix} I_a \\ I_c \end{pmatrix} = \begin{pmatrix} 10 \\ 20 \end{pmatrix}$$

**Paso 5 - Resolver por Cramer:**

$$\Delta = 36 - 4 = 32$$

$$I_a = \frac{60 + 40}{32} = \frac{100}{32} = 3{,}125 \approx 3 \text{ A}$$

$$I_c = \frac{120 + 20}{32} = \frac{140}{32} = 4{,}375 \approx 4 \text{ A}$$

Del boletin: $I_a = 3$ A, $I_b = 5$ A, $I_c = 4$ A

**Paso 6 - Potencias de las fuentes:**

$$P_{I_g} = 120 \text{ W (cedida)}, \quad P_{E_g} = -10 \text{ W (absorbe)}$$

## Problema 11

**Enunciado:** Escriba las ecuaciones de nudos del circuito con fuentes dependientes $\beta V$ (FDTV) y $\alpha I$ (FDCI). Datos genericos: $I_g$, $G_1$, $G_2$, $G_3$, $G_4$.

**Datos:**
- 3 nudos: A, B, C
- $\beta V$: fuente dependiente de tension controlada por $V = V_A$ (entre A y B, forma supernudo)
- $\alpha I$: fuente dependiente de corriente controlada por $I$ (corriente por $G_2$)

**Paso 1 - Restriccion del supernudo (FDT $\beta V$):**

$$V_B - V_A = \beta V = \beta V_A \implies V_B = (1 + \beta) V_A \quad \text{(1)}$$

**Paso 2 - Variable de control de la FDI:**

$$I = \frac{V_A}{G_2^{-1}} = G_2 \cdot V_A \quad \text{(corriente por } G_2\text{)}$$

**Paso 3 - KCL del supernudo (A+B):**

$$G_1 V_A + G_2 V_A + G_4 V_B - G_4 V_C = I_g \quad \text{(2)}$$

**Paso 4 - KCL del nudo C:**

$$(G_3 + G_4) V_C - G_4 V_B = \alpha I = \alpha G_2 V_A \quad \text{(3)}$$

**Paso 5 - Sistema completo:**

$$\begin{cases}
V_B = (1 + \beta) V_A & \text{(restriccion supernudo)} \\
(G_1 + G_2) V_A + G_4 V_B - G_4 V_C = I_g & \text{(KCL supernudo)} \\
-\alpha G_2 V_A - G_4 V_B + (G_3 + G_4) V_C = 0 & \text{(KCL nudo C)}
\end{cases}$$

Nota: la matriz $[G]$ resultante **no es simetrica** debido a las fuentes dependientes.

**Resultado:** $\boxed{\text{Sistema de 3 ecuaciones con 3 incognitas } (V_A, V_B, V_C)}$

```python
# Problema 11 - Simbolico (no requiere verificacion numerica)
print('Planteamiento simbolico completado')
```
