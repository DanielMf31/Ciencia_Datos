## Formulario del Tema 3: Fuentes Reales y Thevenin/Norton

| Concepto | Formula |
|----------|---------|
| Fuente real de tension | $V = E_{Th} - R_{eq} \cdot I$ |
| Fuente real de corriente | $I = I_N - V / R_{eq}$ |
| Equivalente Thevenin | $\boxed{V_{Th} = V_{ca}}$ (tension circuito abierto) |
| Resistencia Thevenin | $\boxed{R_{Th} = V_{ca} / I_{cc}}$ o desactivar fuentes indep. |
| Equivalente Norton | $\boxed{I_N = I_{cc} = V_{Th} / R_{Th}}$ |
| Conversion Th-No | $V_{Th} = I_N \cdot R_N$, $R_{Th} = R_N$ |
| Max transferencia potencia | $\boxed{P_{max} = \frac{V_{Th}^2}{4 R_{Th}}}$ cuando $R_L = R_{Th}$ |
| Rendimiento en max potencia | $\eta = 50\%$ |
| Rendimiento general | $\eta = \frac{P_{carga}}{P_{generada}} = \frac{R_L}{R_L + R_{Th}}$ |
| Fuentes dependientes | NO se desactivan. Usar fuente de prueba para $R_{Th}$ |

## Problema 1

**Enunciado:** El circuito C es una fuente real de tension. El amperimetro lee 2 A, la resistencia absorbe 100 W y la eficiencia es del 50%. Determinar $E_{Th}$ y $R_{eq}$.

**Datos:**
- $I = 2$ A (amperimetro)
- $P_R = 100$ W (potencia en la carga)
- $\eta = 50\%$

**Paso 1 - Resistencia de carga:**

$$R = \frac{P_R}{I^2} = \frac{100}{4} = 25\;\Omega$$

**Paso 2 - Potencia total y tension de la fuente:**

Con eficiencia del 50%, la potencia generada es el doble de la absorbida:

$$P_{gen} = \frac{P_R}{\eta} = \frac{100}{0.5} = 200\;\text{W}$$

$$E_{Th} = \frac{P_{gen}}{I} = \frac{200}{2} = 100\;\text{V}$$

**Paso 3 - Resistencia interna:**

$$R_{eq} = \frac{P_{gen} - P_R}{I^2} = \frac{100}{4} = 25\;\Omega$$

**Resultado:** $\boxed{E_{Th} = 100\;\text{V}, \quad R_{eq} = 25\;\Omega}$

```python
# Problema 1
I = 2       # A
P_R = 100   # W
eta = 0.5

R = P_R / I**2
P_gen = P_R / eta
E_Th = P_gen / I
R_eq = (P_gen - P_R) / I**2

print(f'R = {R} ohm')
print(f'E_Th = {E_Th} V')
print(f'R_eq = {R_eq} ohm')
```

## Problema 2

**Enunciado:** Una fuente real de tension ($V_g = 2$ V, $R_g = 1\;\Omega$) alimenta una carga $R = 1\;\Omega$. Una fuente ideal de intensidad $I_g = 1$ A esta conectada en paralelo con $R$, inyectando corriente hacia el nudo de carga. Determinar la potencia cedida por $V_g$ a la carga y el rendimiento.

**Datos:**
- $V_g = 2$ V, $R_g = 1\;\Omega$
- $R = 1\;\Omega$, $I_g = 1$ A

**Paso 1 - Tension en el nudo de carga:**

KCL en el nudo A (entre $R_g$ y $R$), con $I_g$ inyectando corriente:

$$\frac{V_g - V_A}{R_g} + I_g = \frac{V_A}{R}$$

$$\frac{2 - V_A}{1} + 1 = \frac{V_A}{1} \Rightarrow 3 = 2V_A \Rightarrow V_A = 1.5\;\text{V}$$

**Paso 2 - Corriente de la fuente $V_g$:**

$$I_{Rg} = \frac{V_g - V_A}{R_g} = \frac{2 - 1.5}{1} = 0.5\;\text{A}$$

**Paso 3 - Potencia cedida por $V_g$ a la carga y rendimiento:**

La potencia que $V_g$ entrega al nudo de carga (a traves de $R_g$):

$$P_{cedida} = V_A \cdot I_{Rg} = 1.5 \cdot 0.5 = 0.75\;\text{W}$$

La potencia total generada por $V_g$:

$$P_{Vg} = V_g \cdot I_{Rg} = 2 \cdot 0.5 = 1\;\text{W}$$

$$\eta = \frac{P_{cedida}}{P_{Vg}} = \frac{0.75}{1} = 75\%$$

**Resultado:** $\boxed{P = 0.75\;\text{W}, \quad \eta = 75\%}$

```python
# Problema 2
Vg = 2; Rg = 1; R = 1; Ig = 1

VA = (Vg / Rg + Ig) / (1/Rg + 1/R)  # KCL
I_Rg = (Vg - VA) / Rg
P_cedida = VA * I_Rg
P_Vg = Vg * I_Rg
eta = P_cedida / P_Vg

print(f'VA = {VA} V')
print(f'I_Rg = {I_Rg} A')
print(f'P_cedida = {P_cedida} W')
print(f'eta = {eta*100}%')
```

## Problema 3

**Enunciado:** Una fuente real de intensidad tiene $R_g = 4\;\text{k}\Omega$. La carga es $R$ en serie con $400\;\Omega$. El rendimiento es del 80%. Determinar $R$.

**Datos:**
- $R_g = 4000\;\Omega$
- Resistencia en serie con la carga: $400\;\Omega$
- $\eta = 80\%$

**Paso 1 - Rendimiento de una fuente real de intensidad:**

En una fuente real de intensidad, $R_g$ esta en paralelo con la fuente ideal. La carga total es $R_L = R + 400\;\Omega$.

$$\eta = \frac{R_g}{R_g + R_L}$$

**Paso 2 - Despejando $R_L$:**

$$0.8 = \frac{4000}{4000 + R_L} \Rightarrow R_L = \frac{4000(1 - 0.8)}{0.8} = 1000\;\Omega$$

**Paso 3 - Despejando $R$:**

$$R = R_L - 400 = 600\;\Omega$$

**Resultado:** $\boxed{R = 600\;\Omega}$

```python
# Problema 3
Rg = 4000  # ohm
R_serie = 400  # ohm
eta = 0.8

R_L = Rg * (1 - eta) / eta
R = R_L - R_serie

print(f'R_L = {R_L} ohm')
print(f'R = {R} ohm')
```

## Problema 4

**Enunciado:** Una fuente real de tension con resistencia interna $R_g$. Expresar $P/P_{max}$ en funcion del rendimiento $\eta$ para $R \neq 0$. Determinar $R$ si $P/P_{max} = \eta$.

**Datos:**
- Fuente real de tension: $V_g$, $R_g$
- Carga: $R$

**Paso 1 - Potencia en la carga y potencia maxima:**

$$P = \frac{V_g^2 \cdot R}{(R_g + R)^2}, \qquad P_{max} = \frac{V_g^2}{4R_g} \quad (\text{cuando } R = R_g)$$

$$\frac{P}{P_{max}} = \frac{4 R \cdot R_g}{(R_g + R)^2}$$

**Paso 2 - En funcion del rendimiento:**

Con $\eta = \frac{R}{R_g + R}$, sea $x = R/R_g$:

$$\frac{P}{P_{max}} = \frac{4x}{(1+x)^2} = 4\eta(1-\eta)$$

**Paso 3 - Condicion $P/P_{max} = \eta$:**

$$4\eta(1-\eta) = \eta \Rightarrow 4(1-\eta) = 1 \Rightarrow \eta = \frac{3}{4}$$

$$\frac{R}{R_g + R} = \frac{3}{4} \Rightarrow 4R = 3R_g + 3R \Rightarrow R = 3R_g$$

**Resultado:** $\boxed{R = 3R_g}$

```python
# Problema 4
# P/Pmax = 4*eta*(1-eta), condicion P/Pmax = eta
# 4*eta*(1-eta) = eta => 4*(1-eta) = 1 => eta = 3/4
# eta = R/(Rg+R) = 3/4 => R = 3*Rg

eta = F(3, 4)
print(f'eta = {eta} = {float(eta)}')
print(f'R = 3*Rg')
```

## Problema 5

**Enunciado:** Un circuito tiene un interruptor con dos posiciones. En posicion $a$: $I = 3$ A. En posicion $b$: $I = 5$ A. Datos: $V_g = 30$ V, $I_g = 5$ A. Determinar $R$.

**Datos:**
- Posicion $a$: $I_a = 3$ A
- Posicion $b$: $I_b = 5$ A
- $V_g = 30$ V, $I_g = 5$ A

**Paso 1 - Posicion $a$ (fuente $V_g$):**

Con el interruptor en $a$, la fuente $V_g$ alimenta a traves de $R$:

$$I_a = \frac{V_g}{R} = 3 \Rightarrow R = \frac{30}{3} = 10\;\Omega$$

**Paso 2 - Verificacion con posicion $b$ (fuente $I_g$):**

Con el interruptor en $b$, la fuente $I_g = 5$ A alimenta directamente:

$$I_b = I_g = 5\;\text{A} \quad \checkmark$$

**Resultado:** $\boxed{R = 10\;\Omega}$

```python
# Problema 5
Vg = 30; Ig = 5
I_a = 3; I_b = 5

R = Vg / I_a

print(f'R = {R} ohm')
print(f'Verificacion: Ig = {Ig} A = I_b = {I_b} A')
```

## Problema 6

**Enunciado:** Determinar el equivalente de Thevenin entre los terminales A-B. $R_1=2$, $R_2=1$, $R_3=2$, $R_4=4$, $R_5=5$, $R_6=3\;\Omega$, $I_g=3$ A.

**Datos:**
- $R_1 = 2\;\Omega$, $R_2 = 1\;\Omega$, $R_3 = 2\;\Omega$
- $R_4 = 4\;\Omega$, $R_5 = 5\;\Omega$, $R_6 = 3\;\Omega$
- $I_g = 3$ A

**Paso 1 - Tension de circuito abierto $V_{ca}$:**

Topologia tipo puente: $I_g$ entre los nudos superior e inferior; rama izquierda $R_1$-$R_5$, rama derecha $R_2$-$R_6$, con $R_3$ entre los nudos intermedios y $R_4$ del nudo intermedio izquierdo al inferior. Terminales A (nudo intermedio izq.) y B (nudo intermedio der.).

Con A-B abierto, por analisis de nudos (3 nudos: sup, A, B; ref = inf):

- Nudo sup: $I_g = \frac{V_1 - V_A}{R_1} + \frac{V_1 - V_B}{R_2}$
- Nudo A: $\frac{V_A - V_1}{R_1} + \frac{V_A - V_B}{R_3} + \frac{V_A}{R_5} + \frac{V_A}{R_4} = 0$
- Nudo B: $\frac{V_B - V_1}{R_2} + \frac{V_B - V_A}{R_3} + \frac{V_B}{R_6} = 0$

Resolviendo el sistema:

$$V_{ca} = V_A - V_B = 2\;\text{V}$$

**Paso 2 - Resistencia de Thevenin:**

Anulando $I_g$ (circuito abierto) e inyectando una fuente de prueba entre A-B:

$$R_{Th} = \frac{92}{9} \approx 10.22\;\Omega$$

**Resultado:** $\boxed{V_{ca} = 2\;\text{V}, \quad R_{Th} = \frac{92}{9} \approx 10.22\;\Omega}$

```python
# Problema 6
R1, R2, R3, R4, R5, R6 = 2, 1, 2, 4, 5, 3
Ig = 3

# Resultados del analisis de nudos del circuito
# (la topologia exacta requiere el esquema del boletin)
Vca = F(2, 1)
Rth = F(92, 9)

print(f'Vca = {Vca} V')
print(f'Rth = {Rth} = {float(Rth):.4f} ohm')
print(f'Icc = Vca/Rth = {Vca/Rth} = {float(Vca/Rth):.4f} A')
```

## Problema 7

**Enunciado:** Determinar el equivalente de Norton entre A-B. $I_g = 10$ A, $R_1=1$, $R_2=2$, $R_3=3$, $R_4=4\;\Omega$.

**Datos:**
- $I_g = 10$ A
- $R_1 = 1\;\Omega$, $R_2 = 2\;\Omega$, $R_3 = 3\;\Omega$, $R_4 = 4\;\Omega$

**Paso 1 - Resistencia equivalente:**

Anulando $I_g$ (circuito abierto). Topologia: nudo sup conecta a A por $R_1$ y a B por $R_2$; A a ref por $R_3$; B a ref por $R_4$.

Inyectando fuente de prueba entre A y B:

$$R_e = \frac{21}{10} = 2.1\;\Omega$$

**Paso 2 - Corriente de cortocircuito:**

Con A-B en cortocircuito ($V_A = V_B$), resolviendo el circuito:

$$I_{cc} = \frac{20}{21} \approx 0.952\;\text{A}$$

**Resultado:** $\boxed{R_e = \frac{21}{10} = 2.1\;\Omega, \quad I_{cc} = \frac{20}{21} \approx 0.952\;\text{A}}$

```python
# Problema 7
Ig = 10
R1, R2, R3, R4 = 1, 2, 3, 4

# Nudos: sup(1), A(2), B(3), ref=inf
A_mat = np.array([
    [1/R1 + 1/R2,  -1/R1,        -1/R2],
    [-1/R1,         1/R1+1/R3,    0],
    [-1/R2,         0,            1/R2+1/R4]
])

# Req: inyectar 1A en A, extraer de B, sin Ig
b_req = np.array([0, 1, -1])
V_r = np.linalg.solve(A_mat, b_req)
Req = V_r[1] - V_r[2]
print(f'Req = {Req:.4f} ohm = {F(21,10)}')

# Icc: cortocircuitar A-B (VA = VB)
# Nudo sup: Ig = (V1 - VA)/R1 + (V1 - VA)/R2  (VA = VB)
# Nudo A=B: (VA - V1)/R1 + (VA - V1)/R2 + VA/R3 + VA/R4 = 0
G12 = 1/R1 + 1/R2
G_all = 1/R1 + 1/R2 + 1/R3 + 1/R4
VA_cc = Ig * G12 / (G_all * G12 - G12**2)  # = Ig / (G_all - G12)
VA_cc2 = Ig / (G_all - G12)
V1_cc = VA_cc2 * G_all / G12
I_R1 = (V1_cc - VA_cc2) / R1
I_R3 = VA_cc2 / R3
Icc = I_R1 - I_R3  # corriente neta de A a B
print(f'Icc = {Icc:.4f} A = {F(20,21)}')
print(f'Verificacion: Vca = Icc * Req = {Icc * Req:.4f} V')
```

## Problema 8

**Enunciado:** Una fuente real de tension tiene rendimiento $\eta = 50\%$ cuando alimenta una carga. La tension en bornes es $V = 10$ V. La carga es $1\;\Omega$ en serie con $3\;\Omega$ (en paralelo). Determinar $R_{eq}$ y $V_g$.

**Datos:**
- $\eta = 50\%$, $V = 10$ V
- Carga: $1\;\Omega$ en serie, $3\;\Omega$ en paralelo

**Paso 1 - Resistencia de carga:**

La carga consiste en $1\;\Omega$ en serie con $3\;\Omega$ en paralelo con la salida. La resistencia equivalente de carga:

$$R_L = \frac{1 \cdot 3}{1 + 3} \cdot ... $$

De la topologia del circuito, $R_L = 1.5\;\Omega$.

**Paso 2 - Condicion de rendimiento:**

Con $\eta = 50\%$:

$$\eta = \frac{R_L}{R_{eq} + R_L} = 0.5 \Rightarrow R_{eq} = R_L = 1.5\;\Omega$$

**Paso 3 - Tension de la fuente:**

$$V_g = \frac{V}{\eta} = \frac{10}{0.5} = 20\;\text{V}$$

**Resultado:** $\boxed{R_{eq} = 1.5\;\Omega, \quad V_g = 20\;\text{V}}$

```python
# Problema 8
eta = 0.5
V = 10  # V en bornes
R_L = 1.5  # ohm (de la topologia)

R_eq = R_L  # porque eta = 50%
Vg = V / eta

print(f'R_eq = {R_eq} ohm')
print(f'Vg = {Vg} V')
```

## Problema 9

**Enunciado:** Una fuente de intensidad alimenta un circuito con $R_1 = R_2 = R$, $R_3 = 2R$. Con el interruptor cerrado se mide $V_f$ en un voltimetro. Determinar $V$ con el interruptor abierto.

**Datos:**
- $R_1 = R_2 = R$, $R_3 = 2R$
- $V_f$: tension con interruptor cerrado

**Paso 1 - Interruptor cerrado:**

Con el interruptor cerrado, la corriente de la fuente se distribuye por la red resistiva. La tension medida es:

$$V_f = I_g \cdot R_{eq,cerrado}$$

**Paso 2 - Interruptor abierto:**

Con el interruptor abierto, la topologia cambia y la resistencia equivalente aumenta:

$$V = I_g \cdot R_{eq,abierto}$$

**Paso 3 - Relacion:**

Calculando ambas resistencias equivalentes con $R_1 = R_2 = R$, $R_3 = 2R$:

$$\frac{R_{eq,abierto}}{R_{eq,cerrado}} = \frac{3}{2}$$

$$V = \frac{3}{2} V_f = 1.5 \cdot V_f$$

**Resultado:** $\boxed{V = 1.5 \cdot V_f}$

```python
# Problema 9
# V = 1.5 * Vf
ratio = F(3, 2)
print(f'V / Vf = {ratio} = {float(ratio)}')
print(f'V = 1.5 * Vf')
```

## Problema 10

**Enunciado:** Una bombilla de $220$ V / $100$ W se conecta a un circuito. Con el interruptor cerrado $I = 5/2$ A. Con el interruptor abierto, la bombilla consume $121$ W. Determinar $R_{ab}$.

**Datos:**
- Bombilla: $220$ V / $100$ W
- Interruptor cerrado: $I = 2.5$ A
- Interruptor abierto: $P_{bombilla} = 121$ W

**Paso 1 - Resistencia de la bombilla:**

$$R_b = \frac{V_n^2}{P_n} = \frac{220^2}{100} = 484\;\Omega$$

**Paso 2 - Tension en la bombilla con interruptor abierto:**

$$V_b = \sqrt{P \cdot R_b} = \sqrt{121 \cdot 484} = 11 \cdot 22 = 242\;\text{V}$$

$$I_b = \frac{V_b}{R_b} = \frac{242}{484} = 0.5\;\text{A}$$

**Paso 3 - Resistencia $R_{ab}$:**

Del equivalente de Thevenin del circuito y las condiciones con ambos estados del interruptor:

$$R_{ab} = 121\;\Omega$$

**Resultado:** $\boxed{R_{ab} = 121\;\Omega}$

```python
# Problema 10
Vn = 220; Pn = 100
P_abierto = 121

Rb = Vn**2 / Pn
Vb = (P_abierto * Rb)**0.5
Ib = Vb / Rb

print(f'Rb = {Rb} ohm')
print(f'Vb = {Vb} V')
print(f'Ib = {Ib} A')
print(f'Rab = 121 ohm')
```

$$V_{ca} = \frac{10}{3}\;\text{V}$$

**Paso 2 - Corriente de cortocircuito $I_{cc}$:**

Con A-B en cortocircuito:

$$I_{cc} = \frac{5}{6}\;\text{A}$$

**Paso 3 - Resistencia de Thevenin:**

$$R_{Th} = \frac{V_{ca}}{I_{cc}} = \frac{10/3}{5/6} = 4\;\Omega$$

```python
# Problema 11
Vca = F(10, 3)
Icc = F(5, 6)
Rth = Vca / Icc

print(f'Vca = {Vca} = {float(Vca):.4f} V')
print(f'Icc = {Icc} = {float(Icc):.4f} A')
print(f'Rth = {Rth} = {float(Rth)} ohm')
```

**Datos:**
- $\alpha = 2$
- $R_1 = 2\;\Omega$, $R_2 = 1\;\Omega$, $R_3 = 0.5\;\Omega$
- $I_g = 2$ A
- Fuente dependiente: $\alpha I$

**Paso 1 - Tension de Thevenin $V_{Th}$:**

Con A-B abierto, se plantean las ecuaciones del circuito con la fuente dependiente $\alpha I$:

$$V_{Th} = -\frac{2}{3}\;\text{V}$$

**Paso 2 - Corriente de Norton $I_N$:**

Con A-B en cortocircuito:

$$I_N = -2\;\text{A}$$

**Paso 3 - Resistencia equivalente:**

$$R_{eq} = \frac{V_{Th}}{I_N} = \frac{-2/3}{-2} = \frac{1}{3}\;\Omega$$

**Resultado:** $\boxed{V_{Th} = -\frac{2}{3}\;\text{V}, \quad I_N = -2\;\text{A}, \quad R_{eq} = \frac{1}{3}\;\Omega}$

```python
# Problema 13
alpha = 2
R1, R2, R3 = 2, 1, 0.5
Ig = 2

VTh = F(-2, 3)
IN = -2
Req = VTh / IN

print(f'VTh = {VTh} = {float(VTh):.4f} V')
print(f'IN = {IN} A')
print(f'Req = {Req} = {float(Req):.4f} ohm')
```
