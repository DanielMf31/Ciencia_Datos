## Formulario del Tema 7: Corriente Alterna Sinusoidal

| Concepto | Formula |
|----------|---------|
| Fasor | $v(t) = V_m \cos(\omega t + \varphi) \leftrightarrow \mathbf{V} = V_m \angle \varphi$ |
| Derivada fasorial | $\frac{d}{dt} \to j\omega$ |
| Impedancia resistencia | $\boxed{Z_R = R}$ |
| Impedancia bobina | $\boxed{Z_L = j\omega L}$ |
| Impedancia condensador | $\boxed{Z_C = \frac{1}{j\omega C} = \frac{-j}{\omega C}}$ |
| Ley de Ohm fasorial | $\mathbf{V} = Z \cdot \mathbf{I}$ |
| Valor eficaz (RMS) | $\boxed{V_{rms} = \frac{V_m}{\sqrt{2}}}$ |
| Divisor tension | $\mathbf{V}_2 = \mathbf{V}_{in} \cdot \frac{Z_2}{Z_1 + Z_2}$ |
| Superposicion AC | Resolver cada frecuencia por separado, sumar en dominio temporal |
| Resonancia serie | $\boxed{\omega_0 = \frac{1}{\sqrt{LC}}}$, $Z = R$ (minima) |
| Factor de calidad | $Q = \frac{\omega_0 L}{R} = \frac{1}{\omega_0 C R}$ |

## Problema 1

**Enunciado:** $R=100\ \Omega$ en paralelo con $L=50$ mH. En serie con $R_s=10\ \Omega$ y $C=10\ \mu$F.

a) $Z$ a $\omega=1000$ rad/s. b) $Z$ a $\omega=4000$ rad/s. c) Frecuencia puramente resistiva. d) $Z$ a esa frecuencia.

**Paso 1:** $Z_{RL} = R \cdot j\omega L/(R+j\omega L)$. $Z = R_s + Z_{RL} + 1/(j\omega C)$.

**a)** $\omega=1000$: $Z_{RL} = 100(j50)/(100+j50) = 20+j40$. $Z = 10+20+j40-j100 = 30-j60$.

**b)** $\omega=4000$: $Z_{RL} = 100(j200)/(100+j200) = 80+j40$. $Z = 10+80+j40-j25 = 90+j15$.

**c)** Im$(Z)=0$: resolviendo, $\omega = 2000$ rad/s.

**d)** $Z(2000) = 60\ \Omega$.

**Resultado:** $\boxed{a)\ 30-j60;\ b)\ 90+j15;\ c)\ 2000\ \text{rad/s};\ d)\ 60\ \Omega}$

```python
import numpy as np
R,L,Rs,C = 100, 50e-3, 10, 10e-6
for w in [1000, 4000, 2000]:
    ZRL = R*1j*w*L/(R+1j*w*L)
    Z = Rs + ZRL + 1/(1j*w*C)
    print(f'w={w}: Z = {Z.real:.1f} + {Z.imag:.1f}j')
```

## Problema 2

**Enunciado:** CA permanente. $V_1 = 100$ V. Encontrar $V_2$. Impedancias: $R$, $Rj$, $R$.

**Paso 1:** La rama donde se mide $V_1$: impedancia $R + jR$, modulo $R\sqrt{2}$.

$I = V_1/(R\sqrt{2}) = 100/(R\sqrt{2})$.

**Paso 2:** $V_2$ se mide en el circuito exterior. De la topologia:
$$V_2 = 50\sqrt{10} \approx 158{,}1 \text{ V}$$

**Resultado:** $\boxed{V_2 = 50\sqrt{10} \text{ V}}$

```python
import numpy as np
print(f'V2 = 50*sqrt(10) = {50*np.sqrt(10):.1f} V')
```

## Problema 3

**Enunciado:** $i_g(t)=3\sqrt{2}\cos(1000t)$ A, $C=10\ \mu$F, $L=50$ mH. Valores eficaces en la bobina.

**Datos:** $I_g=3$ A eficaz, $\omega=1000$.

**Paso 1:** $Z_C=-j100$, $Z_L=j50$. Paralelo:
$$Z_p = \frac{Z_C Z_L}{Z_C+Z_L} = \frac{5000}{-j50} = j100\ \Omega$$

**Paso 2:** $V = I_g Z_p = 3(j100) = j300$. $|V|=300$ V.

$I_L = V/Z_L = j300/j50 = 6$ A.

**Resultado:** $\boxed{U_L = 300 \text{ V};\ I_L = 6 \text{ A}}$

```python
import numpy as np
w=1000; ZC=1/(1j*w*10e-6); ZL=1j*w*50e-3
Zp=ZC*ZL/(ZC+ZL); V=3*Zp; IL=V/ZL
print(f'U_L = {abs(V):.0f} V, I_L = {abs(IL):.0f} A')
```

```python
import numpy as np
print(f'Amplitudes: {9.806*np.sqrt(2):.2f} A y {7.96*np.sqrt(2):.2f} A')
```

## Problema 5

**Enunciado:** $u_g(t)=8$ V, $i_g(t)=4\sqrt{2}\cos(100t)$ A. $L=20$ mH, $R=4\ \Omega$. Determinar $i(t)$.

**Paso 1:** CC: $i_{CC} = u_g/R = 2$ A.

**Paso 2:** AC ($\omega=100$): $Z_L=j2$. Divisor de corriente.

$I_L = I_g - I_g \cdot Z_L/(R+Z_L) = I_g \cdot R/(R+Z_L) = 4 \cdot 4/(4+j2) = 16/(4+j2)$.

$= 16(4-j2)/20 = 3{,}2-j1{,}6$. $|I_L|=\sqrt{12{,}8}=3{,}578$ A, fase $= -0{,}464$ rad.

Pero $i(t)$ es la corriente por $L$ (que baja en la figura):
$$i(t) = 2 + 3{,}578\sqrt{2}\cos(100t - 0{,}464)$$

**Resultado:** $\boxed{i(t) = 2 + 3{,}578\sqrt{2}\cos(100t-0{,}464)}$

```python
import numpy as np
Ig=4; R=4; ZL=1j*100*20e-3
IL = Ig*R/(R+ZL)
print(f'|IL| = {abs(IL):.3f} A, fase = {np.angle(IL):.3f} rad')
print(f'i(t) = 2 + {abs(IL):.3f}*sqrt(2)*cos(100t - {abs(np.angle(IL)):.3f})')
```

## Problema 6

**Enunciado:** $v_g(t)=4+4\cos(10t)$, $i_g(t)=4\cos(100t)$. $R_1=R_2=2\ \Omega$, $L=2$ H. Encontrar $i(t)$ por el amperimetro.

**Paso 1:** Tres componentes: CC, $\omega=10$, $\omega=100$.

**CC:** $L\to$ cable. $i_{CC} = v_{g,CC}/R_1 = 4/2 = 2$ A.

**$\omega=10$:** $v_g=4\angle 0$, $i_g=0$. $Z_L=j20$.
De la solucion: $i_{AC1} = 0{,}199\cos(10t-1{,}47)$.

**$\omega=100$:** $v_g=0$, $i_g=4\angle 0$. $Z_L=j200$.
Por la topologia, $i_g$ pasa directamente por el amperimetro con signo negativo: $i_{AC2} = -4\cos(100t)$.

**Resultado:** $\boxed{i(t) = 2 + 0{,}199\cos(10t-1{,}47) - 4\cos(100t)}$

```python
import numpy as np
w=10; R1=R2=2; L=2; ZL=1j*w*L
Zp=R2*ZL/(R2+ZL); I=4/(R1+Zp)
print(f'AC1: |I|={abs(I):.4f}, fase={np.angle(I):.4f} rad')
print(f'i(t) = 2 + {abs(I):.4f}*cos(10t + ({np.angle(I):.2f})) - 4*cos(100t)')
```

## Problema 9

**Enunciado:** Thevenin A-B. $E_g=10$ V, $Z_1=2j$, $Z_2=5j$, $Z_c=-10j$.

**Paso 1:** $Z_{2\|c} = Z_2 Z_c/(Z_2+Z_c) = j5(-j10)/(j5-j10) = 50/(-j5) = j10$.

$E_{Th} = E_g \cdot Z_{2\|c}/(Z_1+Z_{2\|c}) = 10 \cdot j10/(j2+j10) = 100/12 = 50/6\angle 0$.

**Paso 2:** $Z_{eq}$: desde AB con $E_g$ cortocircuitada.
$Z_{eq} = 5j/3\ \Omega$.

**Resultado:** $\boxed{E_{Th} = 50/6\angle 0;\ Z_{eq} = 5j/3\ \Omega}$

```python
import numpy as np
Z1=2j; Z2=5j; Zc=-10j; Eg=10
Z2c=Z2*Zc/(Z2+Zc)
Vth=Eg*Z2c/(Z1+Z2c)
print(f'E_Th = {Vth:.4f} = {abs(Vth):.4f} angle {np.degrees(np.angle(Vth)):.1f} deg')
Zeq = Z1*(Z2+Zc)/(Z1+Z2+Zc)
print(f'Z_eq = {Zeq:.4f}')
```

## Problema 14

**Enunciado:** Interruptores de 1 a 2 en $t=0$. $e(t)=70\cos(40t)$ V. Calcular $i_L(t)$.

Datos: 25 $\Omega$, 0.5 H, 5 mF, fuente 2 A.

**Paso 1:** Circuito de 2do orden. Raices: $s_1=-10$, $s_2=-40$.

**Paso 2:** Forzada: $2{,}4\cos(40t-30{,}96^\circ)$.

**Paso 3:** Transitoria: $-2{,}39e^{-10t} - 1{,}67e^{-40t}$.

$$i_L(t) = -2{,}39e^{-10t} - 1{,}67e^{-40t} + 2{,}4\cos(40t-30{,}96^\circ)$$

**Resultado:** $\boxed{i_L(t) = -2{,}39e^{-10t} - 1{,}67e^{-40t} + 2{,}4\cos(40t-30{,}96^\circ) \text{ A}}$

```python
import numpy as np, matplotlib.pyplot as plt
t=np.linspace(0,0.5,1000)
iL=-2.39*np.exp(-10*t)-1.67*np.exp(-40*t)+2.4*np.cos(40*t-np.radians(30.96))
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(t,iL,'#2171b5',lw=2.5,label=r'$i_L(t)$')
ax.set_xlabel('t (s)'); ax.set_ylabel('$i_L$ (A)')
ax.set_title('Problema 14: Transitorio 2do orden + AC')
ax.legend(); ax.grid(True,alpha=0.3); plt.tight_layout(); plt.show()
```
