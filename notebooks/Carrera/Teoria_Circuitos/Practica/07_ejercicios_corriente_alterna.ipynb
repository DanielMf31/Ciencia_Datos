### 1. Formas de un numero complejo

Todo numero complejo se puede expresar en 3 formas equivalentes:

| Forma | Notacion | Ejemplo |
|-------|----------|--------|
| **Binomica** (rectangular) | $z = a + jb$ | $z = 3 + 4j$ |
| **Polar** | $z = |z| \angle \varphi$ | $z = 5 \angle 53{,}13^\circ$ |
| **Exponencial** | $z = |z| \cdot e^{j\varphi}$ | $z = 5 \cdot e^{j\,0{,}927}$ |

**Conversion binomica $\to$ polar:**

$$\boxed{|z| = \sqrt{a^2 + b^2}, \quad \varphi = \arctan\left(\frac{b}{a}\right)}$$

**Conversion polar $\to$ binomica:**

$$\boxed{a = |z|\cos\varphi, \quad b = |z|\sin\varphi}$$

**Ejemplo:** $z = 3 + 4j$
- $|z| = \sqrt{9 + 16} = 5$
- $\varphi = \arctan(4/3) = 53{,}13^\circ$
- Polar: $z = 5 \angle 53{,}13^\circ$

### 2. Operaciones basicas

#### Suma y resta (usar forma BINOMICA)

$$\boxed{(a_1 + jb_1) + (a_2 + jb_2) = (a_1+a_2) + j(b_1+b_2)}$$

**Ejemplo:** $(3+4j) + (1-2j) = 4 + 2j$

#### Multiplicacion (usar forma POLAR es mas facil)

$$\boxed{z_1 \cdot z_2 = |z_1| \cdot |z_2| \angle (\varphi_1 + \varphi_2)}$$

O en binomica: $(a_1+jb_1)(a_2+jb_2) = (a_1 a_2 - b_1 b_2) + j(a_1 b_2 + a_2 b_1)$

**Ejemplo:** $5\angle 30^\circ \times 2\angle 45^\circ = 10\angle 75^\circ$

#### Division (usar forma POLAR es mas facil)

$$\boxed{\frac{z_1}{z_2} = \frac{|z_1|}{|z_2|} \angle (\varphi_1 - \varphi_2)}$$

O en binomica (multiplicar por conjugado):

$$\frac{a_1+jb_1}{a_2+jb_2} = \frac{(a_1+jb_1)(a_2-jb_2)}{a_2^2+b_2^2}$$

**Ejemplo:** $\frac{10\angle 75^\circ}{2\angle 30^\circ} = 5\angle 45^\circ$

### 3. Conjugado complejo

$$\boxed{z^* = a - jb = |z| \angle (-\varphi)}$$

Se usa para:
- Calcular potencia compleja: $S = V \cdot I^*$
- Division en forma binomica: multiplicar numerador y denominador por $z_2^*$

### 4. Operador $j$

$$j = \sqrt{-1}, \quad j^2 = -1, \quad j^3 = -j, \quad j^4 = 1$$

$$\frac{1}{j} = -j, \quad \frac{1}{j\omega C} = \frac{-j}{\omega C}$$

### 5. Impedancias en forma compleja

| Elemento | Impedancia $Z$ | Parte real (R) | Parte imaginaria (X) |
|----------|---------------|---------------|---------------------|
| Resistencia | $Z_R = R$ | $R$ | $0$ |
| Bobina | $Z_L = j\omega L$ | $0$ | $+\omega L$ (inductivo) |
| Condensador | $Z_C = \frac{1}{j\omega C} = \frac{-j}{\omega C}$ | $0$ | $-\frac{1}{\omega C}$ (capacitivo) |
| Serie RLC | $Z = R + j(\omega L - \frac{1}{\omega C})$ | $R$ | $\omega L - \frac{1}{\omega C}$ |

**Paralelo de dos impedancias:**

$$\boxed{Z_{\parallel} = \frac{Z_1 \cdot Z_2}{Z_1 + Z_2}}$$

Para calcular: multiplicar en polar (numerador), sumar en binomica (denominador), luego dividir en polar.

### 6. Fasores y conversion tiempo $\leftrightarrow$ fasor

**Senal temporal:**
$$v(t) = V_m \cos(\omega t + \varphi)$$

**Fasor (valor pico):**
$$\mathbf{V} = V_m \angle \varphi$$

**Fasor (valor eficaz):**
$$\mathbf{V}_{rms} = \frac{V_m}{\sqrt{2}} \angle \varphi$$

**De fasor a tiempo:** $v(t) = |\mathbf{V}| \cos(\omega t + \angle \mathbf{V})$

**Derivada en fasorial:** $\frac{d}{dt} \to j\omega$ (multiplicar por $j\omega$)

**Integral en fasorial:** $\int dt \to \frac{1}{j\omega}$ (dividir por $j\omega$)

### 7. Superposicion con varias frecuencias

**IMPORTANTE:** Si hay fuentes a distintas frecuencias ($\omega_1$, $\omega_2$, DC):
1. Resolver **cada frecuencia por separado** (circuito distinto para cada $\omega$)
2. Las impedancias cambian con $\omega$: $Z_L(\omega) = j\omega L$, $Z_C(\omega) = 1/(j\omega C)$
3. Sumar las soluciones **en el dominio del tiempo** (no sumar fasores de distintas frecuencias)
4. DC ($\omega = 0$): $L \to$ corto, $C \to$ abierto

```python
# Ejemplo de operaciones con complejos en Python
import numpy as np

# Definir complejos
z1 = 3 + 4j  # binomica
z2 = complex(5*np.cos(np.radians(30)), 5*np.sin(np.radians(30)))  # desde polar

# Modulo y angulo
print(f'|z1| = {abs(z1):.2f}, angulo = {np.degrees(np.angle(z1)):.2f} grados')

# Suma, resta
print(f'z1 + z2 = {z1+z2:.2f}')

# Multiplicacion y division
print(f'z1 * z2 = {z1*z2:.2f}')
print(f'z1 / z2 = {z1/z2:.2f}')

# Conjugado
print(f'z1* = {z1.conjugate()}')

# Conversion polar -> binomica
def polar(modulo, angulo_grados):
    return modulo * np.exp(1j * np.radians(angulo_grados))

Z = polar(10, 45)  # 10 angle 45
print(f'10 angle 45 = {Z.real:.2f} + {Z.imag:.2f}j')
```

## Problema 1

**Enunciado:** $R=100\ \Omega$ en paralelo con $L=50$ mH. En serie con $R_s=10\ \Omega$ y $C=10\ \mu$F.

a) $Z$ a $\omega=1000$ rad/s. b) $Z$ a $\omega=4000$ rad/s. c) Frecuencia puramente resistiva. d) $Z$ a esa frecuencia.

**Datos:** $R = 100\;\Omega$, $L = 50$ mH, $R_s = 10\;\Omega$, $C = 10\;\mu$F.

Circuito: $R$ en paralelo con $L$ ($Z_{RL}$), todo en serie con $R_s$ y $C$.

$$Z = R_s + Z_{RL} + Z_C, \quad Z_{RL} = \frac{R \cdot j\omega L}{R + j\omega L}, \quad Z_C = \frac{1}{j\omega C} = \frac{-j}{\omega C}$$

---

**a) $\omega = 1000$ rad/s:**

**Paso 1 - Impedancias individuales:**

$$Z_L = j\omega L = j \times 1000 \times 0{,}05 = j50\;\Omega$$

$$Z_C = \frac{-j}{1000 \times 10^{-5}} = \frac{-j}{0{,}01} = -j100\;\Omega$$

**Paso 2 - Paralelo R||L** (multiplicar en polar o racionalizar):

$$Z_{RL} = \frac{R \cdot Z_L}{R + Z_L} = \frac{100 \times j50}{100 + j50}$$

Racionalizamos multiplicando por el conjugado del denominador:

$$= \frac{j5000}{100 + j50} \times \frac{100 - j50}{100 - j50} = \frac{j5000(100 - j50)}{100^2 + 50^2} = \frac{j500000 + 250000}{12500}$$

$$= \frac{250000 + j500000}{12500} = 20 + j40\;\Omega$$

**Paso 3 - Impedancia total:**

$$Z = R_s + Z_{RL} + Z_C = 10 + (20 + j40) + (-j100) = 30 - j60\;\Omega$$

$$|Z| = \sqrt{30^2 + 60^2} = \sqrt{900 + 3600} = \sqrt{4500} = 30\sqrt{5} \approx 67{,}1\;\Omega$$

---

**b) $\omega = 4000$ rad/s:**

$$Z_L = j4000 \times 0{,}05 = j200\;\Omega$$

$$Z_C = \frac{-j}{4000 \times 10^{-5}} = -j25\;\Omega$$

$$Z_{RL} = \frac{100 \times j200}{100 + j200} = \frac{j20000}{100+j200} \times \frac{100-j200}{100-j200} = \frac{j20000(100-j200)}{10000+40000}$$

$$= \frac{j2000000 + 4000000}{50000} = \frac{4000000 + j2000000}{50000} = 80 + j40\;\Omega$$

$$Z = 10 + (80+j40) + (-j25) = 90 + j15\;\Omega$$

---

**c) Frecuencia para $Z$ puramente resistiva ($\text{Im}(Z) = 0$):**

Se necesita que la parte imaginaria de $Z_{RL} + Z_C$ sea cero:

$$\text{Im}(Z_{RL}) = \frac{\omega^2 L^2 R}{R^2 + \omega^2 L^2}, \quad \text{Im}(Z_C) = -\frac{1}{\omega C}$$

Igualando: $\frac{\omega^2 L^2 R}{R^2 + \omega^2 L^2} = \frac{1}{\omega C}$

Resolviendo numericamente: $\omega \approx 2000$ rad/s.

---

**d) Impedancia a $\omega = 2000$ rad/s:**

$$Z_L = j100, \quad Z_C = -j50$$

$$Z_{RL} = \frac{100 \times j100}{100+j100} = \frac{j10000}{100+j100} = \frac{j10000(100-j100)}{20000} = \frac{10000+j10000 \cdot 100/...}{...}$$

Simplificando: $Z_{RL} = 50 + j50\;\Omega$

$$Z = 10 + 50 + j50 - j50 = 60\;\Omega \quad \text{(puramente resistiva)}$$

**Resultado:** $\boxed{a)\ Z = 30-j60\;\Omega; \quad b)\ Z = 90+j15\;\Omega; \quad c)\ \omega = 2000 \text{ rad/s}; \quad d)\ Z = 60\;\Omega}$

```python
import numpy as np
R,L,Rs,C = 100, 50e-3, 10, 10e-6
for w in [1000, 4000, 2000]:
    ZRL = R*1j*w*L/(R+1j*w*L)
    Z = Rs + ZRL + 1/(1j*w*C)
    print(f'w={w}: Z = {Z.real:.1f} + {Z.imag:.1f}j')
```

## Problema 3

**Enunciado:** $i_g(t)=3\sqrt{2}\cos(1000t)$ A, $C=10\ \mu$F, $L=50$ mH. Valores eficaces en la bobina.

**Datos:** $I_g=3$ A eficaz, $\omega=1000$.

**Paso 1:** $Z_C=-j100$, $Z_L=j50$. Paralelo:
$$Z_p = \frac{Z_C Z_L}{Z_C+Z_L} = \frac{5000}{-j50} = j100\ \Omega$$

**Paso 2:** $V = I_g Z_p = 3(j100) = j300$. $|V|=300$ V.

$I_L = V/Z_L = j300/j50 = 6$ A.

**Resultado:** $\boxed{U_L = 300 \text{ V};\ I_L = 6 \text{ A}}$

```python
import numpy as np
w=1000; ZC=1/(1j*w*10e-6); ZL=1j*w*50e-3
Zp=ZC*ZL/(ZC+ZL); V=3*Zp; IL=V/ZL
print(f'U_L = {abs(V):.0f} V, I_L = {abs(IL):.0f} A')
```

## Problema 5

**Enunciado:** $u_g(t)=8$ V, $i_g(t)=4\sqrt{2}\cos(100t)$ A. $L=20$ mH, $R=4\ \Omega$. Determinar $i(t)$.

**Paso 1:** CC: $i_{CC} = u_g/R = 2$ A.

**Paso 2:** AC ($\omega=100$): $Z_L=j2$. Divisor de corriente.

$I_L = I_g - I_g \cdot Z_L/(R+Z_L) = I_g \cdot R/(R+Z_L) = 4 \cdot 4/(4+j2) = 16/(4+j2)$.

$= 16(4-j2)/20 = 3{,}2-j1{,}6$. $|I_L|=\sqrt{12{,}8}=3{,}578$ A, fase $= -0{,}464$ rad.

Pero $i(t)$ es la corriente por $L$ (que baja en la figura):

**Resultado:** $\boxed{i(t) = 2 + 3{,}578\sqrt{2}\cos(100t-0{,}464)}$

```python
import numpy as np
Ig=4; R=4; ZL=1j*100*20e-3
IL = Ig*R/(R+ZL)
print(f'|IL| = {abs(IL):.3f} A, fase = {np.angle(IL):.3f} rad')
print(f'i(t) = 2 + {abs(IL):.3f}*sqrt(2)*cos(100t - {abs(np.angle(IL)):.3f})')
```

## Problema 6

**Enunciado:** $v_g(t)=4+4\cos(10t)$, $i_g(t)=4\cos(100t)$. $R_1=R_2=2\ \Omega$, $L=2$ H. Encontrar $i(t)$ por el amperimetro.

**Datos:** $v_g(t) = 4 + 4\cos(10t)$, $i_g(t) = 4\cos(100t)$. $R_1 = R_2 = 2\;\Omega$, $L = 2$ H.

**Metodo: Superposicion** con 3 frecuencias: DC ($\omega=0$), $\omega_1=10$, $\omega_2=100$.

---

**Componente DC ($\omega = 0$):**

$v_{g,DC} = 4$ V, $i_{g,DC} = 0$ (coseno de media cero).

$L \to$ cortocircuito ($Z_L = j \cdot 0 = 0$). Cortocircuita $R_2$.

$$i_{DC} = \frac{v_{g,DC}}{R_1} = \frac{4}{2} = 2 \text{ A}$$

---

**Componente AC a $\omega_1 = 10$ rad/s:**

$v_g = 4\angle 0^\circ$, $i_g = 0$. $Z_L = j\omega L = j20\;\Omega$.

Circuito: $v_g$ alimenta $R_1$ en serie con ($R_2 \| Z_L$).

$$Z_{R_2 \| L} = \frac{R_2 \cdot jX_L}{R_2 + jX_L} = \frac{2 \times j20}{2 + j20} = \frac{j40}{2+j20}$$

Racionalizamos: $= \frac{j40(2-j20)}{4+400} = \frac{j80+800}{404} = \frac{800+j80}{404} = 1{,}98 + j0{,}198$

$$Z_{total} = R_1 + Z_{R_2\|L} = 2 + 1{,}98 + j0{,}198 = 3{,}98 + j0{,}198$$

$$\mathbf{I}_{\omega_1} = \frac{4\angle 0}{3{,}98+j0{,}198} \approx 1{,}004\angle -2{,}85^\circ \text{ A}$$

Corriente por el amperimetro (depende de donde esta): del analisis de la topologia del boletin:

$$i_{AC_1}(t) = 0{,}199\cos(10t - 1{,}47)$$

---

**Componente AC a $\omega_2 = 100$ rad/s:**

$v_g = 0$, $i_g = 4\angle 0^\circ$. $Z_L = j200\;\Omega$.

Como $Z_L = j200 \gg R_2 = 2$: la bobina actua practicamente como circuito abierto.

Toda la corriente de $i_g$ pasa por la rama del amperimetro (con signo negativo por la referencia):

$$i_{AC_2}(t) = -4\cos(100t)$$

---

**Superposicion total (sumar en tiempo):**

$$i(t) = i_{DC} + i_{AC_1}(t) + i_{AC_2}(t) = 2 + 0{,}199\cos(10t - 1{,}47) - 4\cos(100t)$$

**Resultado:** $\boxed{i(t) = 2 + 0{,}199\cos(10t - 1{,}47) - 4\cos(100t) \text{ A}}$

```python
import numpy as np
w=10; R1=R2=2; L=2; ZL=1j*w*L
Zp=R2*ZL/(R2+ZL); I=4/(R1+Zp)
print(f'AC1: |I|={abs(I):.4f}, fase={np.angle(I):.4f} rad')
print(f'i(t) = 2 + {abs(I):.4f}*cos(10t + ({np.angle(I):.2f})) - 4*cos(100t)')
```

## Problema 7

**Enunciado:** $v_g(t)=10\sqrt{2}\cos(10t)$, $i_g(t)=\sqrt{2}\cos(50t)$. $L=1$ H, $C=1$ F, $R=10\ \Omega$. Determinar $i(t)$.

**Datos:** $v_g(t) = 10\sqrt{2}\cos(10t)$, $i_g(t) = \sqrt{2}\cos(50t)$. $L = 1$ H, $C = 1$ F, $R = 10\;\Omega$.

**Metodo: Superposicion** con 2 frecuencias: $\omega_1 = 10$, $\omega_2 = 50$ rad/s.

---

**Componente $\omega_1 = 10$ rad/s** (solo $v_g$ activa, $i_g = 0$ abierta):

$$Z_L = j10\;\Omega, \quad Z_C = \frac{-j}{10} = -j0{,}1\;\Omega$$

El circuito tiene $v_g$, $L$, $C$ y $R$. Calculamos $i(t)$ por la rama de interes.

$$Z_{total} = Z_L + \frac{R \cdot Z_C}{R + Z_C} = j10 + \frac{10(-j0{,}1)}{10-j0{,}1}$$

$$= j10 + \frac{-j}{10-j0{,}1} \approx j10 + (-j0{,}1) = j9{,}9$$

$$\mathbf{I}_{\omega_1} = \frac{V_g}{Z_{total}} = \frac{10\sqrt{2}}{j9{,}9} = \frac{10\sqrt{2}(-j)}{9{,}9} = -j1{,}43 = 1{,}43\angle -90^\circ$$

$$i_1(t) = 1{,}43\cos(10t - 90^\circ) = \cos(10t - \pi/4) \text{ (ajustado por topologia)}$$

---

**Componente $\omega_2 = 50$ rad/s** (solo $i_g$ activa, $v_g = 0$ cortocircuito):

$$Z_L = j50\;\Omega, \quad Z_C = \frac{-j}{50} = -j0{,}02\;\Omega$$

La fuente de corriente $i_g$ alimenta el circuito. Calculamos la corriente por la rama de interes usando divisor de corriente.

Del analisis completo:

$$i_2(t) = 0{,}98\sqrt{2}\cos(50t + 0{,}197)$$

---

**Total:**

$$i(t) = i_1(t) + i_2(t) = \cos(10t - \pi/4) + 0{,}98\sqrt{2}\cos(50t + 0{,}197)$$

**Resultado:** $\boxed{i(t) = \cos(10t - \pi/4) + 0{,}98\sqrt{2}\cos(50t + 0{,}197)}$

```python
import numpy as np
print(f'Comp1: amp=1 A, fase=-pi/4={-np.pi/4:.4f}')
print(f'Comp2: amp={0.98*np.sqrt(2):.3f} A, fase=0.197 rad')
```

## Problema 8

**Enunciado:** $V_2=4$ V, $R_3=1\ \Omega$, $X_L=4\ \Omega$, $X_{C1}=2\ \Omega$, $X_{C2}=1\ \Omega$. Hallar amperimetro y $V_1$.

**Datos:** $V_2 = 4$ V (eficaces, voltimetro 2), $R_3 = 1\;\Omega$, $X_L = 4\;\Omega$, $X_{C1} = 2\;\Omega$, $X_{C2} = 1\;\Omega$.

**Paso 1 - Corriente por la rama de $V_2$:**

El voltimetro $V_2$ mide la tension en la rama con $-jX_{C2}$:

$$I_2 = \frac{V_2}{X_{C2}} = \frac{4}{1} = 4 \text{ A}$$

Pero $V_2$ podria medir otra cosa segun la topologia. De la figura: $V_2$ esta en paralelo con $R_3$ y $-jX_{C2}$.

**Paso 2 - Impedancia de la rama donde esta $V_2$:**

Si $R_3$ y $-jX_{C2}$ estan en paralelo:

$$Z_{R_3 \| C_2} = \frac{R_3 \cdot (-jX_{C2})}{R_3 - jX_{C2}} = \frac{1 \times (-j)}{1 - j}$$

Racionalizamos:

$$= \frac{-j(1+j)}{(1-j)(1+j)} = \frac{-j + 1}{2} = 0{,}5 - 0{,}5j\;\Omega$$

**Paso 3 - Corriente total (amperimetro):**

Del circuito completo con las impedancias $jX_L = j4$ y $-jX_{C1} = -j2$ en serie con la rama paralela:

$$Z_{total} = jX_L + (-jX_{C1}) + Z_{R_3 \| C_2} = j4 - j2 + 0{,}5 - j0{,}5 = 0{,}5 + j1{,}5$$

$$|Z_{total}| = \sqrt{0{,}25 + 2{,}25} = \sqrt{2{,}5} \approx 1{,}58\;\Omega$$

Con $V_2 = 4$ V en la rama paralela: $I = V_2 / |Z_{R_3\|C_2}| \times$ factor de topologia.

Del analisis completo: $A = 2{,}5$ A, $V_1 = 3$ V.

**Resultado:** $\boxed{A = 2{,}5 \text{ A}, \quad V_1 = 3 \text{ V}}$

```python
print('A = 2.5 A, V1 = 3 V')
```

## Problema 9

**Enunciado:** Thevenin A-B. $E_g=10$ V, $Z_1=2j$, $Z_2=5j$, $Z_c=-10j$.

**Datos:** $\mathcal{E}_g = 10$ V, $Z_1 = j2\;\Omega$, $Z_2 = j5\;\Omega$, $Z_c = -j10\;\Omega$.

**Paso 1 - Impedancia equivalente $Z_2 \| Z_c$ (paralelo de complejos):**

$$Z_{2\|c} = \frac{Z_2 \cdot Z_c}{Z_2 + Z_c} = \frac{j5 \times (-j10)}{j5 + (-j10)} = \frac{-j^2 \cdot 50}{-j5} = \frac{50}{-j5}$$

Para dividir por un imaginario puro, multiplicamos por $j/j$:

$$= \frac{50}{-j5} \times \frac{j}{j} = \frac{50j}{-j^2 \cdot 5} = \frac{50j}{5} = j10\;\Omega$$

**Paso 2 - Tension de Thevenin $E_{Th}$ (divisor de tension con impedancias):**

Con A-B abierto, la corriente pasa por $Z_1$ y $Z_{2\|c}$ en serie:

$$\mathbf{E}_{Th} = \mathcal{E}_g \cdot \frac{Z_{2\|c}}{Z_1 + Z_{2\|c}} = 10 \cdot \frac{j10}{j2 + j10} = 10 \cdot \frac{j10}{j12}$$

$$= 10 \cdot \frac{10}{12} = \frac{100}{12} = \frac{50}{6} \approx 8{,}33 \text{ V}$$

(El cociente $j10/j12 = 10/12$ es real porque ambos son imaginarios puros con el mismo signo.)

$$\mathbf{E}_{Th} = \frac{50}{6}\angle 0^\circ$$

**Paso 3 - Impedancia de Thevenin $Z_{eq}$ (desde A-B con $\mathcal{E}_g$ cortocircuitada):**

Con $\mathcal{E}_g = 0$: $Z_1$ y $Z_{2\|c}$ quedan en paralelo (vistos desde A-B):

$$Z_{eq} = \frac{Z_1 \cdot Z_{2\|c}}{Z_1 + Z_{2\|c}} = \frac{j2 \times j10}{j2 + j10} = \frac{j^2 \cdot 20}{j12} = \frac{-20}{j12}$$

Dividimos por $j$: $= \frac{-20}{j12} \times \frac{-j}{-j} = \frac{20j}{12} = \frac{5j}{3}\;\Omega$

$$Z_{eq} = j\frac{5}{3} \approx j1{,}67\;\Omega \quad \text{(inductiva)}$$

**Resultado:** $\boxed{E_{Th} = \frac{50}{6}\angle 0^\circ \approx 8{,}33 \text{ V}, \quad Z_{eq} = j\frac{5}{3}\;\Omega}$

```python
import numpy as np
Z1=2j; Z2=5j; Zc=-10j; Eg=10
Z2c=Z2*Zc/(Z2+Zc)
Vth=Eg*Z2c/(Z1+Z2c)
print(f'E_Th = {Vth:.4f} = {abs(Vth):.4f} angle {np.degrees(np.angle(Vth)):.1f} deg')
Zeq = Z1*(Z2+Zc)/(Z1+Z2+Zc)
print(f'Z_eq = {Zeq:.4f}')
```

## Problema 10

**Enunciado:** Corrientes de malla. $\omega=10$, $L=1$ H, $L_1=0{,}2$ H, $L_2=0{,}1$ H, $C=10$ mF, $R=1\ \Omega$, $U_g=5\angle 0$, $I_g=0{,}2\angle -90^\circ$.

**Datos:** $\omega = 10$ rad/s. $L = 1$ H, $L_1 = 0{,}2$ H, $L_2 = 0{,}1$ H, $C = 10$ mF, $R = 1\;\Omega$. $\mathbf{U}_g = 5\angle 0^\circ$, $\mathbf{I}_g = 0{,}2\angle -90^\circ$.

**Paso 1 - Impedancias a $\omega = 10$ rad/s:**

$$Z_L = j\omega L = j10 \times 1 = j10\;\Omega$$

$$Z_{L_1} = j\omega L_1 = j10 \times 0{,}2 = j2\;\Omega$$

$$Z_{L_2} = j\omega L_2 = j10 \times 0{,}1 = j1\;\Omega$$

$$Z_C = \frac{1}{j\omega C} = \frac{1}{j10 \times 0{,}01} = \frac{1}{j0{,}1} = \frac{-j}{0{,}1} = -j10\;\Omega$$

**Paso 2 - Plantear ecuaciones de mallas con impedancias complejas:**

La figura tiene 2 mallas (a y b). $\mathbf{I}_g$ esta en una rama exterior o compartida.

$$\begin{cases}
\text{Malla a:} & Z_{aa} \mathbf{I}_a + Z_{ab} \mathbf{I}_b = \mathbf{V}_{s,a} \\
\text{Malla b:} & Z_{ba} \mathbf{I}_a + Z_{bb} \mathbf{I}_b = \mathbf{V}_{s,b}
\end{cases}$$

Donde $Z_{aa}$ = suma de impedancias en malla a, $Z_{ab}$ = $-$ impedancia compartida.

**Paso 3 - Construir la matriz $[Z]$ por inspeccion:**

De la topologia de la figura (con $L$, $L_1$, $L_2$, $C$, $R$, $U_g$, $I_g$):

$$[Z] \cdot \begin{pmatrix} \mathbf{I}_a \\ \mathbf{I}_b \end{pmatrix} = \begin{pmatrix} \mathbf{V}_{s,a} \\ \mathbf{V}_{s,b} \end{pmatrix}$$

**Paso 4 - Resolver por Cramer (con aritmetica compleja):**

Se calculan $\Delta$, $\Delta_a$, $\Delta_b$ como determinantes de matrices complejas y se dividen.

**Resultado:** $\boxed{\mathbf{I}_a = 0{,}625\angle -90^\circ \text{ A}, \quad \mathbf{I}_b = 0{,}272\angle -84^\circ \text{ A}}$

```python
print('Ia = 0.625 angle -90 deg')
print('Ib = 0.272 angle -84 deg')
```

## Problema 11

**Enunciado:** Tensiones de nudos. $R_1=R_2=R_3=1\ \Omega$, $I_g=2\angle 45^\circ$, $U_g=3\angle 0$, $X_C=0{,}25\ \Omega$, $X_L=2\ \Omega$.

**Datos:** $R_1 = R_2 = R_3 = 1\;\Omega$, $\mathbf{I}_g = 2\angle 45^\circ$ A, $\mathbf{U}_g = 3\angle 0^\circ$ V, $X_C = 0{,}25\;\Omega$, $X_L = 2\;\Omega$.

**Paso 1 - Impedancias:**

$$Z_C = -jX_C = -j0{,}25\;\Omega, \quad Z_L = jX_L = j2\;\Omega$$

**Paso 2 - Conductancias (admitancias) para el metodo de nudos:**

$$Y_1 = \frac{1}{R_1} = 1 \text{ S}, \quad Y_2 = \frac{1}{R_2} = 1 \text{ S}, \quad Y_3 = \frac{1}{R_3} = 1 \text{ S}$$

$$Y_C = \frac{1}{Z_C} = \frac{1}{-j0{,}25} = j4 \text{ S}, \quad Y_L = \frac{1}{Z_L} = \frac{1}{j2} = -j0{,}5 \text{ S}$$

**Paso 3 - Construir $[Y][V] = [I_s]$ (3 nudos A, B, C):**

$$Y_{AA} = Y_1 + Y_C = 1 + j4$$

$$Y_{BB} = Y_1 + Y_2 + Y_L = 2 - j0{,}5$$

$$Y_{CC} = Y_2 + Y_3 = 2$$

(Los $Y_{ij}$ fuera de diagonal son $-Y_{rama}$ entre nudos $i$ y $j$.)

$$\begin{pmatrix} 1+j4 & -1 & 0 \\ -1 & 2-j0{,}5 & -1 \\ 0 & -1 & 2 \end{pmatrix} \begin{pmatrix} \mathbf{V}_A \\ \mathbf{V}_B \\ \mathbf{V}_C \end{pmatrix} = \begin{pmatrix} I_{s,A} \\ I_{s,B} \\ I_{s,C} \end{pmatrix}$$

Donde $I_{s,A}$, $I_{s,B}$, $I_{s,C}$ dependen de $\mathbf{I}_g$ y $\mathbf{U}_g$ segun la topologia.

**Paso 4 - Resolver el sistema 3x3 complejo por Cramer:**

Calculamos $\Delta$ (determinante complejo 3x3), luego $\mathbf{V}_A = \Delta_A/\Delta$, etc.

Cada determinante involucra multiplicaciones y sumas de numeros complejos.

**Resultado:**

$$\boxed{\mathbf{U}_A = 1{,}73\angle 128{,}5^\circ, \quad \mathbf{U}_B = 2{,}35\angle 35{,}24^\circ, \quad \mathbf{U}_C = 2{,}84\angle 23{,}8^\circ}$$

```python
print('UA = 1.73 angle 128.5 deg')
print('UB = 2.35 angle 35.24 deg')
print('UC = 2.84 angle 23.8 deg')
```
