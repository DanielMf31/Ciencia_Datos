## Formulario del Tema 5: Componentes Dinamicos

| Concepto | Formula |
|----------|---------|
| Condensador | $\boxed{i = C \frac{dv}{dt}}$, $v = \frac{1}{C}\int i\,dt + v(0)$ |
| Energia condensador | $\boxed{E_C = \frac{1}{2} C V^2}$ |
| $C$ en serie | $\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_2} + \ldots$ |
| $C$ en paralelo | $C_{eq} = C_1 + C_2 + \ldots$ |
| Bobina | $\boxed{v = L \frac{di}{dt}}$, $i = \frac{1}{L}\int v\,dt + i(0)$ |
| Energia bobina | $\boxed{E_L = \frac{1}{2} L I^2}$ |
| $L$ en serie | $L_{eq} = L_1 + L_2 \pm 2M$ |
| $L$ en paralelo | $\frac{1}{L_{eq}} = \frac{1}{L_1} + \frac{1}{L_2}$ |
| Inductancia mutua | $v_1 = L_1 \frac{di_1}{dt} \pm M \frac{di_2}{dt}$ |
| Coeficiente acoplamiento | $k = M / \sqrt{L_1 L_2}$, $0 \leq k \leq 1$ |
| Transformador ideal | $\frac{V_1}{V_2} = \frac{N_1}{N_2}$, $\frac{I_1}{I_2} = \frac{N_2}{N_1}$, $Z' = Z \left(\frac{N_1}{N_2}\right)^2$ |

## Problema 1**Enunciado:** Se tienen 3 condensadores de $C = 3\;\mu$F cada uno. Uno de ellos esta en serie con el paralelo de los otros dos. Hallar la capacidad equivalente en serie y en paralelo.

**Datos:**- $C_1 = C_2 = C_3 = 3\;\mu$F**Caso 1: Serie****Paso 1:** Para condensadores en serie, la capacidad equivalente es:$$\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_2} + \frac{1}{C_3}$$$$\frac{1}{C_{eq}} = \frac{1}{3} + \frac{1}{3} + \frac{1}{3} = 1$$$$C_{eq,serie} = 1\;\mu\text{F}$$Pero la configuracion del boletin es: un condensador en serie con el paralelo de los otros dos.**Paso 2:** Paralelo de $C_2$ y $C_3$:$$C_{23} = C_2 + C_3 = 3 + 3 = 6\;\mu\text{F}$$**Paso 3:** Serie de $C_1$ con $C_{23}$:$$\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_{23}} = \frac{1}{3} + \frac{1}{6} = \frac{2+1}{6} = \frac{1}{2}$$$$C_{eq} = 2\;\mu\text{F}$$**Caso 2: Paralelo****Paso 4:** La configuracion complementaria (un condensador en paralelo con la serie de los otros dos):Serie de $C_2$ y $C_3$:$$\frac{1}{C_{23}} = \frac{1}{3} + \frac{1}{3} = \frac{2}{3} \implies C_{23} = 1{,}5\;\mu\text{F}$$Paralelo con $C_1$:$$C_{eq} = C_1 + C_{23} = 3 + 1{,}5 = 4{,}5\;\mu\text{F}$$**Resultado:**$$\boxed{C_{eq,serie} = 2\;\mu\text{F}, \quad C_{eq,paralelo} = 4{,}5\;\mu\text{F}}$$

```python
# Problema 1 - Asociacion de condensadoresC = 3e-6  # F# Caso 1: C1 en serie con (C2 || C3)C23_par = C + C  # paraleloCeq_serie = 1 / (1/C + 1/C23_par)# Caso 2: C1 en paralelo con (C2 -- C3)C23_ser = 1 / (1/C + 1/C)  # serieCeq_paralelo = C + C23_serprint(f'Caso serie: Ceq = {Ceq_serie*1e6:.1f} uF')print(f'Caso paralelo: Ceq = {Ceq_paralelo*1e6:.1f} uF')
```

## Problema 2**Enunciado:** Se tienen 3 inductores de $L = 10$ mH cada uno (sin acoplamiento magnetico). Uno en paralelo con la serie de los otros dos, y uno en serie con el paralelo de los otros dos. Hallar la inductancia equivalente.

**Datos:**- $L_1 = L_2 = L_3 = 10$ mH**Caso 1: Serie (L1 en serie con paralelo de L2, L3)****Paso 1:** Paralelo de $L_2$ y $L_3$ (sin acoplamiento):$$\frac{1}{L_{23}} = \frac{1}{L_2} + \frac{1}{L_3} = \frac{1}{10} + \frac{1}{10} = \frac{2}{10}$$$$L_{23} = 5 \text{ mH}$$**Paso 2:** Serie con $L_1$:$$L_{eq} = L_1 + L_{23} = 10 + 5 = 15 \text{ mH}$$**Caso 2: Paralelo (L1 en paralelo con serie de L2, L3)****Paso 3:** Serie de $L_2$ y $L_3$:$$L_{23} = L_2 + L_3 = 10 + 10 = 20 \text{ mH}$$**Paso 4:** Paralelo con $L_1$:$$\frac{1}{L_{eq}} = \frac{1}{L_1} + \frac{1}{L_{23}} = \frac{1}{10} + \frac{1}{20} = \frac{3}{20}$$$$L_{eq} = \frac{20}{3} = 6{,}66 \text{ mH}$$**Resultado:**$$\boxed{L_{eq,serie} = 15 \text{ mH}, \quad L_{eq,paralelo} = 6{,}66 \text{ mH}}$$

```python
# Problema 2 - Asociacion de inductoresL = 10e-3  # H# Caso serie: L1 + (L2 || L3)L23_par = 1 / (1/L + 1/L)Leq_serie = L + L23_par# Caso paralelo: L1 || (L2 + L3)L23_ser = L + LLeq_paralelo = 1 / (1/L + 1/L23_ser)print(f'Caso serie: Leq = {Leq_serie*1e3:.2f} mH')print(f'Caso paralelo: Leq = {Leq_paralelo*1e3:.2f} mH')
```

## Problema 3**Enunciado:** Dos inductores acoplados con $L_1 = L_2 = 10$ H y mutua inductancia $M = 5$ H. Escribir las ecuaciones de tension.

**Datos:**- $L_1 = L_2 = 10$ H- $M = 5$ H**Paso 1:** Las ecuaciones constitutivas de dos inductores acoplados son:$$v_1 = L_1 \frac{di_1}{dt} \pm M \frac{di_2}{dt}$$$$v_2 = \pm M \frac{di_1}{dt} + L_2 \frac{di_2}{dt}$$El signo de $M$ depende de la convencion de puntos y la orientacion de las corrientes.**Paso 2:** Segun la configuracion del boletin (con la convencion de signos indicada):$$v_1 = 10 \frac{di_1}{dt} - 5 \frac{di_2}{dt}$$$$v_2 = -10 \frac{di_2}{dt} + 5 \frac{di_1}{dt}$$**Paso 3:** Verificacion del coeficiente de acoplamiento:$$k = \frac{M}{\sqrt{L_1 L_2}} = \frac{5}{\sqrt{10 \cdot 10}} = \frac{5}{10} = 0{,}5$$Como $0 \leq k \leq 1$, el acoplamiento es valido.**Resultado:**$$\boxed{v_1 = 10\frac{di_1}{dt} - 5\frac{di_2}{dt}, \quad v_2 = -10\frac{di_2}{dt} + 5\frac{di_1}{dt}}$$

```python
# Problema 3 - Inductores acopladosimport numpy as npL1 = L2 = 10  # HM = 5  # H# Coeficiente de acoplamientok = M / np.sqrt(L1 * L2)print(f'L1 = L2 = {L1} H')print(f'M = {M} H')print(f'k = {k:.2f}')print(f'\nEcuaciones:')print(f'v1 = {L1}*di1/dt - {M}*di2/dt')print(f'v2 = -{L2}*di2/dt + {M}*di1/dt')
```

## Problema 4**Enunciado:** Dos inductores acoplados con $L_1 = L_2 = 10$ H, $M = 10$ H y relacion de espiras $n:1$. Escribir las ecuaciones de tension.

**Datos:**- $L_1 = L_2 = 10$ H- $M = 10$ H, relacion $n:1$**Paso 1:** Coeficiente de acoplamiento:$$k = \frac{M}{\sqrt{L_1 L_2}} = \frac{10}{\sqrt{10 \cdot 10}} = \frac{10}{10} = 1$$El acoplamiento es perfecto ($k = 1$), lo que corresponde a un transformador ideal.**Paso 2:** Ecuaciones con la convencion del boletin:$$v_1 = -10 \frac{di_1}{dt} - 10 \frac{di_2}{dt}$$$$v_2 = 10 \frac{di_2}{dt} + 10 \frac{di_1}{dt}$$**Paso 3:** La convencion de signos depende de la posicion de los puntos respecto a las corrientes. En este caso, con acoplamiento perfecto y la orientacion del boletin, ambos terminos de mutua son positivos dentro de sus respectivas ecuaciones.**Resultado:**$$\boxed{v_1 = -10\frac{di_1}{dt} - 10\frac{di_2}{dt}, \quad v_2 = 10\frac{di_2}{dt} + 10\frac{di_1}{dt}}$$

```python
# Problema 4 - Acoplamiento perfectoL1 = L2 = 10  # HM = 10  # Hk = M / np.sqrt(L1 * L2)print(f'L1 = L2 = {L1} H')print(f'M = {M} H')print(f'k = {k:.2f} (acoplamiento perfecto)')print(f'\nEcuaciones:')print(f'v1 = -{L1}*di1/dt - {M}*di2/dt')print(f'v2 = {L2}*di2/dt + {M}*di1/dt')
```

## Problema 5**Enunciado:** Dos inductores con $L_1 = L_2 = 10$ H y $M = 5$ H. Obtener la inductancia equivalente para 4 configuraciones distintas de conexion.

**Datos:**- $L_1 = L_2 = 10$ H, $M = 5$ H**Configuracion 1: Serie aditiva****Paso 1:** Cuando las corrientes entran por los puntos en ambas bobinas (serie aditiva):$$L_{eq} = L_1 + L_2 + 2M = 10 + 10 + 10 = 30 \text{ H}$$**Configuracion 2: Serie sustractiva****Paso 2:** Cuando una corriente entra por el punto y la otra sale (serie sustractiva):$$L_{eq} = L_1 + L_2 - 2M = 10 + 10 - 10 = 10 \text{ H}$$**Configuracion 3: Paralelo aditivo****Paso 3:** En paralelo con acoplamiento aditivo:$$L_{eq} = \frac{L_1 L_2 - M^2}{L_1 + L_2 - 2M} = \frac{10 \cdot 10 - 25}{10 + 10 - 10} = \frac{75}{10} = 7{,}5 \text{ H}$$**Configuracion 4: Paralelo sustractivo****Paso 4:** En paralelo con acoplamiento sustractivo:$$L_{eq} = \frac{L_1 L_2 - M^2}{L_1 + L_2 + 2M} = \frac{75}{30} = 2{,}5 \text{ H}$$**Resultado:**$$\boxed{L_{eq} = 10 \text{ H},\; 30 \text{ H},\; 7{,}5 \text{ H},\; 2{,}5 \text{ H}}$$

```python
# Problema 5 - 4 configuraciones de inductores acopladosL1 = L2 = 10  # HM = 5  # H# Serie sustractivaLeq_ss = L1 + L2 - 2*M# Serie aditivaLeq_sa = L1 + L2 + 2*M# Paralelo aditivoLeq_pa = (L1*L2 - M**2) / (L1 + L2 - 2*M)# Paralelo sustractivoLeq_ps = (L1*L2 - M**2) / (L1 + L2 + 2*M)print(f'Serie sustractiva:  Leq = {Leq_ss} H')print(f'Serie aditiva:      Leq = {Leq_sa} H')print(f'Paralelo aditivo:   Leq = {Leq_pa} H')print(f'Paralelo sustractivo: Leq = {Leq_ps} H')
```

## Problema 6**Enunciado:** Un condensador cargado con $Q_0 = 0{,}4$ C se descarga a traves de una resistencia $R = 1\;k\Omega$. La corriente inicial es $i(0) = 0{,}1$ A. Obtener $i_R(t)$.

**Datos:**- $Q_0 = 0{,}4$ C- $R = 1\;k\Omega = 1000\;\Omega$- $i(0) = 0{,}1$ A**Paso 1:** Tension inicial del condensador:$$V_0 = R \cdot i(0) = 1000 \cdot 0{,}1 = 100 \text{ V}$$**Paso 2:** Capacidad del condensador:$$C = \frac{Q_0}{V_0} = \frac{0{,}4}{100} = 4 \times 10^{-3} \text{ F} = 4 \text{ mF}$$**Paso 3:** Constante de tiempo:$$\tau = RC = 1000 \cdot 4 \times 10^{-3} = 4 \text{ s}$$**Paso 4:** La corriente de descarga de un circuito RC es:$$i_R(t) = i(0) \cdot e^{-t/\tau} = 0{,}1 \cdot e^{-t/4} \text{ A}$$**Verificacion:** En $t = 0$: $i_R(0) = 0{,}1$ A $\checkmark$Cuando $t \to \infty$: $i_R \to 0$ (el condensador se descarga completamente) $\checkmark$**Resultado:**$$\boxed{i_R(t) = 0{,}1 \cdot e^{-t/4} \text{ A}}$$

```python
# Problema 6 - Descarga RCimport numpy as npimport matplotlib.pyplot as pltQ0 = 0.4    # CR = 1000    # ohmi0 = 0.1    # AV0 = R * i0C = Q0 / V0tau = R * Cprint(f'V0 = {V0} V')print(f'C = {C*1e3} mF')print(f'tau = RC = {tau} s')print(f'iR(t) = {i0} * exp(-t/{tau}) A')# GraficaCOLOR_PRINCIPAL = '#2171b5't = np.linspace(0, 5*tau, 500)iR = i0 * np.exp(-t/tau)fig, ax = plt.subplots(figsize=(8, 4))ax.plot(t, iR*1000, color=COLOR_PRINCIPAL, linewidth=2, label=r'$i_R(t) = 0.1 e^{-t/4}$ A')ax.axhline(y=i0*1000*np.exp(-1), color='#636363', linestyle='--', alpha=0.7, label=rf'$i(\tau) = {i0*np.exp(-1)*1000:.1f}$ mA')ax.axvline(x=tau, color='#636363', linestyle='--', alpha=0.7)ax.set_xlabel('Tiempo (s)')ax.set_ylabel('Corriente (mA)')ax.set_title('Descarga RC')ax.legend()ax.grid(True, alpha=0.3)plt.tight_layout()plt.show()
```

## Problema 7**Enunciado:** Un inductor almacena una energia de $W_L = 0{,}002$ J y se descarga a traves de $R = 1\;k\Omega$. La potencia inicial disipada es $P(0) = 40$ W. Obtener $i_R(t)$.

**Datos:**- $W_L = 0{,}002$ J- $R = 1\;k\Omega = 1000\;\Omega$- $P(0) = 40$ W**Paso 1:** Corriente inicial a partir de la potencia:$$P(0) = i(0)^2 R \implies i(0)^2 = \frac{P(0)}{R} = \frac{40}{1000} = 0{,}04$$$$i(0) = \pm 0{,}2 \text{ A}$$**Paso 2:** Inductancia del inductor:$$W_L = \frac{1}{2} L i(0)^2 \implies L = \frac{2 W_L}{i(0)^2} = \frac{2 \cdot 0{,}002}{0{,}04} = 0{,}1 \text{ H}$$**Paso 3:** Constante de tiempo:$$\tau = \frac{L}{R} = \frac{0{,}1}{1000} = 10^{-4} \text{ s} = 0{,}1 \text{ ms}$$**Paso 4:** La corriente de descarga RL es:$$i_R(t) = \pm 0{,}2 \cdot e^{-t/0{,}0001} \text{ A}$$$$i_R(t) = \pm 0{,}2 \cdot e^{-10000 t} \text{ A}$$**Verificacion:** $P(0) = (0{,}2)^2 \cdot 1000 = 0{,}04 \cdot 1000 = 40$ W $\checkmark$$W_L = \frac{1}{2} \cdot 0{,}1 \cdot (0{,}2)^2 = \frac{1}{2} \cdot 0{,}1 \cdot 0{,}04 = 0{,}002$ J $\checkmark$**Resultado:**$$\boxed{i_R(t) = \pm 0{,}2 \cdot e^{-t/0{,}0001} \text{ A}}$$
