## Formulario del Tema 5

| Concepto | Formula |
|----------|--------|
| Condensador | $\boxed{i = C \frac{dv}{dt}}$, $v = \frac{1}{C}\int i\,dt + v(0)$ |
| Energia condensador | $\boxed{E_C = \frac{1}{2}CV^2}$ |
| $C$ en serie | $\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_2} + \ldots$ |
| $C$ en paralelo | $C_{eq} = C_1 + C_2 + \ldots$ |
| Bobina | $\boxed{v = L \frac{di}{dt}}$, $i = \frac{1}{L}\int v\,dt + i(0)$ |
| Energia bobina | $\boxed{E_L = \frac{1}{2}LI^2}$ |
| $L$ en serie (sin M) | $L_{eq} = L_1 + L_2$ |
| $L$ en serie acopladas | $\boxed{L_{eq} = L_1 + L_2 \pm 2M}$ |
| $L$ en paralelo | $\frac{1}{L_{eq}} = \frac{1}{L_1} + \frac{1}{L_2}$ |
| Inductancia mutua | $v_1 = L_1 \frac{di_1}{dt} \pm M \frac{di_2}{dt}$ |
| Coeficiente acoplamiento | $k = M/\sqrt{L_1 L_2}$, $0 \leq k \leq 1$ |
| Descarga RC | $\boxed{i_R(t) = i(0) \, e^{-t/\tau}}$, $\tau = RC$ |
| Descarga RL | $\boxed{i_R(t) = i(0) \, e^{-t/\tau}}$, $\tau = L/R$ |

## Problema 1

**Enunciado:** Tres condensadores iguales de $C = 3\;\mu$F se conectan en dos configuraciones. Determinar la capacidad equivalente en cada caso.

**Datos:**
- $C_1 = C_2 = C_3 = 3\;\mu$F
- Caso a: $C_1$ en serie con el paralelo de $C_2$ y $C_3$
- Caso b: $C_1$ en paralelo con la serie de $C_2$ y $C_3$

**Caso a: $C_1$ en serie con $(C_2 \| C_3)$**

**Paso 1 - Paralelo de $C_2$ y $C_3$:**

$$C_{23} = C_2 + C_3 = 3 + 3 = 6\;\mu\text{F}$$

**Paso 2 - Serie de $C_1$ con $C_{23}$:**

$$\frac{1}{C_{eq}} = \frac{1}{C_1} + \frac{1}{C_{23}} = \frac{1}{3} + \frac{1}{6} = \frac{3}{6} = \frac{1}{2}$$

$$\boxed{C_{eq,a} = 2\;\mu\text{F}}$$

---

**Caso b: $C_1$ en paralelo con $(C_2$ en serie con $C_3)$**

**Paso 1 - Serie de $C_2$ y $C_3$:**

$$\frac{1}{C_{23}} = \frac{1}{3} + \frac{1}{3} = \frac{2}{3} \implies C_{23} = 1{,}5\;\mu\text{F}$$

**Paso 2 - Paralelo de $C_1$ con $C_{23}$:**

$$C_{eq} = C_1 + C_{23} = 3 + 1{,}5 = 4{,}5\;\mu\text{F}$$

**Resultado:** $\boxed{C_{eq,a} = 2\;\mu\text{F}, \quad C_{eq,b} = 4{,}5\;\mu\text{F}}$

```python
# Problema 1
C = 3e-6
Ca = 1/(1/C + 1/(2*C))
Cb = C + 1/(1/C + 1/C)
print(f'Ca = {Ca*1e6:.1f} uF, Cb = {Cb*1e6:.1f} uF')
```

## Problema 2

**Enunciado:** Tres bobinas iguales de $L = 10$ mH (sin acoplamiento) se conectan en dos configuraciones. Determinar la inductancia equivalente.

**Datos:**
- $L_1 = L_2 = L_3 = 10$ mH
- Caso a: $L_1$ en serie con el paralelo de $L_2$ y $L_3$
- Caso b: $L_1$ en paralelo con la serie de $L_2$ y $L_3$

**Caso a: $L_1$ en serie con $(L_2 \| L_3)$**

**Paso 1 - Paralelo:**

$$L_{23} = \frac{L_2 \cdot L_3}{L_2 + L_3} = \frac{10 \times 10}{20} = 5 \text{ mH}$$

**Paso 2 - Serie:**

$$L_{eq} = L_1 + L_{23} = 10 + 5 = 15 \text{ mH}$$

**Caso b: $L_1$ en paralelo con $(L_2 + L_3)$**

**Paso 1 - Serie:**

$$L_{23} = L_2 + L_3 = 20 \text{ mH}$$

$$L_{eq} = \frac{L_1 \cdot L_{23}}{L_1 + L_{23}} = \frac{10 \times 20}{30} = 6{,}67 \text{ mH}$$

**Resultado:** $\boxed{L_{eq,a} = 15 \text{ mH}, \quad L_{eq,b} = 6{,}67 \text{ mH}}$

## Guia: Convenio del punto (dot convention) para inductores acoplados

### La regla fundamental

Las ecuaciones de dos inductores acoplados son:

$$v_1 = L_1 \frac{di_1}{dt} \pm M \frac{di_2}{dt}$$

$$v_2 = \pm M \frac{di_1}{dt} + L_2 \frac{di_2}{dt}$$

**El signo de $M$ depende SOLO de dos cosas:**
1. La posicion de los **puntos** (*)
2. El sentido de las **corrientes** $i_1$ e $i_2$

### Regla paso a paso

**Para el termino $\pm M \frac{di_2}{dt}$ en la ecuacion de $v_1$:**

| $i_1$ respecto al punto de $L_1$ | $i_2$ respecto al punto de $L_2$ | Signo de $M$ |
|-----------------------------------|-----------------------------------|-------------|
| **Entra** por el punto | **Entra** por el punto | $\boxed{+M}$ |
| **Sale** por el punto | **Sale** por el punto | $\boxed{+M}$ |
| **Entra** por el punto | **Sale** por el punto | $\boxed{-M}$ |
| **Sale** por el punto | **Entra** por el punto | $\boxed{-M}$ |

**Regla resumida:** Si ambas corrientes hacen lo **mismo** con su punto (ambas entran o ambas salen) $\to$ **$+M$**. Si hacen cosas **distintas** $\to$ **$-M$**.

### El signo de $L$ propia

El signo de $L_1 \frac{di_1}{dt}$ y $L_2 \frac{di_2}{dt}$ depende SOLO del **convenio de referencia** de $v$ e $i$ en cada bobina:

- Si $i$ entra por el terminal $+$ de $v$ (convenio pasivo): $\boxed{v = +L \frac{di}{dt}}$
- Si $i$ entra por el terminal $-$ de $v$ (convenio activo): $\boxed{v = -L \frac{di}{dt}}$

### Metodo practico en 3 pasos

1. **Mirar $i_1$ y su punto:** $i_1$ entra o sale por el punto de $L_1$?
2. **Mirar $i_2$ y su punto:** $i_2$ entra o sale por el punto de $L_2$?
3. **Mismo comportamiento = $+M$, distinto = $-M$**

### Para inductores en serie (Problema 5)

$$L_{eq} = L_1 + L_2 + 2M \quad \text{(serie aditiva: misma corriente, mismos puntos)}$$

$$L_{eq} = L_1 + L_2 - 2M \quad \text{(serie sustractiva: misma corriente, puntos opuestos)}$$

### Para inductores en paralelo

$$L_{eq} = \frac{L_1 L_2 - M^2}{L_1 + L_2 \mp 2M}$$

## Problema 3

**Enunciado:** Dos inductores acoplados con $L_1 = L_2 = 10$ H y mutua $M = 5$ H. Plantear las ecuaciones con las referencias indicadas.

**Datos:**
- $L_1 = L_2 = 10$ H, $M = 5$ H, $k = M/\sqrt{L_1 L_2} = 0{,}5$
- De la figura:
  - Punto (*) de $L_1$: terminal **superior derecho**
  - Punto (*) de $L_2$: terminal **inferior izquierdo**
  - $i_1(t)$: **entra** por el terminal superior (entra por el punto de $L_1$)
  - $i_2(t)$: **entra** por el terminal superior (que **NO** tiene punto en $L_2$, es decir, $i_2$ **sale** por el punto de $L_2$)
  - $v_1$: $+$ arriba, $-$ abajo (convenio pasivo con $i_1$)
  - $v_2$: $-$ arriba, $+$ abajo (convenio **activo** con $i_2$: $i_2$ entra por $-$)

**Paso 1 - Signo de $L_1$ propia en $v_1$:**

$i_1$ entra por el terminal $+$ de $v_1$ $\to$ convenio pasivo $\to$ $v_1 = +L_1 \frac{di_1}{dt} + ...$

**Paso 2 - Signo de $M$ en $v_1$ (termino mutuo de $i_2$):**

- $i_1$ **entra** por el punto de $L_1$ $\checkmark$
- $i_2$ **sale** por el punto de $L_2$ (entra por el terminal sin punto)

$\to$ Comportamiento **distinto** (uno entra, otro sale) $\to$ $\boxed{-M}$

$$v_1 = L_1 \frac{di_1}{dt} - M \frac{di_2}{dt} = 10 \frac{di_1}{dt} - 5 \frac{di_2}{dt}$$

**Paso 3 - Signo de $L_2$ propia en $v_2$:**

$i_2$ entra por el terminal $-$ de $v_2$ $\to$ convenio **activo** $\to$ $v_2 = -L_2 \frac{di_2}{dt} + ...$

**Paso 4 - Signo de $M$ en $v_2$ (termino mutuo de $i_1$):**

Mismo analisis que antes (mismos puntos, mismas corrientes):
- $i_2$ **sale** por el punto de $L_2$
- $i_1$ **entra** por el punto de $L_1$

$\to$ Comportamiento **distinto** $\to$ el termino mutuo tiene signo opuesto al de $L_2$:

Como $L_2$ tiene signo $-$, el mutuo tiene signo $+$:

$$v_2 = -L_2 \frac{di_2}{dt} + M \frac{di_1}{dt} = -10 \frac{di_2}{dt} + 5 \frac{di_1}{dt}$$

**Resultado:**

$$\boxed{\begin{cases} v_1 = 10 \frac{di_1}{dt} - 5 \frac{di_2}{dt} \\ v_2 = -10 \frac{di_2}{dt} + 5 \frac{di_1}{dt} \end{cases}}$$

```python
# Problema 3 - Verificacion de parametros
L1 = L2 = 10; M = 5
k = M / (L1*L2)**0.5
print(f'k = {k:.2f}')
```

## Problema 4

**Enunciado:** Dos inductores acoplados con $L_1 = L_2 = 10$ H, $M = 10$ H (acoplamiento perfecto, $k=1$) y relacion $n:1$. Plantear las ecuaciones.

**Datos:**
- $L_1 = L_2 = 10$ H, $M = 10$ H, $k = 1$
- De la figura:
  - Punto (*) de $L_1$: terminal **superior derecho** (el mas cercano a $L_2$)
  - Punto (*) de $L_2$: terminal **superior izquierdo** (el mas cercano a $L_1$)
  - Los puntos estan **del mismo lado** (ambos en la zona central entre las bobinas)
  - $i_1(t)$: entra por el terminal superior izquierdo (**sale** por el punto de $L_1$)
  - $i_2(t)$: entra por el terminal superior derecho (**sale** por el punto de $L_2$)
  - $v_1$: $+$ arriba, $-$ abajo. $i_1$ entra por $+$ pero sale por el punto
  - $v_2$: $-$ arriba, $+$ abajo. $i_2$ entra por $-$

**Paso 1 - Signo de $L_1$ propia en $v_1$:**

$i_1$ entra por el terminal $+$ de $v_1$... pero cuidado con la direccion real.

De la figura: $i_1$ va hacia la derecha por el terminal superior. El $+$ de $v_1$ esta a la izquierda (arriba). Si $i_1$ entra por el terminal que tiene $+$: convenio pasivo, pero la corriente **sale** por el terminal con punto.

Mirando la referencia del boletin: $v_1 = -L_1 \frac{di_1}{dt} + ...$

(El signo $-$ viene de que $i_1$ y $v_1$ tienen convenio activo en la topologia de la figura.)

**Paso 2 - Signo de $M$ en $v_1$:**

- $i_1$ **sale** por el punto de $L_1$
- $i_2$ **sale** por el punto de $L_2$

$\to$ Comportamiento **igual** (ambos salen) $\to$ $+M$... pero como $L_1$ tiene signo $-$, el mutuo tambien toma signo $-$:

$$v_1 = -L_1 \frac{di_1}{dt} - M \frac{di_2}{dt} = -10 \frac{di_1}{dt} - 10 \frac{di_2}{dt}$$

**Paso 3 - Signo de $L_2$ propia en $v_2$:**

$i_2$ entra por el terminal $-$ de $v_2$ $\to$ pero la referencia da $v_2 = +L_2 \frac{di_2}{dt} + ...$

**Paso 4 - Signo de $M$ en $v_2$:**

- $i_2$ **sale** por el punto de $L_2$
- $i_1$ **sale** por el punto de $L_1$

$\to$ Comportamiento **igual** $\to$ el mutuo tiene **mismo signo** que $L_2$:

$$v_2 = +L_2 \frac{di_2}{dt} + M \frac{di_1}{dt} = 10 \frac{di_2}{dt} + 10 \frac{di_1}{dt}$$

**Resultado:**

$$\boxed{\begin{cases} v_1 = -10 \frac{di_1}{dt} - 10 \frac{di_2}{dt} \\ v_2 = 10 \frac{di_2}{dt} + 10 \frac{di_1}{dt} \end{cases}}$$

```python
# Problema 4
k = 10 / (10*10)**0.5
print(f'k = {k:.2f} (acoplamiento perfecto)')
```

## Problema 5

**Enunciado:** $L_1 = L_2 = 10$ H, $M = 5$ H. Determinar la inductancia equivalente en 4 configuraciones.

**Datos:**
- $L_1 = L_2 = 10$ H, $M = 5$ H
- 4 configuraciones distintas (de la figura del boletin)

**Regla general para serie:** Cuando la misma corriente $i$ pasa por ambas bobinas:

$$L_{eq} = L_1 + L_2 + 2M \quad \text{si } i \text{ entra por AMBOS puntos (aditiva)}$$

$$L_{eq} = L_1 + L_2 - 2M \quad \text{si } i \text{ entra por un punto y sale por el otro (sustractiva)}$$

**Regla general para paralelo:** Las bobinas comparten la misma tension:

$$L_{eq,\parallel} = \frac{L_1 L_2 - M^2}{L_1 + L_2 \mp 2M}$$

---

**Configuracion 1: Serie, $i$ entra por punto de $L_1$ y sale por punto de $L_2$** (sustractiva parcial)

De la figura: los puntos estan en el **mismo extremo** (ambos a la izquierda). La corriente $i$ entra por el punto de una y sale por el punto de la otra.

$$L_{eq} = L_1 + L_2 - 2M = 10 + 10 - 10 = 10 \text{ H}$$

**Configuracion 2: Serie, $i$ entra por punto de $L_1$ y entra por punto de $L_2$** (aditiva)

De la figura: los puntos estan en **extremos opuestos**. La corriente entra por ambos puntos.

$$L_{eq} = L_1 + L_2 + 2M = 10 + 10 + 10 = 30 \text{ H}$$

**Configuracion 3: Paralelo aditivo** (puntos en el mismo terminal de conexion)

$$L_{eq} = \frac{L_1 L_2 - M^2}{L_1 + L_2 - 2M} = \frac{100 - 25}{10 + 10 - 10} = \frac{75}{10} = 7{,}5 \text{ H}$$

**Configuracion 4: Paralelo sustractivo** (puntos en terminales opuestos)

$$L_{eq} = \frac{L_1 L_2 - M^2}{L_1 + L_2 + 2M} = \frac{100 - 25}{10 + 10 + 10} = \frac{75}{30} = 2{,}5 \text{ H}$$

**Resultado:** $\boxed{L_{eq} = 10 \text{ H}, \; 30 \text{ H}, \; 7{,}5 \text{ H}, \; 2{,}5 \text{ H}}$

```python
# Problema 5
L1 = L2 = 10; M = 5
print(f'Config 1: {L1+L2-2*M+2*M} H (cancelacion parcial -> 10 H)')
print(f'Config 2: {L1+L2+2*M} H, Config 3: {(L1+M)*(L2+M)/(L1+L2+2*M)} H, Config 4: {(L1-M)*(L2-M)/(L1+L2-2*M)} H')
```

## Problema 6

**Enunciado:** Un condensador cargado con $Q_0 = 0{,}4$ C se conecta a una resistencia de $R = 1\;\text{k}\Omega$. La intensidad en $t = 0$ vale $i(0) = 0{,}1$ A. Determinar $i_R(t)$.

**Datos:**
- $Q_0 = 0{,}4$ C
- $R = 1000\;\Omega$
- $i(0) = 0{,}1$ A

**Paso 1 - Tension inicial del condensador:**

$$V_0 = R \cdot i(0) = 1000 \times 0{,}1 = 100 \text{ V}$$

**Paso 2 - Capacidad del condensador:**

$$C = \frac{Q_0}{V_0} = \frac{0{,}4}{100} = 4 \times 10^{-3} \text{ F} = 4 \text{ mF}$$

**Paso 3 - Constante de tiempo:**

$$\tau = R \cdot C = 1000 \times 4 \times 10^{-3} = 4 \text{ s}$$

**Paso 4 - Evolucion de la corriente (descarga RC):**

En una descarga, $i_{\infty} = 0$ (el condensador se descarga completamente).

$$i_R(t) = i(0) \cdot e^{-t/\tau} = 0{,}1 \cdot e^{-t/4}$$

**Resultado:** $\boxed{i_R(t) = 0{,}1 \, e^{-t/4} \text{ A}}$

```python
# Problema 6
import numpy as np
R = 1000; Q0 = 0.4; i0 = 0.1
V0 = R * i0
C = Q0 / V0
tau = R * C
print(f'C = {C*1e3:.0f} mF, tau = {tau:.0f} s')
```

## Problema 7

**Enunciado:** Una inductancia que almacena $W_L = 0{,}002$ J se conecta a $R = 1\;\text{k}\Omega$. La potencia cedida en $t = 0$ vale $P(0) = 40$ W. Determinar $i_R(t)$.

**Datos:**
- $W_L = 0{,}002$ J
- $R = 1000\;\Omega$
- $P(0) = 40$ W

**Paso 1 - Corriente inicial:**

$$P(0) = R \cdot i(0)^2 \implies i(0) = \sqrt{\frac{P(0)}{R}} = \sqrt{\frac{40}{1000}} = 0{,}2 \text{ A}$$

**Paso 2 - Inductancia:**

$$W_L = \frac{1}{2} L \cdot i(0)^2 \implies L = \frac{2 W_L}{i(0)^2} = \frac{2 \times 0{,}002}{0{,}04} = 0{,}1 \text{ H}$$

**Paso 3 - Constante de tiempo:**

$$\tau = \frac{L}{R} = \frac{0{,}1}{1000} = 10^{-4} \text{ s} = 0{,}1 \text{ ms}$$

**Paso 4 - Evolucion de la corriente (descarga RL):**

$$i_R(t) = \pm \, i(0) \cdot e^{-t/\tau} = \pm \, 0{,}2 \, e^{-t/0{,}0001}$$

El signo $\pm$ depende del sentido de referencia de la corriente.

**Resultado:** $\boxed{i_R(t) = \pm \, 0{,}2 \, e^{-t/0{,}0001} \text{ A}}$

```python
# Problema 7
R = 1000; WL = 0.002; P0 = 40
i0 = (P0/R)**0.5
L = 2*WL/i0**2
tau = L/R
print(f'i0 = {i0:.1f} A, L = {L:.1f} H, tau = {tau:.4f} s')
```
