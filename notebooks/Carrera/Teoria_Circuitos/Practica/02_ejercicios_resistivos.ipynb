```python
import numpy as np
from fractions import Fraction as F
from IPython.display import Markdown as md

# --- colores para print ---
CYAN   = '\033[96m'
YELLOW = '\033[93m'
GREEN  = '\033[92m'
BOLD   = '\033[1m'
RESET  = '\033[0m'

def code(label, val, unit=''):
    print(f'{CYAN}{label}{RESET} = {BOLD}{val}{RESET} {unit}')
```

## Formulario del Tema 2

| Concepto | Formula |
|---|---|
| Ley de Ohm | $V = R \cdot I$ |
| Resistencias en serie | $R_{eq} = R_1 + R_2 + \cdots + R_n$ |
| Resistencias en paralelo | $\dfrac{1}{R_{eq}} = \dfrac{1}{R_1} + \dfrac{1}{R_2} + \cdots$ |
| Dos en paralelo | $R_{eq} = \dfrac{R_1 \cdot R_2}{R_1 + R_2}$ |
| Divisor de tension | $V_k = V \cdot \dfrac{R_k}{R_1 + R_2}$ |
| Divisor de corriente | $I_k = I \cdot \dfrac{R_{otra}}{R_1 + R_2}$ |
| Potencia | $P = V \cdot I = I^2 R = \dfrac{V^2}{R}$ |
| Superposicion | $V = a_1 \cdot V_{g1} + a_2 \cdot I_{g1} + \cdots$ |

## Problema 1

**Enunciado:** Conocidos los datos de tres circuitos con la misma topologia resistiva, determinar $V_{Rc}$.

**Datos:**
- (a): $I_{ga} = 1$ A, $V_{ga} = 2$ V, $V_{Ra} = 2$ V
- (b): $I_{gb} = 2$ A, $V_{Rb} = 6$ V
- (c): $I_{gc} = 5$ A, $V_{gc} = 4$ V

**Paso 1 - Superposicion lineal:**

Los tres circuitos comparten topologia, por linealidad:

$$V_R = a \cdot I_g + b \cdot V_g$$

**Paso 2 - Coeficientes de (a) y (b):**

Del circuito (b) (sin fuente de tension, $V_g = 0$):

$$6 = a \cdot 2 \implies a = 3$$

Del circuito (a):

$$2 = 3 \cdot 1 + b \cdot 2 \implies b = -\frac{1}{2}$$

**Paso 3 - Calcular $V_{Rc}$:**

$$V_{Rc} = 3 \cdot 5 + \left(-\frac{1}{2}\right) \cdot 4 = 15 - 2 = 13 \text{ V}$$

**Resultado:** $\boxed{V_{Rc} = 13 \text{ V}}$

```python
# Problema 1 - Superposicion lineal
a = 6 / 2
b = (2 - a*1) / 2
V_Rc = a * 5 + b * 4
code('a', a)
code('b', b)
code('V_Rc', f'{V_Rc:.0f}', 'V')
```

## Problema 2

**Enunciado:** CA y CB son circuitos sin fuentes independientes ni acoplados. Hallar la lectura del amperimetro y la potencia de la fuente de tension en (c).

**Datos:**
- (a): CA con 5 V, corriente 1 A
- (b): CB con 8 V, corriente 2 A
- (c): CA y CB en paralelo con 16 V

**Paso 1 - Resistencias equivalentes:**

CA y CB son puramente resistivos:

$$R_{CA} = \frac{5}{1} = 5\;\Omega \qquad R_{CB} = \frac{8}{2} = 4\;\Omega$$

**Paso 2 - Amperimetro (corriente por CA):**

$$A = \frac{16}{R_{CA}} = \frac{16}{5} = 3{,}2 \text{ A}$$

**Paso 3 - Potencia de la fuente:**

$$I_{total} = \frac{16}{5} + \frac{16}{4} = 3{,}2 + 4 = 7{,}2 \text{ A}$$

$$P = 16 \times 7{,}2 = 115{,}2 \text{ W}$$

**Resultado:** $\boxed{A = 3{,}2 \text{ A}, \quad P = 115{,}2 \text{ W}}$

```python
# Problema 2
R_CA = 5 / 1
R_CB = 8 / 2
V = 16
I_A = V / R_CA
P = V * (I_A + V / R_CB)
code('A', f'{I_A:.1f}', 'A')
code('P', f'{P:.1f}', 'W')
```

## Problema 3

**Enunciado:** En un circuito resistivo lineal, con $V_g = 1$ V e $I_g = 2$ A se obtiene $V_1 = 1$ V. Con $V_g = 2$ V e $I_g = 0$ A, tambien $V_1 = 1$ V. Cuanto vale $V_1$ con $I_g = 4$ A y $V_g = 0$?

**Datos:**
- Caso 1: $V_g=1$, $I_g=2$ $\Rightarrow$ $V_1=1$
- Caso 2: $V_g=2$, $I_g=0$ $\Rightarrow$ $V_1=1$
- Caso 3: $V_g=0$, $I_g=4$ $\Rightarrow$ $V_1=?$

**Paso 1 - Superposicion:**

$$V_1 = a \cdot V_g + b \cdot I_g$$

**Paso 2 - Sistema de ecuaciones:**

$$\text{Caso 1: } a + 2b = 1$$
$$\text{Caso 2: } 2a = 1 \implies a = \frac{1}{2}$$

Sustituyendo: $\frac{1}{2} + 2b = 1 \implies b = \frac{1}{4}$

**Paso 3 - Caso 3:**

$$V_1 = \frac{1}{2} \cdot 0 + \frac{1}{4} \cdot 4 = 1 \text{ V}$$

**Resultado:** $\boxed{V_1 = 1 \text{ V}}$

```python
# Problema 3 - Superposicion lineal
a = F(1, 2)
b = (1 - a) / 2
V1 = a * 0 + b * 4
code('a', a)
code('b', b)
code('V1', f'{float(V1):.0f}', 'V')
```

## Problema 4

**Enunciado:** Obtener la resistencia equivalente vista desde la fuente de tension y la potencia que cede.

**Datos:** $V_g = 10$ V, $R_1 = 1$ k$\Omega$, $R_2 = 3$ k$\Omega$.

**Paso 1 - Puente de Wheatstone equilibrado:**

El circuito es un puente con dos ramas identicas ($R_1$-$R_2$ y $R_1$-$R_2$), por lo que esta equilibrado: no circula corriente por la resistencia puente.

**Paso 2 - Resistencia equivalente:**

$$R_{eq} = \frac{(R_1 + R_2)}{2} = \frac{1000 + 3000}{2} = 2000\;\Omega$$

**Paso 3 - Potencia:**

$$P = \frac{V_g^2}{R_{eq}} = \frac{10^2}{2000} = 0{,}05 \text{ W}$$

**Resultado:** $\boxed{R_{eq} = 2000\;\Omega, \quad P = 0{,}05 \text{ W}}$

```python
# Problema 4 - Puente equilibrado
R1, R2, Vg = 1e3, 3e3, 10
Req = (R1 + R2) / 2
P = Vg**2 / Req
code('Req', f'{Req:.0f}', 'ohm')
code('P', f'{P:.2f}', 'W')
```

## Problema 5

**Enunciado:** Un magnetofono presenta $R_{eq} = 10\;\Omega$ y funciona con 4 pilas en serie de 1,5 V y 750 mAh. Cuantas horas de funcionamiento?

**Datos:**
- $R = 10\;\Omega$
- 4 pilas: 1,5 V cada una, 750 mAh

**Paso 1 - Tension total:**

$$V = 4 \times 1{,}5 = 6 \text{ V}$$

**Paso 2 - Corriente consumida:**

$$I = \frac{V}{R} = \frac{6}{10} = 0{,}6 \text{ A} = 600 \text{ mA}$$

**Paso 3 - Tiempo de funcionamiento:**

Pilas en serie: misma corriente, capacidad = 750 mAh:

$$t = \frac{750}{600} = 1{,}25 \text{ h}$$

**Resultado:** $\boxed{t = 1{,}25 \text{ h}}$

```python
# Problema 5
V = 4 * 1.5
I = V / 10
t = 750 / (I * 1000)
code('V total', f'{V}', 'V')
code('t', f'{t:.2f}', 'h')
```

## Problema 9

**Enunciado:** Todas las resistencias son iguales. La potencia de la resistencia enmarcada es 100 W. Hallar la potencia total.

**Datos:**
- $R$ (serie) + 3$R$ (paralelo)
- $P_{enmarcada} = 100$ W

**Paso 1 - Division de corriente por simetria:**

La resistencia enmarcada es una de las tres en paralelo:

$$I_{enmarcada} = \frac{I}{3}$$

**Paso 2 - Potencia de la enmarcada:**

$$P_{enmarcada} = \left(\frac{I}{3}\right)^2 R = \frac{I^2 R}{9} = 100 \implies I^2 R = 900 \text{ W}$$

**Paso 3 - Potencia total:**

$$P_{serie} = I^2 R = 900 \text{ W} \qquad P_{paralelo} = 3 \times 100 = 300 \text{ W}$$

$$P_{total} = 900 + 300 = 1200 \text{ W}$$

**Resultado:** $\boxed{P_{total} = 1200 \text{ W}}$

```python
# Problema 9
I2R = 100 * 9
P_total = I2R + 3 * 100
code('I^2*R', I2R, 'W')
code('P_total', P_total, 'W')
```

## Problema 10

**Enunciado:** Calcular la relacion entre $E$ e $I_g$ para que la potencia de la resistencia enmarcada sea nula.

**Datos:** Tres resistencias de $2\;\Omega$, fuente $E$, fuente de corriente $I_g$.

**Paso 1 - Condicion de potencia nula:**

Para $P = 0$ en la resistencia central, su corriente (o tension) debe ser cero $\Rightarrow V_A = 0$.

**Paso 2 - KCL en el nudo intermedio con $V_A = 0$:**

$$\frac{0 - E}{2} + \frac{0}{2} - I_g = 0$$

$$-\frac{E}{2} - I_g = 0$$

**Paso 3 - Despejamos:**

$$E = -2 I_g$$

**Resultado:** $\boxed{E = -2\,I_g}$

```python
# Problema 10
print(f'{CYAN}Relacion:{RESET} E = -2 * Ig')
Ig = 3
E = -2 * Ig
code('Ejemplo: Ig=3 => E', E, 'V')
```
