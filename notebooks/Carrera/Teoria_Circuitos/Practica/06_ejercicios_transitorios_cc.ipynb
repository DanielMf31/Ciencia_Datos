```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Markdown as md

# --- colores para print ---
CYAN   = '\033[96m'
YELLOW = '\033[93m'
GREEN  = '\033[92m'
BOLD   = '\033[1m'
RESET  = '\033[0m'

def code(label, val, unit=''):
    print(f'{CYAN}{label}{RESET} = {BOLD}{val}{RESET} {unit}')
```

## Formulario del Tema 6

| Magnitud | Formula |
|---|---|
| Bobina en CC permanente | Cortocircuito ($v_L=0$) |
| Condensador en CC permanente | Circuito abierto ($i_C=0$) |
| Constante de tiempo RC | $\tau = RC$ |
| Constante de tiempo RL | $\tau = L/R$ |
| Respuesta transitoria | $x(t) = x_{\infty} + (x_0 - x_{\infty})\,e^{-t/\tau}$ |
| Energia en condensador | $E_C = \frac{1}{2}Cv_C^2$ |
| Energia en bobina | $E_L = \frac{1}{2}Li_L^2$ |
| Potencia | $p = vi$ |
| Rendimiento FRT | $\eta = v_{carga}/E_g$ |

## Problema 1

**Enunciado:** El circuito se encuentra en regimen permanente. Determinar $V_{AB}$ y la potencia que cede $I_2$.

**Datos:** $I_1=2$ A, $I_2=4$ A, $R_1=10\ \Omega$, $R_2=5\ \Omega$, $R_3=2\ \Omega$, bobina $L$.

**Paso 1:** En CC permanente, $L$ es cortocircuito. Se unen los nudos superiores en T.

**Paso 2:** Ecuaciones de nudos ($V_B=0$):

$$3V_A - V_T = 20 \quad (1)$$
$$-V_A + 6V_T = 40 \quad (2)$$

**Paso 3:** De (1): $V_T = 3V_A-20$. En (2): $17V_A = 160 \implies V_A = 9{,}41$ V.

**Paso 4:** $V_T = 8{,}24$ V. $P_{I_2} = V_T \cdot I_2 = 32{,}94$ W.

**Resultado:** $\boxed{V_{AB}=9{,}41 \text{ V};\; P_{I_2}=32{,}94 \text{ W}}$

```python
# Problema 1
A = np.array([[3,-1],[-1,6]]); b = np.array([20,40])
VA, VT = np.linalg.solve(A, b)
code('V_AB', f'{VA:.2f}', 'V')
code('P_I2', f'{VT*4:.2f}', 'W')
```

## Problema 2

**Enunciado:** En regimen permanente el circuito almacena $1/3$ J y disipa 2 W. Obtener $R_a$ y $R_b$.

**Datos:** $V_g=2$ V, $I_g=1$ A, $L_1=1$ H, $L_2=2$ H, $C=1/2$ F, $R=3\ \Omega$, $E=1/3$ J, $P=2$ W.

**Paso 1:** CC permanente: $L\to$ cable, $C\to$ abierto. Rama $L_2$-$C$ no conduce. $i_1 = 2/R_a$, $i_{R_b}=1$ A.

**Paso 2:** Energia: $E = \frac{1}{2}L_1 i_1^2 + \frac{1}{2}C v_C^2 = \frac{2}{R_a^2} + \frac{R_b^2}{4} = 1/3$.

**Paso 3:** Potencia disipada: $P = \frac{4}{R_a} + R_b = 2 \implies R_b = 2-4/R_a$.

**Paso 4:** Sustituyendo: $(R_a-3)^2=0 \implies R_a=3\ \Omega$, $R_b=2/3\ \Omega$.

**Resultado:** $\boxed{R_a=3\ \Omega;\; R_b=2/3\ \Omega}$

```python
# Problema 2
Ra = 3.0; Rb = 2 - 4/Ra
code('Ra', f'{Ra:.0f}', 'Ohm')
code('Rb', f'{Rb:.4f} = 2/3', 'Ohm')
```

## Problema 3

**Enunciado:** Condensador conectado a FRT. En $t=0$ se conoce la energia almacenada y el rendimiento. Calcular la energia en regimen permanente.

**Datos:** $E(0)=200$ J, $\eta(0)=25\%$.

**Paso 1:** Rendimiento: $\eta = v_C/E_g = 0{,}25 \implies E_g = 4\,v_C(0)$.

**Paso 2:** En permanente: $v_C(\infty) = E_g = 4\,v_C(0)$.

**Paso 3:** $E(\infty) = \frac{1}{2}C\,v_C(\infty)^2 = \frac{1}{2}C\,(4\,v_C(0))^2 = 16\,E(0) = 3200$ J.

**Resultado:** $\boxed{E(\infty)=3200 \text{ J}}$

```python
# Problema 3
E_inf = 200 / 0.25**2
code('E(inf)', f'{E_inf:.0f}', 'J')
```

## Problema 4

**Enunciado:** CC permanente. Hallar lecturas del voltimetro y amperimetro.

**Datos:** $V_g=4$ V, $R_1=R_2=R_3=2\ \Omega$; $I_g$, $R$, $L_1$, $L_2$ desconocidas.

**Paso 1:** En CC permanente: $L_1,L_2\to$ cables. $L_1$ cortocircuita $R_3$.

**Paso 2:** Circuito equivalente: $V_g$ en serie con $R_1$ y $R_2$.

$$i = \frac{V_g}{R_1+R_2} = \frac{4}{2+2} = 1 \text{ A}$$

**Paso 3:** Voltimetro mide $R_2$: $V = R_2 \cdot i = 2$ V. Amperimetro mide corriente por $L_2$: $A = 1$ A.

**Resultado:** $\boxed{V=2 \text{ V};\; A=1 \text{ A}}$

```python
# Problema 4
i = 4 / (2 + 2)
code('Voltimetro', f'{2*i:.0f}', 'V')
code('Amperimetro', f'{i:.0f}', 'A')
```

## Problema 5

**Enunciado:** FRT con $\eta=0{,}2$ y $E_L=2$ J. Encontrar $V_{ca}$ e $I_{cc}$.

**Datos:** $L=1$ H, $R=2\ \Omega$, $\eta=0{,}2$, $E_L=2$ J.

**Paso 1:** Corriente por la bobina: $i = \sqrt{2E_L/L} = \sqrt{4} = 2$ A.

**Paso 2:** Rendimiento: $\eta = \frac{Ri}{V_{ca}} \implies V_{ca} = \frac{Ri}{\eta} = \frac{2 \cdot 2}{0{,}2} = 20$ V.

**Paso 3:** $R_s = V_{ca}/i - R = 10 - 2 = 8\ \Omega$. $I_{cc} = V_{ca}/R_s = 20/8 = 2{,}5$ A.

**Resultado:** $\boxed{V_{ca}=20 \text{ V};\; I_{cc}=2{,}5 \text{ A}}$

```python
# Problema 5
i = np.sqrt(4); Vca = 2*i/0.2; Rs = Vca/i - 2; Icc = Vca/Rs
code('Vca', f'{Vca:.0f}', 'V')
code('Icc', f'{Icc:.1f}', 'A')
```

## Problema 6

**Enunciado:** Se abre el interruptor. Calcular la potencia cedida por el condensador y por la bobina justo despues de abrir.

**Datos:** $I_g=8$ A, $R=3\ \Omega$, $L_1=1$ H, $C=2$ F.

**Paso 1:** CC permanente ($t<0$): $L\to$ cables, $C\to$ abierto. $i_{L_1}=I_g=8$ A.

**Paso 2:** $v_C(0^-)$: toda la corriente pasa por las bobinas. $v_C = I_g \cdot R = 24$ V (entre ramas).

**Paso 3:** En $0^+$ se abre: las condiciones iniciales se mantienen. Del analisis del nuevo circuito:

$$P_C = v_C \cdot i_C = -48 \text{ W}, \quad P_L = 0 \text{ W}$$

**Resultado:** $\boxed{P_C=-48 \text{ W};\; P_L=0 \text{ W}}$

```python
# Problema 6
code('P_C(0+)', '-48', 'W')
code('P_L(0+)', '0', 'W')
```

## Problema 7

**Enunciado:** Se cierra el interruptor en $t=0$. Calcular $P_{L_1}(0^+)$ y la energia almacenada en las bobinas y condensador en el nuevo permanente.

**Datos:** $L_1=1$ H, $L_2=2$ H, $R=2\ \Omega$, $V_g=6$ V, $I_g=3$ A.

**Paso 1:** Antes del cierre: $i_{L_1}(0^-)=I_g=3$ A, $i_{L_2}(0^-)=0$.

**Paso 2:** En $0^+$: $i_R=i_{L_1}-i_{L_2}=3-0=3$ A pero $v_{L_1}=-V_g=-6$ V.

$$P_{L_1}^{cedida} = -v_{L_1} \cdot i_{L_1} = -6 \times 3 = -18 \text{ W}$$

**Paso 3:** Nuevo permanente: $i_{L_1}=I_g=3$ A, $i_{L_2}=V_g/R=3$ A.

$$E_{L_1}=\frac{1}{2}(1)(3)^2=4{,}5 \text{ J}, \quad E_{L_2}=\frac{1}{2}(2)(3)^2=9 \text{ J}, \quad E_C=0 \text{ J}$$

**Resultado:** $\boxed{P=-18 \text{ W};\; E_{L_1}=4{,}5 \text{ J};\; E_{L_2}=9 \text{ J};\; E_C=0 \text{ J}}$

```python
# Problema 7
L1,L2,R,Vg,Ig = 1,2,2,6,3
code('P_L1(0+)', f'{-Vg*Ig:.0f}', 'W')
code('Energias', f'E_L1={0.5*L1*Ig**2:.1f} J, E_L2={0.5*L2*(Vg/R)**2:.1f} J, E_C=0 J', '')
```

## Problema 8

**Enunciado:** Encontrar la expresion de la intensidad $i$ en la resistencia vertical, sabiendo que la bobina tiene condicion inicial nula.

**Datos:** $I_g=4$ A, $R=4\ \Omega$, $L=1$ H, $i_L(0)=0$.

**Paso 1:** Permanente ($t\to\infty$): $L\to$ cable. $R\|R = 2\ \Omega$. Divisor de corriente:

$$i(\infty) = I_g \cdot \frac{R}{R+R} = 4 \cdot \frac{1}{2} = 2 \text{ A}$$

**Paso 2:** Inicial ($L$ abierta): toda la corriente pasa por la resistencia vertical: $i(0)=I_g=4$ A.

**Paso 3:** $R_{Th}$ vista desde $L$: $R_{Th}=R+R=8\ \Omega$. $\tau=L/R_{Th}=1/8$ s.

$$i(t) = i(\infty) + (i(0)-i(\infty))\,e^{-t/\tau} = 2 + 2\,e^{-8t} \text{ A}$$

**Resultado:** $\boxed{i(t)=2+2\,e^{-8t} \text{ A}}$

```python
# Problema 8
t = np.linspace(0, 0.8, 500); i_t = 2 + 2*np.exp(-8*t)
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(t, i_t, '#2171b5', lw=2.5, label=r'$i(t)=2+2e^{-8t}$ A')
ax.axhline(y=2, color='#cb181d', ls='--', lw=1.5, alpha=0.7, label=r'$i(\infty)=2$ A')
ax.plot(0, 4, 'o', color='#238b45', ms=10, zorder=5)
ax.set_xlabel('t (s)'); ax.set_ylabel('i (A)'); ax.set_title('Problema 8: Transitorio RL')
ax.legend(); ax.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
```

## Problema 9

**Enunciado:** El circuito esta en regimen permanente cuando en $t=0$ se abre el interruptor. Determinar $i_1(t)$ e $i_L(t)$.

**Datos:** $120, 60, 90, 50\ \Omega$; $1, 2, 3$ H; $18$ V.

**Paso 1:** Antes de abrir: $i_{fuente} = 18/(120+60\|90) = 18/(120+36) = 0{,}1154$ A. Divisor: $i_L(0)=i_{fuente}\cdot 60/150 \cdot ...$

**Paso 2:** Al abrir el interruptor, la fuente se desconecta. Transitorio puro.

**Paso 3:** $L_{eq} = 1 + 2\|3 = 1 + 1{,}2 = 2{,}2$ H. $R_{eq} = 60 + 50 = 110\ \Omega$. $\tau = L_{eq}/R_{eq} = 0{,}02$ s.

$$i_1(t) = -0{,}24\,e^{-t/0{,}02} \text{ A}, \quad i_L(t) = 0{,}36\,e^{-t/0{,}02} \text{ A}$$

**Resultado:** $\boxed{i_1(t)=-0{,}24\,e^{-t/0{,}02} \text{ A};\; i_L(t)=0{,}36\,e^{-t/0{,}02} \text{ A}}$

```python
# Problema 9
tau = 0.02; t = np.linspace(0, 0.12, 500)
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(t*1e3, -0.24*np.exp(-t/tau), '#2171b5', lw=2.5, label=r'$i_1(t)=-0.24\,e^{-t/0.02}$')
ax.plot(t*1e3, 0.36*np.exp(-t/tau), '#cb181d', lw=2.5, label=r'$i_L(t)=0.36\,e^{-t/0.02}$')
ax.set_xlabel('t (ms)'); ax.set_ylabel('i (A)'); ax.set_title('Problema 9: Transitorio RL')
ax.legend(); ax.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
```

## Problema 10

**Enunciado:** Encontrar la expresion de $i(t)$ sabiendo que la tension inicial del condensador de 10 mF es 2 V.

**Datos:** $v_g=11$ V, $i_g=3$ A, $C=10$ mF, $v_C(0)=2$ V, $R_1=3\ \Omega$, $R_2=1\ \Omega$, $R_3=2\ \Omega$.

**Paso 1:** $R_{Th}$ vista desde $C$: apagar fuentes. $R_1\|(R_2 + R_3\|...) \to R_{Th} = 3/4\ \Omega$.

$$\tau = R_{Th} \cdot C = \frac{3}{4} \cdot 0{,}01 = 0{,}0075 \text{ s}$$

**Paso 2:** Permanente: $v_C(\infty) \to i(\infty) = 2$ A. Inicial: $i(0) = (v_g - v_C(0))/R_1 = (11-2)/3 = 3$ A.

$$i(t) = i(\infty) + (i(0)-i(\infty))\,e^{-t/\tau} = 2 + e^{-t/0{,}0075} \text{ A}$$

**Resultado:** $\boxed{i(t)=2+e^{-t/0{,}0075} \text{ A}}$

```python
# Problema 10
tau = 0.0075; t = np.linspace(0, 0.05, 500); i_t = 2 + np.exp(-t/tau)
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(t*1e3, i_t, '#2171b5', lw=2.5, label=r'$i(t)=2+e^{-t/0.0075}$ A')
ax.axhline(y=2, color='#cb181d', ls='--', lw=1.5, alpha=0.7, label=r'$i(\infty)=2$ A')
ax.plot(0, 3, 'o', color='#238b45', ms=10, zorder=5)
ax.set_xlabel('t (ms)'); ax.set_ylabel('i (A)'); ax.set_title('Problema 10: Transitorio RC')
ax.legend(); ax.grid(True, alpha=0.3); plt.tight_layout(); plt.show()
```
