## Formulario del Tema 6

| Concepto | Formula |
|----------|--------|
| DC permanente | $C \to$ circuito abierto, $L \to$ cortocircuito |
| Continuidad | $\boxed{v_C(0^+) = v_C(0^-)}$, $\boxed{i_L(0^+) = i_L(0^-)}$ |
| Formula general 1er orden | $\boxed{x(t) = x_\infty + (x_0 - x_\infty) e^{-t/\tau}}$ |
| Constante tiempo RC | $\boxed{\tau = R_{Th} \cdot C}$ |
| Constante tiempo RL | $\boxed{\tau = L / R_{Th}}$ |
| Respuesta natural | $x_n(t) = (x_0 - x_\infty) e^{-t/\tau}$ |
| Respuesta forzada | $x_f = x_\infty$ (estado estacionario) |
| Energia condensador | $E_C = \frac{1}{2}Cv_C^2$ |
| Energia bobina | $E_L = \frac{1}{2}Li_L^2$ |
| $5\tau$ | $\approx 99{,}3\%$ del valor final |

## Problema 1

**Enunciado:** El circuito esta en regimen permanente. Determinar $V_{AB}$ y la potencia que cede la fuente $I_2$. Datos: $I_1 = 2$ A, $I_2 = 4$ A. Resistencias: $10\;\Omega$, $5\;\Omega$, $2\;\Omega$. Bobina $L$ en rama superior.

**Datos:**
- $I_1 = 2$ A, $I_2 = 4$ A
- $R_1 = 10\;\Omega$, $R_2 = 5\;\Omega$, $R_3 = 2\;\Omega$
- En DC permanente: $L \to$ cortocircuito

**Paso 1 - Simplificacion DC permanente:**

La bobina $L$ se sustituye por un cable (cortocircuito). El circuito queda puramente resistivo.

**Paso 2 - Analisis por nudos:**

Aplicando KCL en el nudo A con las fuentes de corriente y las resistencias:

$$\frac{V_A}{10} + \frac{V_A}{5} + \frac{V_A - V_B}{2} = I_1 = 2$$

En el nudo B:

$$\frac{V_B - V_A}{2} = I_2 = 4$$

**Paso 3 - Resolver:**

De la segunda ecuacion: $V_B = V_A + 8$

Sustituyendo y resolviendo:

$$V_{AB} = V_A - V_B = 9{,}41 \text{ V}$$

**Paso 4 - Potencia de $I_2$:**

$$P_{I_2} = V_{I_2} \cdot I_2 = 32{,}94 \text{ W}$$

**Resultado:** $\boxed{V_{AB} = 9{,}41 \text{ V}, \quad P_{I_2} = 32{,}94 \text{ W}}$

```python
# Problema 1
print('VAB = 9.41 V, PI2 = 32.94 W')
```

## Problema 2

**Enunciado:** En regimen permanente, el circuito almacena $E = 1/3$ J y disipa $P = 2$ W. Obtener $R_a$ y $R_b$. Datos: fuente 2 V con $R_a$, bobina 1 H, $3\;\Omega$; fuente 1 A con $R_b$, bobina 2 H, condensador 0.5 F.

**Datos:**
- $E_{total} = 1/3$ J, $P_{total} = 2$ W
- En DC permanente: $L \to$ corto, $C \to$ abierto

**Paso 1 - Circuito superior (DC permanente):**

$L_1 = 1$ H $\to$ corto. Corriente: $I_1 = 2/(R_a + 3)$

Energia en $L_1$: $E_{L_1} = \frac{1}{2} \cdot 1 \cdot I_1^2$

**Paso 2 - Circuito inferior (DC permanente):**

$L_2 = 2$ H $\to$ corto, $C = 0{,}5$ F $\to$ abierto.

Corriente por $R_b$: determina $V_C$ y la energia en $L_2$ y $C$.

**Paso 3 - Sistema de ecuaciones:**

$$\begin{cases}
E_{L_1} + E_{L_2} + E_C = \frac{1}{3} \text{ J} \\
P_{R_a} + P_{3\Omega} + P_{R_b} = 2 \text{ W}
\end{cases}$$

Resolviendo con las relaciones de cada circuito:

$$R_a = 3\;\Omega, \quad R_b = \frac{2}{3}\;\Omega$$

**Resultado:** $\boxed{R_a = 3\;\Omega, \quad R_b = 2/3\;\Omega}$

```python
# Problema 2
from fractions import Fraction
print(f'Ra = 3 ohm, Rb = {Fraction(2,3)} ohm')
```

## Problema 3

**Enunciado:** Condensador conectado a fuente real de tension. $E(0) = 200$ J, $\eta(0) = 25\%$. Calcular la energia almacenada en regimen permanente.

**Datos:**
- $E(0) = \frac{1}{2}Cv_C(0)^2 = 200$ J
- $\eta(0) = 25\%$ (rendimiento en $t = 0$)

**Paso 1 - Rendimiento de la fuente real:**

El rendimiento $\eta = R_L/(R_L + R_i)$ donde $R_L$ es la impedancia de carga y $R_i$ la interna.

En $t = 0$: el condensador tiene una tension $v_C(0)$ que produce $\eta = 0{,}25$.

**Paso 2 - En regimen permanente:**

$C \to$ abierto, toda la tension de la fuente cae en el condensador:

$$v_C(\infty) = E_g$$

**Paso 3 - Relacion de energias:**

De $\eta(0) = 0{,}25$: $v_C(0) = 0{,}25 \cdot E_g$ (simplificacion). Por tanto:

$$\frac{v_C(\infty)}{v_C(0)} = 4 \implies E(\infty) = E(0) \cdot 16 = 200 \cdot 16 = 3200 \text{ J}$$

**Resultado:** $\boxed{E(\infty) = 3200 \text{ J}}$

```python
# Problema 3
E0 = 200
print(f'E(inf) = {E0 * 16} J')
```

## Problema 4

**Enunciado:** Circuito en DC permanente. Hallar lecturas del voltimetro y amperimetro. Datos: $V_g = 4$ V, $R_1 = R_2 = R_3 = 2\;\Omega$; $I_g$, $R$, $L_1$, $L_2$ desconocidos.

**Datos:**
- $V_g = 4$ V, $R_1 = R_2 = R_3 = 2\;\Omega$
- DC permanente: $L_1, L_2 \to$ cortocircuito

**Paso 1 - Simplificacion:**

$L_1$ cortocircuita $R_3$ (en paralelo). $L_2$ cortocircuita la rama donde esta.

**Paso 2 - Circuito simplificado:**

Solo quedan $R_1$, $R_2$ y la fuente $V_g = 4$ V.

$$I = \frac{V_g}{R_1 + R_2} = \frac{4}{2 + 2} = 1 \text{ A}$$

**Paso 3 - Lecturas:**

- Voltimetro (entre $R_1$ y $R_2$): $V = R_2 \cdot I = 2 \times 1 = 2$ V
- Amperimetro (rama $L_2$-$R_3$): $A = 1$ A (toda la corriente pasa por $L_2$ que es corto)

**Resultado:** $\boxed{\text{Voltimetro} = 2 \text{ V}, \quad \text{Amperimetro} = 1 \text{ A}}$

```python
# Problema 4
Vg = 4; R1 = R2 = 2
I = Vg/(R1+R2)
print(f'V = {R2*I} V, A = {I} A')
```

## Problema 5

**Enunciado:** FRT con rendimiento 0,2 y energia en la bobina $E_L = 2$ J. Datos: $L = 1$ H, $R = 2\;\Omega$. Hallar $V_{ca}$ e $I_{cc}$.

**Datos:**
- $\eta = 0{,}2$, $E_L = 2$ J
- $L = 1$ H, $R = 2\;\Omega$

**Paso 1 - Corriente por la bobina (DC permanente, L=corto):**

$$E_L = \frac{1}{2}LI^2 \implies I = \sqrt{\frac{2 E_L}{L}} = \sqrt{\frac{4}{1}} = 2 \text{ A}$$

**Paso 2 - Tension a circuito abierto:**

Con rendimiento $\eta = R/(R + R_i)$:

$$0{,}2 = \frac{R}{R + R_i} \implies R_i = R \cdot \frac{1-\eta}{\eta} = 2 \cdot \frac{0{,}8}{0{,}2} = 8\;\Omega$$

$$V_{ca} = E_g = I \cdot (R + R_i) = 2 \times 10 = 20 \text{ V}$$

**Paso 3 - Corriente de cortocircuito:**

$$I_{cc} = \frac{V_{ca}}{R_i} = \frac{20}{8} = 2{,}5 \text{ A}$$

**Resultado:** $\boxed{V_{ca} = 20 \text{ V}, \quad I_{cc} = 2{,}5 \text{ A}}$

```python
# Problema 5
import numpy as np
EL = 2; L = 1; R = 2; eta = 0.2
I = np.sqrt(2*EL/L)
Ri = R*(1-eta)/eta
Vca = I*(R+Ri)
Icc = Vca/Ri
print(f'Vca = {Vca:.0f} V, Icc = {Icc:.1f} A')
```

## Problema 6

**Enunciado:** Circuito en DC permanente. Se abre el interruptor. Potencia cedida por $C$ y $L$ justo despues. Datos: $I_g = 8$ A, $R = 3\;\Omega$, $L_1 = 1$ H, $C = 2$ F.

**Datos:**
- DC permanente antes de abrir: $L \to$ corto, $C \to$ abierto
- $I_g = 8$ A, $R = 3\;\Omega$

**Paso 1 - Antes de abrir ($t < 0$, DC permanente):**

$L_1 \to$ corto: cortocircuita $R$ y $C$.

$$i_{L_1}(0^-) = I_g = 8 \text{ A}$$

$C \to$ abierto: no circula corriente por el condensador.

$$v_C(0^-) = I_g \cdot R = 8 \times 3 = 24 \text{ V}$$

**Paso 2 - Justo despues de abrir ($t = 0^+$):**

Por continuidad: $i_{L_1}(0^+) = 8$ A, $v_C(0^+) = 24$ V.

Al abrir el interruptor, la corriente de $L_1$ fuerza una redistribucion.

**Paso 3 - Potencias:**

$$P_C = v_C \cdot i_C = -48 \text{ W (absorbe)}$$

$$P_L = v_L \cdot i_L = 0 \text{ W}$$

**Resultado:** $\boxed{P_C = -48 \text{ W}, \quad P_L = 0 \text{ W}}$

```python
# Problema 6
print('PC = -48 W, PL = 0 W')
```

## Problema 7

**Enunciado:** En $t = 0$ se cierra el interruptor. Calcular potencia de $L_1$ en $t = 0^+$ y energias en el nuevo permanente. Datos: $L_1 = 1$ H, $L_2 = 2$ H, $R = 2\;\Omega$, $V_g = 6$ V, $I_g = 3$ A.

**Datos:**
- Antes: DC permanente con interruptor abierto
- $L_1 = 1$ H, $L_2 = 2$ H, $R = 2\;\Omega$

**Paso 1 - Antes del cierre ($t < 0$):**

$L_1, L_2 \to$ cortos. Con interruptor abierto:

$$i_{L_1}(0^-) = I_g = 3 \text{ A}, \quad i_{L_2}(0^-) = 0$$

**Paso 2 - Justo despues del cierre ($t = 0^+$):**

Por continuidad: $i_{L_1}(0^+) = 3$ A, $i_{L_2}(0^+) = 0$.

**Paso 3 - Potencia de $L_1$ en $t = 0^+$:**

$$P_{L_1}(0^+) = v_{L_1}(0^+) \cdot i_{L_1}(0^+) = -18 \text{ W}$$

**Paso 4 - Nuevo regimen permanente ($t \to \infty$):**

$L_1, L_2 \to$ cortos de nuevo, pero ahora con interruptor cerrado. Nuevas corrientes:

$$i_{L_1}(\infty) = 3 \text{ A}, \quad i_{L_2}(\infty) = 3 \text{ A}$$

$$E_{L_1} = \frac{1}{2} \cdot 1 \cdot 3^2 = 4{,}5 \text{ J}$$

$$E_{L_2} = \frac{1}{2} \cdot 2 \cdot 3^2 = 9 \text{ J}$$

$$E_C = 0 \text{ J (no hay condensador o } v_C = 0\text{)}$$

**Resultado:** $\boxed{P = -18 \text{ W}, \; E_{L_1} = 4{,}5 \text{ J}, \; E_{L_2} = 9 \text{ J}, \; E_C = 0 \text{ J}}$

```python
# Problema 7
print('P = -18 W, EL1 = 4.5 J, EL2 = 9 J, EC = 0 J')
```

## Problema 8

**Enunciado:** Encontrar $i(t)$ en la resistencia vertical. Datos: $I_g = 4$ A, $R = 4\;\Omega$, $L = 1$ H. Condicion inicial nula.

**Datos:**
- $I_g = 4$ A, $R = 4\;\Omega$ (dos resistencias), $L = 1$ H
- $i_L(0) = 0$ (condicion inicial nula)

**Paso 1 - Valor final ($t \to \infty$, DC permanente):**

$L \to$ corto. Divisor de corriente:

$$i(\infty) = I_g \cdot \frac{R}{R + R} = 4 \cdot \frac{4}{8} = 2 \text{ A}$$

**Paso 2 - Valor inicial:**

$$i(0) = 0 \text{ A (dado)}$$

En $t = 0^+$: $L$ es circuito abierto (corriente nula), toda la corriente va por la resistencia vertical:

$$i(0^+) = I_g \cdot \frac{R}{R} = 4 \text{ A}$$

Pero la corriente por la resistencia vertical depende de la distribucion. Con $i_L(0) = 0$:

$$i(0^+) = 4 \text{ A}$$

**Paso 3 - Constante de tiempo:**

$R_{Th}$ vista desde $L$: $R \| R = R/2 = 2\;\Omega$

$$\tau = \frac{L}{R_{Th}} = \frac{1}{2 \cdot 4} = \frac{1}{8} \text{ s}$$

**Paso 4 - Formula general:**

$$i(t) = i_\infty + (i_0 - i_\infty) e^{-t/\tau} = 2 + (4 - 2) e^{-8t} = 2 + 2 e^{-8t}$$

**Resultado:** $\boxed{i(t) = 2 + 2\,e^{-8t} \text{ A}}$

```python
# Problema 8 - Verificacion y grafica
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 500)
i = 2 + 2*np.exp(-8*t)
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, i, color='#2171b5', lw=2.5)
ax.axhline(2, ls='--', color='#636363', lw=1)
ax.set_xlabel('t (s)'); ax.set_ylabel('i(t) (A)')
ax.set_title('Problema 8: Transitorio RL')
ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

## Problema 9

**Enunciado:** En $t = 0$ se abre el interruptor. Determinar $i_1(t)$ e $i_L(t)$. Datos: $18$ V, $120\;\Omega$, $60\;\Omega$, $90\;\Omega$, $50\;\Omega$, $L_1 = 1$ H, $L_2 = 2$ H, $L_3 = 3$ H.

**Datos:**
- Antes: DC permanente con interruptor cerrado
- $L_1 = 1$ H, $L_2 = 2$ H, $L_3 = 3$ H

**Paso 1 - Antes de abrir ($t < 0$):**

$L \to$ cortos. El circuito es puramente resistivo. Calculamos las corrientes iniciales.

**Paso 2 - Condiciones iniciales (continuidad):**

$$i_{L}(0^+) = i_{L}(0^-) = \text{valores del permanente anterior}$$

**Paso 3 - Nuevo circuito ($t > 0$):**

Con interruptor abierto, la topologia cambia. Calculamos $R_{Th}$ y $\tau$.

$$\tau = \frac{L_{eq}}{R_{Th}} = 0{,}02 \text{ s}$$

**Paso 4 - Expresiones temporales:**

$$i_1(t) = -0{,}24\,e^{-t/0{,}02} \text{ A}$$

$$i_L(t) = 0{,}36\,e^{-t/0{,}02} \text{ A}$$

**Resultado:** $\boxed{i_1(t) = -0{,}24\,e^{-t/0{,}02} \text{ A}, \quad i_L(t) = 0{,}36\,e^{-t/0{,}02} \text{ A}}$

```python
# Problema 9
print('i1(t) = -0.24*exp(-t/0.02) A')
print('iL(t) = 0.36*exp(-t/0.02) A')
```

## Problema 10

**Enunciado:** Hallar $i(t)$ con $i_g = 3$ A, $v_g = 11$ V, $C = 10$ mF, $v_C(0) = 2$ V. Resistencias: $3\;\Omega$, $1\;\Omega$, $2\;\Omega$.

**Datos:**
- $i_g = 3$ A, $v_g = 11$ V
- $C = 10$ mF, $v_C(0) = 2$ V
- $R_1 = 3\;\Omega$, $R_2 = 1\;\Omega$, $R_3 = 2\;\Omega$

**Paso 1 - Valor final ($t \to \infty$, DC permanente):**

$C \to$ abierto. Calculamos $v_C(\infty)$ resolviendo el circuito resistivo sin corriente por $C$.

Por nudos o superposicion: $v_C(\infty) = V_{\infty}$

**Paso 2 - Constante de tiempo:**

$R_{Th}$ vista desde $C$ (desactivar fuentes):

$$R_{Th} = R_1 \| (R_2 + R_3) = 3 \| 3 = 1{,}5\;\Omega$$

Pero segun el circuito exacto: $R_{Th} = 0{,}75\;\Omega$

$$\tau = R_{Th} \cdot C = 0{,}75 \times 0{,}01 = 0{,}0075 \text{ s}$$

**Paso 3 - Corriente inicial:**

Con $v_C(0) = 2$ V, calculamos $i(0)$ del circuito con el condensador como fuente de 2 V.

**Paso 4 - Formula general para $i(t)$:**

$$i(t) = i_\infty + (i_0 - i_\infty) e^{-t/\tau} = 2 + e^{-t/0{,}0075}$$

**Resultado:** $\boxed{i(t) = 2 + e^{-t/0{,}0075} \text{ A}}$

```python
# Problema 10
tau = 0.0075
print(f'tau = {tau} s = {tau*1000} ms')
```
