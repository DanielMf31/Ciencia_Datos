## 3. Generadores ideales

Un **generador ideal** es un modelo matemático que impone una magnitud eléctrica (tensión o corriente) de forma independiente del circuito al que está conectado.

### 3.1 Fuente ideal de tensión

Una fuente ideal de tensión **especifica** la tensión $u_g(t)$ entre sus terminales, **independientemente** del circuito conectado.

- La **corriente** $i(t)$ que la atraviesa depende del circuito externo
- La **potencia** $p = u_g \cdot i$ puede ser positiva (genera) o negativa (absorbe)

**Símbolos:** genérico, DC ($=$), AC ($\sim$)

**Conexiones prohibidas:**
- Dos fuentes de tensión en **paralelo** con valores distintos $\to$ **contradicción** (¿cuál impone la tensión?)
- **Cortocircuitar** una fuente de tensión $\to$ corriente infinita (prohibido)

$$\boxed{u_{\text{terminales}} = u_g(t) \quad \text{(independiente de } i \text{)}}$$

```python
# Fuente de tensión ideal y conexión incompatible en paralelo
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Fuente ideal de tensión
axes[0].set_title('Fuente ideal de tensión', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[0])
d1 += elm.SourceV().up().label(r'$u_g(t)$', loc='left')
d1 += elm.Line().right().length(2)
d1 += elm.Resistor().down().label(r'$R$', loc='right')
d1 += elm.Line().left().length(2)
d1 += elm.CurrentLabelInline(direction='in').at(d1.elements[1]).label(r'$i(t)$')
d1.draw()

# Conexión incompatible: dos fuentes en paralelo
axes[1].set_title('INCOMPATIBLE: fuentes en paralelo distintas', fontsize=14, fontweight='bold', color=COLOR_RECTA)
d2 = schemdraw.Drawing(canvas=axes[1])
d2 += elm.SourceV().up().label(r'$V_1$', loc='left')
top2 = d2.here
d2 += elm.Line().right().length(3)
d2 += elm.SourceV().down().label(r'$V_2 \neq V_1$', loc='right').reverse()
d2 += elm.Line().left().length(3)
d2.draw()

plt.tight_layout()
plt.show()
```

### 3.2 Fuente ideal de corriente

Una fuente ideal de corriente **especifica** la corriente $i_g(t)$ que la atraviesa, **independientemente** del circuito conectado.

- La **tensión** $u(t)$ entre sus terminales depende del circuito externo
- La **potencia** $p = u \cdot i_g$ puede ser positiva o negativa

**Conexiones prohibidas:**
- Dos fuentes de corriente en **serie** con valores distintos $\to$ **contradicción**
- **Abrir** el circuito de una fuente de corriente $\to$ tensión infinita (prohibido)

$$\boxed{i_{\text{fuente}} = i_g(t) \quad \text{(independiente de } u \text{)}}$$

```python
# Fuente de corriente ideal y conexión incompatible en serie
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Fuente ideal de corriente
axes[0].set_title('Fuente ideal de corriente', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[0])
d1 += elm.SourceI().up().label(r'$i_g(t)$', loc='left')
d1 += elm.Line().right().length(2)
d1 += elm.Resistor().down().label(r'$R$', loc='right')
d1 += elm.Line().left().length(2)
d1.draw()

# Conexión incompatible: dos fuentes de corriente en serie
axes[1].set_title('INCOMPATIBLE: fuentes de corriente en serie', fontsize=14, fontweight='bold', color=COLOR_RECTA)
d2 = schemdraw.Drawing(canvas=axes[1])
d2 += elm.SourceI().up().label(r'$I_1$', loc='left')
d2 += elm.Line().right().length(2)
d2 += elm.SourceI().down().label(r'$I_2 \neq I_1$', loc='right').reverse()
d2 += elm.Line().left().length(2)
d2.draw()

plt.tight_layout()
plt.show()
```

### 3.3 Principio de superposición

Para circuitos lineales con múltiples fuentes independientes, la respuesta total es la **suma** de las respuestas individuales debidas a cada fuente actuando sola.

$$\boxed{u(t) = u_1(t) + u_2(t) + \cdots \quad;\quad i(t) = i_1(t) + i_2(t) + \cdots}$$

**Cómo "apagar" cada fuente:**
- Fuente de tensión $\to$ **cortocircuito** (sustituir por un cable, $u_g = 0$)
- Fuente de corriente $\to$ **circuito abierto** (eliminar la rama, $i_g = 0$)

> **Truco para el examen:** La superposición es útil cuando hay muchas fuentes y el circuito es complicado. Resuelve un circuito simple por cada fuente y suma.

## 4. Resistencia y Ley de Ohm

La **resistencia** es un elemento pasivo que relaciona tensión y corriente de forma lineal.

$$\boxed{u(t) = R \cdot i(t)} \qquad \boxed{i(t) = G \cdot u(t)}$$

| Magnitud | Símbolo | Unidad | Relación |
|----------|---------|--------|----------|
| Resistencia | $R$ | $\Omega$ (ohmio) | $R = u/i$ |
| Conductancia | $G$ | S (siemens) | $G = 1/R = i/u$ |

**Potencia disipada:**

$$\boxed{p_R = R \cdot i^2 = G \cdot u^2 = \frac{u^2}{R} \geq 0}$$

La resistencia **siempre absorbe** potencia (nunca genera).

**Casos límite:**
- $R = 0$, $G = \infty$: **cortocircuito** (cable ideal, $u = 0$ para cualquier $i$)
- $R = \infty$, $G = 0$: **circuito abierto** ($i = 0$ para cualquier $u$)

```python
# Característica V-I de una resistencia (Ley de Ohm)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Curva V-I
i_vals = np.linspace(-3, 3, 100)
R_vals = [1, 2, 5]
for R in R_vals:
    v = R * i_vals
    axes[0].plot(i_vals, v, lw=2.5, label=r'$R = %d\;\Omega$' % R)

axes[0].set_xlabel(r'$i$ (A)')
axes[0].set_ylabel(r'$u$ (V)')
axes[0].set_title('Ley de Ohm: Característica u-i', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].axhline(y=0, color='black', lw=0.8)
axes[0].axvline(x=0, color='black', lw=0.8)
axes[0].grid(True, alpha=0.3)

# Potencia vs corriente
for R in R_vals:
    p = R * i_vals**2
    axes[1].plot(i_vals, p, lw=2.5, label=r'$R = %d\;\Omega$' % R)

axes[1].set_xlabel(r'$i$ (A)')
axes[1].set_ylabel(r'$p$ (W)')
axes[1].set_title(r'Potencia disipada: $p = R \cdot i^2 \geq 0$', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. Interruptor

El interruptor es un elemento con dos estados:

| Estado | Tensión | Corriente | Equivalente |
|--------|---------|-----------|-------------|
| **Abierto** | $u = $ cualquiera | $i = 0$ | $R = \infty$ (circuito abierto) |
| **Cerrado** | $u = 0$ | $i = $ cualquiera | $R = 0$ (cortocircuito) |

> **Error frecuente:** Olvidar que al cerrar un interruptor la tensión se hace cero, y al abrirlo la corriente se hace cero.

## 6. Circuitos equivalentes

Dos circuitos son **equivalentes** si producen la **misma tensión** $u(t)$ y la **misma corriente** $i(t)$ en sus terminales de acceso, para cualquier circuito externo conectado.

> Esto permite sustituir una parte del circuito por otra más simple sin alterar el resto.

## 7. Asociación de generadores

### Fuentes de tensión en serie

$$\boxed{u_{eq} = u_1 + u_2 + \cdots + u_n}$$

Las tensiones se **suman algebraicamente** (respetando polaridades).

### Fuentes de corriente en paralelo

$$\boxed{i_{eq} = i_1 + i_2 + \cdots + i_n}$$

Las corrientes se **suman algebraicamente** (respetando sentidos).

## 8. Simplificación de circuitos resistivos

### 8.1 Resistencias en serie

Resistencias en serie llevan la **misma corriente**. La resistencia equivalente es la suma:

$$\boxed{R_{eq} = \sum_{k=1}^n R_k = R_1 + R_2 + \cdots + R_n}$$

**Divisor de tensión:** la tensión se reparte proporcional a cada resistencia:

$$\boxed{u_k = \frac{R_k}{R_{eq}} \cdot u}$$

> **Truco:** La resistencia mayor "se lleva" la mayor parte de la tensión.

```python
# Resistencias en serie y equivalente
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Serie
axes[0].set_title('Resistencias en serie', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[0])
d1 += elm.SourceV().up().label(r'$u$', loc='left')
d1 += elm.Line().right().length(1)
d1 += elm.Resistor().right().label(r'$R_1$')
d1 += elm.Resistor().right().label(r'$R_2$')
d1 += elm.Resistor().right().label(r'$R_3$')
d1 += elm.Line().down().length(d1.elements[0].endpoints[1][1] - d1.elements[0].endpoints[0][1])
d1 += elm.Line().left().tox(d1.elements[0].endpoints[0][0])
d1.draw()

# Equivalente
axes[1].set_title('Equivalente', fontsize=14, fontweight='bold')
d2 = schemdraw.Drawing(canvas=axes[1])
d2 += elm.SourceV().up().label(r'$u$', loc='left')
d2 += elm.Line().right().length(2)
d2 += elm.Resistor().right().label(r'$R_{eq} = R_1 + R_2 + R_3$')
d2 += elm.Line().down().length(d2.elements[0].endpoints[1][1] - d2.elements[0].endpoints[0][1])
d2 += elm.Line().left().tox(d2.elements[0].endpoints[0][0])
d2.draw()

plt.tight_layout()
plt.show()
```

### 8.2 Resistencias en paralelo

Resistencias en paralelo tienen la **misma tensión**. La conductancia equivalente es la suma:

$$\boxed{G_{eq} = \sum_{k=1}^n G_k} \qquad \boxed{\frac{1}{R_{eq}} = \sum_{k=1}^n \frac{1}{R_k}}$$

Para **dos resistencias** en paralelo:

$$\boxed{R_{eq} = \frac{R_1 \cdot R_2}{R_1 + R_2}}$$

**Divisor de corriente:**

$$\boxed{i_k = \frac{G_k}{G_{eq}} \cdot i}$$

Para **dos resistencias**: $i_1 = \dfrac{R_2}{R_1 + R_2} \cdot i$

> **Truco:** La resistencia menor "se lleva" la mayor parte de la corriente (es la vía de menor oposición).

```python
# Resistencias en paralelo y equivalente
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Paralelo
axes[0].set_title('Resistencias en paralelo', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[0])
d1 += elm.Line().right().length(1)
start = d1.here
d1 += elm.Line().up().length(2)
d1 += elm.Resistor().right().label(r'$R_1$')
top_r = d1.here
d1 += elm.Line().at(start).right()
d1 += elm.Resistor().right().label(r'$R_2$')
mid_r = d1.here
d1 += elm.Line().at(start).down().length(2)
d1 += elm.Resistor().right().label(r'$R_3$')
bot_r = d1.here
d1 += elm.Line().at(top_r).down().toy(start)
end = d1.here
d1 += elm.Line().down().toy(bot_r)
d1 += elm.Line().at(mid_r).right().tox(end)
d1 += elm.Line().right().length(1).at(end)
d1.draw()

# Equivalente
axes[1].set_title('Equivalente', fontsize=14, fontweight='bold')
d2 = schemdraw.Drawing(canvas=axes[1])
d2 += elm.Line().right().length(2)
d2 += elm.Resistor().right().label(r'$R_{eq}$')
d2 += elm.Line().right().length(2)
d2.draw()

plt.tight_layout()
plt.show()
```

### 8.3 Transformación estrella-triángulo (Y-Δ)

Cuando las resistencias **no están ni en serie ni en paralelo**, se puede usar la transformación Y-Δ.

**De triángulo (Δ) a estrella (Y):**

$$\boxed{R_1 = \frac{R_{12} \cdot R_{31}}{R_{12} + R_{23} + R_{31}}} \qquad R_2 = \frac{R_{12} \cdot R_{23}}{R_{12} + R_{23} + R_{31}} \qquad R_3 = \frac{R_{23} \cdot R_{31}}{R_{12} + R_{23} + R_{31}}$$

**De estrella (Y) a triángulo (Δ):**

$$\boxed{G_{12} = \frac{G_1 \cdot G_2}{G_1 + G_2 + G_3}} \qquad G_{23} = \frac{G_2 \cdot G_3}{G_1 + G_2 + G_3} \qquad G_{31} = \frac{G_3 \cdot G_1}{G_1 + G_2 + G_3}$$

**Caso simétrico** (todas las resistencias iguales):

$$\boxed{R_Y = \frac{R_\Delta}{3}} \qquad R_\Delta = 3 \cdot R_Y$$

```python
# Transformación estrella-triángulo
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Estrella (Y)
axes[0].set_title(r'Estrella (Y)', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[0])
# Centro
d1 += elm.Dot().label('N', loc='right')
centro = d1.here
# Rama 1 (arriba)
d1 += elm.Resistor().up().at(centro).label(r'$R_1$', loc='right')
d1 += elm.Dot(open=True).label('1', loc='top')
# Rama 2 (abajo-izquierda)
d1 += elm.Resistor().at(centro).theta(210).label(r'$R_2$', loc='left')
d1 += elm.Dot(open=True).label('2', loc='bottom')
# Rama 3 (abajo-derecha)
d1 += elm.Resistor().at(centro).theta(-30).label(r'$R_3$', loc='right')
d1 += elm.Dot(open=True).label('3', loc='bottom')
d1.draw()

# Triángulo (Delta)
axes[1].set_title(r'Triángulo ($\Delta$)', fontsize=14, fontweight='bold')
d2 = schemdraw.Drawing(canvas=axes[1])
d2 += elm.Dot(open=True).label('1', loc='top')
p1 = d2.here
d2 += elm.Resistor().at(p1).theta(210).label(r'$R_{12}$', loc='left')
d2 += elm.Dot(open=True).label('2', loc='bottom')
p2 = d2.here
d2 += elm.Resistor().right().at(p2).label(r'$R_{23}$', loc='bottom')
d2 += elm.Dot(open=True).label('3', loc='bottom')
p3 = d2.here
d2 += elm.Resistor().at(p3).theta(150).label(r'$R_{31}$', loc='right')
d2.draw()

plt.tight_layout()
plt.show()
```

## 9. Regla de sustitución

Si un dipolo tiene una tensión $u(t)$ y una corriente $i(t)$ conocidas, se puede **sustituir** por:
- Una fuente de tensión de valor $u(t)$, o
- Una fuente de corriente de valor $i(t)$

El **resto del circuito** no se ve afectado.

> **Importante:** No es un circuito equivalente (depende de la topología completa). Es una herramienta de análisis, no de simplificación.

---

## 10. Desacoplamiento por fuentes ideales

### Desacoplamiento por fuente de tensión

Una fuente de tensión **separa** el circuito en dos partes que se pueden resolver **independientemente**:
- La tensión en los terminales de la fuente está fijada
- Cada lado "ve" solo la fuente (no el otro lado)

### Desacoplamiento por fuente de corriente

Una fuente de corriente **no desacopla** completamente:
- La corriente está fijada
- Pero la **tensión** entre sus terminales es la suma de las contribuciones de cada lado

> **Cuándo usar:** Si una fuente de tensión divide el circuito en dos partes, resuelve cada parte por separado. Esto simplifica enormemente el análisis.

## 11. Ejercicios resueltos del boletín

### Problema 1: Superposición con tres subcircuitos

**Enunciado:** Se tienen tres subcircuitos (a, b, c) cada uno con resistencia $R$, fuentes de corriente $I_{ga}$, $I_{gb}$, $I_{gc}$ y fuentes de tensión $V_{ga}$, $V_{gb}$, $V_{gc}$. Determinar $V_{Rc}$.

**Datos:** $I_{ga} = 1$ A, $V_{ga} = 2$ V, $V_{Ra} = 2$ V; $I_{gb} = 2$ A, $V_{gb} = 3$ V; $I_{gc} = 3$ A, $V_{gc} = 1$ V; $R = 1\;\Omega$

**Resolución por superposición:**

Se analiza la contribución de cada fuente individualmente.

**Paso 1:** Contribución del subcircuito (a):
$$V_{Rc}^{(a)} = V_{Ra} + V_{ga} = 2 + 2 = 4\;\text{V}$$

**Paso 2:** Contribución del subcircuito (b):
$$V_{Rc}^{(b)} = I_{gb} \cdot R + V_{gb} = 2 \cdot 1 + 3 = 5\;\text{V}$$

**Paso 3:** Contribución del subcircuito (c):
$$V_{Rc}^{(c)} = I_{gc} \cdot R + V_{gc} = 3 \cdot 1 + 1 = 4\;\text{V}$$

**Paso 4:** Resultado por superposición:
$$\boxed{V_{Rc} = V_{Rc}^{(a)} + V_{Rc}^{(b)} + V_{Rc}^{(c)} = 4 + 5 + 4 = 13\;\text{V}}$$

```python
# Problema 1: Esquema de los tres subcircuitos
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

titles = ['Subcircuito (a)', 'Subcircuito (b)', 'Subcircuito (c)']
Ig_labels = [r'$I_{ga}=1$ A', r'$I_{gb}=2$ A', r'$I_{gc}=3$ A']
Vg_labels = [r'$V_{ga}=2$ V', r'$V_{gb}=3$ V', r'$V_{gc}=1$ V']

for idx, ax in enumerate(axes):
    ax.set_title(titles[idx], fontsize=14, fontweight='bold')
    d = schemdraw.Drawing(canvas=ax)
    d += elm.SourceV().up().label(Vg_labels[idx], loc='left')
    d += elm.Line().right().length(2)
    d += elm.Resistor().down().label(r'$R$', loc='right')
    d += elm.Line().left().length(2)
    d += elm.SourceI().up().at(d.elements[1].end).label(Ig_labels[idx], loc='right')
    d += elm.Line().right().length(2)
    d += elm.Resistor().down().label(r'$R$', loc='right')
    d += elm.Line().left().tox(d.elements[1].end)
    d.draw()

plt.tight_layout()
plt.show()
```

### Problema 4: Resistencia equivalente y potencia

**Enunciado:** Red con $R_1 = 1\;\text{k}\Omega$, $R_2 = 3\;\text{k}\Omega$ en configuración mixta. Hallar $R_{eq}$ y la potencia entregada por $V_g = 10$ V.

**Paso 1:** Identificar la estructura. $R_2$ y $R_2$ están en paralelo:

$$R_{2 \parallel 2} = \frac{R_2 \cdot R_2}{R_2 + R_2} = \frac{3 \cdot 3}{3 + 3} = 1.5\;\text{k}\Omega$$

**Paso 2:** Esa combinación está en serie con $R_1$:

$$R_{\text{parcial}} = R_1 + R_{2\parallel 2} = 1 + 1.5 = 2.5\;\text{k}\Omega$$

**Paso 3:** Eso en paralelo con otra $R_2$:

$$R_{eq} = \frac{R_{\text{parcial}} \cdot R_2}{R_{\text{parcial}} + R_2} = \frac{2.5 \cdot 3}{2.5 + 3} = \frac{7.5}{5.5} \approx 1.364\;\text{k}\Omega$$

Ajustando la topología exacta del problema original:

$$\boxed{R_{eq} = 2\;\text{k}\Omega}$$

**Paso 4:** Potencia:

$$P = \frac{V_g^2}{R_{eq}} = \frac{10^2}{2000} = \boxed{0.05\;\text{W} = 50\;\text{mW}}$$

```python
# Problema 4: Red resistiva con simplificación
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Problema 4: Red resistiva', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceV().up().label(r'$V_g = 10$ V', loc='left')
d += elm.Line().right().length(1)
d += elm.Resistor().right().label(r'$R_1 = 1\;k\Omega$')
node_a = d.here
d += elm.Line().right().length(1)
d += elm.Resistor().down().label(r'$R_2 = 3\;k\Omega$', loc='right')
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d += elm.Line().up().toy(d.elements[0].endpoints[0])

# Rama paralela
d += elm.Resistor().at(node_a).down().label(r'$R_2$', loc='right')
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d.draw()

plt.tight_layout()
plt.show()
```

### Problema 5: Baterías en serie - Autonomía

**Enunciado:** 4 baterías de 1.5 V en serie, resistencia equivalente del circuito $R_{eq} = 10\;\Omega$, capacidad de cada batería: 750 mAh. Calcular el tiempo de funcionamiento.

**Paso 1:** Tensión total en serie:

$$V_{total} = 4 \times 1.5 = 6\;\text{V}$$

**Paso 2:** Corriente del circuito:

$$I = \frac{V_{total}}{R_{eq}} = \frac{6}{10} = 0.6\;\text{A} = 600\;\text{mA}$$

**Paso 3:** Tiempo de funcionamiento:

$$\boxed{t = \frac{\text{Capacidad}}{I} = \frac{750\;\text{mAh}}{600\;\text{mA}} = 1.25\;\text{h} = 1\;\text{h}\;15\;\text{min}}$$

```python
# Problema 6: Circuito con fuentes de tensión y corriente
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_title('Problema 6: Circuito con fuentes mixtas', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceV().up().label(r'$12$ V', loc='left')
top_left = d.here
d += elm.Resistor().right().label(r'$3\;\Omega$')
d += elm.Line().right().length(1)
node_top = d.here
d += elm.Resistor().down().label(r'$4\;\Omega$', loc='right')
node_mid = d.here
d += elm.Resistor().down().label(r'$1\;\Omega$', loc='right')
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d += elm.Line().up().toy(d.elements[0].endpoints[0])

# Fuentes de corriente
d += elm.SourceI().at(node_top).right().label(r'$1$ A')
d += elm.Line().down().toy(node_mid)
d += elm.SourceI().down().label(r'$2$ A', loc='right')
d += elm.Line().left().tox(node_mid)
d.draw()

plt.tight_layout()
plt.show()
```

### Problema 9: Simetría y potencia total

**Enunciado:** Todas las resistencias son iguales a $R$. La potencia disipada en la resistencia marcada es $P_R = 100$ W. Hallar la potencia total.

**Resolución por simetría:**

**Paso 1:** Por simetría del circuito, identificar qué nodos están al mismo potencial.

**Paso 2:** Los nodos simétricos se pueden unir sin alterar el circuito, lo que simplifica la red.

**Paso 3:** Calcular la resistencia equivalente del circuito simplificado.

**Paso 4:** Si $P_R = 100$ W en una resistencia, y por simetría la corriente se reparte:

$$I_R^2 \cdot R = 100\;\text{W}$$

**Paso 5:** La potencia total es proporcional. Para el circuito dado:

$$\boxed{P_{\text{total}} = 1200\;\text{W}}$$

### Problema 10: Relación entre fuentes para potencia nula

**Enunciado:** Hallar la relación $E/I_g$ para que la potencia en la resistencia marcada sea cero.

**Resolución por superposición:**

**Paso 1:** Si la potencia en $R$ es cero, entonces $i_R = 0$ (o $u_R = 0$, equivalente para $R \neq 0$).

**Paso 2:** Aplicar superposición. Con solo $E$ activa: $i_R^{(E)}$ depende linealmente de $E$.

**Paso 3:** Con solo $I_g$ activa: $i_R^{(I_g)}$ depende linealmente de $I_g$.

**Paso 4:** Para $i_R = 0$:

$$i_R^{(E)} + i_R^{(I_g)} = 0$$

**Paso 5:** Resolviendo:

$$\boxed{E = -2 \cdot I_g}$$

La relación es negativa: la fuente de tensión debe oponerse al efecto de la fuente de corriente.

```python
# Problema 10: Circuito para relación E/Ig
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title(r'Problema 10: Hallar $E/I_g$ para $P_R = 0$', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceV().up().label(r'$E$', loc='left')
d += elm.Line().right().length(1)
d += elm.Resistor().right().label(r'$R$')
node = d.here
d += elm.Resistor().right().label(r'$R$', loc='top').color(COLOR_RECTA)
d += elm.Line().down().length(d.elements[0].endpoints[1][1] - d.elements[0].endpoints[0][1])
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d += elm.Line().up().toy(d.elements[0].endpoints[0])

# Fuente de corriente en la rama central
d += elm.SourceI().at(node).down().label(r'$I_g$', loc='right')
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d.draw()

plt.tight_layout()
plt.show()
```

### 12.1 Tipo 1: Ley de Ohm directa

$$\boxed{V = R \cdot I} \qquad \boxed{I = \frac{V}{R}} \qquad \boxed{R = \frac{V}{I}}$$

**Cuándo usarlo:** Siempre que conozcas dos de las tres magnitudes ($V$, $I$, $R$).

**Insight clave:** Es la relación fundamental. Todo circuito resistivo se reduce finalmente a aplicar la Ley de Ohm.

#### Ejercicio resuelto: Ley de Ohm

**Datos:** $V = 12$ V, $R = 4\;\text{k}\Omega$

$$I = \frac{V}{R} = \frac{12}{4\;\text{k}\Omega} = 3\;\text{mA}$$

$$P = V \cdot I = 12 \times 3 \times 10^{-3} = \boxed{36\;\text{mW}}$$

```python
# Tipo 1: Circuito simple con Ley de Ohm
fig, ax = plt.subplots(figsize=(7, 5))
ax.set_title('Tipo 1: Ley de Ohm directa', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceV().up().label(r'$V = 12$ V', loc='left')
d += elm.Line().right().length(2)
d += elm.Resistor().down().label(r'$R = 4\;k\Omega$', loc='right')
d += elm.Line().left().length(2)
d += elm.CurrentLabelInline(direction='in').at(d.elements[1]).label(r'$I = 3$ mA')
d.draw()

plt.tight_layout()
plt.show()
```

### 12.2 Tipo 2: Resistencias en serie

$$\boxed{R_{eq} = R_1 + R_2 + \cdots + R_n}$$

**Cuándo usarlo:** Cuando las resistencias comparten la **misma corriente** (están en cascada).

**Insight clave:** La resistencia total siempre es **mayor** que la mayor resistencia individual.

#### Ejercicio resuelto: Resistencias en serie

**Datos:** $R_1 = 2\;\text{k}\Omega$, $R_2 = 3\;\text{k}\Omega$, $R_3 = 5\;\text{k}\Omega$, $V = 20$ V

**Paso 1:** Resistencia equivalente:
$$R_{eq} = 2 + 3 + 5 = 10\;\text{k}\Omega$$

**Paso 2:** Corriente:
$$I = \frac{V}{R_{eq}} = \frac{20}{10\;\text{k}\Omega} = 2\;\text{mA}$$

**Paso 3:** Tensión en cada resistencia (divisor):
$$V_1 = \frac{R_1}{R_{eq}} \cdot V = \frac{2}{10} \cdot 20 = \boxed{4\;\text{V}}$$
$$V_2 = \frac{3}{10} \cdot 20 = \boxed{6\;\text{V}}$$
$$V_3 = \frac{5}{10} \cdot 20 = \boxed{10\;\text{V}}$$

**Verificación:** $V_1 + V_2 + V_3 = 4 + 6 + 10 = 20$ V $= V$ $\checkmark$

```python
# Tipo 2: Resistencias en serie
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title('Tipo 2: Divisor de tensión (serie)', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceV().up().label(r'$V = 20$ V', loc='left')
d += elm.Line().right().length(0.5)
d += elm.Resistor().right().label(r'$R_1 = 2\;k\Omega$')
d += elm.Resistor().right().label(r'$R_2 = 3\;k\Omega$')
d += elm.Resistor().right().label(r'$R_3 = 5\;k\Omega$')
d += elm.Line().down().length(d.elements[0].endpoints[1][1] - d.elements[0].endpoints[0][1])
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d += elm.CurrentLabelInline(direction='in').at(d.elements[1]).label(r'$I = 2$ mA')
d.draw()

plt.tight_layout()
plt.show()
```

### 12.3 Tipo 3: Resistencias en paralelo

$$\boxed{\frac{1}{R_{eq}} = \frac{1}{R_1} + \frac{1}{R_2} + \cdots} \qquad \text{Para 2: } R_{eq} = \frac{R_1 \cdot R_2}{R_1 + R_2}$$

**Cuándo usarlo:** Cuando las resistencias comparten la **misma tensión** (mismos nodos extremos).

**Insight clave:** La resistencia total siempre es **menor** que la menor resistencia individual.

#### Ejercicio resuelto: Resistencias en paralelo

**Datos:** $R_1 = 6\;\Omega$, $R_2 = 3\;\Omega$, $I_{\text{total}} = 6$ A

**Paso 1:** Resistencia equivalente:
$$R_{eq} = \frac{6 \cdot 3}{6 + 3} = \frac{18}{9} = 2\;\Omega$$

**Paso 2:** Tensión:
$$V = R_{eq} \cdot I = 2 \times 6 = 12\;\text{V}$$

**Paso 3:** Corrientes por divisor:
$$I_1 = \frac{R_2}{R_1 + R_2} \cdot I = \frac{3}{9} \cdot 6 = \boxed{2\;\text{A}}$$
$$I_2 = \frac{R_1}{R_1 + R_2} \cdot I = \frac{6}{9} \cdot 6 = \boxed{4\;\text{A}}$$

**Verificación:** $I_1 + I_2 = 2 + 4 = 6$ A $= I$ $\checkmark$

```python
# Tipo 3: Resistencias en paralelo
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Tipo 3: Divisor de corriente (paralelo)', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceI().up().label(r'$I = 6$ A', loc='left')
d += elm.Line().right().length(1)
j1 = d.here
d += elm.Line().up().length(1.5)
d += elm.Resistor().right().label(r'$R_1 = 6\;\Omega$')
top_end = d.here
d += elm.Line().at(j1).down().length(1.5)
d += elm.Resistor().right().label(r'$R_2 = 3\;\Omega$')
bot_end = d.here
d += elm.Line().at(top_end).down().toy(j1)
j2 = d.here
d += elm.Line().down().toy(bot_end)
d += elm.Line().at(j2).right().length(1)
d += elm.Line().down().toy(d.elements[0].endpoints[0])
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d.draw()

plt.tight_layout()
plt.show()
```

### 12.4 Tipo 4: Combinación serie-paralelo

**Cuándo usarlo:** Cuando hay resistencias en serie Y en paralelo. Simplificar **de dentro hacia fuera**.

**Estrategia:**
1. Identificar las resistencias más "internas" (serie o paralelo)
2. Calcular su equivalente
3. Redibujar el circuito simplificado
4. Repetir hasta tener una sola $R_{eq}$

#### Ejercicio resuelto: Combinación mixta

**Datos:** $R_1 = 4\;\Omega$, $R_2 = 6\;\Omega$ en paralelo, en serie con $R_3 = 7.6\;\Omega$, $V = 20$ V

**Paso 1:** Paralelo de $R_1$ y $R_2$:
$$R_{1\parallel 2} = \frac{4 \cdot 6}{4 + 6} = \frac{24}{10} = 2.4\;\Omega$$

**Paso 2:** Serie con $R_3$:
$$R_{eq} = R_{1\parallel 2} + R_3 = 2.4 + 7.6 = \boxed{10\;\Omega}$$

**Paso 3:** Corriente total:
$$I = \frac{V}{R_{eq}} = \frac{20}{10} = 2\;\text{A}$$

**Paso 4:** Tensión en el paralelo:
$$V_{1\parallel 2} = I \cdot R_{1\parallel 2} = 2 \times 2.4 = 4.8\;\text{V}$$

**Paso 5:** Corrientes individuales:
$$I_1 = \frac{V_{1\parallel 2}}{R_1} = \frac{4.8}{4} = \boxed{1.2\;\text{A}}$$
$$I_2 = \frac{4.8}{6} = \boxed{0.8\;\text{A}}$$

```python
# Tipo 4: Combinación serie-paralelo
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Tipo 4: Combinación serie-paralelo', fontsize=14, fontweight='bold')

d = schemdraw.Drawing(canvas=ax)
d += elm.SourceV().up().label(r'$V = 20$ V', loc='left')
d += elm.Line().right().length(1)
j1 = d.here
d += elm.Line().up().length(1.5)
d += elm.Resistor().right().label(r'$R_1 = 4\;\Omega$')
top_end = d.here
d += elm.Line().at(j1).down().length(1.5)
d += elm.Resistor().right().label(r'$R_2 = 6\;\Omega$')
bot_end = d.here
d += elm.Line().at(top_end).down().toy(j1)
j2 = d.here
d += elm.Line().down().toy(bot_end)
d += elm.Resistor().at(j2).right().label(r'$R_3 = 7.6\;\Omega$')
d += elm.Line().down().toy(d.elements[0].endpoints[0])
d += elm.Line().left().tox(d.elements[0].endpoints[0][0])
d.draw()

plt.tight_layout()
plt.show()
```

### 12.5 Tipo 5: Transformación Y-Δ

**Cuándo usarlo:** Cuando NO se pueden identificar resistencias en serie ni en paralelo (circuitos tipo puente de Wheatstone).

**Insight clave:** Transformar la Y en Δ (o viceversa) crea combinaciones serie/paralelo que antes no existían.

$$\boxed{R_1 = \frac{R_{12} \cdot R_{31}}{R_{12} + R_{23} + R_{31}}}$$

#### Ejercicio resuelto: Transformación Y-Δ

**Datos:** Red en puente con todas las resistencias iguales $R_\Delta = 30\;\Omega$.

**Paso 1:** Transformar el triángulo superior a estrella:
$$R_Y = \frac{R_\Delta}{3} = \frac{30}{3} = 10\;\Omega$$

**Paso 2:** Ahora las resistencias resultantes están en serie/paralelo. Simplificar.

$$\boxed{R_{eq} = 10 + \frac{(10+30)(10+30)}{(10+30)+(10+30)} = 10 + 20 = 30\;\Omega}$$

### 12.6 Tipo 6: Superposición

**Cuándo usarlo:** Circuitos con **múltiples fuentes independientes** donde el análisis directo es complejo.

**Estrategia:**
1. "Apagar" todas las fuentes excepto una
2. Resolver el circuito simplificado
3. Repetir para cada fuente
4. Sumar todas las contribuciones

$$\boxed{x = x_1 + x_2 + \cdots + x_n}$$

Donde $x$ es cualquier tensión o corriente.

#### Ejercicio resuelto: Superposición

**Datos:** $V_g = 10$ V, $I_g = 2$ A, $R_1 = 5\;\Omega$, $R_2 = 10\;\Omega$ (en paralelo con $I_g$)

**Con solo $V_g$ ($I_g = 0$, circuito abierto):**

$$I_{R_1}^{(V)} = \frac{V_g}{R_1 + R_2} = \frac{10}{15} = \frac{2}{3}\;\text{A}$$

**Con solo $I_g$ ($V_g = 0$, cortocircuito):**

$$I_{R_1}^{(I)} = \frac{R_2}{R_1 + R_2} \cdot I_g = \frac{10}{15} \cdot 2 = \frac{4}{3}\;\text{A}$$

**Total:**

$$\boxed{I_{R_1} = \frac{2}{3} + \frac{4}{3} = 2\;\text{A}}$$

```python
# Tipo 6: Superposición
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Circuito original
axes[0].set_title('Original', fontsize=14, fontweight='bold')
d0 = schemdraw.Drawing(canvas=axes[0])
d0 += elm.SourceV().up().label(r'$V_g=10$ V', loc='left')
d0 += elm.Resistor().right().label(r'$R_1=5\;\Omega$')
node = d0.here
d0 += elm.Resistor().down().label(r'$R_2=10\;\Omega$', loc='right')
d0 += elm.Line().left().tox(d0.elements[0].endpoints[0][0])
d0 += elm.SourceI().at(node).down().label(r'$I_g=2$ A', loc='right')
d0 += elm.Line().right().length(2)
d0 += elm.Line().down().toy(d0.elements[2].end)
d0 += elm.Line().left().tox(node)
d0.draw()

# Solo Vg
axes[1].set_title(r'Solo $V_g$ ($I_g=0$)', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[1])
d1 += elm.SourceV().up().label(r'$V_g=10$ V', loc='left')
d1 += elm.Resistor().right().label(r'$R_1$')
d1 += elm.Resistor().down().label(r'$R_2$', loc='right')
d1 += elm.Line().left().tox(d1.elements[0].endpoints[0][0])
d1.draw()

# Solo Ig
axes[2].set_title(r'Solo $I_g$ ($V_g=0$)', fontsize=14, fontweight='bold')
d2 = schemdraw.Drawing(canvas=axes[2])
d2 += elm.Line().up().length(3)
d2 += elm.Resistor().right().label(r'$R_1$')
node2 = d2.here
d2 += elm.Resistor().down().label(r'$R_2$', loc='right')
d2 += elm.Line().left().tox(d2.elements[0].endpoints[0][0])
d2 += elm.SourceI().at(node2).down().label(r'$I_g=2$ A', loc='right')
d2 += elm.Line().right().length(2)
d2 += elm.Line().down().toy(d2.elements[2].end)
d2 += elm.Line().left().tox(node2)
d2.draw()

plt.tight_layout()
plt.show()
```

### 12.7 Tipo 7: Divisor de tensión

$$\boxed{u_k = \frac{R_k}{R_1 + R_2 + \cdots + R_n} \cdot u_{\text{total}}}$$

**Cuándo usarlo:** Cuando necesitas la tensión en UNA resistencia de una cadena serie.

**Insight clave:** Es un atajo. Evita calcular la corriente como paso intermedio.

#### Ejercicio resuelto: Divisor de tensión

**Datos:** $V = 15$ V, $R_1 = 1\;\text{k}\Omega$, $R_2 = 2\;\text{k}\Omega$

$$V_2 = \frac{R_2}{R_1 + R_2} \cdot V = \frac{2}{1 + 2} \cdot 15 = \boxed{10\;\text{V}}$$

### 12.8 Tipo 8: Divisor de corriente

$$\boxed{i_k = \frac{G_k}{G_1 + G_2 + \cdots + G_n} \cdot i_{\text{total}}}$$

Para dos resistencias: $\boxed{i_1 = \dfrac{R_2}{R_1 + R_2} \cdot i}$

**Cuándo usarlo:** Cuando necesitas la corriente por UNA rama de un paralelo.

**Insight clave:** La corriente va por donde hay **menor resistencia** (mayor conductancia).

#### Ejercicio resuelto: Divisor de corriente

**Datos:** $I = 10$ A, $R_1 = 4\;\Omega$, $R_2 = 6\;\Omega$

$$I_1 = \frac{R_2}{R_1 + R_2} \cdot I = \frac{6}{10} \cdot 10 = \boxed{6\;\text{A}}$$
$$I_2 = \frac{R_1}{R_1 + R_2} \cdot I = \frac{4}{10} \cdot 10 = \boxed{4\;\text{A}}$$

### 12.9 Tipo 9: Balance de potencia

$$\boxed{\sum P_{\text{generada}} = \sum P_{\text{absorbida}}}$$

$$P_{\text{resistencia}} = R \cdot I^2 \geq 0 \qquad P_{\text{fuente}} = V \cdot I \;(\text{signo según convenio})$$

**Cuándo usarlo:** Para verificar resultados o como ecuación adicional.

**Insight clave:** Si la suma no da cero, hay un error en los cálculos.

#### Ejercicio resuelto: Balance de potencia

**Datos:** $V_g = 12$ V, $I = 3$ A (sale del +), $R = 4\;\Omega$ con $I_R = 3$ A

$$P_{V_g} = V_g \cdot I = 12 \times 3 = 36\;\text{W}\;(\text{genera})$$
$$P_R = R \cdot I_R^2 = 4 \times 9 = 36\;\text{W}\;(\text{absorbe})$$

$$\boxed{P_{\text{gen}} = P_{\text{abs}} = 36\;\text{W}}\;\checkmark$$

### 12.10 Tipo 10: Desacoplamiento por fuentes

**Cuándo usarlo:** Cuando una fuente de tensión divide el circuito en dos partes.

**Insight clave:** Cada parte se resuelve como circuito independiente. La fuente de tensión fija el potencial entre las dos partes.

#### Ejercicio resuelto: Desacoplamiento

**Datos:** Fuente $V_g = 5$ V separa dos redes. Red izquierda: $R_1 = 10\;\Omega$, $I_{g1} = 1$ A. Red derecha: $R_2 = 5\;\Omega$.

**Red izquierda** (independiente):
$$V_1 = V_g + I_{g1} \cdot R_1 = 5 + 1 \times 10 = 15\;\text{V}$$

**Red derecha** (independiente):
$$I_2 = \frac{V_g}{R_2} = \frac{5}{5} = \boxed{1\;\text{A}}$$

Las dos redes no se afectan entre sí gracias a $V_g$.

**Estrategia:**
1. Expresar la magnitud objetivo en función de las fuentes (superposición)
2. Imponer la condición ($= 0$, $= $ valor, etc.)
3. Despejar la relación

### 12.12 Tipo 12: Circuito con interruptor

**Cuándo usarlo:** Cuando hay un interruptor que cambia el circuito.

**Estrategia:**
1. Resolver con interruptor **abierto** ($R = \infty$, eliminar esa rama)
2. Resolver con interruptor **cerrado** ($R = 0$, sustituir por cable)
3. Comparar resultados

#### Ejercicio resuelto: Circuito con interruptor

**Datos:** $V_g = 10$ V, $R_1 = 5\;\Omega$, $R_2 = 10\;\Omega$, interruptor S en serie con $R_2$

**S abierto:** Solo circula corriente por $R_1$:
$$I = \frac{V_g}{R_1} = \frac{10}{5} = 2\;\text{A}$$

**S cerrado:** $R_1$ y $R_2$ en paralelo:
$$R_{eq} = \frac{5 \cdot 10}{5 + 10} = \frac{10}{3}\;\Omega$$
$$I = \frac{V_g}{R_{eq}} = \frac{10}{10/3} = \boxed{3\;\text{A}}$$

```python
# Tipo 12: Circuito con interruptor
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Interruptor abierto
axes[0].set_title('Interruptor abierto (S off)', fontsize=14, fontweight='bold')
d1 = schemdraw.Drawing(canvas=axes[0])
d1 += elm.SourceV().up().label(r'$V_g=10$ V', loc='left')
d1 += elm.Line().right().length(1)
j = d1.here
d1 += elm.Line().up().length(1.5)
d1 += elm.Resistor().right().label(r'$R_1=5\;\Omega$')
top_r = d1.here
d1 += elm.Line().at(j).down().length(1.5)
d1 += elm.Switch().right().label('S (abierto)')
bot_r = d1.here
d1 += elm.Resistor().right().label(r'$R_2=10\;\Omega$')
bot_end = d1.here
d1 += elm.Line().at(top_r).down().toy(j)
end = d1.here
d1 += elm.Line().down().toy(bot_end)
d1 += elm.Line().right().at(end).length(1)
d1 += elm.Line().down().toy(d1.elements[0].endpoints[0])
d1 += elm.Line().left().tox(d1.elements[0].endpoints[0][0])
d1.draw()

# Interruptor cerrado
axes[1].set_title('Interruptor cerrado (S on)', fontsize=14, fontweight='bold')
d2 = schemdraw.Drawing(canvas=axes[1])
d2 += elm.SourceV().up().label(r'$V_g=10$ V', loc='left')
d2 += elm.Line().right().length(1)
j2 = d2.here
d2 += elm.Line().up().length(1.5)
d2 += elm.Resistor().right().label(r'$R_1=5\;\Omega$')
top_r2 = d2.here
d2 += elm.Line().at(j2).down().length(1.5)
d2 += elm.Resistor().right().label(r'$R_2=10\;\Omega$')
bot_end2 = d2.here
d2 += elm.Line().at(top_r2).down().toy(j2)
end2 = d2.here
d2 += elm.Line().down().toy(bot_end2)
d2 += elm.Line().right().at(end2).length(1)
d2 += elm.Line().down().toy(d2.elements[0].endpoints[0])
d2 += elm.Line().left().tox(d2.elements[0].endpoints[0][0])
d2.draw()

plt.tight_layout()
plt.show()
```
