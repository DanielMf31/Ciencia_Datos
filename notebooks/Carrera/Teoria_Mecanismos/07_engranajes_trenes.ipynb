## 1. Formulario del Tema

### Geometria del diente (sistema metrico)

| Parametro | Formula | Descripcion |
|---|---|---|
| Modulo | $m = d / Z$ | Parametro fundamental |
| Paso circular | $p = \pi\,m$ | Distancia entre dientes consecutivos |
| Diametro primitivo | $d = m\,Z$ | Circunferencia de rodadura |
| Adendum (cabeza) | $h_a = m$ | Altura sobre primitiva |
| Dedendum (pie) | $h_f = 1.25\,m$ | Profundidad bajo primitiva |
| Altura total | $h = h_a + h_f = 2.25\,m$ | Altura completa del diente |
| Diametro de cabeza | $d_a = m\,(Z + 2)$ | Circunferencia exterior |
| Diametro de pie | $d_f = m\,(Z - 2.5)$ | Circunferencia interior |
| Distancia entre centros | $a_c = m\,(Z_1 + Z_2)/2$ | Para engranajes externos |

### Relacion de transmision

$$\boxed{i = \frac{\omega_1}{\omega_2} = \frac{Z_2}{Z_1} = \frac{d_2}{d_1}}$$

### Condicion de conjugacion

$$\boxed{\eta = -\frac{d\psi_2}{d\psi_3} = \text{constante}}$$

El perfil que cumple esta condicion es la **evolvente** (involuta) de circulo.

### Perfil de evolvente

$$\boxed{x(\theta) = r_b(\cos\theta + \theta\sin\theta)}$$
$$\boxed{y(\theta) = r_b(\sin\theta - \theta\cos\theta)}$$

donde $r_b = r\cos\alpha$ (radio de base), $\alpha$ = angulo de presion (tipicamente $20^\circ$).

### Ecuacion de Willis (trenes epicicloidales)

$$\boxed{i^{\text{brazo}} = \frac{\omega_L - \omega_B}{\omega_F - \omega_B} = (-1)^n \prod \frac{Z_{\text{motoras}}}{Z_{\text{conducidas}}}}$$

donde $B$ = brazo, $L$ = ultimo, $F$ = primero, $n$ = numero de contactos externos.

### Engranajes helicoidales

$$\boxed{m_n = m_t \cos\beta} \qquad \boxed{p_n = p_t \cos\beta}$$

$$\boxed{Z_{\min} = \frac{2}{\sin^2\alpha}\;(\text{para evitar interferencia, dientes rectos})}$$

## 2. Clasificacion de Engranajes

| Tipo | Ejes | Caracteristicas |
|---|---|---|
| **Cilindricos rectos** | Paralelos | Dientes rectos, el mas simple |
| **Cilindricos helicoidales** | Paralelos | Dientes inclinados $\beta$, mas suaves |
| **Conicos** | Concurrentes | Ejes que se cortan, tipicamente 90 grados |
| **Tornillo sin fin** | Cruzados (90 grados) | Gran reduccion, irreversible |
| **Cremallera** | - | Engranaje con $Z \to \infty$ (plano) |

### Ventajas del perfil de evolvente

- Relacion de transmision constante (condicion de conjugacion)
- Insensible a pequenos errores en la distancia entre centros
- Fabricacion sencilla (herramienta tipo cremallera generatriz)
- La linea de accion es recta

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 6))

# â”€â”€ Left: Involute curve â”€â”€
rb = 1.0  # radio de base
theta = np.linspace(0, 1.2, 300)
x_inv = rb * (np.cos(theta) + theta * np.sin(theta))
y_inv = rb * (np.sin(theta) - theta * np.cos(theta))

# Base circle
theta_circ = np.linspace(0, 2 * np.pi, 200)
ax1.plot(rb * np.cos(theta_circ), rb * np.sin(theta_circ), '--', color=COLOR_FIJO, lw=1.5, label='Circulo base $r_b$')

# Involute
ax1.plot(x_inv, y_inv, color=COLOR_PRINCIPAL, lw=2.5, label='Evolvente')

# Mirror involute for the other flank
ax1.plot(x_inv, -y_inv, color=COLOR_CLARO, lw=2, label='Otro flanco')

# Pitch circle
alpha_press = np.radians(20)
rp = rb / np.cos(alpha_press)
ax1.plot(rp * np.cos(theta_circ), rp * np.sin(theta_circ), ':', color=COLOR_PUNTO, lw=1.5, label=f'Circulo primitivo $r_p$ ($\\alpha={20}^\\circ$)')

# String unwinding
for t_show in [0.3, 0.6, 0.9]:
    xp = rb * (np.cos(t_show) + t_show * np.sin(t_show))
    yp = rb * (np.sin(t_show) - t_show * np.cos(t_show))
    xt = rb * np.cos(t_show)
    yt = rb * np.sin(t_show)
    ax1.plot([xt, xp], [yt, yp], '-', color=COLOR_NARANJA, lw=1, alpha=0.6)
    ax1.plot(xp, yp, 'o', color=COLOR_NARANJA, ms=4)

ax1.set_xlim(-1.8, 2.2)
ax1.set_ylim(-1.8, 1.8)
ax1.set_aspect('equal')
ax1.set_title('Perfil de Evolvente (Involuta)', fontsize=12, fontweight='bold')
ax1.legend(fontsize=9, loc='lower left')
ax1.grid(True, alpha=0.3)
ax1.plot(0, 0, '+', color=COLOR_FIJO, ms=10, mew=2)

# â”€â”€ Right: Gear tooth geometry â”€â”€
Z = 20
m_mod = 1
r_prim = m_mod * Z / 2
r_base = r_prim * np.cos(alpha_press)
ha = m_mod
hf = 1.25 * m_mod
ra = r_prim + ha
rf = r_prim - hf

# Draw circles
for r_val, col, ls, lbl in [(r_prim, COLOR_PRINCIPAL, '-', f'Primitivo $d = {2*r_prim}$'),
                              (ra, COLOR_ROJO, '--', f'Cabeza $d_a = {2*ra}$'),
                              (rf, COLOR_PUNTO, '--', f'Pie $d_f = {2*rf}$'),
                              (r_base, COLOR_NARANJA, ':', f'Base $d_b = {2*r_base:.2f}$')]:
    ax2.plot(r_val * np.cos(theta_circ), r_val * np.sin(theta_circ), ls=ls, color=col, lw=1.5, label=lbl)

# Draw a few teeth (simplified as involute sectors)
for k in range(4):
    ang_offset = k * 2 * np.pi / Z
    theta_tooth = np.linspace(0, np.sqrt((ra/r_base)**2 - 1), 100)
    xt = r_base * (np.cos(theta_tooth + ang_offset) + theta_tooth * np.sin(theta_tooth + ang_offset))
    yt = r_base * (np.sin(theta_tooth + ang_offset) - theta_tooth * np.cos(theta_tooth + ang_offset))
    mask = (xt**2 + yt**2) <= ra**2
    ax2.plot(xt[mask], yt[mask], color=COLOR_MORADO, lw=1.5)

ax2.set_xlim(6, 13)
ax2.set_ylim(-3.5, 3.5)
ax2.set_aspect('equal')
ax2.set_title(f'Geometria del diente ($Z={Z}$, $m={m_mod}$)', fontsize=12, fontweight='bold')
ax2.legend(fontsize=9, loc='lower right')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 3. Geometria del Diente

### Diente normalizado (sistema metrico ISO)

El **modulo** $m$ es el parametro que define completamente la geometria del diente:

- $h_a = 1.0\,m$ (adendum)
- $h_f = 1.25\,m$ (dedendum, incluye juego radial $0.25\,m$)
- $h = 2.25\,m$ (altura total)
- Espesor del diente $= \pi\,m / 2$ (en la primitiva)

Modulos normalizados: 0.5, 0.75, 1, 1.25, 1.5, 2, 2.5, 3, 4, 5, 6, 8, 10...

### Angulo de presion $\alpha$

- Valor estandar: $\alpha = 20^\circ$ (antiguo: $14.5^\circ$)
- Define la inclinacion de la linea de accion
- $r_b = r\cos\alpha$: radio de base

### Interferencia

Se produce cuando el diente del pinon penetra por debajo de la base del engranaje.

Numero minimo de dientes para evitar interferencia (con cremallera):

$$Z_{\min} = \frac{2}{\sin^2\alpha} = \frac{2}{\sin^2(20^\circ)} \approx 17$$

```python
fig, ax = plt.subplots(figsize=(10, 8))

# Parameters
Z1, Z2 = 15, 30
m_g = 1
r1 = m_g * Z1 / 2
r2 = m_g * Z2 / 2
alpha_p = np.radians(20)
rb1, rb2 = r1 * np.cos(alpha_p), r2 * np.cos(alpha_p)
ra1 = r1 + m_g
ra2 = r2 + m_g
rf1 = r1 - 1.25 * m_g
rf2 = r2 - 1.25 * m_g
a_c = r1 + r2  # center distance

theta_c = np.linspace(0, 2 * np.pi, 300)

# Gear 1 (left)
for r_val, col, ls in [(r1, COLOR_PRINCIPAL, '-'), (ra1, COLOR_ROJO, '--'), (rf1, COLOR_PUNTO, '--'), (rb1, COLOR_NARANJA, ':')]:
    ax.plot(r_val * np.cos(theta_c), r_val * np.sin(theta_c), ls=ls, color=col, lw=1.2)

# Gear 2 (right)
for r_val, col, ls in [(r2, COLOR_PRINCIPAL, '-'), (ra2, COLOR_ROJO, '--'), (rf2, COLOR_PUNTO, '--'), (rb2, COLOR_NARANJA, ':')]:
    ax.plot(a_c + r_val * np.cos(theta_c), r_val * np.sin(theta_c), ls=ls, color=col, lw=1.2)

# Centers
ax.plot(0, 0, '+', color=COLOR_FIJO, ms=12, mew=2)
ax.plot(a_c, 0, '+', color=COLOR_FIJO, ms=12, mew=2)
ax.text(0, -r1 - 2, f'Pinon\n$Z_1 = {Z1}$', fontsize=11, ha='center', color=COLOR_PRINCIPAL, fontweight='bold')
ax.text(a_c, -r2 - 2, f'Rueda\n$Z_2 = {Z2}$', fontsize=11, ha='center', color=COLOR_PRINCIPAL, fontweight='bold')

# Line of action
tangent_angle = np.arcsin((rb2 - rb1) / a_c) if rb2 > rb1 else 0
la_start_x = rb1 * np.sin(alpha_p)
la_start_y = rb1 * np.cos(alpha_p)
la_end_x = a_c - rb2 * np.sin(alpha_p)
la_end_y = rb2 * np.cos(alpha_p)
ax.plot([la_start_x, la_end_x], [la_start_y, la_end_y], '-', color=COLOR_MORADO, lw=2, label='Linea de accion')
ax.plot([la_start_x, la_end_x], [-la_start_y, -la_end_y], '-', color=COLOR_MORADO, lw=2)

# Pitch point
ax.plot(r1, 0, 'o', color=COLOR_ROJO, ms=8, zorder=5)
ax.text(r1 + 0.5, -0.8, 'Punto\nde paso', fontsize=9, color=COLOR_ROJO)

# Center distance
ax.annotate('', xy=(a_c, -r2 - 3.5), xytext=(0, -r1 - 3.5), arrowprops=dict(arrowstyle='<->', color=COLOR_FIJO, lw=1.5))
ax.text(a_c / 2, -r2 - 4.2, f'$a_c = m(Z_1+Z_2)/2 = {a_c}$', fontsize=10, ha='center', color=COLOR_FIJO)

# Rotation arrows
from matplotlib.patches import Arc as ArcPatch
arc1 = ArcPatch((0, 0), 2 * (r1 + 2), 2 * (r1 + 2), angle=0, theta1=30, theta2=100, color=COLOR_PRINCIPAL, lw=1.5)
ax.add_patch(arc1)
ax.annotate('', xy=(0 + (r1+2)*np.cos(np.radians(100)), (r1+2)*np.sin(np.radians(100))),
            xytext=(0 + (r1+2)*np.cos(np.radians(95)), (r1+2)*np.sin(np.radians(95))),
            arrowprops=dict(arrowstyle='->', color=COLOR_PRINCIPAL, lw=1.5))
ax.text(-3, r1 + 1, '$\\omega_1$', fontsize=13, color=COLOR_PRINCIPAL)

arc2 = ArcPatch((a_c, 0), 2 * (r2 + 2), 2 * (r2 + 2), angle=0, theta1=80, theta2=150, color=COLOR_PUNTO, lw=1.5)
ax.add_patch(arc2)
ax.text(a_c - 6, r2 + 1, '$\\omega_2$', fontsize=13, color=COLOR_PUNTO)

# Legend
ax.plot([], [], '-', color=COLOR_PRINCIPAL, lw=1.5, label='Primitivo')
ax.plot([], [], '--', color=COLOR_ROJO, lw=1.5, label='Cabeza')
ax.plot([], [], '--', color=COLOR_PUNTO, lw=1.5, label='Pie')
ax.plot([], [], ':', color=COLOR_NARANJA, lw=1.5, label='Base')
ax.legend(fontsize=9, loc='upper right')

ax.set_aspect('equal')
ax.set_title(f'Engranaje - Pinon y Rueda ($m = {m_g}$, $\\alpha = 20^\\circ$)', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

## 4. Trenes de Engranajes

### 4.1 Tren simple

Engranajes en serie, cada eje tiene un solo engranaje:

$$i = \frac{\omega_1}{\omega_n} = (-1)^{n-1}\frac{Z_n}{Z_1}$$

Los engranajes intermedios (**locos**) no afectan la relacion, solo el signo.

### 4.2 Tren compuesto

Algunos ejes tienen 2 engranajes solidarios:

$$i = \frac{\omega_{\text{entrada}}}{\omega_{\text{salida}}} = \frac{\text{producto } Z_{\text{conducidas}}}{\text{producto } Z_{\text{motoras}}}$$

### 4.3 Tren epicicloidal (planetario)

Contiene un **brazo** giratorio que porta engranajes **planetarios** (satelites) que orbitan alrededor de un engranaje **solar** (central).

Componentes:
- **Solar** (S): engranaje central fijo o movil
- **Planetario** (P): engranaje(s) montado(s) en el brazo
- **Corona** (C): engranaje anular exterior
- **Brazo** (B): porta-planetarios

### Ecuacion de Willis

$$\boxed{i_{FL}^B = \frac{\omega_L - \omega_B}{\omega_F - \omega_B}}$$

donde $F$ = primer engranaje, $L$ = ultimo engranaje, $B$ = brazo.

El valor de $i_{FL}^B$ se calcula como si el brazo estuviera fijo (tren ordinario):

$$i_{FL}^B = (-1)^{n_{ext}} \prod \frac{Z_{\text{motoras}}}{Z_{\text{conducidas}}}$$

$n_{ext}$ = numero de contactos externos.

```python
fig, ax = plt.subplots(figsize=(9, 9))
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('Tren Epicicloidal (Planetario)', fontsize=14, fontweight='bold')

# Solar gear
r_solar = 1.5
theta_c = np.linspace(0, 2 * np.pi, 200)
ax.plot(r_solar * np.cos(theta_c), r_solar * np.sin(theta_c), color=COLOR_PRINCIPAL, lw=2.5)
ax.fill(r_solar * np.cos(theta_c), r_solar * np.sin(theta_c), color=COLOR_PRINCIPAL, alpha=0.15)
ax.text(0, 0, 'SOLAR\n$Z_S$', fontsize=11, ha='center', va='center', fontweight='bold', color=COLOR_PRINCIPAL)

# Ring gear (corona)
r_ring_inner = 4.0
r_ring_outer = 4.7
ax.plot(r_ring_inner * np.cos(theta_c), r_ring_inner * np.sin(theta_c), color=COLOR_ROJO, lw=2.5)
ax.plot(r_ring_outer * np.cos(theta_c), r_ring_outer * np.sin(theta_c), color=COLOR_ROJO, lw=2.5)
ax.fill_between(r_ring_inner * np.cos(theta_c), r_ring_inner * np.sin(theta_c),
                r_ring_outer * np.cos(theta_c), r_ring_outer * np.sin(theta_c),
                color=COLOR_ROJO, alpha=0.1)
ax.text(0, 4.35, 'CORONA $Z_C$', fontsize=10, ha='center', color=COLOR_ROJO, fontweight='bold')

# Planet gears (3)
r_planet = 1.0
r_orbit = (r_solar + r_ring_inner) / 2
for i in range(3):
    angle = i * 2 * np.pi / 3 + np.pi / 6
    cx = r_orbit * np.cos(angle)
    cy = r_orbit * np.sin(angle)
    ax.plot(cx + r_planet * np.cos(theta_c), cy + r_planet * np.sin(theta_c), color=COLOR_PUNTO, lw=2)
    ax.fill(cx + r_planet * np.cos(theta_c), cy + r_planet * np.sin(theta_c), color=COLOR_PUNTO, alpha=0.15)
    ax.text(cx, cy, f'P{i+1}', fontsize=10, ha='center', va='center', fontweight='bold', color=COLOR_PUNTO)
    # Arm line
    ax.plot([0, cx], [0, cy], color=COLOR_NARANJA, lw=2, ls='--')
    ax.plot(cx, cy, 'o', color=COLOR_NARANJA, ms=5)

# Center
ax.plot(0, 0, 'o', color=COLOR_FIJO, ms=8)

# Arm label
ax.annotate('BRAZO (B)', xy=(r_orbit * 0.5 * np.cos(np.pi/6), r_orbit * 0.5 * np.sin(np.pi/6)),
            fontsize=11, color=COLOR_NARANJA, fontweight='bold',
            xytext=(3.5, -3), arrowprops=dict(arrowstyle='->', color=COLOR_NARANJA, lw=1.5))

# Formulas
ax.text(-5.5, -5, 'Willis: $\\frac{\\omega_C - \\omega_B}{\\omega_S - \\omega_B} = -\\frac{Z_S}{Z_C}$',
        fontsize=12, color=COLOR_MORADO, bbox=dict(boxstyle='round', facecolor='#f8f0ff', edgecolor=COLOR_MORADO))
ax.text(-5.5, -5.8, '$Z_C = Z_S + 2\\,Z_P$ (condicion geometrica)',
        fontsize=10, color=COLOR_FIJO)

plt.tight_layout()
plt.show()
```

## 5. Tipos Especiales de Engranajes

### 5.1 Engranajes helicoidales

Dientes inclinados un angulo $\beta$ respecto al eje:
- Modulo normal: $m_n = m_t \cos\beta$
- Paso normal: $p_n = p_t \cos\beta$
- $Z$ virtual: $Z_v = Z / \cos^3\beta$ (para calculo de interferencia)
- Fuerza axial: $F_a = F_t \tan\beta$

### 5.2 Engranajes conicos

- Ejes concurrentes (tipicamente $\Sigma = 90^\circ$)
- $i = Z_2/Z_1 = \tan\delta_2 / \tan\delta_1$ con $\delta_1 + \delta_2 = \Sigma$
- Modulo y paso varian a lo largo del ancho

### 5.3 Tornillo sin fin

- Gran reduccion en una etapa: $i = Z_{\text{rueda}} / Z_{\text{tornillo}}$
- $Z_{\text{tornillo}} = 1, 2, 3$ o $4$ (numero de entradas)
- Generalmente **irreversible** (autoretension) si angulo de avance < angulo de friccion

```python
def gear_params(m, Z, alpha_deg=20):
    """Calcula todos los parametros geometricos de un engranaje."""
    alpha = np.radians(alpha_deg)
    d = m * Z              # diametro primitivo
    da = m * (Z + 2)       # diametro de cabeza
    df = m * (Z - 2.5)     # diametro de pie
    db = d * np.cos(alpha)  # diametro de base
    ha = m                  # adendum
    hf = 1.25 * m           # dedendum
    p = np.pi * m           # paso
    return {'Z': Z, 'm': m, 'd': d, 'da': da, 'df': df, 'db': db,
            'ha': ha, 'hf': hf, 'p': p, 'alpha': alpha_deg}
```

## 6. Ejercicios Resueltos

### Ejercicio 1 - Parametros del engranaje

Un par de engranajes rectos tiene modulo $m = 3$ mm. El pinon tiene $Z_1 = 18$ dientes y la rueda $Z_2 = 54$ dientes. Angulo de presion $\alpha = 20^\circ$.

**Calcular:** todos los parametros geometricos y la relacion de transmision.

---

**Datos:** $m = 3$ mm, $Z_1 = 18$, $Z_2 = 54$, $\alpha = 20^\circ$

**Paso 1:** Parametros del pinon ($Z_1 = 18$)

| Parametro | Formula | Valor |
|---|---|---|
| Diametro primitivo | $d_1 = m \cdot Z_1 = 3 \times 18$ | $54$ mm |
| Diametro de cabeza | $d_{a1} = m(Z_1 + 2) = 3 \times 20$ | $60$ mm |
| Diametro de pie | $d_{f1} = m(Z_1 - 2.5) = 3 \times 15.5$ | $46.5$ mm |
| Diametro de base | $d_{b1} = d_1 \cos\alpha = 54 \cos 20^\circ$ | $50.743$ mm |
| Adendum | $h_a = m$ | $3$ mm |
| Dedendum | $h_f = 1.25\,m$ | $3.75$ mm |
| Paso circular | $p = \pi\,m = 3\pi$ | $9.425$ mm |

**Paso 2:** Parametros de la rueda ($Z_2 = 54$)

| Parametro | Formula | Valor |
|---|---|---|
| Diametro primitivo | $d_2 = m \cdot Z_2 = 3 \times 54$ | $162$ mm |
| Diametro de cabeza | $d_{a2} = m(Z_2 + 2) = 3 \times 56$ | $168$ mm |
| Diametro de pie | $d_{f2} = m(Z_2 - 2.5) = 3 \times 51.5$ | $154.5$ mm |
| Diametro de base | $d_{b2} = d_2 \cos\alpha = 162 \cos 20^\circ$ | $152.230$ mm |

**Paso 3:** Relacion de transmision y distancia entre centros

$$i = \frac{Z_2}{Z_1} = \frac{54}{18} = \boxed{3.0}$$

$$a_c = \frac{m(Z_1 + Z_2)}{2} = \frac{3 \times 72}{2} = \boxed{108.0 \text{ mm}}$$

**Paso 4:** Verificacion de interferencia

$$Z_{\min} = \frac{2}{\sin^2 \alpha} = \frac{2}{\sin^2 20^\circ} \approx 17.1$$

$Z_1 = 18 \geq 17.1$ $\Rightarrow$ **No hay interferencia**

### Ejercicio 2 - Tren simple

Un tren de engranajes simple tiene 4 engranajes: $Z_1 = 20$, $Z_2 = 40$, $Z_3 = 30$, $Z_4 = 60$. El eje de entrada gira a $\omega_1 = 1500$ rpm.

**Calcular:** velocidad de salida $\omega_4$ y sentido de giro.

---

**Datos:** $Z_1 = 20$, $Z_2 = 40$, $Z_3 = 30$, $Z_4 = 60$, $\omega_1 = 1500$ rpm

**Paso 1:** Identificar tipo de tren

Engranajes en serie (cada eje tiene un solo engranaje) $\Rightarrow$ **tren simple**.

$Z_2$ y $Z_3$ son engranajes **locos** (intermedios): no afectan a $|i|$, solo al signo.

**Paso 2:** Relacion de transmision

$$i = (-1)^{n-1} \frac{Z_n}{Z_1}$$

Con $n = 4$ engranajes:

$$i = (-1)^{3} \cdot \frac{Z_4}{Z_1} = (-1) \cdot \frac{60}{20} = -3.0$$

**Paso 3:** Velocidad de salida

$$\omega_4 = \frac{\omega_1}{|i|} = \frac{1500}{3} = \boxed{500 \text{ rpm}}$$

**Resultado:** $i = -3$ (signo negativo $\Rightarrow$ sentido **opuesto** a $\omega_1$, numero impar de contactos externos)

### Ejercicio 3 - Tren compuesto

Un reductor de dos etapas: $Z_1 = 15$ engrana con $Z_2 = 45$ (primera etapa); en el mismo eje de $Z_2$ va solidario $Z_3 = 18$ que engrana con $Z_4 = 72$ (segunda etapa). Entrada a $\omega_1 = 3000$ rpm.

**Calcular:** relacion total y velocidad de salida.

---

**Datos:** $Z_1 = 15$, $Z_2 = 45$, $Z_3 = 18$, $Z_4 = 72$, $\omega_1 = 3000$ rpm

($Z_2$ y $Z_3$ son solidarios en el mismo eje)

**Paso 1:** Primera etapa ($Z_1 \to Z_2$)

$$i_1 = \frac{Z_2}{Z_1} = \frac{45}{15} = 3.00$$

**Paso 2:** Segunda etapa ($Z_3 \to Z_4$)

$$i_2 = \frac{Z_4}{Z_3} = \frac{72}{18} = 4.00$$

**Paso 3:** Relacion total

$$i_{\text{total}} = i_1 \cdot i_2 = 3 \times 4 = \boxed{12}$$

(Reduccion $12:1$)

**Paso 4:** Velocidad de salida

$$\omega_4 = \frac{\omega_1}{i_{\text{total}}} = \frac{3000}{12} = \boxed{250 \text{ rpm}}$$

Sentido: **mismo** que $\omega_1$ (dos contactos externos, ambos invierten el sentido)

### Ejercicio 4 - Tren planetario (Willis)

Un tren epicicloidal tiene: solar $Z_S = 30$, planetario $Z_P = 20$, corona $Z_C = 70$. El solar gira a $\omega_S = 1000$ rpm y la corona esta fija ($\omega_C = 0$).

**Calcular:** velocidad del brazo $\omega_B$.

---

**Datos:** $Z_S = 30$, $Z_P = 20$, $Z_C = 70$, $\omega_S = 1000$ rpm, $\omega_C = 0$ rpm

**Paso 1:** Verificacion geometrica

$$Z_C = Z_S + 2\,Z_P \quad \Rightarrow \quad 70 = 30 + 2 \times 20 = 70 \quad \checkmark$$

**Paso 2:** Calculo de $i_{SC}^B$ (con brazo fijo)

Cadena $S \to P$ (contacto externo) $\to C$ (contacto interno):

$$i_{SC}^B = \frac{-Z_S}{Z_P} \cdot \frac{Z_P}{Z_C} = -\frac{Z_S}{Z_C} = -\frac{30}{70} = -0.4286$$

**Paso 3:** Aplicar ecuacion de Willis

$$\frac{\omega_C - \omega_B}{\omega_S - \omega_B} = i_{SC}^B$$

$$\frac{0 - \omega_B}{1000 - \omega_B} = -0.4286$$

**Paso 4:** Despejar $\omega_B$

$$\omega_B = \frac{\omega_C - i_{SC}^B \cdot \omega_S}{1 - i_{SC}^B} = \frac{0 - (-0.4286)(1000)}{1 - (-0.4286)} = \frac{428.6}{1.4286}$$

$$\boxed{\omega_B = 300 \text{ rpm}}$$

**Verificacion:**

$$\frac{\omega_C - \omega_B}{\omega_S - \omega_B} = \frac{0 - 300}{1000 - 300} = \frac{-300}{700} = -0.4286 = i_{SC}^B \quad \checkmark$$

### Estrategia general de resolucion

1. **Identificar tipo de engranaje** y sistema (metrico/imperial)
2. **Geometria**: a partir de $m$ y $Z$ calcular todos los diametros
3. **Trenes simples/compuestos**: multiplicar relaciones por etapa, cuidar signos
4. **Planetarios**: 
   - Identificar $F$ (primer engranaje), $L$ (ultimo), $B$ (brazo)
   - Calcular $i_{FL}^B$ como tren ordinario (brazo fijo)
   - Aplicar Willis: $(\omega_L - \omega_B)/(\omega_F - \omega_B) = i_{FL}^B$
   - Imponer la condicion de contorno (elemento fijo o velocidad conocida)
5. **Verificar interferencia** si $Z$ es pequeno

### Tornillo sin fin - Formulas clave

| Parametro | Formula |
|---|---|
| Relacion de transmision | $i = Z_{\text{rueda}} / Z_{\text{tornillo}}$ |
| Angulo de avance | $\gamma = \arctan(Z_t \cdot m / d_w)$ |
| Eficiencia | $\eta = \frac{\tan\gamma}{\tan(\gamma + \rho)}$ |
| Autoretension | $\gamma < \rho$ (angulo de friccion) |

Para $Z_t = 1$ entrada: $i = Z_r / 1 = Z_r$ (reducciones de 20:1 a 100:1 comunes).

### Configuraciones tipicas de trenes planetarios

| Configuracion | Fijo | Entrada | Salida | $i$ (tipico) |
|---|---|---|---|---|
| Solar fija | Solar | Corona | Brazo | $1 + Z_S/Z_C < 1$ |
| Corona fija | Corona | Solar | Brazo | $1 + Z_C/Z_S > 1$ |
| Brazo fijo | Brazo | Solar | Corona | $-Z_S/Z_C$ (tren ordinario) |

**Ejemplo numerico:** $Z_S = 30$, $Z_P = 20$, $Z_C = 70$

- Corona fija, $S \to B$: $\;i = 1 + Z_C/Z_S = 1 + 70/30 = 3.333$
- Solar fija, $C \to B$: $\;i = 1 + Z_S/Z_C = 1 + 30/70 = 1.429$
- Brazo fijo, $S \to C$: $\;i = -Z_S/Z_C = -30/70 = -0.4286$

**Notas finales:**
- El modulo debe ser **igual** en ambos engranajes para que engranen
- Willis siempre se aplica entre el **primer** y **ultimo** engranaje del tren (no el brazo)
- La condicion geometrica $Z_C = Z_S + 2Z_P$ es necesaria para el montaje
- Para gran reduccion en poco espacio: trenes planetarios o tornillo sin fin
