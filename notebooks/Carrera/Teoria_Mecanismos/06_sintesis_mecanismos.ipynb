## 1. Formulario del Tema

### Sintesis estructural - Ecuacion de Grubler

$$\boxed{G = 3\,(N-1) - 2\,P_I - P_{II}}$$

donde $N$ = eslabones, $P_I$ = pares de clase I (1 GDL), $P_{II}$ = pares de clase II (2 GDL), $G$ = grados de libertad.

### Relacion tipo-numero

$$\boxed{N - (G + 3) = T + 2\,Q + 3\,P}$$

donde $T$ = ternarios, $Q$ = cuaternarios, $P$ = pentagonales.

Con eslabones binarios ($B$):

$$\boxed{B = G + 3 + T + 2\,Q + 3\,P}$$

### Ecuacion de Freudenstein

$$\boxed{R_1 \cos\phi - R_2 \cos\psi + R_3 = \cos(\phi - \psi)}$$

donde:

$$R_1 = \frac{d}{a}, \quad R_2 = \frac{d}{c}, \quad R_3 = \frac{a^2 - b^2 + c^2 + d^2}{2\,a\,c}$$

$(a, b, c, d)$: manivela, biela, balancin, bancada.

### Espaciado de Chebyshev ($n$ puntos de precision)

$$\boxed{\phi_j = \frac{\phi_s + \phi_f}{2} - \frac{\phi_f - \phi_s}{2}\cos\left(\frac{2j-1}{2n}\pi\right), \quad j = 1,2,\ldots,n}$$

### Error estructural

Diferencia entre la funcion deseada $\psi_d(\phi)$ y la generada $\psi_g(\phi)$:

$$e(\phi) = \psi_d(\phi) - \psi_g(\phi)$$

## 4. Ecuacion de Freudenstein

### Derivacion

Para un cuadrilatero articulado (4 barras) con eslabones $a$ (manivela), $b$ (biela), $c$ (balancin), $d$ (bancada):

Cerramiento vectorial:
$$a\,e^{i\phi} + b\,e^{i\lambda} = d + c\,e^{i\psi}$$

Eliminando el angulo de la biela $\lambda$:

$$\boxed{R_1 \cos\phi - R_2 \cos\psi + R_3 = \cos(\phi - \psi)}$$

con:
$$R_1 = \frac{d}{a}, \quad R_2 = \frac{d}{c}, \quad R_3 = \frac{a^2 - b^2 + c^2 + d^2}{2\,a\,c}$$

### Uso para sintesis de generacion de funcion

Dados **3 puntos de precision** $(\phi_j, \psi_j)$ para $j = 1, 2, 3$, se obtiene un sistema lineal $3 \times 3$ en las incognitas $(R_1, R_2, R_3)$:

$$\begin{bmatrix} \cos\phi_1 & -\cos\psi_1 & 1 \\ \cos\phi_2 & -\cos\psi_2 & 1 \\ \cos\phi_3 & -\cos\psi_3 & 1 \end{bmatrix} \begin{bmatrix} R_1 \\ R_2 \\ R_3 \end{bmatrix} = \begin{bmatrix} \cos(\phi_1 - \psi_1) \\ \cos(\phi_2 - \psi_2) \\ \cos(\phi_3 - \psi_3) \end{bmatrix}$$

De $R_1, R_2, R_3$ se recuperan las longitudes $(a, b, c, d)$ fijando una (normalmente $d = 1$).

## 5. Espaciado de Chebyshev

Para minimizar el **error estructural** maximo al generar una funcion, los puntos de precision deben distribuirse segun Chebyshev:

$$\phi_j = \frac{\phi_s + \phi_f}{2} - \frac{\phi_f - \phi_s}{2}\cos\left(\frac{2j - 1}{2n}\pi\right), \quad j = 1, \ldots, n$$

Donde $\phi_s$ y $\phi_f$ son los extremos del rango de la variable de entrada.

Analogas formulas aplican para $\psi_j$ si se conoce la funcion deseada $\psi = f(\phi)$.

```python
def freudenstein_synthesis(phi_pts, psi_pts, d=1.0):
    """Sintesis de Freudenstein para 3 puntos de precision.
    phi_pts, psi_pts: angulos en radianes.
    d: longitud de bancada (fijada).
    Retorna: a, b, c (longitudes de manivela, biela, balancin)."""
    A = np.array([
        [np.cos(phi_pts[i]), -np.cos(psi_pts[i]), 1]
        for i in range(3)
    ])
    rhs = np.array([np.cos(phi_pts[i] - psi_pts[i]) for i in range(3)])
    R = np.linalg.solve(A, rhs)
    R1, R2, R3 = R
    # d/a = R1 => a = d/R1
    a = d / R1
    c = d / R2
    # R3 = (a^2 - b^2 + c^2 + d^2) / (2*a*c)
    b2 = a**2 + c**2 + d**2 - 2 * a * c * R3
    b = np.sqrt(abs(b2))
    return a, b, c, R1, R2, R3

def chebyshev_spacing(xs, xf, n):
    """Espaciado de Chebyshev para n puntos de precision."""
    j = np.arange(1, n + 1)
    return (xs + xf) / 2 - (xf - xs) / 2 * np.cos((2 * j - 1) / (2 * n) * np.pi)
```

### Ejemplo: Generar $\psi = f(\phi)$ con 3 puntos Chebyshev

Disenar un cuadrilatero articulado que genere $\psi = 60 + 40\sin(\phi - 30)$ (grados) para $\phi \in [30, 120]$ grados, usando 3 puntos de precision de Chebyshev.

```python
# Ejemplo de Freudenstein con Chebyshev
phi_s, phi_f = np.radians(30), np.radians(120)
phi_cheb = chebyshev_spacing(phi_s, phi_f, 3)

# Funcion deseada
def psi_desired(phi):
    return np.radians(60 + 40 * np.sin(np.degrees(phi) - 30))

psi_cheb = psi_desired(phi_cheb)

for j, (p, ps) in enumerate(zip(phi_cheb, psi_cheb), 1):

# Resolver Freudenstein
d = 1.0  # Fijar bancada
a, b, c, R1, R2, R3 = freudenstein_synthesis(phi_cheb, psi_cheb, d)
```

## 6. Curva del Acoplador

La **curva del acoplador** es la trayectoria de un punto $P$ fijo en la biela del cuadrilatero articulado. Estas curvas pueden aproximar trayectorias complejas (rectas, circulos, figuras de 8, etc.).

```python
def four_bar_positions(a, b, c, d, phi, coupler_x=0.5, coupler_y=0.3):
    """Calcula posiciones del cuadrilatero para un array de phi.
    Devuelve posiciones de A, B, P (punto del acoplador)."""
    Ax = a * np.cos(phi)
    Ay = a * np.sin(phi)
    # Resolver posicion de B usando cerramiento
    Bx_list, By_list, Px_list, Py_list = [], [], [], []
    for i in range(len(phi)):
        ax_i, ay_i = Ax[i], Ay[i]
        # Distancia de A a D fijo en (d, 0)
        dx = d - ax_i
        dy = -ay_i
        dist_AD = np.sqrt(dx**2 + dy**2)
        if dist_AD > b + c or dist_AD < abs(b - c):
            Bx_list.append(np.nan)
            By_list.append(np.nan)
            Px_list.append(np.nan)
            Py_list.append(np.nan)
            continue
        # Triangle ABD: find angle at D
        cos_alpha = (c**2 + dist_AD**2 - b**2) / (2 * c * dist_AD)
        cos_alpha = np.clip(cos_alpha, -1, 1)
        alpha = np.arccos(cos_alpha)
        beta = np.arctan2(ay_i - 0, ax_i - d)
        psi_i = beta + alpha  # or beta - alpha
        bx = d + c * np.cos(psi_i)
        by = c * np.sin(psi_i)
        Bx_list.append(bx)
        By_list.append(by)
        # Coupler point
        ab_x, ab_y = bx - ax_i, by - ay_i
        ab_len = np.sqrt(ab_x**2 + ab_y**2)
        ux, uy = ab_x / ab_len, ab_y / ab_len
        perp_x, perp_y = -uy, ux
        px = ax_i + coupler_x * b * ux + coupler_y * b * perp_x
        py = ay_i + coupler_x * b * uy + coupler_y * b * perp_y
        Px_list.append(px)
        Py_list.append(py)
    return Ax, Ay, np.array(Bx_list), np.array(By_list), np.array(Px_list), np.array(Py_list)

# Example four-bar
a_ex, b_ex, c_ex, d_ex = 1.0, 2.5, 2.0, 2.8
phi_range = np.linspace(0, 2 * np.pi, 500)
Ax, Ay, Bx, By, Px, Py = four_bar_positions(a_ex, b_ex, c_ex, d_ex, phi_range, 0.5, 0.4)

fig, ax = plt.subplots(figsize=(9, 7))
# Coupler curve
ax.plot(Px, Py, color=COLOR_MORADO, lw=2, label='Curva del acoplador')

# Draw one position
phi0 = np.radians(45)
Ax0, Ay0, Bx0, By0, Px0, Py0 = four_bar_positions(a_ex, b_ex, c_ex, d_ex, np.array([phi0]), 0.5, 0.4)
ax.plot([0, Ax0[0]], [0, Ay0[0]], 'o-', color=COLOR_PRINCIPAL, lw=2.5, ms=8, label='Manivela (a)')
ax.plot([Ax0[0], Bx0[0]], [Ay0[0], By0[0]], 'o-', color=COLOR_ROJO, lw=2.5, ms=8, label='Biela (b)')
ax.plot([d_ex, Bx0[0]], [0, By0[0]], 'o-', color=COLOR_PUNTO, lw=2.5, ms=8, label='Balancin (c)')
ax.plot([0, d_ex], [0, 0], 's-', color=COLOR_FIJO, lw=2.5, ms=10, mfc='white', mew=2, label='Bancada (d)')
ax.plot(Px0[0], Py0[0], '*', color=COLOR_NARANJA, ms=15, zorder=5, label='Punto P')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('Curva del Acoplador de un Cuadrilatero Articulado', fontsize=13, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Generacion de Movimiento

En este problema, un cuerpo rigido (acoplador) debe pasar por **posiciones prescritas** (posicion + orientacion).

- Con **2 posiciones**: solucion infinita (se eligen las articulaciones)
- Con **3 posiciones**: solucion finita (sistema determinado)
- Con **4 posiciones**: sistema sobredeterminado (se usa optimizacion)

### Metodo grafico (2 posiciones)

1. Elegir dos puntos $A$ y $B$ en el acoplador
2. Dibujar las posiciones $A_1, A_2$ y $B_1, B_2$
3. Las articulaciones fijas se ubican en las **mediatrices** de $A_1A_2$ y $B_1B_2$

$$3\,(N-1) - 2\,P_I = G \implies P_I = \frac{3 \cdot 7 - 1}{2} = 10$$

**Paso 2:** Relacion tipo-numero

$$N - (G + 3) = T + 2Q + 3P \implies 8 - 4 = 4$$

**Paso 3:** Combinaciones posibles $(T, Q, P)$ con $B = G + 3 + T + 2Q + 3P$

| T | Q | P | B | N total |
|---|---|---|---|---|
| 4 | 0 | 0 | 8 | 8 |
| 2 | 1 | 0 | 7 | 8 |
| 0 | 2 | 0 | 6 | 8 |
| 1 | 0 | 1 | 5 | 8 |

**Resultado:** $\boxed{4 \text{ combinaciones posibles de cadenas cinematicas}}$

```python
# Ejercicio 1: Sintesis estructural N=8, G=1
N = 8
G = 1
P_I = (3 * (N - 1) - G) // 2
residuo = N - (G + 3)

soluciones = []
for P in range(residuo // 3 + 1):
    for Q in range((residuo - 3*P) // 2 + 1):
        T = residuo - 2*Q - 3*P
        if T >= 0:
            B = G + 3 + T + 2*Q + 3*P
            soluciones.append((T, Q, P, B))
```

Disenar un mecanismo de manivela-corredera con **relacion de tiempos** $Q = t_{avance}/t_{retorno} = 1.4$. Calcular el angulo de presion $\alpha$.

**Datos:** $Q = 1.4$

**Paso 1:** Relacion entre $Q$ y $\alpha$

$$Q = \frac{180° + \alpha}{180° - \alpha}$$

**Paso 2:** Despejando $\alpha$

$$\alpha = \frac{180°\,(Q - 1)}{1 + Q} = \frac{180° \times 0.4}{2.4} = 30°$$

**Paso 3:** Verificacion

- Angulo de avance: $180° + 30° = 210°$
- Angulo de retorno: $180° - 30° = 150°$
- $Q = 210° / 150° = 1.4$ \checkmark

**Resultado:** $\boxed{\alpha = 30°}$

## 10. Criterio de Grashof (Recordatorio)

Para que un eslabon pueda dar vueltas completas:

$$\boxed{s + l \leq p + q}$$

donde $s$ = eslabon mas corto, $l$ = mas largo, $p$ y $q$ los otros dos.

- **Cumple Grashof**: al menos un eslabon puede rotar completamente
- **No cumple**: ninguno puede; todos son balancines (doble balancin)
