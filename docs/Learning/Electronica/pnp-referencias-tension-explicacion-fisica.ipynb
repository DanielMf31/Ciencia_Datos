## 6. Resumen visual: NPN vs PNP -- la "chuleta" definitiva

### Regla de oro

> **PNP = NPN dado la vuelta, con las mismas ecuaciones pero usando VEB y VEC en vez de VBE y VCE.**

### Tabla resumen completa

| Propiedad | NPN | PNP |
|:--|:--|:--|
| **Estructura** | N - P - N | P - N - P |
| **Portadores del emisor** | Electrones (e-, negativos) | Huecos (h+, positivos) |
| **Flujo de portadores** | E -> B -> C | E -> B -> C |
| **Corriente convencional** | Entra por C y B, sale por E | Entra por E, sale por B y C |
| **Flecha del emisor** | Sale del transistor | Entra al transistor |
| **VCC conecta a...** | Colector (via RC) | Emisor |
| **GND conecta a...** | Emisor | Colector (via RC) |
| **Tension union activa** | $V_{BE} = +0.7\,V$ | $V_{EB} = +0.7\,V$ |
| **Tension saturacion** | $V_{CE,sat} \approx 0.2\,V$ | $V_{EC,sat} \approx 0.2\,V$ |
| **Malla entrada** | $V_{CC} = I_B R_B + V_{BE}$ | $V_{CC} = I_B R_B + V_{EB}$ |
| **Malla salida** | $V_{CC} = I_C R_C + V_{CE}$ | $V_{CC} = I_C R_C + V_{EC}$ |

### La cadena logica completa

1. El emisor PNP es tipo P -> portadores mayoritarios = huecos (cargas +)
2. La corriente convencional sigue a las cargas + -> corriente entra por emisor
3. Para que los huecos fluyan E -> B, necesitamos VE > VB -> usamos VEB para que sea positivo
4. VCC debe estar en el emisor (punto mas alto) para que los huecos fluyan "cuesta abajo"
5. Las ecuaciones son identicas a NPN, solo con VEB y VEC

### Conexion con otros temas

- **Truco RB/RC vs beta:** El truco para determinar si esta en activa o saturacion funciona IGUAL en PNP. Solo recuerda usar VEB=0.7 y VEC,sat=0.2. Ver [notebook del truco](./truco-rb-rc-beta-estado-bjt.ipynb).
- **Diodos:** La union EB de un PNP en directa es identica a un diodo con el anodo en E y catodo en B (caida de 0.7V). En NPN el anodo esta en B y el catodo en E.
- **MOSFET:** En los PMOS ocurre algo analogo: la tension de puerta debe ser mas BAJA que la fuente para conducir, y se usa $V_{SG}$ en vez de $V_{GS}$.

````python
# ============================================================
# Comparacion: resolver PNP con VEB vs con VBE
# ============================================================
print("=" * 60)
print("  PNP resuelto con VEB (convencion comoda)")
print("=" * 60)
VEB = 0.7  # positivo
IB = (VCC - VEB) / RB
IC = beta * IB
VEC = VCC - IC * RC
print(f"  VEB = +{VEB} V")
print(f"  IB  = (VCC - VEB) / RB = ({VCC} - {VEB}) / {RB/1e3:.0f}k = {IB*1e6:.2f} uA")
print(f"  IC  = {IC*1e3:.3f} mA")
print(f"  VEC = VCC - IC*RC = {VEC:.2f} V")
print(f"  --> Todo positivo, facil de leer\n")

print("=" * 60)
print("  PNP resuelto con VBE (convencion incomoda)")
print("=" * 60)
VBE = -0.7  # negativo!
IB2 = (VCC + VBE) / RB  # nota: VCC + VBE = VCC + (-0.7) = VCC - 0.7
IC2 = beta * IB2
VCE = -(VCC - IC2 * RC)  # negativo!
print(f"  VBE = {VBE} V  <-- negativo!")
print(f"  IB  = (VCC + VBE) / RB = ({VCC} + ({VBE})) / {RB/1e3:.0f}k = {IB2*1e6:.2f} uA")
print(f"  IC  = {IC2*1e3:.3f} mA")
print(f"  VCE = -(VCC - IC*RC) = {VCE:.2f} V  <-- negativo!")
print(f"  --> Mismos numeros, pero con signos negativos que confunden\n")

print("=" * 60)
print("  Verificacion: |VEC| == |VCE|?")
print(f"  |VEC| = {abs(VEC_pnp):.2f} V")
print(f"  |VCE| = {abs(VCE):.2f} V")
print(f"  Iguales? {np.isclose(abs(VEC_pnp), abs(VCE))}")
print("=" * 60)
````

## 5. Que pasa si usas VBE en un PNP en vez de VEB?

Puedes hacerlo perfectamente. La fisica no se rompe. Pero los signos se vuelven negativos:

### Ejemplo: PNP en activa usando VBE (en vez de VEB)

Con los mismos datos ($V_{CC} = 12\,V$, $R_B = 240\,k\Omega$):

La malla de entrada se escribe como:

$$V_{CC} + V_{BE} = I_B \cdot R_B$$

Nota: aqui $V_{BE} = -0.7\,V$ (negativo, porque en PNP la base esta por debajo del emisor).

$$I_B = \frac{V_{CC} + V_{BE}}{R_B} = \frac{12 + (-0.7)}{240\,000} = \frac{11.3}{240\,000} = 47.08\,\mu A$$

El resultado es el mismo! Pero fijate que tuviste que pensar en $V_{BE} = -0.7$ y sumar $V_{CC} + V_{BE}$ en vez de restar $V_{CC} - V_{EB}$.

**Comparacion de ecuaciones:**

| Usando VEB (convencion PNP) | Usando VBE (convencion NPN) |
|:--|:--|
| $I_B = \frac{V_{CC} - V_{EB}}{R_B}$ | $I_B = \frac{V_{CC} + V_{BE}}{R_B}$ |
| $V_{EB} = +0.7\,V$ | $V_{BE} = -0.7\,V$ |
| $V_{EC} = V_{CC} - I_C R_C$ | $V_{CE} = -(V_{CC} - I_C R_C)$ |
| Todo positivo | Signos negativos por todas partes |

**Conclusion:** Usar $V_{EB}$ y $V_{EC}$ en PNP es simplemente mas comodo. Es como elegir entre conducir por la derecha o por la izquierda -- llegas al mismo sitio, pero uno de los dos te resulta mas natural.

````python
# ============================================================
# Calculo numerico: NPN vs PNP con los mismos componentes
# ============================================================
VCC = 12.0      # V
RB = 240e3      # Ohm
RC = 2.2e3      # Ohm
beta = 100

print("=" * 60)
print("  DEMOSTRACION: NPN vs PNP con mismos componentes")
print("=" * 60)
print(f"\n  Datos: VCC={VCC}V, RB={RB/1e3:.0f}kOhm, RC={RC/1e3:.1f}kOhm, beta={beta}")

# --- NPN ---
print("\n" + "-" * 60)
print("  NPN")
print("-" * 60)
VBE_npn = 0.7
IB_npn = (VCC - VBE_npn) / RB
IC_npn = beta * IB_npn
VCE_npn = VCC - IC_npn * RC

print(f"  Paso 1: VBE = {VBE_npn:.1f} V (positivo)")
print(f"  Paso 2: IB = (VCC - VBE) / RB = ({VCC} - {VBE_npn}) / {RB/1e3:.0f}k = {IB_npn*1e6:.2f} uA")
print(f"  Paso 3: IC = beta * IB = {beta} * {IB_npn*1e6:.2f}u = {IC_npn*1e3:.3f} mA")
print(f"  Paso 4: VCE = VCC - IC*RC = {VCC} - {IC_npn*1e3:.3f}m * {RC/1e3:.1f}k = {VCE_npn:.2f} V")

# --- PNP ---
print("\n" + "-" * 60)
print("  PNP")
print("-" * 60)
VEB_pnp = 0.7
IB_pnp = (VCC - VEB_pnp) / RB
IC_pnp = beta * IB_pnp
VEC_pnp = VCC - IC_pnp * RC

print(f"  Paso 1: VEB = {VEB_pnp:.1f} V (positivo)")
print(f"  Paso 2: IB = (VCC - VEB) / RB = ({VCC} - {VEB_pnp}) / {RB/1e3:.0f}k = {IB_pnp*1e6:.2f} uA")
print(f"  Paso 3: IC = beta * IB = {beta} * {IB_pnp*1e6:.2f}u = {IC_pnp*1e3:.3f} mA")
print(f"  Paso 4: VEC = VCC - IC*RC = {VCC} - {IC_pnp*1e3:.3f}m * {RC/1e3:.1f}k = {VEC_pnp:.2f} V")

# --- Comparacion ---
print("\n" + "=" * 60)
print("  COMPARACION")
print("=" * 60)
print(f"  IB_npn  = {IB_npn*1e6:.2f} uA    IB_pnp  = {IB_pnp*1e6:.2f} uA    Iguales? {np.isclose(IB_npn, IB_pnp)}")
print(f"  IC_npn  = {IC_npn*1e3:.3f} mA   IC_pnp  = {IC_pnp*1e3:.3f} mA   Iguales? {np.isclose(IC_npn, IC_pnp)}")
print(f"  VCE_npn = {VCE_npn:.2f} V       VEC_pnp = {VEC_pnp:.2f} V       Iguales? {np.isclose(VCE_npn, VEC_pnp)}")
print(f"\n  --> Los numeros son IDENTICOS. Solo cambian los nombres.")
````

## 4. Demostracion numerica: mismo circuito, NPN vs PNP

Vamos a resolver el **mismo circuito** (mismos valores de componentes) una vez como NPN y otra como PNP, para demostrar que los **numeros son identicos** y solo cambian los nombres de las variables.

### Datos del circuito
- $V_{CC} = 12\,V$
- $R_B = 240\,k\Omega$
- $R_C = 2.2\,k\Omega$
- $\beta = 100$

### Resolucion como NPN

**Malla de entrada** (base-emisor):

$$V_{CC} = I_B \cdot R_B + V_{BE}$$

Donde:
- $V_{CC} = 12\,V$: fuente de alimentacion
- $I_B$: corriente de base (lo que queremos encontrar)
- $R_B = 240\,k\Omega$: resistencia de base
- $V_{BE} = 0.7\,V$: caida en la union base-emisor (diodo en directa)

Despejando $I_B$:

$$I_B = \frac{V_{CC} - V_{BE}}{R_B} = \frac{12 - 0.7}{240\,000} = \frac{11.3}{240\,000} = 47.08\,\mu A$$

**Corriente de colector:**

$$I_C = \beta \cdot I_B = 100 \times 47.08\,\mu A = 4.708\,mA$$

**Malla de salida** (colector-emisor):

$$V_{CC} = I_C \cdot R_C + V_{CE}$$

$$V_{CE} = V_{CC} - I_C \cdot R_C = 12 - 4.708 \times 10^{-3} \times 2200 = 12 - 10.36 = 1.64\,V$$

$$V_{CC} = I_B \cdot R_B + V_{EB}$$

Donde:
- $V_{EB} = 0.7\,V$: caida en la union emisor-base (en PNP, VEB es positivo)

Despejando $I_B$:

$$I_B = \frac{V_{CC} - V_{EB}}{R_B} = \frac{12 - 0.7}{240\,000} = \frac{11.3}{240\,000} = 47.08\,\mu A$$

**Corriente de colector:**

$$I_C = \beta \cdot I_B = 100 \times 47.08\,\mu A = 4.708\,mA$$

**Malla de salida** (emisor-colector):

$$V_{CC} = I_C \cdot R_C + V_{EC}$$

$$V_{EC} = V_{CC} - I_C \cdot R_C = 12 - 4.708 \times 10^{-3} \times 2200 = 12 - 10.36 = 1.64\,V$$

### Comparacion directa

| Variable | NPN | PNP | Valor |

| Tension union activa | $V_{BE} = 0.7\,V$ | $V_{EB} = 0.7\,V$ | 0.7 V |
| Corriente de base | $I_B$ | $I_B$ | 47.08 uA |

## 3. Por que VCC se conecta al emisor en PNP (y al colector en NPN)

### Analogia: el emisor es la "fuente" de portadores

Piensa en el emisor como un grifo de agua:
- En **NPN**, el emisor "emite" electrones. Los electrones tienen carga negativa, asi que necesitan ir desde una tension baja (GND) hacia una tension alta. El emisor se conecta a GND y el colector (donde llegan los electrones) a VCC a traves de RC.
- En **PNP**, el emisor "emite" huecos. Los huecos tienen carga positiva, asi que necesitan ir desde una tension alta (VCC) hacia una tension baja. El emisor se conecta a VCC y el colector (donde llegan los huecos) a GND a traves de RC.

### Circuito NPN tipico (polarizacion fija)

```
        VCC = 12V
           |
          [RC]
           |
           C
     RB    |
VCC--[  ]--B   (NPN)
           |
           E
           |
          GND
```

Camino de la corriente: VCC -> RC -> C -> E -> GND

La corriente convencional entra por el colector y sale por el emisor.


### Circuito PNP tipico (polarizacion fija)

```
        VCC = 12V
           |
           E
     RB    |
GND--[  ]--B   (PNP)
           |
           C
           |
          [RC]
           |
          GND
```

Camino de la corriente: VCC -> E -> C -> RC -> GND

La corriente convencional entra por el emisor y sale por el colector.

### Por que funciona asi?

1. Para que los **huecos** fluyan del emisor a la base, necesitamos $V_E > V_B$ (union EB en directa).
2. Para que los huecos sean "recogidos" por el colector, necesitamos $V_E > V_C$ (union CB en inversa, que en PNP significa que el colector esta a menor tension que la base).
3. Si VCC esta en el emisor: $V_E = V_{CC}$ es el punto mas alto del circuito. $V_B$ estara un poco por debajo ($V_E - 0.7$). $V_C$ estara bastante por debajo. Todo cuadra.

**Es literalmente "dar la vuelta" al circuito NPN:** lo que antes estaba arriba (colector) ahora esta abajo, y viceversa.

````python
# ============================================================
# Visualizacion: VBE vs VEB segun el tipo de transistor
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- NPN ---
ax = axes[0]
ax.set_title('NPN: tensiones positivas con VBE, VCE', fontsize=12, fontweight='bold', color=C_DARK)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Barras de tension
# VE = 0V (referencia)
# VB = 0.7V
# VC = 5V (ejemplo)
tensiones_npn = {'E': 0, 'B': 0.7, 'C': 5.0}
colores = {'E': C_SUCCESS, 'B': C_ORANGE, 'C': C_PRIMARY}
x_pos = {'E': 2, 'B': 5, 'C': 8}

for term, v in tensiones_npn.items():
    bar_h = v * 1.2 + 0.3  # escala visual
    rect = plt.Rectangle((x_pos[term]-0.5, 1), 1, bar_h, color=colores[term], alpha=0.6)
    ax.add_patch(rect)
    ax.text(x_pos[term], 1 + bar_h + 0.3, f'{term}\n{v:.1f}V', ha='center',
            fontsize=11, fontweight='bold', color=C_DARK)

# Indicar VBE y VCE
ax.annotate('', xy=(5, 0.5), xytext=(2, 0.5),
            arrowprops=dict(arrowstyle='<->', color=C_DANGER, lw=2))
ax.text(3.5, 0.1, r'$V_{BE}$ = +0.7V', ha='center', fontsize=10, color=C_DANGER, fontweight='bold')

ax.annotate('', xy=(8, 8.5), xytext=(2, 8.5),
            arrowprops=dict(arrowstyle='<->', color=C_PURPLE, lw=2))
ax.text(5, 8.9, r'$V_{CE}$ = +5.0V', ha='center', fontsize=10, color=C_PURPLE, fontweight='bold')

# --- PNP ---
ax = axes[1]
ax.set_title('PNP: tensiones positivas con VEB, VEC', fontsize=12, fontweight='bold', color=C_DARK)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# VE = 12V (alto)
# VB = 11.3V
# VC = 7V (ejemplo)
tensiones_pnp = {'E': 12.0, 'B': 11.3, 'C': 7.0}

max_v = 12.0
for term, v in tensiones_pnp.items():
    bar_h = (v / max_v) * 6 + 0.3
    rect = plt.Rectangle((x_pos[term]-0.5, 1), 1, bar_h, color=colores[term], alpha=0.6)
    ax.add_patch(rect)
    ax.text(x_pos[term], 1 + bar_h + 0.3, f'{term}\n{v:.1f}V', ha='center',
            fontsize=11, fontweight='bold', color=C_DARK)

# Indicar VEB y VEC
ax.annotate('', xy=(2, 0.5), xytext=(5, 0.5),
            arrowprops=dict(arrowstyle='<->', color=C_DANGER, lw=2))
ax.text(3.5, 0.1, r'$V_{EB}$ = +0.7V', ha='center', fontsize=10, color=C_DANGER, fontweight='bold')

ax.annotate('', xy=(2, 8.5), xytext=(8, 8.5),
            arrowprops=dict(arrowstyle='<->', color=C_PURPLE, lw=2))
ax.text(5, 8.9, r'$V_{EC}$ = +5.0V', ha='center', fontsize=10, color=C_PURPLE, fontweight='bold')

fig.suptitle('Convencion de tensiones: siempre trabajamos con numeros positivos',
             fontsize=13, fontweight='bold', color=C_DARK, y=1.02)
plt.tight_layout()
plt.show()

print("NPN: VBE = VB - VE = 0.7 - 0.0 = +0.7V  (positivo, comodo)")
print("PNP: VEB = VE - VB = 12.0 - 11.3 = +0.7V  (positivo, comodo)")
print()
print("Si usaramos VBE en PNP: VBE = 11.3 - 12.0 = -0.7V  (negativo, incomodo)")
````

## 2. Por que usamos VEB en vez de VBE en PNP

### La razon: queremos trabajar con numeros POSITIVOS

Recordemos la definicion de tension entre dos puntos:

$$V_{XY} = V_X - V_Y$$

Donde:
- $V_X$: tension en el nodo X (primer subindice)
- $V_Y$: tension en el nodo Y (segundo subindice)
- $V_{XY}$: diferencia de potencial desde X hasta Y

Entonces:
- $V_{BE} = V_B - V_E$ (tension de la base respecto al emisor)
- $V_{EB} = V_E - V_B$ (tension del emisor respecto a la base)
- Relacion: $V_{EB} = -V_{BE}$ (siempre son opuestas en signo)

### Ejemplo numerico: NPN en activa

Para que un NPN conduzca, la base debe estar **mas alta** que el emisor:
- Supongamos: $V_B = 3.7\,V$, $V_E = 3.0\,V$
- $V_{BE} = 3.7 - 3.0 = +0.7\,V$ (positivo, comodo)
- $V_{EB} = 3.0 - 3.7 = -0.7\,V$ (negativo, incomodo)
- Usamos $V_{BE}$ porque sale positivo cuando conduce.

### Ejemplo numerico: PNP en activa

Para que un PNP conduzca, el emisor debe estar **mas alto** que la base:
- Supongamos: $V_E = 12.0\,V$, $V_B = 11.3\,V$
- $V_{BE} = 11.3 - 12.0 = -0.7\,V$ (negativo, incomodo)
- $V_{EB} = 12.0 - 11.3 = +0.7\,V$ (positivo, comodo)
- Usamos $V_{EB}$ porque sale positivo cuando conduce.

### Lo mismo pasa con VCE vs VEC

- NPN en activa: colector a tension mas alta que emisor -> $V_{CE} > 0$ (usamos $V_{CE}$)
- PNP en activa: emisor a tension mas alta que colector -> $V_{CE} < 0$, pero $V_{EC} > 0$ (usamos $V_{EC}$)

### Conclusion

| Variable | NPN (positiva) | PNP (positiva) | Relacion |
|:--|:--|:--|:--|
| Union E-B | $V_{BE} = +0.7\,V$ | $V_{EB} = +0.7\,V$ | $V_{EB} = -V_{BE}$ |
| Union C-E | $V_{CE} > 0$ | $V_{EC} > 0$ | $V_{EC} = -V_{CE}$ |

**Es SOLO una convencion para evitar signos negativos. La fisica no cambia. Si quieres usar $V_{BE}$ en un PNP, puedes hacerlo, pero sera $-0.7\,V$ en activa y te sera mas facil equivocarte con los signos.**

## 1. El origen fisico: Que ocurre DENTRO del semiconductor

### 1.1 NPN: los electrones son los protagonistas

En un transistor NPN, las tres regiones son:

```
  Emisor (N)  --  Base (P)  --  Colector (N)
  [muchos e-]    [delgada]     [muchos e-]
```

- El **emisor tipo N** tiene una enorme cantidad de **electrones libres** (portadores mayoritarios).
- Cuando polarizamos la union Base-Emisor en **directa** (VB > VE), los electrones del emisor **cruzan hacia la base**.
- La base es muy delgada y poco dopada, asi que la mayoria de esos electrones **no se recombinan** con los huecos de la base.
- En cambio, son **atraidos** por el campo electrico de la union Base-Colector (polarizada en inversa, VC > VB), y pasan al colector.

**Flujo de electrones:** E --> B --> C (de izquierda a derecha)

**Corriente convencional** (de + a -): Como los electrones son cargas negativas, la corriente convencional va en sentido CONTRARIO al flujo de electrones:

**Corriente convencional:** C --> B --> E (de derecha a izquierda, ENTRA por C y B, SALE por E)

Esto coincide con la flecha del emisor en el simbolo NPN (apunta hacia fuera del emisor).

### 1.2 PNP: los HUECOS son los protagonistas

En un transistor PNP, las tres regiones se invierten:

```
  Emisor (P)  --  Base (N)  --  Colector (P)
  [muchos h+]    [delgada]     [muchos h+]
```

- El **emisor tipo P** tiene una enorme cantidad de **huecos** (portadores mayoritarios).
- Los huecos son "ausencias de electron" que se comportan como **cargas positivas**.
- Cuando polarizamos la union Emisor-Base en directa (VE > VB), los huecos del emisor **cruzan hacia la base**.
- La base (tipo N) es delgada, asi que la mayoria de los huecos **no se recombinan** y son atraidos hacia el colector.

**Flujo de huecos:** E --> B --> C (igual que los electrones en NPN, pero ahora son cargas POSITIVAS)

**Corriente convencional** (de + a -): Como los huecos SON cargas positivas, la corriente convencional va en la **MISMA** direccion que los huecos:

**Corriente convencional:** E --> B --> C (ENTRA por E, SALE por B y C)

Esto es EXACTAMENTE AL REVES que NPN. Y la flecha del emisor en el simbolo PNP apunta **hacia dentro** del transistor (hacia la base), indicando que la corriente entra por el emisor.

### Resumen de la diferencia fisica

| Propiedad | NPN | PNP |
|:--|:--|:--|
| Portadores del emisor | Electrones (e-) | Huecos (h+) |
| Flujo de portadores | E -> B -> C | E -> B -> C |
| Carga de los portadores | Negativa | Positiva |
| Corriente convencional | C,B -> E (sale por E) | E -> B,C (entra por E) |
| Flecha del emisor | Apunta hacia fuera | Apunta hacia dentro |
| UE en directa requiere | VB > VE | VE > VB |

**Conclusion clave:** La corriente va al reves en PNP porque los portadores son cargas POSITIVAS (huecos), y la corriente convencional sigue la misma direccion que las cargas positivas.
