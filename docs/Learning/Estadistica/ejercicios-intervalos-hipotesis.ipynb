## Formulas de referencia

---

### 1. Intervalo de confianza para la media con varianza conocida (IC z)

$$\bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$$

- $\bar{x}$: media muestral.
- $z_{\alpha/2}$: valor critico de la distribucion normal estandar para un nivel de confianza $1 - \alpha$. Por ejemplo, para 95% de confianza, $z_{0.025} = 1.96$.
- $\sigma$: desviacion estandar poblacional (conocida).
- $n$: tamano de la muestra.

---

### 2. Intervalo de confianza para la media con varianza desconocida (IC t)

$$\bar{x} \pm t_{\alpha/2, \, n-1} \cdot \frac{s}{\sqrt{n}}$$

- $\bar{x}$: media muestral.
- $t_{\alpha/2, \, n-1}$: valor critico de la distribucion t de Student con $n - 1$ grados de libertad.
- $s$: desviacion estandar muestral.
- $n$: tamano de la muestra.

---

### 3. Intervalo de confianza para una proporcion (IC proporcion)

$$\hat{p} \pm z_{\alpha/2} \cdot \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}$$

- $\hat{p}$: proporcion muestral, calculada como $\hat{p} = x / n$ donde $x$ es el numero de exitos.
- $z_{\alpha/2}$: valor critico de la distribucion normal estandar.
- $n$: tamano de la muestra.

---

### 4. Intervalo de confianza para la diferencia de medias (IC diferencia de medias)

$$(\bar{x}_1 - \bar{x}_2) \pm t_{\alpha/2, \, \nu} \cdot \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}$$

- $\bar{x}_1, \bar{x}_2$: medias muestrales de los grupos 1 y 2.
- $s_1, s_2$: desviaciones estandar muestrales de los grupos 1 y 2.
- $n_1, n_2$: tamanos de las muestras de los grupos 1 y 2.
- $\nu$: grados de libertad calculados con la aproximacion de Welch:

$$\nu = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1 - 1} + \frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2 - 1}}$$

- $t_{\alpha/2, \, \nu}$: valor critico de la distribucion t de Student con $\nu$ grados de libertad.

---

### 5. Intervalo de confianza para la varianza (IC varianza con Chi-cuadrado)

$$\left[\frac{(n-1) s^2}{\chi^2_{\alpha/2, \, n-1}}, \quad \frac{(n-1) s^2}{\chi^2_{1-\alpha/2, \, n-1}}\right]$$

- $s^2$: varianza muestral.
- $n$: tamano de la muestra.
- $\chi^2_{\alpha/2, \, n-1}$: valor critico superior de la distribucion Chi-cuadrado con $n - 1$ grados de libertad.
- $\chi^2_{1-\alpha/2, \, n-1}$: valor critico inferior de la distribucion Chi-cuadrado con $n - 1$ grados de libertad.

---

### 6. t-test para una muestra

$$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$$

- $\bar{x}$: media muestral.
- $\mu_0$: valor hipotetico de la media poblacional bajo $H_0$.
- $s$: desviacion estandar muestral.
- $n$: tamano de la muestra.
- Se compara $t$ con la distribucion $t$ de Student con $n - 1$ grados de libertad.

---

### 7. t-test para dos muestras independientes (Welch)

$$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

- $\bar{x}_1, \bar{x}_2$: medias muestrales de los grupos 1 y 2.
- $s_1, s_2$: desviaciones estandar muestrales.
- $n_1, n_2$: tamanos muestrales.
- Los grados de libertad se calculan con la aproximacion de Welch (ver formula en seccion 4).

---

### 8. ANOVA de un factor (estadistico F)

$$F = \frac{MS_{entre}}{MS_{dentro}}$$

donde:

$$MS_{entre} = \frac{SS_{entre}}{k - 1}, \qquad MS_{dentro} = \frac{SS_{dentro}}{N - k}$$

$$SS_{entre} = \sum_{i=1}^{k} n_i (\bar{x}_i - \bar{x})^2, \qquad SS_{dentro} = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (x_{ij} - \bar{x}_i)^2$$

- $k$: numero de grupos.
- $n_i$: tamano del grupo $i$.
- $N$: tamano total de la muestra ($N = \sum n_i$).
- $\bar{x}_i$: media del grupo $i$.
- $\bar{x}$: media global.
- $SS_{entre}$: suma de cuadrados entre grupos.
- $SS_{dentro}$: suma de cuadrados dentro de los grupos.
- $MS_{entre}$: cuadrado medio entre grupos.
- $MS_{dentro}$: cuadrado medio dentro de los grupos.
- Se compara $F$ con la distribucion $F$ de Fisher con $(k-1, N-k)$ grados de libertad.

---

### 9. Chi-cuadrado de independencia

$$\chi^2 = \sum_{i} \sum_{j} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$

donde:

$$E_{ij} = \frac{(\text{total fila } i) \times (\text{total columna } j)}{N}$$

- $O_{ij}$: frecuencia observada en la celda $(i, j)$.
- $E_{ij}$: frecuencia esperada en la celda $(i, j)$ bajo independencia.
- $N$: total de observaciones.
- Grados de libertad: $(r - 1)(c - 1)$ donde $r$ es el numero de filas y $c$ el de columnas.

---

### 10. Tamano de muestra para estimar la media

$$n = \left(\frac{z_{\alpha/2} \cdot \sigma}{E}\right)^2$$

- $z_{\alpha/2}$: valor critico de la distribucion normal estandar.
- $\sigma$: desviacion estandar poblacional (o una estimacion de ella).
- $E$: error maximo permitido (margen de error).

---

### 11. Tamano de muestra para estimar una proporcion

$$n = \frac{z_{\alpha/2}^2 \cdot \hat{p}(1 - \hat{p})}{E^2}$$

- $z_{\alpha/2}$: valor critico de la distribucion normal estandar.
- $\hat{p}$: proporcion estimada (si no se conoce, se usa 0.5 como caso mas conservador).
- $E$: margen de error deseado.

**Ejercicio 1:** Una fabrica produce tornillos cuyo diametro tiene una desviacion estandar conocida de sigma=0.5mm. Se toma una muestra de 50 tornillos y se obtiene una media muestral de 10.2mm. Construir un intervalo de confianza al 99% para el diametro medio.

**Resolucion:**

Como sigma es conocida, usamos la distribucion **Normal** para construir el IC.

Parametros:
- **x_barra = 10.2** mm (media muestral)
- **sigma = 0.5** mm (desviacion estandar poblacional conocida)
- **n = 50** (tamano muestral)
- **alfa = 0.01** (1 - 0.99)

**Formula:** $\bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$

**Calculo paso a paso:**

1. Valor critico: $z_{0.005} = 2.576$
2. Error estandar: $\frac{\sigma}{\sqrt{n}} = \frac{0.5}{\sqrt{50}} = \frac{0.5}{7.071} = 0.0707$
3. Margen de error: $2.576 \times 0.0707 = 0.1822$
4. IC al 99%: $10.2 \pm 0.1822 = [10.018, \; 10.382]$

```python
x_bar, sigma, n = 10.2, 0.5, 50
alfa = 0.01

z_crit = stats.norm.ppf(1 - alfa/2)
se = sigma / np.sqrt(n)
margen = z_crit * se
ic_inf, ic_sup = x_bar - margen, x_bar + margen

print(f"z_critico(0.005) = {z_crit:.4f}")
print(f"Error estandar: sigma/sqrt(n) = {sigma}/sqrt({n}) = {se:.4f}")
print(f"Margen de error: {z_crit:.4f} * {se:.4f} = {margen:.4f}")
print(f"IC al 99%: [{ic_inf:.4f}, {ic_sup:.4f}]")

# Grafica: distribucion muestral de la media con IC
x_vals = np.linspace(x_bar - 4*se, x_bar + 4*se, 500)
dist_media = stats.norm(x_bar, se)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_vals, dist_media.pdf(x_vals), color=C_DARK, linewidth=2)

# Area del IC
x_ic = np.linspace(ic_inf, ic_sup, 300)
ax.fill_between(x_ic, dist_media.pdf(x_ic), alpha=0.4, color=C_PRIMARY,
                label=f'IC 99%: [{ic_inf:.3f}, {ic_sup:.3f}]')

ax.axvline(x=x_bar, color=C_ORANGE, linestyle='-', linewidth=2, label=f'x_barra = {x_bar}')
ax.axvline(x=ic_inf, color=C_DANGER, linestyle='--', linewidth=1.5)
ax.axvline(x=ic_sup, color=C_DANGER, linestyle='--', linewidth=1.5)

ax.set_xlabel('Diametro medio (mm)')
ax.set_ylabel('Densidad')
ax.set_title('Distribucion muestral de la media - IC al 99%')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 2:** Se mide la presion arterial de 25 pacientes, obteniendo una media de 130 mmHg y una desviacion estandar muestral de 15 mmHg. Construir un intervalo de confianza al 95% para la presion arterial media poblacional.

**Resolucion:**

Como sigma es desconocida, usamos la distribucion **t-Student**.

Parametros:
- **x_barra = 130** mmHg (media muestral)
- **s = 15** mmHg (desviacion estandar muestral)
- **n = 25** (tamano de muestra)
- **gl = n - 1 = 24** (grados de libertad)
- **alfa = 0.05**

**Formula:** $\bar{x} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}}$

**Calculo paso a paso:**

1. Valor critico: $t_{0.025, 24} = 2.0639$
2. Error estandar: $\frac{s}{\sqrt{n}} = \frac{15}{\sqrt{25}} = \frac{15}{5} = 3.0$
3. Margen de error: $2.0639 \times 3.0 = 6.192$
4. IC al 95%: $130 \pm 6.192 = [123.81, \; 136.19]$

```python
x_bar, s, n = 130, 15, 25
gl = n - 1
alfa = 0.05

t_crit = stats.t.ppf(1 - alfa/2, gl)
se = s / np.sqrt(n)
margen = t_crit * se
ic_inf, ic_sup = x_bar - margen, x_bar + margen

print(f"t_critico(0.025, {gl}) = {t_crit:.4f}")
print(f"Error estandar: s/sqrt(n) = {s}/sqrt({n}) = {se:.4f}")
print(f"Margen de error: {t_crit:.4f} * {se:.4f} = {margen:.4f}")
print(f"IC al 95%: [{ic_inf:.2f}, {ic_sup:.2f}]")

# Grafica: distribucion t con IC
x_vals = np.linspace(-4, 4, 500)
t_dist = stats.t(gl)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_vals, t_dist.pdf(x_vals), color=C_DARK, linewidth=2, label=f't-Student (gl={gl})')

# Area central 95%
x_ic = np.linspace(-t_crit, t_crit, 300)
ax.fill_between(x_ic, t_dist.pdf(x_ic), alpha=0.3, color=C_PRIMARY, label=f'95% central')

# Colas
x_left = np.linspace(-4, -t_crit, 100)
x_right = np.linspace(t_crit, 4, 100)
ax.fill_between(x_left, t_dist.pdf(x_left), alpha=0.4, color=C_DANGER)
ax.fill_between(x_right, t_dist.pdf(x_right), alpha=0.4, color=C_DANGER, label='2.5% cada cola')

ax.set_xlabel('t')
ax.set_ylabel('f(t)')
ax.set_title(f'IC al 95% para la media: [{ic_inf:.2f}, {ic_sup:.2f}] mmHg')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 3:** En una encuesta a 500 personas, 280 responden afirmativamente. Construir un intervalo de confianza al 95% para la proporcion poblacional.

**Resolucion:**

IC para una **proporcion** usando la aproximacion Normal.

Parametros:
- **x = 280** (numero de exitos)
- **n = 500** (tamano de muestra)
- **p_hat = x/n = 280/500 = 0.56** (proporcion muestral)
- **q_hat = 1 - p_hat = 0.44**
- **alfa = 0.05**

**Condicion:** $np\hat{} \geq 5$ y $n(1-\hat{p}) \geq 5$ --> $500 \times 0.56 = 280 \geq 5$ y $500 \times 0.44 = 220 \geq 5$. Se cumple.

**Formula:** $\hat{p} \pm z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$

**Calculo paso a paso:**

1. Valor critico: $z_{0.025} = 1.96$
2. Error estandar: $\sqrt{\frac{0.56 \times 0.44}{500}} = \sqrt{\frac{0.2464}{500}} = \sqrt{0.0004928} = 0.02220$
3. Margen de error: $1.96 \times 0.02220 = 0.04351$
4. IC al 95%: $0.56 \pm 0.04351 = [0.5165, \; 0.6035]$

```python
x_exitos, n = 280, 500
p_hat = x_exitos / n
q_hat = 1 - p_hat
alfa = 0.05

z_crit = stats.norm.ppf(1 - alfa/2)
se = np.sqrt(p_hat * q_hat / n)
margen = z_crit * se
ic_inf, ic_sup = p_hat - margen, p_hat + margen

print(f"p_hat = {x_exitos}/{n} = {p_hat:.4f}")
print(f"Error estandar: sqrt({p_hat:.2f}*{q_hat:.2f}/{n}) = {se:.5f}")
print(f"Margen de error: {z_crit:.4f} * {se:.5f} = {margen:.5f}")
print(f"IC al 95%: [{ic_inf:.4f}, {ic_sup:.4f}]")
print(f"IC al 95%: [{ic_inf*100:.2f}%, {ic_sup*100:.2f}%]")

# Grafica: distribucion muestral de la proporcion con IC
x_vals = np.linspace(p_hat - 4*se, p_hat + 4*se, 500)
dist_p = stats.norm(p_hat, se)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_vals, dist_p.pdf(x_vals), color=C_DARK, linewidth=2)

# Area del IC
x_ic = np.linspace(ic_inf, ic_sup, 300)
ax.fill_between(x_ic, dist_p.pdf(x_ic), alpha=0.4, color=C_SUCCESS,
                label=f'IC 95%: [{ic_inf:.4f}, {ic_sup:.4f}]')

ax.axvline(x=p_hat, color=C_ORANGE, linewidth=2, label=f'p_hat = {p_hat:.2f}')
ax.axvline(x=ic_inf, color=C_DANGER, linestyle='--', linewidth=1.5)
ax.axvline(x=ic_sup, color=C_DANGER, linestyle='--', linewidth=1.5)
ax.axvline(x=0.5, color=C_DARK, linestyle=':', linewidth=1, alpha=0.5, label='p = 0.5 (referencia)')

ax.set_xlabel('Proporcion')
ax.set_ylabel('Densidad')
ax.set_title('IC al 95% para la proporcion poblacional')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 4:** Un fabricante afirma que sus pilas duran en promedio 100 horas. Se toma una muestra de 36 pilas y se obtiene una media de 97 horas con desviacion estandar de 12 horas. Con un nivel de significacion alfa=0.05, hay evidencia para rechazar la afirmacion del fabricante?

**Resolucion:**

**t-test de una muestra** (bilateral).

- **H0:** $\mu = 100$ (la media es 100 horas)
- **H1:** $\mu \neq 100$ (la media es diferente de 100)

Parametros:
- **x_barra = 97**, **mu_0 = 100**, **s = 12**, **n = 36**, **gl = 35**
- **alfa = 0.05**

**Estadistico:** $t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$

**Calculo paso a paso:**

$$t = \frac{97 - 100}{12 / \sqrt{36}} = \frac{-3}{12/6} = \frac{-3}{2} = -1.5$$

**Valores criticos:** $t_{0.025, 35} = \pm 2.030$

Como $|{-1.5}| = 1.5 < 2.030$, **no se rechaza H0**. No hay evidencia suficiente para afirmar que las pilas no duran 100 horas.

```python
x_bar, mu_0, s, n = 97, 100, 12, 36
gl = n - 1
alfa = 0.05

t_stat = (x_bar - mu_0) / (s / np.sqrt(n))
t_crit = stats.t.ppf(1 - alfa/2, gl)
p_value = 2 * stats.t.cdf(t_stat, gl)  # bilateral, t_stat es negativo

print(f"Estadistico t = ({x_bar} - {mu_0}) / ({s}/sqrt({n})) = {t_stat:.4f}")
print(f"t_critico = +/- {t_crit:.4f}")
print(f"p-value = {p_value:.4f}")
print(f"|t| = {abs(t_stat):.4f} {'>' if abs(t_stat) > t_crit else '<'} t_crit = {t_crit:.4f}")
print(f"Decision: {'Rechazar H0' if abs(t_stat) > t_crit else 'No rechazar H0'}")

# Grafica
x_vals = np.linspace(-4, 4, 500)
t_dist = stats.t(gl)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_vals, t_dist.pdf(x_vals), color=C_DARK, linewidth=2)

# Region de no rechazo
x_nr = np.linspace(-t_crit, t_crit, 300)
ax.fill_between(x_nr, t_dist.pdf(x_nr), alpha=0.2, color=C_PRIMARY, label='No rechazo')

# Regiones de rechazo
x_left = np.linspace(-4, -t_crit, 100)
x_right = np.linspace(t_crit, 4, 100)
ax.fill_between(x_left, t_dist.pdf(x_left), alpha=0.4, color=C_DANGER)
ax.fill_between(x_right, t_dist.pdf(x_right), alpha=0.4, color=C_DANGER, label='Region rechazo')

# Estadistico observado
ax.axvline(x=t_stat, color=C_SUCCESS, linewidth=2.5, label=f't obs = {t_stat:.2f} (p={p_value:.4f})')
ax.axvline(x=-t_crit, color=C_DANGER, linestyle='--', linewidth=1)
ax.axvline(x=t_crit, color=C_DANGER, linestyle='--', linewidth=1)

ax.set_xlabel('t')
ax.set_ylabel('f(t)')
ax.set_title(f't-test de 1 muestra (gl={gl}) - H0: mu = {mu_0}')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 5:** Se comparan dos metodos de ensenanza. El Grupo A (n=30) obtiene una media de 75 con s=10, y el Grupo B (n=35) obtiene una media de 70 con s=12. Hay una diferencia significativa entre las medias de ambos grupos al nivel alfa=0.05?

**Resolucion:**

**t-test de Welch** para dos muestras independientes (no se asumen varianzas iguales).

- **H0:** $\mu_A = \mu_B$ (no hay diferencia)
- **H1:** $\mu_A \neq \mu_B$ (hay diferencia)

Parametros:
- Grupo A: $\bar{x}_A = 75$, $s_A = 10$, $n_A = 30$
- Grupo B: $\bar{x}_B = 70$, $s_B = 12$, $n_B = 35$

**Estadistico de Welch:** $t = \frac{\bar{x}_A - \bar{x}_B}{\sqrt{\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}}}$

**Calculo paso a paso:**

$$t = \frac{75 - 70}{\sqrt{\frac{100}{30} + \frac{144}{35}}} = \frac{5}{\sqrt{3.333 + 4.114}} = \frac{5}{\sqrt{7.448}} = \frac{5}{2.729} = 1.833$$

**Grados de libertad (Welch-Satterthwaite):**

$$df = \frac{\left(\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}\right)^2}{\frac{(s_A^2/n_A)^2}{n_A-1} + \frac{(s_B^2/n_B)^2}{n_B-1}} = \frac{(7.448)^2}{\frac{(3.333)^2}{29} + \frac{(4.114)^2}{34}} \approx 62.8$$

```python
x_a, s_a, n_a = 75, 10, 30
x_b, s_b, n_b = 70, 12, 35
alfa = 0.05

# Calculo manual del estadistico de Welch
se_welch = np.sqrt(s_a**2/n_a + s_b**2/n_b)
t_stat = (x_a - x_b) / se_welch

# Grados de libertad Welch-Satterthwaite
num_df = (s_a**2/n_a + s_b**2/n_b)**2
den_df = (s_a**2/n_a)**2/(n_a-1) + (s_b**2/n_b)**2/(n_b-1)
df_welch = num_df / den_df

t_crit = stats.t.ppf(1 - alfa/2, df_welch)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df_welch))

print(f"Error estandar de Welch: sqrt({s_a}^2/{n_a} + {s_b}^2/{n_b}) = {se_welch:.4f}")
print(f"Estadistico t = ({x_a} - {x_b}) / {se_welch:.4f} = {t_stat:.4f}")
print(f"Grados de libertad (Welch): {df_welch:.2f}")
print(f"t_critico = +/- {t_crit:.4f}")
print(f"p-value = {p_value:.4f}")
print(f"Decision: {'Rechazar H0' if p_value < alfa else 'No rechazar H0'}")

# Verificacion con scipy
np.random.seed(42)
muestra_a = np.random.normal(x_a, s_a, n_a)
muestra_b = np.random.normal(x_b, s_b, n_b)
# Usamos nuestros valores calculados directamente

# Grafica
x_vals = np.linspace(-4, 4, 500)
t_dist = stats.t(df_welch)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_vals, t_dist.pdf(x_vals), color=C_DARK, linewidth=2)

# Region de no rechazo
x_nr = np.linspace(-t_crit, t_crit, 300)
ax.fill_between(x_nr, t_dist.pdf(x_nr), alpha=0.2, color=C_PRIMARY, label='No rechazo')

# Regiones de rechazo
x_left = np.linspace(-4, -t_crit, 100)
x_right = np.linspace(t_crit, 4, 100)
ax.fill_between(x_left, t_dist.pdf(x_left), alpha=0.4, color=C_DANGER)
ax.fill_between(x_right, t_dist.pdf(x_right), alpha=0.4, color=C_DANGER, label='Region rechazo')

ax.axvline(x=t_stat, color=C_SUCCESS, linewidth=2.5, label=f't obs = {t_stat:.3f} (p={p_value:.4f})')

ax.set_xlabel('t')
ax.set_ylabel('f(t)')
ax.set_title(f'Welch t-test (gl={df_welch:.1f}) - H0: mu_A = mu_B')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 6:** Se prueban 3 fertilizantes en parcelas de cultivo. Los rendimientos (kg/m2) obtenidos son:
- Fertilizante A: [22, 24, 25, 23, 21]
- Fertilizante B: [28, 30, 27, 29, 31]
- Fertilizante C: [25, 26, 24, 27, 25]

Hay diferencia significativa entre los rendimientos medios de los tres fertilizantes? Usar alfa=0.05.

**Resolucion:**

**ANOVA de un factor** (one-way ANOVA).

- **H0:** $\mu_A = \mu_B = \mu_C$ (las medias de los tres grupos son iguales)
- **H1:** Al menos una media es diferente

**Formulas:**

- Media global: $\bar{X} = \frac{\sum_{i,j} X_{ij}}{N}$
- SCE (entre grupos): $\sum_i n_i (\bar{X}_i - \bar{X})^2$
- SCD (dentro de grupos): $\sum_i \sum_j (X_{ij} - \bar{X}_i)^2$
- CME = SCE / (k-1), CMD = SCD / (N-k)
- **F = CME / CMD**

donde k = numero de grupos, N = total de observaciones.

Si F > F_critico o p-value < alfa, se rechaza H0 y se procede con comparaciones post-hoc (Tukey HSD).

```python
fert_a = np.array([22, 24, 25, 23, 21])
fert_b = np.array([28, 30, 27, 29, 31])
fert_c = np.array([25, 26, 24, 27, 25])
alfa = 0.05

# Calculos manuales
grupos = [fert_a, fert_b, fert_c]
nombres = ['Fert. A', 'Fert. B', 'Fert. C']
k = len(grupos)
N = sum(len(g) for g in grupos)
media_global = np.concatenate(grupos).mean()

print(f"Medias por grupo:")
for nombre, g in zip(nombres, grupos):
    print(f"  {nombre}: media = {g.mean():.1f}, s = {g.std(ddof=1):.2f}")
print(f"Media global: {media_global:.2f}")
print(f"k = {k} grupos, N = {N} observaciones\n")

# SCE (entre grupos) y SCD (dentro de grupos)
SCE = sum(len(g) * (g.mean() - media_global)**2 for g in grupos)
SCD = sum(np.sum((g - g.mean())**2) for g in grupos)
CME = SCE / (k - 1)
CMD = SCD / (N - k)
F_stat = CME / CMD

print(f"SCE (entre grupos) = {SCE:.2f}")
print(f"SCD (dentro grupos) = {SCD:.2f}")
print(f"CME = SCE/{k-1} = {CME:.2f}")
print(f"CMD = SCD/{N-k} = {CMD:.2f}")
print(f"F = CME/CMD = {F_stat:.4f}")

# Verificacion con scipy
F_scipy, p_scipy = stats.f_oneway(fert_a, fert_b, fert_c)
F_crit = stats.f.ppf(1 - alfa, k-1, N-k)
print(f"\nVerificacion scipy: F = {F_scipy:.4f}, p-value = {p_scipy:.6f}")
print(f"F_critico({k-1}, {N-k}) = {F_crit:.4f}")
print(f"Decision: {'Rechazar H0' if p_scipy < alfa else 'No rechazar H0'}")

# Grafica
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot de los grupos
bp = ax1.boxplot([fert_a, fert_b, fert_c], labels=nombres, patch_artist=True,
                  boxprops=dict(alpha=0.7))
colors_box = [C_PRIMARY, C_DANGER, C_SUCCESS]
for patch, color in zip(bp['boxes'], colors_box):
    patch.set_facecolor(color)
ax1.axhline(y=media_global, color=C_ORANGE, linestyle='--', linewidth=1.5, label=f'Media global = {media_global:.1f}')
ax1.set_ylabel('Rendimiento (kg/m2)')
ax1.set_title('Rendimiento por fertilizante')
ax1.legend()

# Distribucion F con region de rechazo
x_f = np.linspace(0, 30, 500)
f_dist = stats.f(k-1, N-k)
ax2.plot(x_f, f_dist.pdf(x_f), color=C_DARK, linewidth=2)

x_nr = np.linspace(0, F_crit, 300)
ax2.fill_between(x_nr, f_dist.pdf(x_nr), alpha=0.2, color=C_PRIMARY, label='No rechazo')

x_rej = np.linspace(F_crit, 30, 200)
ax2.fill_between(x_rej, f_dist.pdf(x_rej), alpha=0.4, color=C_DANGER, label='Rechazo')

ax2.axvline(x=F_stat, color=C_SUCCESS, linewidth=2.5, label=f'F obs = {F_stat:.2f}')
ax2.set_xlabel('F')
ax2.set_ylabel('f(F)')
ax2.set_title(f'ANOVA F({k-1},{N-k}) - p = {p_scipy:.6f}')
ax2.legend()

plt.tight_layout()
plt.show()

# Post-hoc Tukey si se rechaza H0
if p_scipy < alfa:
    print("\n--- Comparaciones post-hoc (Tukey HSD) ---")
    from itertools import combinations
    for (i, gi), (j, gj) in combinations(enumerate(grupos), 2):
        t_val, p_val = stats.ttest_ind(gi, gj)
        print(f"  {nombres[i]} vs {nombres[j]}: t = {t_val:.3f}, p = {p_val:.4f} {'*' if p_val < alfa else ''}")
```

**Ejercicio 7:** Se realiza una encuesta para estudiar si el genero (Hombre/Mujer) es independiente de la preferencia de transporte (Auto/Bus/Bici). Los resultados son:

|  | Auto | Bus | Bici |
|---|---|---|---|
| Hombre | 40 | 30 | 20 |
| Mujer | 25 | 35 | 30 |

Son independientes el genero y la preferencia de transporte? Usar alfa=0.05.

**Resolucion:**

**Test Chi-cuadrado de independencia.**

- **H0:** El genero y la preferencia de transporte son independientes
- **H1:** Existe asociacion entre genero y preferencia de transporte

**Formula del estadistico:** $\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$

donde $E_{ij} = \frac{\text{Total fila}_i \times \text{Total columna}_j}{N}$ son las frecuencias esperadas bajo independencia.

**Calculo paso a paso:**

| | Auto | Bus | Bici | Total |
|---|---|---|---|---|
| Hombre | 40 | 30 | 20 | 90 |
| Mujer | 25 | 35 | 30 | 90 |
| Total | 65 | 65 | 50 | 180 |

Frecuencias esperadas: $E_{ij} = \frac{\text{fila}_i \times \text{col}_j}{180}$

| | Auto | Bus | Bici |
|---|---|---|---|
| Hombre | 32.5 | 32.5 | 25 |
| Mujer | 32.5 | 32.5 | 25 |

$$\chi^2 = \frac{(40-32.5)^2}{32.5} + \frac{(30-32.5)^2}{32.5} + \frac{(20-25)^2}{25} + \frac{(25-32.5)^2}{32.5} + \frac{(35-32.5)^2}{32.5} + \frac{(30-25)^2}{25}$$

Grados de libertad: $(r-1)(c-1) = (2-1)(3-1) = 2$

```python
observados = np.array([[40, 30, 20],
                       [25, 35, 30]])
alfa = 0.05

# Calculos manuales
totales_fila = observados.sum(axis=1)
totales_col = observados.sum(axis=0)
N = observados.sum()

# Frecuencias esperadas
esperados = np.outer(totales_fila, totales_col) / N

print("Tabla de frecuencias observadas:")
print(observados)
print(f"\nTotales fila: {totales_fila}")
print(f"Totales columna: {totales_col}")
print(f"Total: {N}")

print(f"\nFrecuencias esperadas:")
print(esperados)

# Estadistico chi2
chi2_manual = np.sum((observados - esperados)**2 / esperados)
gl = (observados.shape[0] - 1) * (observados.shape[1] - 1)

print(f"\nContribucion de cada celda al chi2:")
contribuciones = (observados - esperados)**2 / esperados
print(contribuciones.round(4))

print(f"\nchi2 = {chi2_manual:.4f}")
print(f"gl = ({observados.shape[0]}-1)*({observados.shape[1]}-1) = {gl}")

# Verificacion con scipy
chi2_scipy, p_value, gl_scipy, esperados_scipy = stats.chi2_contingency(observados)
chi2_crit = stats.chi2.ppf(1 - alfa, gl)

print(f"\nVerificacion scipy: chi2 = {chi2_scipy:.4f}, p-value = {p_value:.4f}")
print(f"chi2_critico({gl}) = {chi2_crit:.4f}")
print(f"Decision: {'Rechazar H0 (no son independientes)' if p_value < alfa else 'No rechazar H0 (son independientes)'}")

# Grafica
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Grafica de barras agrupadas
x_pos = np.arange(3)
ancho = 0.35
ax1.bar(x_pos - ancho/2, observados[0], ancho, color=C_PRIMARY, alpha=0.8, label='Hombre')
ax1.bar(x_pos + ancho/2, observados[1], ancho, color=C_DANGER, alpha=0.8, label='Mujer')
ax1.bar(x_pos - ancho/2, esperados[0], ancho, fill=False, edgecolor=C_DARK, linewidth=2, linestyle='--')
ax1.bar(x_pos + ancho/2, esperados[1], ancho, fill=False, edgecolor=C_DARK, linewidth=2, linestyle='--', label='Esperado')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(['Auto', 'Bus', 'Bici'])
ax1.set_ylabel('Frecuencia')
ax1.set_title('Observados vs Esperados')
ax1.legend()

# Distribucion Chi-cuadrado con region de rechazo
x_vals = np.linspace(0, 15, 500)
chi2_dist = stats.chi2(gl)
ax2.plot(x_vals, chi2_dist.pdf(x_vals), color=C_DARK, linewidth=2)

x_nr = np.linspace(0, chi2_crit, 300)
ax2.fill_between(x_nr, chi2_dist.pdf(x_nr), alpha=0.2, color=C_PRIMARY, label='No rechazo')
x_rej = np.linspace(chi2_crit, 15, 200)
ax2.fill_between(x_rej, chi2_dist.pdf(x_rej), alpha=0.4, color=C_DANGER, label='Rechazo')

ax2.axvline(x=chi2_scipy, color=C_SUCCESS, linewidth=2.5, label=f'chi2 obs = {chi2_scipy:.2f}')
ax2.set_xlabel('chi2')
ax2.set_ylabel('f(chi2)')
ax2.set_title(f'Chi-cuadrado (gl={gl}) - p = {p_value:.4f}')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 8:** Se desea estimar el diametro medio de unos rodamientos con un intervalo de confianza de ancho total 2mm (es decir, error maximo de 1mm). La desviacion estandar poblacional es sigma=5mm y se quiere un nivel de confianza del 95%. Cuantas muestras se necesitan?

**Resolucion:**

**Calculo del tamano de muestra** para un IC de la media con sigma conocida.

Parametros:
- **E = 1** mm (error maximo, mitad del ancho del IC)
- **sigma = 5** mm (desviacion estandar poblacional)
- **confianza = 95%** --> alfa = 0.05

**Formula:** $n = \left(\frac{z_{\alpha/2} \cdot \sigma}{E}\right)^2$

**Calculo paso a paso:**

$$n = \left(\frac{z_{0.025} \cdot \sigma}{E}\right)^2 = \left(\frac{1.96 \times 5}{1}\right)^2 = (9.8)^2 = 96.04$$

Como n debe ser entero, se redondea hacia arriba: **n = 97 muestras**.

Verificacion: con n=97, el ancho del IC seria $2 \times 1.96 \times 5 / \sqrt{97} = 1.988$ mm, que es menor que 2mm.

```python
E_max = 1       # Error maximo (mm)
sigma = 5       # Desviacion estandar poblacional (mm)
confianza = 0.95
alfa = 1 - confianza

z_crit = stats.norm.ppf(1 - alfa/2)
n_exacto = (z_crit * sigma / E_max)**2
n_necesario = int(np.ceil(n_exacto))

print(f"z_critico = {z_crit:.4f}")
print(f"n = (z * sigma / E)^2 = ({z_crit:.4f} * {sigma} / {E_max})^2 = {n_exacto:.2f}")
print(f"n redondeado (ceil) = {n_necesario}")

# Verificacion
ancho_ic = 2 * z_crit * sigma / np.sqrt(n_necesario)
print(f"\nVerificacion con n={n_necesario}:")
print(f"  Ancho del IC = 2 * {z_crit:.4f} * {sigma} / sqrt({n_necesario}) = {ancho_ic:.4f} mm")
print(f"  Ancho objetivo: 2 mm")
print(f"  Cumple: {ancho_ic <= 2.0}")

# Grafica: como varia el ancho del IC con n
n_vals = np.arange(10, 300)
anchos = 2 * z_crit * sigma / np.sqrt(n_vals)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(n_vals, anchos, color=C_PRIMARY, linewidth=2)
ax.axhline(y=2.0, color=C_DANGER, linestyle='--', linewidth=1.5, label='Ancho objetivo = 2 mm')
ax.axvline(x=n_necesario, color=C_SUCCESS, linestyle='--', linewidth=1.5, label=f'n necesario = {n_necesario}')
ax.scatter([n_necesario], [ancho_ic], color=C_SUCCESS, s=100, zorder=5)

# Sombrear region donde se cumple el requisito
ax.fill_between(n_vals, anchos, 0, where=(anchos <= 2.0), alpha=0.2, color=C_SUCCESS,
                label='Cumple requisito')

ax.set_xlabel('Tamano de muestra (n)')
ax.set_ylabel('Ancho del IC (mm)')
ax.set_title(f'Ancho del IC al {confianza*100:.0f}% en funcion de n (sigma={sigma}mm)')
ax.legend()
ax.set_ylim(0, 10)
plt.tight_layout()
plt.show()

# Tabla de tamanos de muestra para distintos errores y confianzas
print("\nTabla de tamano de muestra necesario (sigma=5mm):")
print(f"{'Error (mm)':<12}", end='')
for conf in [0.90, 0.95, 0.99]:
    print(f"{'IC '+str(int(conf*100))+'%':>10}", end='')
print()
print("-" * 42)
for E in [0.5, 1.0, 1.5, 2.0, 3.0]:
    print(f"{E:<12.1f}", end='')
    for conf in [0.90, 0.95, 0.99]:
        z = stats.norm.ppf(1 - (1-conf)/2)
        n_req = int(np.ceil((z * sigma / E)**2))
        print(f"{n_req:>10}", end='')
    print()
```

**Ejercicio 9 (IC diferencia de medias):** Se comparan las notas de dos grupos de estudiantes. El Grupo A tiene $n_A = 40$ estudiantes con media $\bar{x}_A = 78$ y desviacion estandar $s_A = 12$. El Grupo B tiene $n_B = 35$ estudiantes con media $\bar{x}_B = 72$ y desviacion estandar $s_B = 10$. Construir un intervalo de confianza al 95% para la diferencia de medias $\mu_A - \mu_B$.

**Formula:**

$$(\bar{x}_A - \bar{x}_B) \pm t_{\alpha/2, \, \nu} \cdot \sqrt{\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}}$$

- $\bar{x}_A, \bar{x}_B$: medias muestrales de cada grupo.
- $s_A, s_B$: desviaciones estandar muestrales.
- $n_A, n_B$: tamanos muestrales.
- $\nu$: grados de libertad de Welch.
- $t_{\alpha/2, \, \nu}$: valor critico de la distribucion t de Student.

**Resolucion:**

```python
# Ejercicio 9: IC para la diferencia de medias (Welch)

x_a, s_a, n_a = 78, 12, 40
x_b, s_b, n_b = 72, 10, 35

# Diferencia de medias
diff = x_a - x_b

# Error estandar de la diferencia
se_diff = np.sqrt(s_a**2 / n_a + s_b**2 / n_b)

# Grados de libertad de Welch
numerador_gl = (s_a**2 / n_a + s_b**2 / n_b)**2
denominador_gl = (s_a**2 / n_a)**2 / (n_a - 1) + (s_b**2 / n_b)**2 / (n_b - 1)
gl = numerador_gl / denominador_gl

# Valor critico t al 95%
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha / 2, gl)

# Intervalo de confianza
ic_lower = diff - t_crit * se_diff
ic_upper = diff + t_crit * se_diff

print(f"Diferencia de medias: {diff}")
print(f"Error estandar: {se_diff:.4f}")
print(f"Grados de libertad (Welch): {gl:.2f}")
print(f"Valor critico t: {t_crit:.4f}")
print(f"IC 95%: ({ic_lower:.4f}, {ic_upper:.4f})")
print(f"\nInterpretacion: Con un 95% de confianza, la diferencia real entre")
print(f"las medias de los dos grupos esta entre {ic_lower:.2f} y {ic_upper:.2f} puntos.")
```

**Ejercicio 10 (IC varianza):** Se toman 25 mediciones de un proceso industrial, obteniendo una desviacion estandar muestral de $s = 3.5$. Construir un intervalo de confianza al 95% para la varianza poblacional $\sigma^2$ usando la distribucion Chi-cuadrado.

**Formula:**

$$\left[\frac{(n-1) s^2}{\chi^2_{\alpha/2, \, n-1}}, \quad \frac{(n-1) s^2}{\chi^2_{1-\alpha/2, \, n-1}}\right]$$

- $s^2$: varianza muestral ($s^2 = 3.5^2 = 12.25$).
- $n$: tamano de la muestra (25).
- $\chi^2_{\alpha/2, \, n-1}$: percentil superior de la distribucion Chi-cuadrado con $n-1 = 24$ grados de libertad.
- $\chi^2_{1-\alpha/2, \, n-1}$: percentil inferior de la distribucion Chi-cuadrado con $n-1 = 24$ grados de libertad.

**Resolucion:**

```python
# Ejercicio 10: IC para la varianza usando Chi-cuadrado

s = 3.5
n = 25
s2 = s**2        # Varianza muestral
gl = n - 1        # Grados de libertad
alpha = 0.05

# Valores criticos de Chi-cuadrado
chi2_upper = stats.chi2.ppf(1 - alpha / 2, gl)   # Percentil superior
chi2_lower = stats.chi2.ppf(alpha / 2, gl)        # Percentil inferior

# Intervalo de confianza para la varianza
ic_var_lower = (gl * s2) / chi2_upper
ic_var_upper = (gl * s2) / chi2_lower

# Intervalo de confianza para la desviacion estandar
ic_sd_lower = np.sqrt(ic_var_lower)
ic_sd_upper = np.sqrt(ic_var_upper)

print(f"Varianza muestral s^2: {s2}")
print(f"Grados de libertad: {gl}")
print(f"Chi2 inferior ({alpha/2}): {chi2_lower:.4f}")
print(f"Chi2 superior ({1-alpha/2}): {chi2_upper:.4f}")
print(f"\nIC 95% para la varianza sigma^2: ({ic_var_lower:.4f}, {ic_var_upper:.4f})")
print(f"IC 95% para la desviacion estandar sigma: ({ic_sd_lower:.4f}, {ic_sd_upper:.4f})")
print(f"\nInterpretacion: Con un 95% de confianza, la varianza poblacional")
print(f"se encuentra entre {ic_var_lower:.2f} y {ic_var_upper:.2f}.")
```

**Ejercicio 11 (Test de proporcion):** En una encuesta a 300 personas, 120 prefieren el producto A. Testear si la proporcion poblacional que prefiere el producto A es significativamente diferente de 0.5 (es decir, si hay preferencia real o no).

**Hipotesis:**
- $H_0: p = 0.5$ (no hay preferencia, la proporcion es 50%).
- $H_1: p \neq 0.5$ (existe preferencia significativa).

**Estadistico de prueba (test z para proporcion):**

$$z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0 (1 - p_0)}{n}}}$$

- $\hat{p}$: proporcion muestral ($\hat{p} = 120/300 = 0.4$).
- $p_0$: proporcion hipotetica bajo $H_0$ ($p_0 = 0.5$).
- $n$: tamano de la muestra (300).
- Se compara $|z|$ con $z_{\alpha/2}$ para un test bilateral.

**Resolucion:**

```python
# Ejercicio 11: Test de proporcion (z-test)

x_exitos = 120
n = 300
p_hat = x_exitos / n   # Proporcion muestral
p_0 = 0.5              # Proporcion bajo H0
alpha = 0.05

# Estadistico z
se_p = np.sqrt(p_0 * (1 - p_0) / n)
z_stat = (p_hat - p_0) / se_p

# p-valor bilateral
p_valor = 2 * stats.norm.sf(abs(z_stat))

# Valor critico
z_crit = stats.norm.ppf(1 - alpha / 2)

print(f"Proporcion muestral p_hat: {p_hat:.4f}")
print(f"Proporcion bajo H0: {p_0}")
print(f"Error estandar: {se_p:.4f}")
print(f"Estadistico z: {z_stat:.4f}")
print(f"Valor critico z (alpha={alpha}): +/-{z_crit:.4f}")
print(f"p-valor: {p_valor:.6f}")

if p_valor < alpha:
    print(f"\nDecision: Se rechaza H0 (p-valor {p_valor:.6f} < {alpha}).")
    print("La proporcion es significativamente diferente de 0.5.")
else:
    print(f"\nDecision: No se rechaza H0 (p-valor {p_valor:.6f} >= {alpha}).")
    print("No hay evidencia suficiente de que la proporcion difiera de 0.5.")
```

**Estadistico de prueba (t-test pareado):**

**Ejercicio 13 (ANOVA):** Se evaluan 4 metodos de estudio diferentes. Se asignan 15 alumnos a cada metodo y se registran sus notas finales. Determinar si existe una diferencia significativa en las notas medias entre los 4 metodos (alpha = 0.05).

**Hipotesis:**
- $H_0: \mu_1 = \mu_2 = \mu_3 = \mu_4$ (todos los metodos producen la misma nota media).
- $H_1$: al menos una media difiere.

**Estadistico F (ANOVA de un factor):**

$$F = \frac{MS_{entre}}{MS_{dentro}}$$

- $MS_{entre} = SS_{entre} / (k - 1)$: cuadrado medio entre grupos. $k = 4$ es el numero de grupos.
- $MS_{dentro} = SS_{dentro} / (N - k)$: cuadrado medio dentro de los grupos. $N = 60$ es el total de observaciones.
- $SS_{entre} = \sum_{i=1}^{k} n_i (\bar{x}_i - \bar{x})^2$: suma de cuadrados entre grupos.
- $SS_{dentro} = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (x_{ij} - \bar{x}_i)^2$: suma de cuadrados dentro de los grupos.
- Se compara $F$ con la distribucion F de Fisher con $(k-1, N-k) = (3, 56)$ grados de libertad.

**Resolucion:**

```python
# Ejercicio 13: ANOVA con 4 metodos de estudio

np.random.seed(123)

# Generar notas para 4 metodos (15 alumnos cada uno)
metodo_1 = np.random.normal(loc=70, scale=8, size=15).round(1)
metodo_2 = np.random.normal(loc=75, scale=7, size=15).round(1)
metodo_3 = np.random.normal(loc=72, scale=9, size=15).round(1)
metodo_4 = np.random.normal(loc=78, scale=6, size=15).round(1)

alpha = 0.05

# ANOVA de un factor
F_stat, p_valor = stats.f_oneway(metodo_1, metodo_2, metodo_3, metodo_4)

# Calculos manuales para entender
medias = [np.mean(m) for m in [metodo_1, metodo_2, metodo_3, metodo_4]]
media_global = np.mean(np.concatenate([metodo_1, metodo_2, metodo_3, metodo_4]))

print(f"Medias por metodo:")
for i, m in enumerate(medias, 1):
    print(f"  Metodo {i}: {m:.2f}")
print(f"Media global: {media_global:.2f}")
print(f"\nEstadistico F: {F_stat:.4f}")
print(f"p-valor: {p_valor:.6f}")
print(f"Grados de libertad: (3, 56)")

if p_valor < alpha:
    print(f"\nDecision: Se rechaza H0 (p-valor {p_valor:.6f} < {alpha}).")
    print("Existe diferencia significativa entre al menos dos metodos de estudio.")
else:
    print(f"\nDecision: No se rechaza H0 (p-valor {p_valor:.6f} >= {alpha}).")
    print("No hay evidencia de diferencias significativas entre los metodos.")
```

**Ejercicio 14 (Chi-cuadrado bondad de ajuste):** Un dado se lanza 120 veces y se obtienen las siguientes frecuencias observadas: [15, 22, 17, 25, 19, 22]. Testear si el dado es justo (es decir, si cada cara tiene la misma probabilidad 1/6).

**Hipotesis:**
- $H_0$: el dado es justo ($p_i = 1/6$ para $i = 1, \ldots, 6$).
- $H_1$: el dado no es justo (al menos una probabilidad difiere de 1/6).

**Estadistico de prueba (Chi-cuadrado bondad de ajuste):**

$$\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

- $O_i$: frecuencia observada de la cara $i$.
- $E_i$: frecuencia esperada de la cara $i$ bajo $H_0$ ($E_i = 120 / 6 = 20$).
- $k$: numero de categorias (6 caras).
- Grados de libertad: $k - 1 = 5$.

**Resolucion:**

```python
# Ejercicio 14: Chi-cuadrado bondad de ajuste (dado justo)

observados = np.array([15, 22, 17, 25, 19, 22])
n_lanzamientos = observados.sum()
k = len(observados)          # Numero de caras
esperados = np.full(k, n_lanzamientos / k)   # Frecuencia esperada uniforme

alpha = 0.05

# Estadistico Chi-cuadrado manual
chi2_manual = np.sum((observados - esperados)**2 / esperados)

# Usando scipy
chi2_stat, p_valor = stats.chisquare(observados, f_exp=esperados)

print(f"Frecuencias observadas: {observados}")
print(f"Frecuencias esperadas:  {esperados}")
print(f"Total de lanzamientos: {n_lanzamientos}")
print(f"\nEstadistico Chi-cuadrado: {chi2_stat:.4f}")
print(f"Grados de libertad: {k - 1}")
print(f"p-valor: {p_valor:.6f}")

if p_valor < alpha:
    print(f"\nDecision: Se rechaza H0 (p-valor {p_valor:.6f} < {alpha}).")
    print("Hay evidencia de que el dado no es justo.")
else:
    print(f"\nDecision: No se rechaza H0 (p-valor {p_valor:.6f} >= {alpha}).")
    print("No hay evidencia suficiente para afirmar que el dado no es justo.")
```

**Ejercicio 15 (IC bootstrap):** Se tiene una muestra de 50 datos que provienen de una distribucion exponencial (no normal). Construir un intervalo de confianza al 95% para la media usando el metodo bootstrap con $B = 5000$ remuestreos.

**Metodo bootstrap:**

El bootstrap no requiere supuestos distribucionales. Se basa en remuestrear con reemplazo de la muestra original:

1. Generar $B$ muestras bootstrap (remuestreos con reemplazo de tamano $n$).
2. Calcular la media de cada muestra bootstrap: $\bar{x}_1^*, \bar{x}_2^*, \ldots, \bar{x}_B^*$.
3. El IC al $(1 - \alpha)\%$ se obtiene tomando los percentiles $\alpha/2$ y $1 - \alpha/2$ de las $B$ medias bootstrap.

- $B$: numero de remuestreos bootstrap (5000).
- $n$: tamano de la muestra original (50).
- $\bar{x}_b^*$: media de la $b$-esima muestra bootstrap.
- Percentiles: $P_{2.5}$ y $P_{97.5}$ de las medias bootstrap para un IC al 95%.

**Resolucion:**

```python
# Ejercicio 15: IC bootstrap para datos no normales (exponenciales)

np.random.seed(77)

# Generar datos de una distribucion exponencial (media real = 5)
datos = np.random.exponential(scale=5, size=50)

B = 5000           # Numero de remuestreos
n = len(datos)
alpha = 0.05

# Bootstrap: remuestrear B veces y calcular la media de cada muestra
medias_bootstrap = np.array([
    np.mean(np.random.choice(datos, size=n, replace=True))
    for _ in range(B)
])

# IC por percentiles
ic_lower = np.percentile(medias_bootstrap, 100 * alpha / 2)
ic_upper = np.percentile(medias_bootstrap, 100 * (1 - alpha / 2))

print(f"Tamano de muestra: {n}")
print(f"Media muestral: {np.mean(datos):.4f}")
print(f"Mediana muestral: {np.median(datos):.4f}")
print(f"(La distribucion exponencial es asimetrica, no normal)")
print(f"\nNumero de remuestreos bootstrap (B): {B}")
print(f"Media de las medias bootstrap: {np.mean(medias_bootstrap):.4f}")
print(f"Desviacion estandar bootstrap: {np.std(medias_bootstrap):.4f}")
print(f"\nIC 95% bootstrap (metodo de percentiles): ({ic_lower:.4f}, {ic_upper:.4f})")
print(f"\nInterpretacion: Con un 95% de confianza, la media poblacional")
print(f"se encuentra entre {ic_lower:.2f} y {ic_upper:.2f}.")
```

**Ejercicio 16 (Test de Wilcoxon):** Se tienen dos muestras de datos que no siguen una distribucion normal. Usar el test de Wilcoxon (suma de rangos) como alternativa no parametrica al t-test de dos muestras independientes.

**Hipotesis:**
- $H_0$: las dos poblaciones tienen la misma distribucion (misma mediana).
- $H_1$: las distribuciones difieren (medianas diferentes).

**Estadistico de prueba (Wilcoxon rank-sum / Mann-Whitney U):**

El test de Wilcoxon-Mann-Whitney funciona asi:
1. Se combinan ambas muestras y se asignan rangos a todas las observaciones.
2. Se calcula la suma de rangos $W$ para una de las muestras.
3. Se compara $W$ con la distribucion esperada bajo $H_0$.

- $W$: suma de rangos de la primera muestra.
- $U$: estadistico de Mann-Whitney, $U = W - n_1(n_1 + 1)/2$.
- $n_1, n_2$: tamanos de las muestras.
- No se requiere normalidad de los datos.

**Resolucion:**

```python
# Ejercicio 16: Test de Wilcoxon (Mann-Whitney U) - test no parametrico

np.random.seed(55)

# Generar datos no normales (distribucion exponencial)
grupo_1 = np.random.exponential(scale=5, size=20)
grupo_2 = np.random.exponential(scale=8, size=20)

alpha = 0.05

# Test de Mann-Whitney U (equivalente a Wilcoxon rank-sum)
U_stat, p_valor = stats.mannwhitneyu(grupo_1, grupo_2, alternative='two-sided')

# Comparar con t-test (que no es apropiado aqui, pero sirve para ver la diferencia)
t_stat, p_valor_ttest = stats.ttest_ind(grupo_1, grupo_2)

print(f"Grupo 1 - Mediana: {np.median(grupo_1):.4f}, Media: {np.mean(grupo_1):.4f}")
print(f"Grupo 2 - Mediana: {np.median(grupo_2):.4f}, Media: {np.mean(grupo_2):.4f}")
print(f"\n--- Test de Mann-Whitney U (no parametrico) ---")
print(f"Estadistico U: {U_stat:.4f}")
print(f"p-valor: {p_valor:.6f}")

if p_valor < alpha:
    print(f"Decision: Se rechaza H0. Las distribuciones difieren significativamente.")
else:
    print(f"Decision: No se rechaza H0. No hay evidencia de diferencia.")

print(f"\n--- t-test (parametrico, para comparacion) ---")
print(f"Estadistico t: {t_stat:.4f}")
print(f"p-valor: {p_valor_ttest:.6f}")
print(f"\nNota: El test de Wilcoxon es mas apropiado cuando los datos no son normales.")
```

**Ejercicio 17 (Potencia del test):** Retomando el ejercicio 4 (pilas), donde $H_0: \mu = 100$, $s = 12$, $n = 36$ y $\alpha = 0.05$ (test bilateral), calcular la potencia del test si la media real es $\mu_1 = 95$ en vez de 100.

**Formula de la potencia:**

$$\text{Potencia} = 1 - \beta = P(\text{rechazar } H_0 \mid \mu = \mu_1)$$

Para un test bilateral con estadistico z:

$$\text{Potencia} = P\left(Z > z_{\alpha/2} - \frac{\mu_1 - \mu_0}{\sigma / \sqrt{n}}\right) + P\left(Z < -z_{\alpha/2} - \frac{\mu_1 - \mu_0}{\sigma / \sqrt{n}}\right)$$

- $\mu_0$: media bajo $H_0$ (100 horas).
- $\mu_1$: media real alternativa (95 horas).
- $\sigma$: desviacion estandar (12, se usa $s$ como estimacion).
- $n$: tamano de la muestra (36).
- $z_{\alpha/2}$: valor critico de la normal estandar (1.96 para $\alpha = 0.05$).
- $\beta$: probabilidad de error tipo II (no rechazar $H_0$ cuando es falsa).
- Potencia = $1 - \beta$: probabilidad de rechazar $H_0$ cuando realmente es falsa.

**Resolucion:**

```python
# Ejercicio 17: Potencia del test (ejercicio 4 - pilas)

mu_0 = 100    # Media bajo H0
mu_1 = 95     # Media real alternativa
s = 12        # Desviacion estandar
n = 36        # Tamano de muestra
alpha = 0.05

# Error estandar
se = s / np.sqrt(n)

# Valor critico z
z_crit = stats.norm.ppf(1 - alpha / 2)

# Limites criticos en la escala original
lim_inferior = mu_0 - z_crit * se
lim_superior = mu_0 + z_crit * se

# Potencia: probabilidad de rechazar H0 cuando mu = mu_1
# Bajo mu_1, la distribucion de x_bar es N(mu_1, se^2)
# Rechazamos si x_bar < lim_inferior o x_bar > lim_superior
potencia_izq = stats.norm.cdf(lim_inferior, loc=mu_1, scale=se)
potencia_der = 1 - stats.norm.cdf(lim_superior, loc=mu_1, scale=se)
potencia = potencia_izq + potencia_der
beta = 1 - potencia

print(f"Parametros del test:")
print(f"  H0: mu = {mu_0}, H1: mu != {mu_0}")
print(f"  Media real alternativa: mu_1 = {mu_1}")
print(f"  s = {s}, n = {n}, alpha = {alpha}")
print(f"  Error estandar: {se:.2f}")
print(f"\nLimites criticos (escala original):")
print(f"  Inferior: {lim_inferior:.2f}")
print(f"  Superior: {lim_superior:.2f}")
print(f"\nPotencia del test: {potencia:.4f} ({potencia*100:.1f}%)")
print(f"Beta (error tipo II): {beta:.4f} ({beta*100:.1f}%)")
print(f"\nInterpretacion: Si la media real fuera {mu_1} horas, el test")
print(f"detectaria la diferencia el {potencia*100:.1f}% de las veces.")
```

**Ejercicio 18 (Correccion de Bonferroni):** Se realizan 5 tests de hipotesis simultaneos, cada uno con $\alpha = 0.05$. Aplicar la correccion de Bonferroni para controlar el error tipo I global (FWER).

**Problema de las comparaciones multiples:**

Cuando se realizan $m$ tests independientes con nivel $\alpha$ cada uno, la probabilidad de cometer al menos un error tipo I (rechazar incorrectamente al menos una $H_0$ verdadera) crece:

$$P(\text{al menos un error tipo I}) = 1 - (1 - \alpha)^m$$

**Correccion de Bonferroni:**

$$\alpha_{ajustado} = \frac{\alpha}{m}$$

Alternativamente, se ajustan los p-valores:

$$p_{ajustado,i} = \min(m \cdot p_i, \, 1)$$

- $m$: numero de tests simultaneos (5).
- $\alpha$: nivel de significancia individual original (0.05).
- $\alpha_{ajustado}$: nivel de significancia corregido por Bonferroni ($0.05 / 5 = 0.01$).
- $p_i$: p-valor del test $i$.
- $p_{ajustado,i}$: p-valor ajustado del test $i$.
- FWER: family-wise error rate, probabilidad de al menos un falso positivo.

**Resolucion:**

```python
# Ejercicio 18: Correccion de Bonferroni para comparaciones multiples

m = 5          # Numero de tests simultaneos
alpha = 0.05   # Nivel de significancia individual

# Correccion de Bonferroni
alpha_bonferroni = alpha / m

# Probabilidad de al menos un error tipo I sin correccion
prob_error_sin = 1 - (1 - alpha)**m

# Probabilidad de al menos un error tipo I con correccion
prob_error_con = 1 - (1 - alpha_bonferroni)**m

# Ejemplo con p-valores simulados de 5 tests
np.random.seed(99)
p_valores = np.array([0.003, 0.012, 0.048, 0.15, 0.72])

# p-valores ajustados por Bonferroni
p_ajustados = np.minimum(p_valores * m, 1.0)

print(f"Numero de tests simultaneos (m): {m}")
print(f"Alpha individual original: {alpha}")
print(f"Alpha corregido (Bonferroni): {alpha_bonferroni}")
print(f"\nP(al menos un error tipo I) sin correccion: {prob_error_sin:.4f} ({prob_error_sin*100:.1f}%)")
print(f"P(al menos un error tipo I) con Bonferroni:  {prob_error_con:.4f} ({prob_error_con*100:.1f}%)")

print(f"\n--- Resultados de los 5 tests ---")
print(f"{'Test':<8} {'p-valor':<12} {'p-ajustado':<12} {'Sin corr.':<15} {'Con Bonferroni'}")
for i in range(m):
    sig_sin = "Rechaza H0" if p_valores[i] < alpha else "No rechaza"
    sig_con = "Rechaza H0" if p_ajustados[i] < alpha else "No rechaza"
    print(f"Test {i+1:<3} {p_valores[i]:<12.4f} {p_ajustados[i]:<12.4f} {sig_sin:<15} {sig_con}")

print(f"\nInterpretacion: La correccion de Bonferroni es conservadora pero")
print(f"garantiza que la probabilidad global de falsos positivos no supere {alpha}.")
```

**Ejercicio 19 (Tamano de muestra para proporcion):** Se desea realizar una encuesta con un margen de error del 3% y un nivel de confianza del 95%. Determinar cuantas personas se deben encuestar.

**Formula:**

$$n = \frac{z_{\alpha/2}^2 \cdot \hat{p}(1 - \hat{p})}{E^2}$$

- $z_{\alpha/2}$: valor critico de la distribucion normal estandar. Para 95% de confianza, $z_{0.025} = 1.96$.
- $\hat{p}$: proporcion estimada. Cuando no se tiene informacion previa, se usa $\hat{p} = 0.5$ (caso mas conservador que maximiza el tamano de muestra).
- $E$: margen de error deseado (0.03, es decir, 3%).
- $n$: tamano de muestra necesario (se redondea hacia arriba).

**Resolucion:**

```python
# Ejercicio 19: Tamano de muestra para estimar una proporcion

E = 0.03       # Margen de error (3%)
confianza = 0.95
alpha = 1 - confianza
p_hat = 0.5    # Caso mas conservador (maximiza n)

# Valor critico z
z_crit = stats.norm.ppf(1 - alpha / 2)

# Tamano de muestra
n_exacto = (z_crit**2 * p_hat * (1 - p_hat)) / E**2
n_redondeado = int(np.ceil(n_exacto))

print(f"Parametros:")
print(f"  Margen de error (E): {E} ({E*100:.0f}%)")
print(f"  Nivel de confianza: {confianza*100:.0f}%")
print(f"  Proporcion estimada (p_hat): {p_hat}")
print(f"  Valor critico z: {z_crit:.4f}")
print(f"\nCalculo: n = ({z_crit:.4f})^2 * {p_hat} * {1-p_hat} / ({E})^2")
print(f"         n = {n_exacto:.2f}")
print(f"\nTamano de muestra necesario: {n_redondeado} personas")
print(f"\nNota: Si se tuviera una estimacion previa de p diferente de 0.5,")
print(f"el tamano de muestra seria menor. Por ejemplo:")
for p_est in [0.3, 0.4, 0.5]:
    n_alt = int(np.ceil((z_crit**2 * p_est * (1 - p_est)) / E**2))
    print(f"  Con p_hat = {p_est}: n = {n_alt}")
```
