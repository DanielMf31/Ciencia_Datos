## <span style="color:#2c3e50;">Formulas de referencia</span>

---

### <span style="color:#3498db;">1. Factorial</span>

$$n! = n \times (n-1) \times (n-2) \times \cdots \times 2 \times 1$$

- $n$: numero total de elementos.
- $0! = 1$ por convencion.

---

### <span style="color:#3498db;">2. Permutaciones sin repeticion</span>

$$P(n, r) = \frac{n!}{(n-r)!}$$

- $n$: numero total de elementos disponibles.
- $r$: numero de elementos que se eligen/ordenan.
- $(n-r)!$: factorial de los elementos que quedan sin elegir.
- El **orden importa**: elegir A luego B es distinto de B luego A.

---

### <span style="color:#3498db;">3. Permutaciones con repeticion (elementos repetidos)</span>

$$P_r = \frac{n!}{n_1! \cdot n_2! \cdot \ldots \cdot n_k!}$$

- $n$: numero total de elementos (incluyendo repetidos).
- $n_1, n_2, \ldots, n_k$: cantidad de veces que se repite cada elemento distinto.
- $k$: numero de elementos distintos.
- Se divide por los factoriales de las repeticiones para eliminar ordenaciones indistinguibles.

---

### <span style="color:#3498db;">4. Combinaciones (sin repeticion)</span>

$$C(n, r) = \binom{n}{r} = \frac{n!}{r! \cdot (n-r)!}$$

- $n$: numero total de elementos disponibles.
- $r$: numero de elementos que se eligen.
- $r!$: corrige el hecho de que el orden no importa (elimina las permutaciones internas).
- $(n-r)!$: factorial de los elementos no elegidos.

---

### <span style="color:#3498db;">5. Probabilidad clasica (Laplace)</span>

$$P(A) = \frac{|A|}{|\Omega|} = \frac{\text{casos favorables}}{\text{casos posibles}}$$

- $P(A)$: probabilidad del evento $A$.
- $|A|$: numero de resultados favorables al evento $A$.
- $|\Omega|$: numero total de resultados posibles (espacio muestral).

---

### <span style="color:#3498db;">6. Probabilidad del complemento</span>

$$P(A) = 1 - P(\bar{A})$$

- $P(A)$: probabilidad de que ocurra $A$.
- $P(\bar{A})$: probabilidad de que **no** ocurra $A$.
- Util cuando es mas facil calcular lo contrario de lo que buscamos.

---

### <span style="color:#3498db;">7. Probabilidad condicional</span>

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

- $P(A|B)$: probabilidad de $A$ dado que ya ocurrio $B$.
- $P(A \cap B)$: probabilidad de que ocurran $A$ y $B$ simultaneamente.
- $P(B)$: probabilidad del evento condicionante (debe ser $> 0$).

---

### <span style="color:#3498db;">8. Teorema de Bayes</span>

$$P(A_i|B) = \frac{P(B|A_i) \cdot P(A_i)}{\sum_{j=1}^{n} P(B|A_j) \cdot P(A_j)}$$

- $P(A_i|B)$: probabilidad posterior de la hipotesis $A_i$ dado que se observo $B$.
- $P(B|A_i)$: verosimilitud, probabilidad de observar $B$ si $A_i$ es cierta.
- $P(A_i)$: probabilidad a priori de la hipotesis $A_i$.
- $\sum_{j=1}^{n} P(B|A_j) \cdot P(A_j)$: probabilidad total de $B$ (denominador normalizador).
- $n$: numero de hipotesis mutuamente excluyentes y exhaustivas.

---

### <span style="color:#3498db;">9. Distribucion binomial</span>

$$P(X = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$$

- $X$: variable aleatoria que cuenta el numero de exitos.
- $n$: numero de ensayos independientes.
- $k$: numero de exitos deseados.
- $p$: probabilidad de exito en cada ensayo.
- $(1-p)$: probabilidad de fracaso en cada ensayo.
- $\binom{n}{k}$: numero de formas de elegir cuales $k$ ensayos son exitos.

---

### <span style="color:#3498db;">10. Distribucion hipergeometrica</span>

$$P(X = k) = \frac{\binom{K}{k} \cdot \binom{N-K}{n-k}}{\binom{N}{n}}$$

- $N$: tamano total de la poblacion.
- $K$: numero de elementos de tipo "exito" en la poblacion.
- $n$: tamano de la muestra extraida (sin reemplazo).
- $k$: numero de exitos deseados en la muestra.
- $\binom{K}{k}$: formas de elegir $k$ exitos de los $K$ disponibles.
- $\binom{N-K}{n-k}$: formas de elegir los restantes $n-k$ de los $N-K$ no-exitos.
- $\binom{N}{n}$: formas totales de elegir la muestra.
