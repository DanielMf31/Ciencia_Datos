## Formulas fundamentales de Esperanza y Varianza

---

### Esperanza (valor esperado)

**Caso discreto:** La esperanza de una variable aleatoria discreta $X$ es la suma ponderada de todos sus posibles valores por sus respectivas probabilidades.

$$E[X] = \sum_{i} x_i \cdot P(X = x_i)$$

- $E[X]$: esperanza (valor esperado) de la variable aleatoria $X$, tambien denotada $\mu$ (mu). Representa el "centro de masa" de la distribucion.
- $x_i$: cada uno de los posibles valores que puede tomar la variable aleatoria $X$.
- $P(X = x_i)$: probabilidad de que $X$ tome el valor $x_i$. Cada probabilidad esta entre 0 y 1, y la suma de todas es 1.
- $\sum_{i}$: sumatoria sobre todos los posibles valores $i$ del soporte de $X$.

**Caso continuo:** La esperanza de una variable aleatoria continua $X$ con funcion de densidad $f(x)$ es la integral ponderada sobre todo el soporte.

$$E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx$$

- $E[X]$: esperanza de la variable aleatoria continua $X$.
- $x$: variable de integracion que recorre todos los valores reales del soporte de $X$.
- $f(x)$: funcion de densidad de probabilidad (PDF). Cumple $f(x) \geq 0$ y $\int_{-\infty}^{\infty} f(x) \, dx = 1$.
- $dx$: diferencial de integracion, indica que se integra respecto a $x$.
- $\int_{-\infty}^{\infty}$: integral sobre todo el eje real (en la practica, solo el soporte de $X$ contribuye).

---

### Varianza

**Definicion:** La varianza mide la dispersion de $X$ respecto a su media $\mu = E[X]$.

$$Var(X) = E\left[(X - \mu)^2\right] = \sum_{i} (x_i - \mu)^2 \cdot P(X = x_i)$$

- $Var(X)$: varianza de $X$, tambien denotada $\sigma^2$. Siempre es no negativa.
- $X$: variable aleatoria cuya dispersion se mide.
- $\mu$: media de $X$, es decir $\mu = E[X]$.
- $(X - \mu)^2$: desviacion cuadratica de $X$ respecto a su media. Se eleva al cuadrado para evitar cancelaciones entre desviaciones positivas y negativas.
- $x_i$: cada posible valor de $X$.
- $P(X = x_i)$: probabilidad asociada al valor $x_i$.

**Formula alternativa (computacionalmente mas eficiente):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $E[X^2]$: esperanza del cuadrado de $X$, es decir $\sum_i x_i^2 \cdot P(X = x_i)$ (discreto) o $\int x^2 f(x) dx$ (continuo). Este es el **segundo momento** de $X$.
- $(E[X])^2$: el cuadrado de la esperanza de $X$. No es lo mismo que $E[X^2]$.
- La diferencia $E[X^2] - (E[X])^2$ siempre es $\geq 0$.

**Desviacion estandar:**

$$\sigma = \sqrt{Var(X)}$$

- $\sigma$: desviacion estandar de $X$. Tiene las mismas unidades que $X$, a diferencia de la varianza que tiene unidades al cuadrado.

---

### Propiedades de la Esperanza

$$E[aX + b] = a \cdot E[X] + b$$

- $a$: constante multiplicativa (escalar). Escala la esperanza proporcionalmente.
- $b$: constante aditiva. Se suma directamente a la esperanza.
- $X$: variable aleatoria.
- Esta propiedad se llama **linealidad de la esperanza** y siempre vale, sin importar la distribucion.

$$E[X + Y] = E[X] + E[Y]$$

- $X, Y$: dos variables aleatorias cualesquiera (no necesitan ser independientes).
- La esperanza de la suma siempre es la suma de las esperanzas.

$$E[g(X)] = \sum_{i} g(x_i) \cdot P(X = x_i) \quad \text{(LOTUS, caso discreto)}$$

- $g(X)$: funcion aplicada a la variable aleatoria $X$ (por ejemplo, $g(X) = X^2$).
- LOTUS (Law of the Unconscious Statistician): permite calcular $E[g(X)]$ sin conocer la distribucion de $g(X)$.

---

### Propiedades de la Varianza

$$Var(aX + b) = a^2 \cdot Var(X)$$

- $a^2$: la constante multiplicativa se eleva al cuadrado porque la varianza es una medida cuadratica.
- $b$: la constante aditiva desaparece porque trasladar la distribucion no cambia su dispersion.

$$Var(X + Y) = Var(X) + Var(Y) + 2 \cdot Cov(X, Y)$$

- $Cov(X,Y)$: covarianza entre $X$ e $Y$. Mide la relacion lineal entre ambas variables.
- Si $X$ e $Y$ son **independientes**, entonces $Cov(X,Y) = 0$ y la formula se simplifica a $Var(X) + Var(Y)$.

$$Cov(X, Y) = E[XY] - E[X] \cdot E[Y]$$

- $E[XY]$: esperanza del producto de $X$ e $Y$.
- $E[X] \cdot E[Y]$: producto de las esperanzas individuales.
- Si $X$ e $Y$ son independientes, $E[XY] = E[X] \cdot E[Y]$ y la covarianza es cero.

---

### Funcion Generadora de Momentos (MGF)

$$M_X(t) = E[e^{tX}]$$

- $M_X(t)$: funcion generadora de momentos de la variable aleatoria $X$, evaluada en $t$.
- $t$: parametro real. La MGF debe existir en un entorno abierto de $t = 0$.
- $e^{tX}$: funcion exponencial aplicada al producto $tX$.
- La MGF, cuando existe, caracteriza unicamente la distribucion de $X$.

**Obtencion de momentos:**

$$E[X^n] = M_X^{(n)}(0) = \left. \frac{d^n}{dt^n} M_X(t) \right|_{t=0}$$

- $E[X^n]$: momento de orden $n$ de $X$.
- $M_X^{(n)}(0)$: la $n$-esima derivada de la MGF evaluada en $t = 0$.
- $n = 1$: da $E[X]$ (primer momento, la media).
- $n = 2$: da $E[X^2]$ (segundo momento, usado para calcular la varianza).

**Ejercicio 1:** Un dado justo de 6 caras. Calcular la esperanza E[X] y la varianza Var(X).

**Resolucion:**

**Esperanza:** Es el valor promedio ponderado por las probabilidades.

$$E[X] = \sum_{i=1}^{6} x_i \cdot P(x_i) = \frac{1}{6}(1 + 2 + 3 + 4 + 5 + 6) = \frac{21}{6} = 3.5$$

**Varianza:** Mide la dispersion respecto a la media.

$$Var(X) = E[X^2] - (E[X])^2$$

Primero calculamos $E[X^2]$:

$$E[X^2] = \frac{1}{6}(1^2 + 2^2 + 3^2 + 4^2 + 5^2 + 6^2) = \frac{1 + 4 + 9 + 16 + 25 + 36}{6} = \frac{91}{6} = 15.1\overline{6}$$

$$Var(X) = 15.1\overline{6} - (3.5)^2 = 15.1\overline{6} - 12.25 = 2.91\overline{6}$$

```python
x = np.array([1, 2, 3, 4, 5, 6])
p = np.ones(6) / 6

E_X = np.sum(x * p)
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2

print(f"E[X] = {E_X:.4f}")
print(f"E[X^2] = {E_X2:.4f}")
print(f"Var(X) = E[X^2] - (E[X])^2 = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")
print(f"Desviacion estandar = {np.sqrt(Var_X):.4f}")

# Grafica: PMF del dado con E[X] marcada
fig, ax = plt.subplots()
ax.bar(x, p, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=x)')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.axvspan(E_X - np.sqrt(Var_X), E_X + np.sqrt(Var_X), alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {np.sqrt(Var_X):.2f}')
ax.set_xlabel('Valor del dado')
ax.set_ylabel('Probabilidad')
ax.set_title('PMF del dado justo con E[X] y desviacion estandar', color=C_DARK, fontweight='bold')
ax.set_xticks(x)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 2:** Sea X una variable aleatoria con distribucion Bernoulli(p=0.3). Calcular E[X] y Var(X).

**Resolucion:**

La Bernoulli toma valores 0 (fracaso) y 1 (exito) con probabilidad $p$.

**Esperanza:**

$$E[X] = 0 \cdot (1-p) + 1 \cdot p = p = 0.3$$

**Varianza:**

$$Var(X) = E[X^2] - (E[X])^2$$

Como $X$ solo toma 0 y 1: $X^2 = X$, entonces $E[X^2] = E[X] = p$.

$$Var(X) = p - p^2 = p(1-p) = 0.3 \times 0.7 = 0.21$$

```python
p = 0.3
E_X = p
Var_X = p * (1 - p)

print(f"E[X] = p = {E_X}")
print(f"Var(X) = p(1-p) = {p} * {1-p} = {Var_X}")

# Grafica: PMF Bernoulli y como varia Var(X) con p
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# PMF
ax1.bar([0, 1], [1-p, p], color=[C_PRIMARY, C_DANGER], edgecolor=C_DARK, alpha=0.85, width=0.4)
ax1.set_xlabel('X')
ax1.set_ylabel('P(X)')
ax1.set_title('PMF Bernoulli(p=0.3)', color=C_DARK, fontweight='bold')
ax1.set_xticks([0, 1])
ax1.set_xticklabels(['0 (fracaso)', '1 (exito)'])
for i, v in enumerate([1-p, p]):
    ax1.text(i, v + 0.02, f'{v:.1f}', ha='center', fontweight='bold', color=C_DARK)

# Varianza en funcion de p
ps = np.linspace(0, 1, 100)
vars_p = ps * (1 - ps)
ax2.plot(ps, vars_p, color=C_PRIMARY, linewidth=2)
ax2.axvline(x=0.3, color=C_DANGER, linestyle='--', label=f'p=0.3, Var={Var_X}')
ax2.scatter([0.3], [Var_X], color=C_DANGER, s=80, zorder=5)
ax2.set_xlabel('p')
ax2.set_ylabel('Var(X) = p(1-p)')
ax2.set_title('Varianza de Bernoulli segun p', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 3:** Sea X ~ Binomial(n=10, p=0.4). Calcular E[X], Var(X) y P(X=4).

**Resolucion:**

Para $X \sim \text{Binomial}(n, p)$:

**Esperanza:**

$$E[X] = n \cdot p = 10 \times 0.4 = 4$$

**Varianza:**

$$Var(X) = n \cdot p \cdot (1-p) = 10 \times 0.4 \times 0.6 = 2.4$$

**P(X=4):** Usando la formula de la PMF binomial:

$$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$$

$$P(X=4) = \binom{10}{4} (0.4)^4 (0.6)^6 = 210 \times 0.0256 \times 0.046656$$

$$= 210 \times 0.001194 = 0.2508$$

```python
n, p = 10, 0.4
E_X = n * p
Var_X = n * p * (1 - p)
P_X4 = stats.binom.pmf(4, n, p)

print(f"E[X] = n*p = {n}*{p} = {E_X}")
print(f"Var(X) = n*p*(1-p) = {n}*{p}*{1-p} = {Var_X}")
print(f"P(X=4) = C(10,4) * 0.4^4 * 0.6^6 = {comb(10,4)} * {0.4**4:.4f} * {0.6**6:.6f} = {P_X4:.4f}")

# Grafica: PMF binomial con E[X] y area de P(X=4)
x_vals = np.arange(0, n+1)
pmf_vals = stats.binom.pmf(x_vals, n, p)

fig, ax = plt.subplots()
colores = [C_DANGER if x == 4 else C_PRIMARY for x in x_vals]
ax.bar(x_vals, pmf_vals, color=colores, edgecolor=C_DARK, alpha=0.85)
ax.axvline(x=E_X, color=C_ORANGE, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
for x, v in zip(x_vals, pmf_vals):
    ax.text(x, v + 0.005, f'{v:.3f}', ha='center', fontsize=8, color=C_DARK)
ax.set_xlabel('k')
ax.set_ylabel('P(X = k)')
ax.set_title('PMF Binomial(n=10, p=0.4)', color=C_DARK, fontweight='bold')
ax.set_xticks(x_vals)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 4:** Sea X una variable aleatoria continua con funcion de densidad f(x) = 2x para x en [0, 1] y f(x) = 0 en otro caso. Calcular E[X] y Var(X) mediante integrales.

**Resolucion:**

**Verificacion de que es densidad valida:**

$$\int_0^1 2x \, dx = \left[ x^2 \right]_0^1 = 1$$

**Esperanza:**

$$E[X] = \int_0^1 x \cdot f(x) \, dx = \int_0^1 x \cdot 2x \, dx = \int_0^1 2x^2 \, dx = \left[ \frac{2x^3}{3} \right]_0^1 = \frac{2}{3} \approx 0.6667$$

**E[X^2]:**

$$E[X^2] = \int_0^1 x^2 \cdot 2x \, dx = \int_0^1 2x^3 \, dx = \left[ \frac{2x^4}{4} \right]_0^1 = \frac{1}{2} = 0.5$$

**Varianza:**

$$Var(X) = E[X^2] - (E[X])^2 = \frac{1}{2} - \left(\frac{2}{3}\right)^2 = \frac{1}{2} - \frac{4}{9} = \frac{9 - 8}{18} = \frac{1}{18} \approx 0.0556$$

```python
f = lambda x: 2 * x

# Verificacion de densidad
integral_total, _ = quad(f, 0, 1)
print(f"Integral de f(x) en [0,1] = {integral_total} (debe ser 1)")

# E[X]
E_X, _ = quad(lambda x: x * f(x), 0, 1)
print(f"E[X] = integral(x * 2x, 0, 1) = {E_X:.4f} = 2/3")

# E[X^2]
E_X2, _ = quad(lambda x: x**2 * f(x), 0, 1)
print(f"E[X^2] = integral(x^2 * 2x, 0, 1) = {E_X2:.4f} = 1/2")

# Var(X)
Var_X = E_X2 - E_X**2
print(f"Var(X) = {E_X2:.4f} - ({E_X:.4f})^2 = {Var_X:.4f} = 1/18")

# Grafica: PDF con E[X] y area sombreada +/- 1 std
x_plot = np.linspace(0, 1, 200)
y_plot = 2 * x_plot
std = np.sqrt(Var_X)

fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, color=C_PRIMARY, linewidth=2.5, label='f(x) = 2x')
ax.fill_between(x_plot, y_plot, alpha=0.2, color=C_PRIMARY)

# Sombrear region E[X] +/- std
mask = (x_plot >= E_X - std) & (x_plot <= E_X + std)
ax.fill_between(x_plot[mask], y_plot[mask], alpha=0.4, color=C_ORANGE, label=f'E[X] +/- std')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X:.4f}')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF f(x) = 2x en [0, 1]', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 5:** Sea X ~ Poisson(lambda=5). Calcular E[X], Var(X) y P(X >= 3).

**Resolucion:**

Para $X \sim \text{Poisson}(\lambda)$, la esperanza y la varianza son iguales:

$$E[X] = \lambda = 5$$

$$Var(X) = \lambda = 5$$

**P(X >= 3):** Usamos el complemento.

$$P(X \geq 3) = 1 - P(X \leq 2) = 1 - \sum_{k=0}^{2} \frac{e^{-\lambda} \lambda^k}{k!}$$

$$P(X=0) = \frac{e^{-5} \cdot 5^0}{0!} = e^{-5} = 0.00674$$

$$P(X=1) = \frac{e^{-5} \cdot 5^1}{1!} = 5e^{-5} = 0.03369$$

$$P(X=2) = \frac{e^{-5} \cdot 5^2}{2!} = \frac{25e^{-5}}{2} = 0.08422$$

$$P(X \geq 3) = 1 - (0.00674 + 0.03369 + 0.08422) = 1 - 0.12465 = 0.8753$$

```python
lam = 5
E_X = lam
Var_X = lam

print(f"E[X] = lambda = {E_X}")
print(f"Var(X) = lambda = {Var_X}")

# P(X >= 3) por complemento
p0 = stats.poisson.pmf(0, lam)
p1 = stats.poisson.pmf(1, lam)
p2 = stats.poisson.pmf(2, lam)
p_ge3 = 1 - (p0 + p1 + p2)
print(f"\nP(X=0) = {p0:.5f}")
print(f"P(X=1) = {p1:.5f}")
print(f"P(X=2) = {p2:.5f}")
print(f"P(X<=2) = {p0+p1+p2:.5f}")
print(f"P(X>=3) = 1 - {p0+p1+p2:.5f} = {p_ge3:.4f}")

# Verificacion con scipy
print(f"Verificacion: 1 - CDF(2) = {1 - stats.poisson.cdf(2, lam):.4f}")

# Grafica: PMF con area sombreada para X>=3
x_vals = np.arange(0, 16)
pmf_vals = stats.poisson.pmf(x_vals, lam)

fig, ax = plt.subplots()
colores = [C_DANGER if x < 3 else C_SUCCESS for x in x_vals]
ax.bar(x_vals, pmf_vals, color=colores, edgecolor=C_DARK, alpha=0.85)
ax.axvline(x=E_X, color=C_ORANGE, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.set_xlabel('k')
ax.set_ylabel('P(X = k)')
ax.set_title(f'Poisson(lambda=5) - P(X>=3) = {p_ge3:.4f} (verde)', color=C_DARK, fontweight='bold')
ax.set_xticks(x_vals)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 6:** Si E[X]=10, E[Y]=20, Var(X)=4, Var(Y)=9, y X e Y son independientes, calcular E[3X + 2Y] y Var(3X + 2Y).

**Resolucion:**

**Propiedad de linealidad de la esperanza** (siempre vale, no requiere independencia):

$$E[aX + bY] = a \cdot E[X] + b \cdot E[Y]$$

$$E[3X + 2Y] = 3 \times 10 + 2 \times 20 = 30 + 40 = 70$$

**Propiedad de la varianza** (requiere independencia para que no haya termino de covarianza):

$$Var(aX + bY) = a^2 \cdot Var(X) + b^2 \cdot Var(Y)$$

Donde:
- $a = 3$, $b = 2$
- La constante se eleva al cuadrado porque la varianza mide dispersion cuadratica

$$Var(3X + 2Y) = 3^2 \times 4 + 2^2 \times 9 = 9 \times 4 + 4 \times 9 = 36 + 36 = 72$$

**Ejercicio 8:** Demostrar que Var(X) = E[X^2] - (E[X])^2 usando un ejemplo discreto: un dado de 3 caras con valores {1, 2, 3} equiprobables. Calcular la varianza por ambos metodos y verificar que dan el mismo resultado.

**Resolucion:**

Dado de 3 caras: $X \in \{1, 2, 3\}$, cada uno con $P = 1/3$.

**Metodo 1 - Definicion directa:**

$$Var(X) = E[(X - \mu)^2] = \sum_{i} (x_i - \mu)^2 \cdot P(x_i)$$

Primero: $\mu = E[X] = \frac{1+2+3}{3} = 2$

$$Var(X) = \frac{1}{3}(1-2)^2 + \frac{1}{3}(2-2)^2 + \frac{1}{3}(3-2)^2$$

$$= \frac{1}{3}(1 + 0 + 1) = \frac{2}{3} \approx 0.6667$$

**Metodo 2 - Formula alternativa:**

$$Var(X) = E[X^2] - (E[X])^2$$

$$E[X^2] = \frac{1}{3}(1^2 + 2^2 + 3^2) = \frac{1 + 4 + 9}{3} = \frac{14}{3} \approx 4.6667$$

$$Var(X) = \frac{14}{3} - 2^2 = \frac{14}{3} - 4 = \frac{14 - 12}{3} = \frac{2}{3} \approx 0.6667$$

Ambos metodos dan el mismo resultado: $\frac{2}{3}$.

```python
x = np.array([1, 2, 3])
p = np.array([1/3, 1/3, 1/3])

# Metodo 1: definicion directa
E_X = np.sum(x * p)
var_metodo1 = np.sum((x - E_X)**2 * p)

# Metodo 2: formula alternativa
E_X2 = np.sum(x**2 * p)
var_metodo2 = E_X2 - E_X**2

print("=== Metodo 1: Var(X) = E[(X - mu)^2] ===")
print(f"E[X] = mu = {E_X}")
for xi, pi in zip(x, p):
    print(f"  ({xi} - {E_X})^2 * {pi:.4f} = {(xi-E_X)**2 * pi:.4f}")
print(f"Var(X) = {var_metodo1:.4f}")

print(f"\n=== Metodo 2: Var(X) = E[X^2] - (E[X])^2 ===")
print(f"E[X^2] = {E_X2:.4f}")
print(f"(E[X])^2 = {E_X**2:.4f}")
print(f"Var(X) = {E_X2:.4f} - {E_X**2:.4f} = {var_metodo2:.4f}")

print(f"\nAmbos metodos iguales: {np.isclose(var_metodo1, var_metodo2)}")
print(f"Valor exacto: 2/3 = {2/3:.4f}")

# Grafica: descomposicion visual de la varianza
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Metodo 1: desviaciones al cuadrado
desviaciones = (x - E_X)**2
contribuciones = desviaciones * p
ax1.bar(x, contribuciones, color=[C_DANGER, C_SUCCESS, C_DANGER], edgecolor=C_DARK, alpha=0.85, width=0.5)
ax1.axhline(y=0, color=C_DARK, linewidth=0.5)
for xi, c in zip(x, contribuciones):
    ax1.text(xi, c + 0.02, f'{c:.4f}', ha='center', fontsize=10, color=C_DARK)
ax1.set_xlabel('x')
ax1.set_ylabel('(x - mu)^2 * P(x)')
ax1.set_title(f'Metodo 1: contribuciones a Var = {var_metodo1:.4f}', color=C_DARK, fontweight='bold')
ax1.set_xticks(x)

# Metodo 2: E[X^2] vs (E[X])^2
categorias = ['E[X^2]', '(E[X])^2', 'Var(X)']
valores = [E_X2, E_X**2, var_metodo2]
colores_bar = [C_PRIMARY, C_ORANGE, C_SUCCESS]
ax2.bar(categorias, valores, color=colores_bar, edgecolor=C_DARK, alpha=0.85, width=0.5)
for i, v in enumerate(valores):
    ax2.text(i, v + 0.05, f'{v:.4f}', ha='center', fontweight='bold', color=C_DARK)
ax2.set_title(f'Metodo 2: E[X^2] - (E[X])^2 = Var', color=C_DARK, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Ejercicio 9:** X tiene PMF: P(X=1)=0.1, P(X=2)=0.3, P(X=3)=0.4, P(X=4)=0.2. Calcular E[X], E[X^2], Var(X).

**Resolucion:**

Se tiene una variable aleatoria discreta $X$ con funcion de masa de probabilidad (PMF) dada por:

| $x_i$ | $P(X = x_i)$ |
|--------|---------------|
| 1      | 0.1           |
| 2      | 0.3           |
| 3      | 0.4           |
| 4      | 0.2           |

Donde:
- $X$: variable aleatoria discreta que toma valores en $\{1, 2, 3, 4\}$.
- $x_i$: cada uno de los posibles valores de $X$. Aqui $x_1 = 1$, $x_2 = 2$, $x_3 = 3$, $x_4 = 4$.
- $P(X = x_i)$: probabilidad de que $X$ tome el valor $x_i$. La suma de todas las probabilidades es $0.1 + 0.3 + 0.4 + 0.2 = 1$.

**Esperanza E[X]:**

La esperanza es el valor promedio ponderado por las probabilidades:

$$E[X] = \sum_{i=1}^{4} x_i \cdot P(X = x_i)$$

- $E[X]$: esperanza de $X$, tambien llamada valor esperado o media $\mu$.
- $\sum_{i=1}^{4}$: sumatoria sobre los 4 valores posibles de $X$.
- $x_i \cdot P(X = x_i)$: cada valor multiplicado por su probabilidad.

$$E[X] = 1 \times 0.1 + 2 \times 0.3 + 3 \times 0.4 + 4 \times 0.2$$
$$= 0.1 + 0.6 + 1.2 + 0.8 = 2.7$$

**Segundo momento E[X^2]:**

$$E[X^2] = \sum_{i=1}^{4} x_i^2 \cdot P(X = x_i)$$

- $E[X^2]$: esperanza del cuadrado de $X$, es decir el segundo momento. Se calcula aplicando LOTUS (Law of the Unconscious Statistician): no necesitamos la distribucion de $X^2$, solo la de $X$.
- $x_i^2$: el cuadrado de cada valor posible de $X$.

$$E[X^2] = 1^2 \times 0.1 + 2^2 \times 0.3 + 3^2 \times 0.4 + 4^2 \times 0.2$$
$$= 0.1 + 1.2 + 3.6 + 3.2 = 8.1$$

**Varianza Var(X):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $Var(X)$: varianza de $X$, denotada $\sigma^2$. Mide la dispersion de $X$ respecto a su media.
- $E[X^2]$: segundo momento de $X$, calculado arriba como $8.1$.
- $(E[X])^2$: cuadrado de la esperanza, es decir $\mu^2 = (2.7)^2 = 7.29$.

$$Var(X) = 8.1 - (2.7)^2 = 8.1 - 7.29 = 0.81$$

La desviacion estandar es $\sigma = \sqrt{Var(X)} = \sqrt{0.81} = 0.9$.

```python
x = np.array([1, 2, 3, 4])
p = np.array([0.1, 0.3, 0.4, 0.2])

# Verificacion de PMF valida
print(f"Suma de probabilidades = {np.sum(p)} (debe ser 1)")

# E[X]
E_X = np.sum(x * p)
print(f"\nE[X] = {' + '.join([f'{xi}*{pi}' for xi, pi in zip(x, p)])}")
print(f"E[X] = {' + '.join([f'{xi*pi:.1f}' for xi, pi in zip(x, p)])} = {E_X}")

# E[X^2]
E_X2 = np.sum(x**2 * p)
print(f"\nE[X^2] = {' + '.join([f'{xi}^2*{pi}' for xi, pi in zip(x, p)])}")
print(f"E[X^2] = {' + '.join([f'{xi**2*pi:.1f}' for xi, pi in zip(x, p)])} = {E_X2}")

# Var(X)
Var_X = E_X2 - E_X**2
std_X = np.sqrt(Var_X)
print(f"\nVar(X) = E[X^2] - (E[X])^2 = {E_X2} - ({E_X})^2 = {E_X2} - {E_X**2} = {Var_X}")
print(f"Desviacion estandar = sqrt({Var_X}) = {std_X}")

# Grafica: PMF con E[X] y banda de +/- 1 std
fig, ax = plt.subplots()
ax.bar(x, p, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, width=0.5, label='P(X = x)')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.axvspan(E_X - std_X, E_X + std_X, alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {std_X:.2f}')

for xi, pi in zip(x, p):
    ax.text(xi, pi + 0.015, f'{pi}', ha='center', fontweight='bold', color=C_DARK)

ax.set_xlabel('x')
ax.set_ylabel('P(X = x)')
ax.set_title('PMF discreta con E[X] y desviacion estandar', color=C_DARK, fontweight='bold')
ax.set_xticks(x)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 10:** Variable continua f(x)=3x^2 para x en [0,1]. Calcular E[X] y Var(X) con integrales.

**Resolucion:**

Se tiene una variable aleatoria continua $X$ con funcion de densidad de probabilidad (PDF):

$$f(x) = 3x^2, \quad x \in [0, 1]$$

Donde:
- $f(x)$: funcion de densidad de probabilidad. Asigna una "densidad" a cada punto $x$, de modo que la probabilidad de un intervalo es el area bajo la curva.
- $x$: variable continua que toma valores en el intervalo $[0, 1]$.
- $3x^2$: funcion cuadratica que crece con $x$, lo que significa que valores cercanos a 1 son mas probables que valores cercanos a 0.

**Verificacion de densidad valida:**

$$\int_0^1 3x^2 \, dx = \left[ x^3 \right]_0^1 = 1^3 - 0^3 = 1$$

- $\int_0^1$: integral definida en el soporte $[0, 1]$.
- $dx$: diferencial de integracion.
- El resultado es 1, confirmando que $f(x)$ es una densidad valida.

**Esperanza E[X]:**

$$E[X] = \int_0^1 x \cdot f(x) \, dx = \int_0^1 x \cdot 3x^2 \, dx = \int_0^1 3x^3 \, dx$$

- $E[X]$: esperanza de $X$, el "centro de masa" de la distribucion.
- $x \cdot f(x)$: el integrando es el valor $x$ ponderado por su densidad $f(x)$.
- $3x^3$: resultado de multiplicar $x$ por $3x^2$.

$$E[X] = \left[ \frac{3x^4}{4} \right]_0^1 = \frac{3}{4} = 0.75$$

**Segundo momento E[X^2]:**

$$E[X^2] = \int_0^1 x^2 \cdot f(x) \, dx = \int_0^1 x^2 \cdot 3x^2 \, dx = \int_0^1 3x^4 \, dx$$

- $E[X^2]$: segundo momento de $X$. Se usa para calcular la varianza.
- $x^2 \cdot f(x)$: el cuadrado de $x$ ponderado por la densidad.

$$E[X^2] = \left[ \frac{3x^5}{5} \right]_0^1 = \frac{3}{5} = 0.6$$

**Varianza Var(X):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $Var(X)$: varianza, mide la dispersion de $X$ respecto a su media $\mu = E[X] = 0.75$.
- $E[X^2] = 0.6$: segundo momento calculado arriba.
- $(E[X])^2 = (0.75)^2 = 0.5625$: cuadrado de la esperanza.

$$Var(X) = 0.6 - (0.75)^2 = 0.6 - 0.5625 = 0.0375$$

La desviacion estandar es $\sigma = \sqrt{0.0375} \approx 0.1936$.

```python
f = lambda x: 3 * x**2

# Verificacion de densidad
integral_total, _ = quad(f, 0, 1)
print(f"Integral de f(x) en [0,1] = {integral_total} (debe ser 1)")

# E[X]
E_X, _ = quad(lambda x: x * f(x), 0, 1)
print(f"\nE[X] = integral(x * 3x^2, 0, 1) = integral(3x^3, 0, 1) = {E_X:.4f} = 3/4")

# E[X^2]
E_X2, _ = quad(lambda x: x**2 * f(x), 0, 1)
print(f"E[X^2] = integral(x^2 * 3x^2, 0, 1) = integral(3x^4, 0, 1) = {E_X2:.4f} = 3/5")

# Var(X)
Var_X = E_X2 - E_X**2
std_X = np.sqrt(Var_X)
print(f"\nVar(X) = E[X^2] - (E[X])^2 = {E_X2:.4f} - ({E_X:.4f})^2 = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")
print(f"Desviacion estandar = sqrt({Var_X:.4f}) = {std_X:.4f}")

# Grafica: PDF con E[X] y region de +/- 1 std
x_plot = np.linspace(0, 1, 300)
y_plot = 3 * x_plot**2

fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, color=C_PRIMARY, linewidth=2.5, label='f(x) = 3x^2')
ax.fill_between(x_plot, y_plot, alpha=0.2, color=C_PRIMARY)

# Sombrear region E[X] +/- std
mask = (x_plot >= E_X - std_X) & (x_plot <= E_X + std_X)
ax.fill_between(x_plot[mask], y_plot[mask], alpha=0.4, color=C_ORANGE, label=f'E[X] +/- std')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X:.4f}')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF f(x) = 3x^2 en [0, 1]', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 11:** X ~ Geometrica(p=0.2). Calcular E[X] y Var(X). Cuantos intentos esperas para el primer exito?

**Resolucion:**

La distribucion geometrica modela el numero de ensayos independientes de Bernoulli necesarios para obtener el primer exito.

$$P(X = k) = (1 - p)^{k-1} \cdot p, \quad k = 1, 2, 3, \ldots$$

Donde:
- $X$: variable aleatoria que cuenta el numero de ensayos hasta el primer exito. Toma valores en $\{1, 2, 3, \ldots\}$.
- $k$: numero de ensayos realizados (incluyendo el exitoso).
- $p$: probabilidad de exito en cada ensayo individual. Aqui $p = 0.2$.
- $(1 - p)$: probabilidad de fracaso en cada ensayo. Aqui $1 - p = 0.8$.
- $(1 - p)^{k-1}$: probabilidad de obtener $k-1$ fracasos consecutivos antes del exito.
- La multiplicacion por $p$ al final representa el exito en el ensayo $k$-esimo.

**Esperanza E[X]:**

$$E[X] = \frac{1}{p}$$

- $E[X]$: numero esperado de ensayos hasta el primer exito.
- $\frac{1}{p}$: inverso de la probabilidad de exito. A menor probabilidad de exito, mas ensayos se esperan.

$$E[X] = \frac{1}{0.2} = 5$$

Interpretacion: en promedio, se esperan 5 intentos para obtener el primer exito.

**Varianza Var(X):**

$$Var(X) = \frac{1 - p}{p^2}$$

- $Var(X)$: varianza de $X$, mide la dispersion del numero de ensayos respecto a la media.
- $1 - p$: probabilidad de fracaso. Aqui $1 - p = 0.8$.
- $p^2$: cuadrado de la probabilidad de exito. Aqui $p^2 = 0.04$.

$$Var(X) = \frac{1 - 0.2}{(0.2)^2} = \frac{0.8}{0.04} = 20$$

La desviacion estandar es $\sigma = \sqrt{20} \approx 4.47$, indicando alta variabilidad en el numero de intentos.

```python
p = 0.2

# Formulas teoricas
E_X = 1 / p
Var_X = (1 - p) / p**2
std_X = np.sqrt(Var_X)

print(f"E[X] = 1/p = 1/{p} = {E_X}")
print(f"Var(X) = (1-p)/p^2 = {1-p}/{p**2} = {Var_X}")
print(f"Desviacion estandar = sqrt({Var_X}) = {std_X:.4f}")
print(f"\nInterpretacion: se esperan {E_X:.0f} intentos para el primer exito")

# Verificacion con scipy (geom en scipy cuenta desde 1)
dist = stats.geom(p)
print(f"\nVerificacion scipy: E[X] = {dist.mean()}, Var(X) = {dist.var()}")

# Simulacion
np.random.seed(42)
n_sim = 50_000
simulaciones = stats.geom.rvs(p, size=n_sim)
print(f"\nSimulacion ({n_sim:,} muestras):")
print(f"Media simulada = {np.mean(simulaciones):.2f} (teorica: {E_X})")
print(f"Varianza simulada = {np.var(simulaciones):.2f} (teorica: {Var_X})")

# Grafica: PMF geometrica
k_vals = np.arange(1, 26)
pmf_vals = stats.geom.pmf(k_vals, p)

fig, ax = plt.subplots()
ax.bar(k_vals, pmf_vals, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X = k)')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X:.0f}')
ax.axvspan(E_X - std_X, E_X + std_X, alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {std_X:.2f}')

ax.set_xlabel('k (numero de intentos)')
ax.set_ylabel('P(X = k)')
ax.set_title(f'Geometrica(p={p}): intentos hasta el primer exito', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 12:** E[aX+b] = aE[X]+b. Verificar con X ~ Binomial(10, 0.3), a=2, b=5.

**Resolucion:**

Se quiere verificar la propiedad de linealidad de la esperanza:

$$E[aX + b] = a \cdot E[X] + b$$

Donde:
- $E[aX + b]$: esperanza de la transformacion lineal $aX + b$.
- $a$: constante multiplicativa (escalar). Aqui $a = 2$.
- $b$: constante aditiva. Aqui $b = 5$.
- $X$: variable aleatoria. Aqui $X \sim \text{Binomial}(n=10, p=0.3)$.
- $E[X]$: esperanza de $X$.
- Esta propiedad se cumple siempre, independientemente de la distribucion de $X$.

**Paso 1: Calcular E[X] para X ~ Binomial(n, p):**

$$E[X] = n \cdot p$$

- $n$: numero de ensayos. Aqui $n = 10$.
- $p$: probabilidad de exito en cada ensayo. Aqui $p = 0.3$.

$$E[X] = 10 \times 0.3 = 3$$

**Paso 2: Calcular E[aX + b] usando la propiedad:**

$$E[aX + b] = a \cdot E[X] + b = 2 \times 3 + 5 = 6 + 5 = 11$$

**Paso 3: Verificar directamente con la definicion:**

$$E[aX + b] = \sum_{k=0}^{n} (a \cdot k + b) \cdot P(X = k)$$

- $k$: cada posible valor de $X$, desde 0 hasta $n = 10$.
- $a \cdot k + b$: transformacion lineal aplicada al valor $k$. Aqui $2k + 5$.
- $P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$: probabilidad binomial de obtener exactamente $k$ exitos.
- $\binom{n}{k}$: coeficiente binomial, numero de formas de elegir $k$ exitos entre $n$ ensayos.

**Tambien verificamos la varianza de la transformacion:**

$$Var(aX + b) = a^2 \cdot Var(X)$$

- $Var(aX + b)$: varianza de la transformacion lineal.
- $a^2$: la constante multiplicativa se eleva al cuadrado porque la varianza mide dispersion cuadratica. Aqui $a^2 = 4$.
- $Var(X) = n \cdot p \cdot (1-p) = 10 \times 0.3 \times 0.7 = 2.1$.
- $b$ no aparece porque trasladar una distribucion no cambia su dispersion.

$$Var(2X + 5) = 2^2 \times 2.1 = 4 \times 2.1 = 8.4$$

```python
n, p = 10, 0.3
a, b = 2, 5

# E[X] y Var(X) de la Binomial
E_X = n * p
Var_X = n * p * (1 - p)

# Propiedad de linealidad
E_aXb_propiedad = a * E_X + b
Var_aXb_propiedad = a**2 * Var_X

print("=== Usando propiedades ===")
print(f"E[X] = n*p = {n}*{p} = {E_X}")
print(f"E[aX+b] = a*E[X]+b = {a}*{E_X}+{b} = {E_aXb_propiedad}")
print(f"Var(X) = n*p*(1-p) = {Var_X}")
print(f"Var(aX+b) = a^2*Var(X) = {a**2}*{Var_X} = {Var_aXb_propiedad}")

# Verificacion directa con la definicion
k_vals = np.arange(0, n + 1)
pmf_vals = stats.binom.pmf(k_vals, n, p)
transformados = a * k_vals + b

E_aXb_directo = np.sum(transformados * pmf_vals)
E_aXb2_directo = np.sum(transformados**2 * pmf_vals)
Var_aXb_directo = E_aXb2_directo - E_aXb_directo**2

print(f"\n=== Verificacion directa (definicion) ===")
print(f"E[aX+b] = sum((2k+5)*P(k)) = {E_aXb_directo}")
print(f"Var(aX+b) = E[(aX+b)^2] - (E[aX+b])^2 = {Var_aXb_directo:.4f}")
print(f"\nCoinciden: E -> {np.isclose(E_aXb_propiedad, E_aXb_directo)}, Var -> {np.isclose(Var_aXb_propiedad, Var_aXb_directo)}")

# Grafica: comparar PMF de X vs PMF de aX+b
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(k_vals, pmf_vals, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=k)')
ax1.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title('X ~ Binomial(10, 0.3)', color=C_DARK, fontweight='bold')
ax1.legend()

ax2.bar(transformados, pmf_vals, color=C_SUCCESS, edgecolor=C_DARK, alpha=0.85, width=1.5, label='P(2X+5=v)')
ax2.axvline(x=E_aXb_propiedad, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[2X+5] = {E_aXb_propiedad}')
ax2.set_xlabel('v = 2k + 5')
ax2.set_ylabel('P(2X+5 = v)')
ax2.set_title('Y = 2X + 5 (transformacion lineal)', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 13:** X ~ Uniforme discreta {1,2,...,n}. Derivar E[X]=(n+1)/2 y Var(X)=(n^2-1)/12 para n=6.

**Resolucion:**

La distribucion uniforme discreta asigna la misma probabilidad a cada valor del conjunto $\{1, 2, \ldots, n\}$.

$$P(X = k) = \frac{1}{n}, \quad k = 1, 2, \ldots, n$$

Donde:
- $X$: variable aleatoria con distribucion uniforme discreta sobre $\{1, 2, \ldots, n\}$.
- $k$: cada uno de los $n$ valores posibles.
- $\frac{1}{n}$: probabilidad de cada valor. Todos tienen la misma probabilidad.
- $n$: numero total de valores posibles. Aqui $n = 6$.

**Derivacion de E[X]:**

$$E[X] = \sum_{k=1}^{n} k \cdot \frac{1}{n} = \frac{1}{n} \sum_{k=1}^{n} k$$

- $E[X]$: esperanza de $X$, el valor promedio.
- $\sum_{k=1}^{n} k$: suma de los primeros $n$ enteros positivos.
- $\frac{1}{n}$: factor constante que sale de la sumatoria.

Usando la formula conocida $\sum_{k=1}^{n} k = \frac{n(n+1)}{2}$:

$$E[X] = \frac{1}{n} \cdot \frac{n(n+1)}{2} = \frac{n+1}{2}$$

Para $n = 6$: $E[X] = \frac{6+1}{2} = \frac{7}{2} = 3.5$

**Derivacion de Var(X):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $Var(X)$: varianza de $X$, mide la dispersion respecto a la media.
- $E[X^2]$: segundo momento de $X$.
- $(E[X])^2$: cuadrado de la esperanza.

Primero calculamos $E[X^2]$:

$$E[X^2] = \sum_{k=1}^{n} k^2 \cdot \frac{1}{n} = \frac{1}{n} \sum_{k=1}^{n} k^2$$

- $\sum_{k=1}^{n} k^2$: suma de los cuadrados de los primeros $n$ enteros.

Usando la formula conocida $\sum_{k=1}^{n} k^2 = \frac{n(n+1)(2n+1)}{6}$:

$$E[X^2] = \frac{1}{n} \cdot \frac{n(n+1)(2n+1)}{6} = \frac{(n+1)(2n+1)}{6}$$

Entonces:

$$Var(X) = \frac{(n+1)(2n+1)}{6} - \left(\frac{n+1}{2}\right)^2 = \frac{(n+1)(2n+1)}{6} - \frac{(n+1)^2}{4}$$

$$= (n+1) \left[\frac{2n+1}{6} - \frac{n+1}{4}\right] = (n+1) \cdot \frac{2(2n+1) - 3(n+1)}{12}$$

$$= (n+1) \cdot \frac{4n + 2 - 3n - 3}{12} = (n+1) \cdot \frac{n - 1}{12} = \frac{n^2 - 1}{12}$$

Para $n = 6$: $Var(X) = \frac{36 - 1}{12} = \frac{35}{12} \approx 2.9167$

```python
n = 6
x = np.arange(1, n + 1)
p = np.ones(n) / n

# Formulas teoricas
E_X_teorica = (n + 1) / 2
Var_X_teorica = (n**2 - 1) / 12

# Calculo directo
E_X = np.sum(x * p)
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2

print(f"n = {n}")
print(f"\n=== Formula teorica ===")
print(f"E[X] = (n+1)/2 = ({n}+1)/2 = {E_X_teorica}")
print(f"Var(X) = (n^2-1)/12 = ({n}^2-1)/12 = ({n**2}-1)/12 = {Var_X_teorica:.4f}")

print(f"\n=== Calculo directo ===")
print(f"E[X] = (1/n) * sum(k) = (1/{n}) * {int(np.sum(x))} = {E_X}")
print(f"E[X^2] = (1/n) * sum(k^2) = (1/{n}) * {int(np.sum(x**2))} = {E_X2:.4f}")
print(f"Var(X) = E[X^2] - (E[X])^2 = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")

print(f"\nCoinciden: E -> {np.isclose(E_X, E_X_teorica)}, Var -> {np.isclose(Var_X, Var_X_teorica)}")

# Grafica: PMF uniforme con formulas para distintos n
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# PMF para n=6
ax1.bar(x, p, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=k) = 1/6')
ax1.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax1.axvspan(E_X - np.sqrt(Var_X), E_X + np.sqrt(Var_X), alpha=0.15, color=C_ORANGE,
            label=f'+/- 1 std = {np.sqrt(Var_X):.2f}')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title(f'Uniforme discreta n={n}', color=C_DARK, fontweight='bold')
ax1.set_xticks(x)
ax1.legend(fontsize=9)

# E[X] y Var(X) para distintos n
n_vals = np.arange(2, 21)
e_vals = (n_vals + 1) / 2
var_vals = (n_vals**2 - 1) / 12

ax2.plot(n_vals, e_vals, color=C_PRIMARY, linewidth=2, marker='o', markersize=4, label='E[X] = (n+1)/2')
ax2.plot(n_vals, var_vals, color=C_DANGER, linewidth=2, marker='s', markersize=4, label='Var(X) = (n^2-1)/12')
ax2.axvline(x=6, color=C_DARK, linestyle=':', alpha=0.5, label='n=6')
ax2.set_xlabel('n')
ax2.set_ylabel('Valor')
ax2.set_title('E[X] y Var(X) segun n', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 14:** X ~ Normal(50, 10). Calcular P(40<X<60), P(X>65), E[X], Var(X).

**Resolucion:**

Se tiene $X \sim \mathcal{N}(\mu, \sigma)$ con $\mu = 50$ y $\sigma = 10$.

La funcion de densidad de la distribucion normal es:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)$$

Donde:
- $f(x)$: funcion de densidad de probabilidad (PDF) de la normal. Es la famosa "campana de Gauss".
- $\mu$: media de la distribucion. Aqui $\mu = 50$. Es el centro de la campana.
- $\sigma$: desviacion estandar. Aqui $\sigma = 10$. Controla el ancho de la campana.
- $\sigma^2$: varianza, $\sigma^2 = 100$.
- $\pi$: constante pi ($\approx 3.14159$).
- $\exp(\cdot)$: funcion exponencial $e^{(\cdot)}$.
- $(x - \mu)^2$: distancia cuadratica de $x$ a la media.
- $2\sigma^2$: factor de escala en el exponente.
- $\frac{1}{\sigma\sqrt{2\pi}}$: constante de normalizacion que asegura que el area total bajo la curva sea 1.

**Esperanza y Varianza:**

$$E[X] = \mu = 50$$

- $E[X]$: esperanza de $X$. Para la distribucion normal, la esperanza coincide con el parametro $\mu$.

$$Var(X) = \sigma^2 = 10^2 = 100$$

- $Var(X)$: varianza de $X$. Para la normal, es el cuadrado del parametro $\sigma$.

**P(40 < X < 60):**

Estandarizamos usando $Z = \frac{X - \mu}{\sigma}$:

- $Z$: variable normal estandar, $Z \sim \mathcal{N}(0, 1)$.
- $\frac{X - \mu}{\sigma}$: transformacion que convierte $X$ a la escala estandar (media 0, desviacion 1).

$$P(40 < X < 60) = P\left(\frac{40 - 50}{10} < Z < \frac{60 - 50}{10}\right) = P(-1 < Z < 1)$$

$$= \Phi(1) - \Phi(-1) \approx 0.8413 - 0.1587 = 0.6827$$

- $\Phi(\cdot)$: funcion de distribucion acumulada (CDF) de la normal estandar $\mathcal{N}(0,1)$.
- $\Phi(1) \approx 0.8413$: probabilidad de que $Z \leq 1$.
- $\Phi(-1) \approx 0.1587$: probabilidad de que $Z \leq -1$.

Esto confirma la regla empirica: aproximadamente el 68.27% de los datos caen dentro de $\pm 1\sigma$ de la media.

**P(X > 65):**

$$P(X > 65) = P\left(Z > \frac{65 - 50}{10}\right) = P(Z > 1.5) = 1 - \Phi(1.5)$$

$$= 1 - 0.9332 = 0.0668$$

- $\Phi(1.5) \approx 0.9332$: probabilidad acumulada hasta $z = 1.5$.
- $1 - \Phi(1.5)$: cola derecha, probabilidad de estar a mas de 1.5 desviaciones estandar por encima de la media.

```python
mu, sigma = 50, 10
dist = stats.norm(loc=mu, scale=sigma)

E_X = dist.mean()
Var_X = dist.var()

# Probabilidades
p_40_60 = dist.cdf(60) - dist.cdf(40)
p_gt65 = 1 - dist.cdf(65)

# Estandarizacion
z_40 = (40 - mu) / sigma
z_60 = (60 - mu) / sigma
z_65 = (65 - mu) / sigma

print(f"E[X] = mu = {E_X}")
print(f"Var(X) = sigma^2 = {sigma}^2 = {Var_X}")
print(f"sigma = {sigma}")

print(f"\nP(40 < X < 60):")
print(f"  Z_40 = (40 - {mu}) / {sigma} = {z_40}")
print(f"  Z_60 = (60 - {mu}) / {sigma} = {z_60}")
print(f"  P(-1 < Z < 1) = Phi(1) - Phi(-1) = {stats.norm.cdf(1):.4f} - {stats.norm.cdf(-1):.4f} = {p_40_60:.4f}")

print(f"\nP(X > 65):")
print(f"  Z_65 = (65 - {mu}) / {sigma} = {z_65}")
print(f"  P(Z > 1.5) = 1 - Phi(1.5) = 1 - {stats.norm.cdf(1.5):.4f} = {p_gt65:.4f}")

# Grafica: PDF normal con regiones sombreadas
x_plot = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)
y_plot = dist.pdf(x_plot)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_plot, y_plot, color=C_DARK, linewidth=2, label=f'N({mu}, {sigma})')

# Sombrear P(40 < X < 60)
mask1 = (x_plot >= 40) & (x_plot <= 60)
ax.fill_between(x_plot[mask1], y_plot[mask1], alpha=0.4, color=C_PRIMARY,
                label=f'P(40<X<60) = {p_40_60:.4f}')

# Sombrear P(X > 65)
mask2 = x_plot >= 65
ax.fill_between(x_plot[mask2], y_plot[mask2], alpha=0.5, color=C_DANGER,
                label=f'P(X>65) = {p_gt65:.4f}')

ax.axvline(x=mu, color=C_ORANGE, linewidth=2, linestyle='--', label=f'mu = {mu}')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(f'Normal(mu={mu}, sigma={sigma}): probabilidades sombreadas', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 15:** Dos dados: Y=X1+X2. Calcular E[Y] y Var(Y) usando propiedades de linealidad.

**Resolucion:**

Se lanzan dos dados justos de 6 caras independientes. Definimos:

- $X_1$: resultado del primer dado. $X_1 \in \{1, 2, 3, 4, 5, 6\}$, cada valor con probabilidad $\frac{1}{6}$.
- $X_2$: resultado del segundo dado. $X_2 \in \{1, 2, 3, 4, 5, 6\}$, cada valor con probabilidad $\frac{1}{6}$.
- $Y = X_1 + X_2$: suma de ambos dados. $Y \in \{2, 3, \ldots, 12\}$.
- $X_1$ y $X_2$ son **independientes**: el resultado de un dado no afecta al otro.

**E[Y] usando linealidad de la esperanza:**

$$E[Y] = E[X_1 + X_2] = E[X_1] + E[X_2]$$

- $E[Y]$: esperanza de la suma $Y$.
- $E[X_1 + X_2]$: esperanza de la suma de dos variables aleatorias.
- $E[X_1] + E[X_2]$: la esperanza de la suma siempre es la suma de las esperanzas (no requiere independencia).

Para un dado justo (uniforme discreta en $\{1, \ldots, 6\}$):

$$E[X_1] = E[X_2] = \frac{n+1}{2} = \frac{6+1}{2} = 3.5$$

- $\frac{n+1}{2}$: formula de la esperanza de la uniforme discreta con $n = 6$.

$$E[Y] = 3.5 + 3.5 = 7$$

**Var(Y) usando propiedades de la varianza:**

$$Var(Y) = Var(X_1 + X_2) = Var(X_1) + Var(X_2) + 2 \cdot Cov(X_1, X_2)$$

- $Var(Y)$: varianza de la suma.
- $Var(X_1) + Var(X_2)$: suma de las varianzas individuales.
- $Cov(X_1, X_2)$: covarianza entre $X_1$ y $X_2$. Mide la dependencia lineal.
- Como $X_1$ y $X_2$ son **independientes**, $Cov(X_1, X_2) = 0$.

$$Var(Y) = Var(X_1) + Var(X_2)$$

Para un dado justo:

$$Var(X_1) = Var(X_2) = \frac{n^2 - 1}{12} = \frac{36 - 1}{12} = \frac{35}{12} \approx 2.9167$$

- $\frac{n^2-1}{12}$: formula de la varianza de la uniforme discreta con $n = 6$.

$$Var(Y) = \frac{35}{12} + \frac{35}{12} = \frac{70}{12} = \frac{35}{6} \approx 5.8333$$

La desviacion estandar de $Y$ es $\sigma_Y = \sqrt{35/6} \approx 2.4152$.

```python
# Propiedades teoricas
E_X1 = 3.5
E_X2 = 3.5
Var_X1 = 35 / 12
Var_X2 = 35 / 12

E_Y = E_X1 + E_X2
Var_Y = Var_X1 + Var_X2  # independientes
std_Y = np.sqrt(Var_Y)

print(f"E[X1] = E[X2] = {E_X1}")
print(f"Var(X1) = Var(X2) = 35/12 = {Var_X1:.4f}")
print(f"\nE[Y] = E[X1] + E[X2] = {E_X1} + {E_X2} = {E_Y}")
print(f"Var(Y) = Var(X1) + Var(X2) = {Var_X1:.4f} + {Var_X2:.4f} = {Var_Y:.4f}")
print(f"Desviacion estandar de Y = {std_Y:.4f}")

# Verificacion: construir PMF exacta de Y = X1 + X2
y_vals = np.arange(2, 13)
pmf_Y = np.zeros(11)
for x1 in range(1, 7):
    for x2 in range(1, 7):
        y = x1 + x2
        pmf_Y[y - 2] += 1 / 36

E_Y_directo = np.sum(y_vals * pmf_Y)
E_Y2_directo = np.sum(y_vals**2 * pmf_Y)
Var_Y_directo = E_Y2_directo - E_Y_directo**2

print(f"\n=== Verificacion directa (36 combinaciones) ===")
print(f"E[Y] = {E_Y_directo}")
print(f"Var(Y) = {Var_Y_directo:.4f}")

# Grafica: PMF de Y = X1 + X2
fig, ax = plt.subplots()
ax.bar(y_vals, pmf_Y, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(Y = y)')
ax.axvline(x=E_Y, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[Y] = {E_Y}')
ax.axvspan(E_Y - std_Y, E_Y + std_Y, alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {std_Y:.2f}')

for yi, pi in zip(y_vals, pmf_Y):
    ax.text(yi, pi + 0.005, f'{pi:.3f}', ha='center', fontsize=9, color=C_DARK)

ax.set_xlabel('Y = X1 + X2')
ax.set_ylabel('P(Y = y)')
ax.set_title('PMF de la suma de dos dados justos', color=C_DARK, fontweight='bold')
ax.set_xticks(y_vals)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 16:** X ~ Exponencial(lambda=0.5). Calcular E[X], Var(X), P(X>3) con integral.

**Resolucion:**

La distribucion exponencial modela tiempos de espera entre eventos en un proceso de Poisson.

$$f(x) = \lambda e^{-\lambda x}, \quad x \geq 0$$

Donde:
- $f(x)$: funcion de densidad de probabilidad (PDF) de la exponencial.
- $\lambda$: parametro de tasa (rate). Aqui $\lambda = 0.5$. Representa el numero promedio de eventos por unidad de tiempo.
- $e^{-\lambda x}$: decaimiento exponencial. La densidad decrece rapidamente.
- $x$: variable continua no negativa (tiempo de espera, por ejemplo).

**Esperanza E[X]:**

$$E[X] = \frac{1}{\lambda}$$

- $E[X]$: tiempo de espera promedio hasta el proximo evento.
- $\frac{1}{\lambda}$: inverso de la tasa. A mayor tasa, menor tiempo de espera promedio.

$$E[X] = \frac{1}{0.5} = 2$$

**Varianza Var(X):**

$$Var(X) = \frac{1}{\lambda^2}$$

- $Var(X)$: varianza del tiempo de espera.
- $\frac{1}{\lambda^2}$: cuadrado del inverso de la tasa.
- $\lambda^2 = (0.5)^2 = 0.25$.

$$Var(X) = \frac{1}{(0.5)^2} = \frac{1}{0.25} = 4$$

La desviacion estandar es $\sigma = \sqrt{4} = 2$, que coincide con la media (propiedad de la exponencial).

**P(X > 3) con integral:**

$$P(X > 3) = \int_3^{\infty} \lambda e^{-\lambda x} \, dx = \int_3^{\infty} 0.5 \cdot e^{-0.5x} \, dx$$

- $\int_3^{\infty}$: integral desde $x = 3$ hasta infinito (cola derecha).
- $\lambda e^{-\lambda x}$: la PDF evaluada en el integrando.

$$= \left[ -e^{-0.5x} \right]_3^{\infty} = 0 - (-e^{-1.5}) = e^{-1.5} \approx 0.2231$$

- $e^{-1.5}$: resultado de evaluar $-e^{-0.5x}$ en los limites. En $x \to \infty$, $e^{-0.5x} \to 0$. En $x = 3$, $e^{-0.5 \cdot 3} = e^{-1.5}$.

```python
lam = 0.5
dist = stats.expon(scale=1/lam)

E_X = 1 / lam
Var_X = 1 / lam**2
std_X = np.sqrt(Var_X)

print(f"E[X] = 1/lambda = 1/{lam} = {E_X}")
print(f"Var(X) = 1/lambda^2 = 1/{lam**2} = {Var_X}")
print(f"Desviacion estandar = {std_X}")

# P(X > 3) con integral
p_gt3_integral, _ = quad(lambda x: lam * np.exp(-lam * x), 3, np.inf)
p_gt3_formula = np.exp(-lam * 3)
p_gt3_scipy = 1 - dist.cdf(3)

print(f"\nP(X > 3):")
print(f"  Por integral numerica: {p_gt3_integral:.4f}")
print(f"  Por formula e^(-lambda*3) = e^(-{lam*3}) = {p_gt3_formula:.4f}")
print(f"  Por scipy: {p_gt3_scipy:.4f}")

# Grafica: PDF con area sombreada para P(X > 3)
x_plot = np.linspace(0, 10, 300)
y_plot = dist.pdf(x_plot)

fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, color=C_PRIMARY, linewidth=2.5, label=f'f(x) = {lam}*exp(-{lam}x)')
ax.fill_between(x_plot, y_plot, where=(x_plot <= 3), alpha=0.3, color=C_PRIMARY,
                label=f'P(X<=3) = {1-p_gt3_formula:.4f}')
ax.fill_between(x_plot, y_plot, where=(x_plot >= 3), alpha=0.4, color=C_DANGER,
                label=f'P(X>3) = {p_gt3_formula:.4f}')
ax.axvline(x=E_X, color=C_ORANGE, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.axvline(x=3, color=C_DARK, linewidth=1, linestyle=':')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(f'Exponencial(lambda={lam}): P(X > 3) sombreada en rojo', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 17:** Desigualdad de Chebyshev: P(|X-mu|>=k*sigma) <= 1/k^2. Verificar con X ~ Normal para k=1,2,3.

**Resolucion:**

La desigualdad de Chebyshev establece una cota superior para la probabilidad de que una variable aleatoria se aleje de su media mas de $k$ desviaciones estandar:

$$P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2}$$

Donde:
- $P(|X - \mu| \geq k\sigma)$: probabilidad de que $X$ este a una distancia de al menos $k\sigma$ de su media $\mu$.
- $X$: variable aleatoria con media $\mu$ y varianza $\sigma^2$ finitas.
- $\mu$: media (esperanza) de $X$, $\mu = E[X]$.
- $\sigma$: desviacion estandar de $X$, $\sigma = \sqrt{Var(X)}$.
- $k$: numero de desviaciones estandar. Debe ser $k > 0$.
- $k\sigma$: distancia medida en unidades de desviacion estandar.
- $|X - \mu|$: valor absoluto de la desviacion de $X$ respecto a su media.
- $\frac{1}{k^2}$: cota superior. Es una cota conservadora que vale para **cualquier** distribucion.

**Equivalentemente**, la probabilidad de estar dentro de $k$ desviaciones:

$$P(|X - \mu| < k\sigma) \geq 1 - \frac{1}{k^2}$$

**Verificacion con X ~ Normal(0, 1):**

Para la normal estandar $Z \sim \mathcal{N}(0, 1)$, donde $\mu = 0$ y $\sigma = 1$:

- $Z$: variable normal estandar.
- $\mathcal{N}(0, 1)$: distribucion normal con media 0 y desviacion estandar 1.

| $k$ | Cota Chebyshev $\frac{1}{k^2}$ | P real Normal $P(\|Z\| \geq k)$ | La cota se cumple? |
|-----|-------------------------------|--------------------------------|-------------------|
| 1   | $\frac{1}{1} = 1.0000$       | $2(1 - \Phi(1)) = 0.3173$     | Si (0.3173 <= 1)  |
| 2   | $\frac{1}{4} = 0.2500$       | $2(1 - \Phi(2)) = 0.0455$     | Si (0.0455 <= 0.25)|
| 3   | $\frac{1}{9} = 0.1111$       | $2(1 - \Phi(3)) = 0.0027$     | Si (0.0027 <= 0.11)|

- $\Phi(k)$: funcion de distribucion acumulada (CDF) de la normal estandar evaluada en $k$.
- $2(1 - \Phi(k))$: probabilidad en ambas colas (simetria de la normal).
- La cota de Chebyshev es siempre mayor que la probabilidad real, confirmando que es una cota **conservadora**.
- La normal concentra mucha mas probabilidad cerca de la media que lo que Chebyshev garantiza.

```python
k_vals = [1, 2, 3]

print(f"{'k':>3} | {'Chebyshev 1/k^2':>16} | {'P real Normal':>16} | {'Cumple?':>10}")
print("-" * 55)

chebyshev_cotas = []
p_reales = []

for k in k_vals:
    cota = 1 / k**2
    p_real = 2 * (1 - stats.norm.cdf(k))
    cumple = p_real <= cota
    chebyshev_cotas.append(cota)
    p_reales.append(p_real)
    print(f"{k:>3} | {cota:>16.4f} | {p_real:>16.4f} | {'Si' if cumple else 'No':>10}")

print(f"\nLa cota de Chebyshev siempre se cumple.")
print(f"Para la Normal, la probabilidad real es mucho menor que la cota.")

# Grafica: comparacion Chebyshev vs Normal real
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# Barras comparativas
x_pos = np.arange(len(k_vals))
width = 0.35
ax1.bar(x_pos - width/2, chebyshev_cotas, width, color=C_DANGER, edgecolor=C_DARK,
        alpha=0.85, label='Cota Chebyshev (1/k^2)')
ax1.bar(x_pos + width/2, p_reales, width, color=C_PRIMARY, edgecolor=C_DARK,
        alpha=0.85, label='P real Normal')

for i, (c, r) in enumerate(zip(chebyshev_cotas, p_reales)):
    ax1.text(i - width/2, c + 0.02, f'{c:.4f}', ha='center', fontsize=9, color=C_DARK)
    ax1.text(i + width/2, r + 0.02, f'{r:.4f}', ha='center', fontsize=9, color=C_DARK)

ax1.set_xlabel('k (desviaciones estandar)')
ax1.set_ylabel('Probabilidad')
ax1.set_title('Chebyshev vs Normal: P(|X-mu| >= k*sigma)', color=C_DARK, fontweight='bold')
ax1.set_xticks(x_pos)
ax1.set_xticklabels([f'k={k}' for k in k_vals])
ax1.legend()

# PDF normal con regiones sombreadas
z_plot = np.linspace(-4, 4, 400)
y_plot = stats.norm.pdf(z_plot)
ax2.plot(z_plot, y_plot, color=C_DARK, linewidth=2)
ax2.fill_between(z_plot, y_plot, alpha=0.15, color=C_PRIMARY)

colores_k = [C_SUCCESS, C_ORANGE, C_DANGER]
for k, color in zip(k_vals, colores_k):
    ax2.axvline(x=k, color=color, linewidth=1.5, linestyle='--', alpha=0.8)
    ax2.axvline(x=-k, color=color, linewidth=1.5, linestyle='--', alpha=0.8)
    mask = (z_plot >= k) | (z_plot <= -k)
    ax2.fill_between(z_plot[mask], y_plot[mask], alpha=0.2, color=color, label=f'k={k}')

ax2.set_xlabel('z')
ax2.set_ylabel('f(z)')
ax2.set_title('Normal estandar: regiones fuera de k*sigma', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 18:** Covarianza: Cov(X,Y)=E[XY]-E[X]E[Y]. Calcular con datos discretos conjuntos.

**Resolucion:**

Consideremos dos variables aleatorias discretas $X$ e $Y$ con la siguiente distribucion conjunta:

| | Y=0 | Y=1 | Y=2 |
|---|---|---|---|
| X=0 | 0.10 | 0.05 | 0.05 |
| X=1 | 0.05 | 0.20 | 0.10 |
| X=2 | 0.05 | 0.10 | 0.30 |

Donde:
- $X$: variable aleatoria discreta que toma valores en $\{0, 1, 2\}$.
- $Y$: variable aleatoria discreta que toma valores en $\{0, 1, 2\}$.
- Cada entrada $P(X = x_i, Y = y_j)$: probabilidad conjunta de que $X = x_i$ y $Y = y_j$ simultaneamente.
- La suma de todas las entradas debe ser 1.

**Distribuciones marginales:**

$$P(X = x_i) = \sum_{j} P(X = x_i, Y = y_j)$$

- $P(X = x_i)$: probabilidad marginal de $X$. Se obtiene sumando la fila correspondiente.

$P(X=0) = 0.10 + 0.05 + 0.05 = 0.20$, $P(X=1) = 0.35$, $P(X=2) = 0.45$

$$P(Y = y_j) = \sum_{i} P(X = x_i, Y = y_j)$$

- $P(Y = y_j)$: probabilidad marginal de $Y$. Se obtiene sumando la columna correspondiente.

$P(Y=0) = 0.10 + 0.05 + 0.05 = 0.20$, $P(Y=1) = 0.35$, $P(Y=2) = 0.45$

**Esperanzas:**

$$E[X] = \sum_{i} x_i \cdot P(X = x_i) = 0 \times 0.20 + 1 \times 0.35 + 2 \times 0.45 = 1.25$$

- $E[X]$: esperanza de $X$, calculada con la distribucion marginal de $X$.

$$E[Y] = \sum_{j} y_j \cdot P(Y = y_j) = 0 \times 0.20 + 1 \times 0.35 + 2 \times 0.45 = 1.25$$

- $E[Y]$: esperanza de $Y$, calculada con la distribucion marginal de $Y$.

**E[XY]:**

$$E[XY] = \sum_{i} \sum_{j} x_i \cdot y_j \cdot P(X = x_i, Y = y_j)$$

- $E[XY]$: esperanza del producto $XY$. Se calcula usando la distribucion conjunta.
- $x_i \cdot y_j$: producto de los valores de $X$ e $Y$.
- $\sum_i \sum_j$: doble sumatoria sobre todos los pares $(x_i, y_j)$.

$$E[XY] = 0 \cdot 0 \cdot 0.10 + 0 \cdot 1 \cdot 0.05 + \ldots + 2 \cdot 2 \cdot 0.30$$

**Covarianza:**

$$Cov(X, Y) = E[XY] - E[X] \cdot E[Y]$$

- $Cov(X, Y)$: covarianza entre $X$ e $Y$. Mide la relacion lineal entre ambas variables.
- $E[XY]$: esperanza del producto.
- $E[X] \cdot E[Y]$: producto de las esperanzas individuales.
- Si $Cov(X, Y) > 0$: $X$ e $Y$ tienden a crecer juntas.
- Si $Cov(X, Y) < 0$: cuando una crece, la otra tiende a decrecer.
- Si $Cov(X, Y) = 0$: no hay relacion lineal (pero puede haber relacion no lineal).

**Correlacion:**

$$\rho(X, Y) = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$

- $\rho(X, Y)$: coeficiente de correlacion de Pearson. Siempre esta entre $-1$ y $1$.
- $\sigma_X = \sqrt{Var(X)}$: desviacion estandar de $X$.
- $\sigma_Y = \sqrt{Var(Y)}$: desviacion estandar de $Y$.
- Normaliza la covarianza para que sea comparable entre distintas escalas.

```python
# Distribucion conjunta P(X=xi, Y=yj)
# Filas: X=0,1,2; Columnas: Y=0,1,2
P_conjunta = np.array([
    [0.10, 0.05, 0.05],
    [0.05, 0.20, 0.10],
    [0.05, 0.10, 0.30]
])

x_vals = np.array([0, 1, 2])
y_vals = np.array([0, 1, 2])

print("Distribucion conjunta:")
print(P_conjunta)
print(f"Suma total = {P_conjunta.sum():.2f} (debe ser 1)")

# Marginales
P_X = P_conjunta.sum(axis=1)  # sumar filas
P_Y = P_conjunta.sum(axis=0)  # sumar columnas
print(f"\nMarginales:")
print(f"P(X) = {P_X}")
print(f"P(Y) = {P_Y}")

# Esperanzas
E_X = np.sum(x_vals * P_X)
E_Y = np.sum(y_vals * P_Y)
print(f"\nE[X] = {E_X}")
print(f"E[Y] = {E_Y}")

# E[XY]
E_XY = 0
print(f"\nCalculo de E[XY]:")
for i, xi in enumerate(x_vals):
    for j, yj in enumerate(y_vals):
        contrib = xi * yj * P_conjunta[i, j]
        if contrib != 0:
            print(f"  {xi}*{yj}*{P_conjunta[i,j]} = {contrib:.4f}")
        E_XY += contrib
print(f"E[XY] = {E_XY:.4f}")

# Covarianza
Cov_XY = E_XY - E_X * E_Y
print(f"\nCov(X,Y) = E[XY] - E[X]*E[Y] = {E_XY:.4f} - {E_X}*{E_Y} = {E_XY:.4f} - {E_X*E_Y:.4f} = {Cov_XY:.4f}")

# Varianzas y correlacion
E_X2 = np.sum(x_vals**2 * P_X)
E_Y2 = np.sum(y_vals**2 * P_Y)
Var_X = E_X2 - E_X**2
Var_Y = E_Y2 - E_Y**2
rho = Cov_XY / (np.sqrt(Var_X) * np.sqrt(Var_Y))

print(f"\nVar(X) = {Var_X:.4f}")
print(f"Var(Y) = {Var_Y:.4f}")
print(f"Correlacion rho = Cov(X,Y) / (sigma_X * sigma_Y) = {Cov_XY:.4f} / ({np.sqrt(Var_X):.4f} * {np.sqrt(Var_Y):.4f}) = {rho:.4f}")
print(f"Cov > 0: X e Y tienden a crecer juntas")

# Grafica: distribucion conjunta como mapa de calor y scatter de muestras
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Mapa de calor
im = ax1.imshow(P_conjunta, cmap='Blues', aspect='auto', origin='lower')
for i in range(3):
    for j in range(3):
        ax1.text(j, i, f'{P_conjunta[i,j]:.2f}', ha='center', va='center',
                fontsize=14, fontweight='bold', color='black' if P_conjunta[i,j] < 0.2 else 'white')

ax1.set_xticks([0, 1, 2])
ax1.set_yticks([0, 1, 2])
ax1.set_xlabel('Y')
ax1.set_ylabel('X')
ax1.set_title('Distribucion conjunta P(X, Y)', color=C_DARK, fontweight='bold')
plt.colorbar(im, ax=ax1, label='Probabilidad')

# Simulacion y scatter
np.random.seed(42)
n_sim = 5000
pares = []
probs = P_conjunta.flatten()
indices = [(i, j) for i in range(3) for j in range(3)]
elegidos = np.random.choice(len(indices), size=n_sim, p=probs)
X_sim = np.array([x_vals[indices[e][0]] for e in elegidos])
Y_sim = np.array([y_vals[indices[e][1]] for e in elegidos])

# Agregar jitter para visualizacion
jitter = 0.1
ax2.scatter(X_sim + np.random.uniform(-jitter, jitter, n_sim),
           Y_sim + np.random.uniform(-jitter, jitter, n_sim),
           alpha=0.15, s=10, color=C_PRIMARY)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title(f'Muestras simuladas (rho = {rho:.4f})', color=C_DARK, fontweight='bold')
ax2.set_xticks([0, 1, 2])
ax2.set_yticks([0, 1, 2])

plt.tight_layout()
plt.show()
```

**Ejercicio 19:** X ~ Poisson(lambda=3). Calcular los primeros 4 momentos: E[X], E[X^2], E[X^3], E[X^4].

**Resolucion:**

Para $X \sim \text{Poisson}(\lambda)$ con $\lambda = 3$, la PMF es:

$$P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}, \quad k = 0, 1, 2, \ldots$$

Donde:
- $X$: variable aleatoria con distribucion de Poisson. Modela el numero de eventos en un intervalo fijo.
- $k$: numero de eventos observados (entero no negativo).
- $\lambda$: parametro de la Poisson. Aqui $\lambda = 3$. Representa la tasa promedio de eventos.
- $e^{-\lambda}$: factor exponencial de normalizacion. $e \approx 2.71828$.
- $\lambda^k$: $\lambda$ elevado a la potencia $k$.
- $k!$: factorial de $k$, es decir $k! = k \times (k-1) \times \ldots \times 1$.

Los **momentos** $E[X^n]$ se pueden obtener usando la funcion generadora de momentos (MGF) de la Poisson:

$$M_X(t) = E[e^{tX}] = \exp\left(\lambda(e^t - 1)\right)$$

- $M_X(t)$: MGF de la Poisson evaluada en $t$.
- $\exp(\cdot)$: funcion exponencial.
- $\lambda(e^t - 1)$: argumento de la exponencial.
- $t$: parametro real de la MGF.

Los momentos se obtienen derivando la MGF y evaluando en $t = 0$:

$$E[X^n] = M_X^{(n)}(0)$$

- $M_X^{(n)}(0)$: $n$-esima derivada de la MGF evaluada en $t = 0$.

**Primer momento E[X]:**

$$E[X] = \lambda = 3$$

- $E[X]$: esperanza de $X$. Para la Poisson, coincide con $\lambda$.

**Segundo momento E[X^2]:**

$$E[X^2] = \lambda^2 + \lambda = 9 + 3 = 12$$

- $E[X^2]$: segundo momento. Se obtiene de $Var(X) + (E[X])^2 = \lambda + \lambda^2$.
- $Var(X) = \lambda = 3$: la varianza de la Poisson es igual a $\lambda$.

**Tercer momento E[X^3]:**

$$E[X^3] = \lambda^3 + 3\lambda^2 + \lambda = 27 + 27 + 3 = 57$$

- $E[X^3]$: tercer momento. Se puede derivar de la MGF o usando la relacion con momentos centrales.
- $\lambda^3 = 27$, $3\lambda^2 = 27$, $\lambda = 3$.

**Cuarto momento E[X^4]:**

$$E[X^4] = \lambda^4 + 6\lambda^3 + 7\lambda^2 + \lambda = 81 + 162 + 63 + 3 = 309$$

- $E[X^4]$: cuarto momento.
- $\lambda^4 = 81$, $6\lambda^3 = 162$, $7\lambda^2 = 63$, $\lambda = 3$.

```python
lam = 3

# Formulas teoricas para momentos de Poisson
E_X1_teorico = lam                                    # lambda
E_X2_teorico = lam**2 + lam                           # lambda^2 + lambda
E_X3_teorico = lam**3 + 3*lam**2 + lam               # lambda^3 + 3*lambda^2 + lambda
E_X4_teorico = lam**4 + 6*lam**3 + 7*lam**2 + lam    # lambda^4 + 6*lambda^3 + 7*lambda^2 + lambda

print(f"lambda = {lam}")
print(f"\n=== Momentos teoricos (formulas) ===")
print(f"E[X]   = lambda = {E_X1_teorico}")
print(f"E[X^2] = lambda^2 + lambda = {lam**2} + {lam} = {E_X2_teorico}")
print(f"E[X^3] = lambda^3 + 3*lambda^2 + lambda = {lam**3} + {3*lam**2} + {lam} = {E_X3_teorico}")
print(f"E[X^4] = lambda^4 + 6*lambda^3 + 7*lambda^2 + lambda = {lam**4} + {6*lam**3} + {7*lam**2} + {lam} = {E_X4_teorico}")

# Verificacion numerica directa (sumando la PMF)
K = 50  # truncar la suma en k=50 (suficiente para lambda=3)
k_vals = np.arange(0, K + 1)
pmf_vals = stats.poisson.pmf(k_vals, lam)

E_X1_num = np.sum(k_vals * pmf_vals)
E_X2_num = np.sum(k_vals**2 * pmf_vals)
E_X3_num = np.sum(k_vals**3 * pmf_vals)
E_X4_num = np.sum(k_vals**4 * pmf_vals)

print(f"\n=== Verificacion numerica (suma directa hasta k={K}) ===")
print(f"E[X]   = {E_X1_num:.4f} (teorico: {E_X1_teorico})")
print(f"E[X^2] = {E_X2_num:.4f} (teorico: {E_X2_teorico})")
print(f"E[X^3] = {E_X3_num:.4f} (teorico: {E_X3_teorico})")
print(f"E[X^4] = {E_X4_num:.4f} (teorico: {E_X4_teorico})")

# Momentos centrales
Var_X = E_X2_teorico - E_X1_teorico**2
skew = (E_X3_num - 3*E_X1_num*Var_X - E_X1_num**3) / Var_X**1.5
kurt = (E_X4_num - 4*E_X3_num*E_X1_num + 6*E_X2_num*E_X1_num**2 - 3*E_X1_num**4) / Var_X**2

print(f"\nVar(X) = E[X^2] - (E[X])^2 = {E_X2_teorico} - {E_X1_teorico**2} = {Var_X}")
print(f"Skewness (asimetria) = {skew:.4f}")
print(f"Kurtosis (exceso) = {kurt:.4f}")

# Grafica: PMF con contribuciones de cada momento
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# PMF
k_plot = np.arange(0, 15)
pmf_plot = stats.poisson.pmf(k_plot, lam)
ax1.bar(k_plot, pmf_plot, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=k)')
ax1.axvline(x=E_X1_teorico, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X1_teorico}')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title(f'Poisson(lambda={lam})', color=C_DARK, fontweight='bold')
ax1.legend()

# Momentos como barras
momentos = [E_X1_teorico, E_X2_teorico, E_X3_teorico, E_X4_teorico]
etiquetas = ['E[X]', 'E[X^2]', 'E[X^3]', 'E[X^4]']
colores = [C_PRIMARY, C_SUCCESS, C_ORANGE, C_DANGER]
bars = ax2.bar(etiquetas, momentos, color=colores, edgecolor=C_DARK, alpha=0.85)
for bar, val in zip(bars, momentos):
    ax2.text(bar.get_x() + bar.get_width()/2, val + 5, f'{val}',
            ha='center', fontweight='bold', fontsize=12, color=C_DARK)
ax2.set_ylabel('Valor del momento')
ax2.set_title(f'Primeros 4 momentos de Poisson(lambda={lam})', color=C_DARK, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Ejercicio 20:** Ley de los grandes numeros: Simular N lanzamientos de moneda y mostrar como la media converge a 0.5.

**Resolucion:**

La **Ley de los Grandes Numeros (LGN)** establece que la media muestral converge a la esperanza poblacional a medida que el tamano de muestra crece:

$$\bar{X}_N = \frac{1}{N} \sum_{i=1}^{N} X_i \xrightarrow{N \to \infty} E[X] = \mu$$

Donde:
- $\bar{X}_N$: media muestral de las primeras $N$ observaciones. Es una variable aleatoria que depende de la muestra.
- $N$: numero total de observaciones (tamano de muestra).
- $X_i$: la $i$-esima observacion. Cada $X_i$ es una variable aleatoria independiente con la misma distribucion (iid).
- $\sum_{i=1}^{N} X_i$: suma de las $N$ observaciones.
- $\frac{1}{N}$: factor de promediado.
- $E[X] = \mu$: esperanza de la distribucion comun de los $X_i$. Es el valor al que converge la media muestral.
- $\xrightarrow{N \to \infty}$: convergencia cuando $N$ tiende a infinito (en probabilidad para la LGN debil, casi seguramente para la LGN fuerte).

**Aplicacion: lanzamiento de moneda justa**

Cada lanzamiento $X_i \sim \text{Bernoulli}(p = 0.5)$:

- $X_i = 1$ si sale cara, $X_i = 0$ si sale cruz.
- $P(X_i = 1) = p = 0.5$: probabilidad de cara.
- $P(X_i = 0) = 1 - p = 0.5$: probabilidad de cruz.
- $E[X_i] = p = 0.5$: esperanza de cada lanzamiento.
- $Var(X_i) = p(1-p) = 0.25$: varianza de cada lanzamiento.

La media muestral $\bar{X}_N$ representa la proporcion de caras en $N$ lanzamientos:

$$\bar{X}_N = \frac{\text{numero de caras}}{N}$$

**Propiedades de la media muestral:**

$$E[\bar{X}_N] = \mu = 0.5$$

- $E[\bar{X}_N]$: esperanza de la media muestral. Es igual a $\mu$ para cualquier $N$ (la media muestral es un estimador insesgado).

$$Var(\bar{X}_N) = \frac{Var(X)}{N} = \frac{p(1-p)}{N} = \frac{0.25}{N}$$

- $Var(\bar{X}_N)$: varianza de la media muestral. Decrece con $N$.
- $\frac{Var(X)}{N}$: la varianza se divide entre $N$ porque promediamos $N$ variables independientes.
- A medida que $N$ crece, $Var(\bar{X}_N) \to 0$, lo que significa que $\bar{X}_N$ se concentra cada vez mas alrededor de $\mu$.

```python
np.random.seed(42)

p = 0.5
N_max = 10_000

# Simular N_max lanzamientos
lanzamientos = np.random.binomial(1, p, size=N_max)

# Media acumulada: media muestral despues de cada lanzamiento
n_vals = np.arange(1, N_max + 1)
media_acumulada = np.cumsum(lanzamientos) / n_vals

# Banda teorica de +/- 2*std de la media muestral
std_media = np.sqrt(p * (1 - p) / n_vals)
banda_sup = p + 2 * std_media
banda_inf = p - 2 * std_media

print(f"Moneda justa: p = {p}")
print(f"E[X] = p = {p}")
print(f"Var(X) = p(1-p) = {p*(1-p)}")
print(f"\nMedia muestral en distintos N:")
for n in [10, 100, 1000, 5000, 10000]:
    media_n = media_acumulada[n - 1]
    var_media_n = p * (1 - p) / n
    print(f"  N={n:>5}: media = {media_n:.4f}, Var(media) = {var_media_n:.6f}, +/- 2*std = {2*np.sqrt(var_media_n):.4f}")

# Grafica: convergencia de la media muestral
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Panel 1: media acumulada con bandas
ax1.plot(n_vals, media_acumulada, color=C_PRIMARY, linewidth=0.8, alpha=0.8, label='Media muestral')
ax1.axhline(y=p, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {p}')
ax1.fill_between(n_vals, banda_inf, banda_sup, alpha=0.15, color=C_ORANGE,
                 label='+/- 2*std de la media')
ax1.set_xlabel('N (numero de lanzamientos)')
ax1.set_ylabel('Media muestral')
ax1.set_title('Ley de los Grandes Numeros: convergencia de la media muestral', color=C_DARK, fontweight='bold')
ax1.set_xscale('log')
ax1.set_xlim(1, N_max)
ax1.legend(loc='upper right')

# Panel 2: error absoluto |media - 0.5|
error = np.abs(media_acumulada - p)
ax2.plot(n_vals, error, color=C_DANGER, linewidth=0.8, alpha=0.7, label='|media - 0.5|')
ax2.plot(n_vals, 2 * std_media, color=C_ORANGE, linewidth=2, linestyle='--',
         label='2*sqrt(0.25/N) (cota teorica)')
ax2.set_xlabel('N (numero de lanzamientos)')
ax2.set_ylabel('Error absoluto')
ax2.set_title('Error |media muestral - E[X]| decrece con N', color=C_DARK, fontweight='bold')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlim(1, N_max)
ax2.legend()

plt.tight_layout()
plt.show()
```
