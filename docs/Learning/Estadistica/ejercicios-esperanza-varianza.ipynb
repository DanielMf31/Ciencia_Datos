## Formulas fundamentales de Esperanza y Varianza

---

### Esperanza (valor esperado)

**Caso discreto:** La esperanza de una variable aleatoria discreta $X$ es la suma ponderada de todos sus posibles valores por sus respectivas probabilidades.

$$E[X] = \sum_{i} x_i \cdot P(X = x_i)$$

- $E[X]$: esperanza (valor esperado) de la variable aleatoria $X$, tambien denotada $\mu$ (mu). Representa el "centro de masa" de la distribucion.
- $x_i$: cada uno de los posibles valores que puede tomar la variable aleatoria $X$.
- $P(X = x_i)$: probabilidad de que $X$ tome el valor $x_i$. Cada probabilidad esta entre 0 y 1, y la suma de todas es 1.
- $\sum_{i}$: sumatoria sobre todos los posibles valores $i$ del soporte de $X$.

**Caso continuo:** La esperanza de una variable aleatoria continua $X$ con funcion de densidad $f(x)$ es la integral ponderada sobre todo el soporte.

$$E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx$$

- $E[X]$: esperanza de la variable aleatoria continua $X$.
- $x$: variable de integracion que recorre todos los valores reales del soporte de $X$.
- $f(x)$: funcion de densidad de probabilidad (PDF). Cumple $f(x) \geq 0$ y $\int_{-\infty}^{\infty} f(x) \, dx = 1$.
- $dx$: diferencial de integracion, indica que se integra respecto a $x$.
- $\int_{-\infty}^{\infty}$: integral sobre todo el eje real (en la practica, solo el soporte de $X$ contribuye).

---

### Varianza

**Definicion:** La varianza mide la dispersion de $X$ respecto a su media $\mu = E[X]$.

$$Var(X) = E\left[(X - \mu)^2\right] = \sum_{i} (x_i - \mu)^2 \cdot P(X = x_i)$$

- $Var(X)$: varianza de $X$, tambien denotada $\sigma^2$. Siempre es no negativa.
- $X$: variable aleatoria cuya dispersion se mide.
- $\mu$: media de $X$, es decir $\mu = E[X]$.
- $(X - \mu)^2$: desviacion cuadratica de $X$ respecto a su media. Se eleva al cuadrado para evitar cancelaciones entre desviaciones positivas y negativas.
- $x_i$: cada posible valor de $X$.
- $P(X = x_i)$: probabilidad asociada al valor $x_i$.

**Formula alternativa (computacionalmente mas eficiente):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $E[X^2]$: esperanza del cuadrado de $X$, es decir $\sum_i x_i^2 \cdot P(X = x_i)$ (discreto) o $\int x^2 f(x) dx$ (continuo). Este es el **segundo momento** de $X$.
- $(E[X])^2$: el cuadrado de la esperanza de $X$. No es lo mismo que $E[X^2]$.
- La diferencia $E[X^2] - (E[X])^2$ siempre es $\geq 0$.

**Desviacion estandar:**

$$\sigma = \sqrt{Var(X)}$$

- $\sigma$: desviacion estandar de $X$. Tiene las mismas unidades que $X$, a diferencia de la varianza que tiene unidades al cuadrado.

---

### Propiedades de la Esperanza

$$E[aX + b] = a \cdot E[X] + b$$

- $a$: constante multiplicativa (escalar). Escala la esperanza proporcionalmente.
- $b$: constante aditiva. Se suma directamente a la esperanza.
- $X$: variable aleatoria.
- Esta propiedad se llama **linealidad de la esperanza** y siempre vale, sin importar la distribucion.

$$E[X + Y] = E[X] + E[Y]$$

- $X, Y$: dos variables aleatorias cualesquiera (no necesitan ser independientes).
- La esperanza de la suma siempre es la suma de las esperanzas.

$$E[g(X)] = \sum_{i} g(x_i) \cdot P(X = x_i) \quad \text{(LOTUS, caso discreto)}$$

- $g(X)$: funcion aplicada a la variable aleatoria $X$ (por ejemplo, $g(X) = X^2$).
- LOTUS (Law of the Unconscious Statistician): permite calcular $E[g(X)]$ sin conocer la distribucion de $g(X)$.

---

### Propiedades de la Varianza

$$Var(aX + b) = a^2 \cdot Var(X)$$

- $a^2$: la constante multiplicativa se eleva al cuadrado porque la varianza es una medida cuadratica.
- $b$: la constante aditiva desaparece porque trasladar la distribucion no cambia su dispersion.

$$Var(X + Y) = Var(X) + Var(Y) + 2 \cdot Cov(X, Y)$$

- $Cov(X,Y)$: covarianza entre $X$ e $Y$. Mide la relacion lineal entre ambas variables.
- Si $X$ e $Y$ son **independientes**, entonces $Cov(X,Y) = 0$ y la formula se simplifica a $Var(X) + Var(Y)$.

$$Cov(X, Y) = E[XY] - E[X] \cdot E[Y]$$

- $E[XY]$: esperanza del producto de $X$ e $Y$.
- $E[X] \cdot E[Y]$: producto de las esperanzas individuales.
- Si $X$ e $Y$ son independientes, $E[XY] = E[X] \cdot E[Y]$ y la covarianza es cero.

---

### Funcion Generadora de Momentos (MGF)

$$M_X(t) = E[e^{tX}]$$

- $M_X(t)$: funcion generadora de momentos de la variable aleatoria $X$, evaluada en $t$.
- $t$: parametro real. La MGF debe existir en un entorno abierto de $t = 0$.
- $e^{tX}$: funcion exponencial aplicada al producto $tX$.
- La MGF, cuando existe, caracteriza unicamente la distribucion de $X$.

**Obtencion de momentos:**

$$E[X^n] = M_X^{(n)}(0) = \left. \frac{d^n}{dt^n} M_X(t) \right|_{t=0}$$

- $E[X^n]$: momento de orden $n$ de $X$.
- $M_X^{(n)}(0)$: la $n$-esima derivada de la MGF evaluada en $t = 0$.
- $n = 1$: da $E[X]$ (primer momento, la media).
- $n = 2$: da $E[X^2]$ (segundo momento, usado para calcular la varianza).

**Ejercicio 1:** Un dado justo de 6 caras. Calcular la esperanza E[X] y la varianza Var(X).

**Resolucion:**

**Esperanza:** Es el valor promedio ponderado por las probabilidades.

$$E[X] = \sum_{i=1}^{6} x_i \cdot P(x_i) = \frac{1}{6}(1 + 2 + 3 + 4 + 5 + 6) = \frac{21}{6} = 3.5$$

**Varianza:** Mide la dispersion respecto a la media.

$$Var(X) = E[X^2] - (E[X])^2$$

Primero calculamos $E[X^2]$:

$$E[X^2] = \frac{1}{6}(1^2 + 2^2 + 3^2 + 4^2 + 5^2 + 6^2) = \frac{1 + 4 + 9 + 16 + 25 + 36}{6} = \frac{91}{6} = 15.1\overline{6}$$

$$Var(X) = 15.1\overline{6} - (3.5)^2 = 15.1\overline{6} - 12.25 = 2.91\overline{6}$$

```python
x = np.array([1, 2, 3, 4, 5, 6])
p = np.ones(6) / 6

E_X = np.sum(x * p)
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2

print(f"E[X] = {E_X:.4f}")
print(f"E[X^2] = {E_X2:.4f}")
print(f"Var(X) = E[X^2] - (E[X])^2 = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")
print(f"Desviacion estandar = {np.sqrt(Var_X):.4f}")

# Grafica: PMF del dado con E[X] marcada
fig, ax = plt.subplots()
ax.bar(x, p, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=x)')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.axvspan(E_X - np.sqrt(Var_X), E_X + np.sqrt(Var_X), alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {np.sqrt(Var_X):.2f}')
ax.set_xlabel('Valor del dado')
ax.set_ylabel('Probabilidad')
ax.set_title('PMF del dado justo con E[X] y desviacion estandar', color=C_DARK, fontweight='bold')
ax.set_xticks(x)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 2:** Sea X una variable aleatoria con distribucion Bernoulli(p=0.3). Calcular E[X] y Var(X).

**Resolucion:**

La Bernoulli toma valores 0 (fracaso) y 1 (exito) con probabilidad $p$.

**Esperanza:**

$$E[X] = 0 \cdot (1-p) + 1 \cdot p = p = 0.3$$

**Varianza:**

$$Var(X) = E[X^2] - (E[X])^2$$

Como $X$ solo toma 0 y 1: $X^2 = X$, entonces $E[X^2] = E[X] = p$.

$$Var(X) = p - p^2 = p(1-p) = 0.3 \times 0.7 = 0.21$$

```python
p = 0.3
E_X = p
Var_X = p * (1 - p)

print(f"E[X] = p = {E_X}")
print(f"Var(X) = p(1-p) = {p} * {1-p} = {Var_X}")

# Grafica: PMF Bernoulli y como varia Var(X) con p
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# PMF
ax1.bar([0, 1], [1-p, p], color=[C_PRIMARY, C_DANGER], edgecolor=C_DARK, alpha=0.85, width=0.4)
ax1.set_xlabel('X')
ax1.set_ylabel('P(X)')
ax1.set_title('PMF Bernoulli(p=0.3)', color=C_DARK, fontweight='bold')
ax1.set_xticks([0, 1])
ax1.set_xticklabels(['0 (fracaso)', '1 (exito)'])
for i, v in enumerate([1-p, p]):
    ax1.text(i, v + 0.02, f'{v:.1f}', ha='center', fontweight='bold', color=C_DARK)

# Varianza en funcion de p
ps = np.linspace(0, 1, 100)
vars_p = ps * (1 - ps)
ax2.plot(ps, vars_p, color=C_PRIMARY, linewidth=2)
ax2.axvline(x=0.3, color=C_DANGER, linestyle='--', label=f'p=0.3, Var={Var_X}')
ax2.scatter([0.3], [Var_X], color=C_DANGER, s=80, zorder=5)
ax2.set_xlabel('p')
ax2.set_ylabel('Var(X) = p(1-p)')
ax2.set_title('Varianza de Bernoulli segun p', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 3:** Sea X ~ Binomial(n=10, p=0.4). Calcular E[X], Var(X) y P(X=4).

**Resolucion:**

Para $X \sim \text{Binomial}(n, p)$:

**Esperanza:**

$$E[X] = n \cdot p = 10 \times 0.4 = 4$$

**Varianza:**

$$Var(X) = n \cdot p \cdot (1-p) = 10 \times 0.4 \times 0.6 = 2.4$$

**P(X=4):** Usando la formula de la PMF binomial:

$$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$$

$$P(X=4) = \binom{10}{4} (0.4)^4 (0.6)^6 = 210 \times 0.0256 \times 0.046656$$

$$= 210 \times 0.001194 = 0.2508$$

```python
n, p = 10, 0.4
E_X = n * p
Var_X = n * p * (1 - p)
P_X4 = stats.binom.pmf(4, n, p)

print(f"E[X] = n*p = {n}*{p} = {E_X}")
print(f"Var(X) = n*p*(1-p) = {n}*{p}*{1-p} = {Var_X}")
print(f"P(X=4) = C(10,4) * 0.4^4 * 0.6^6 = {comb(10,4)} * {0.4**4:.4f} * {0.6**6:.6f} = {P_X4:.4f}")

# Grafica: PMF binomial con E[X] y area de P(X=4)
x_vals = np.arange(0, n+1)
pmf_vals = stats.binom.pmf(x_vals, n, p)

fig, ax = plt.subplots()
colores = [C_DANGER if x == 4 else C_PRIMARY for x in x_vals]
ax.bar(x_vals, pmf_vals, color=colores, edgecolor=C_DARK, alpha=0.85)
ax.axvline(x=E_X, color=C_ORANGE, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
for x, v in zip(x_vals, pmf_vals):
    ax.text(x, v + 0.005, f'{v:.3f}', ha='center', fontsize=8, color=C_DARK)
ax.set_xlabel('k')
ax.set_ylabel('P(X = k)')
ax.set_title('PMF Binomial(n=10, p=0.4)', color=C_DARK, fontweight='bold')
ax.set_xticks(x_vals)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 4:** Sea X una variable aleatoria continua con funcion de densidad f(x) = 2x para x en [0, 1] y f(x) = 0 en otro caso. Calcular E[X] y Var(X) mediante integrales.

**Resolucion:**

**Verificacion de que es densidad valida:**

$$\int_0^1 2x \, dx = \left[ x^2 \right]_0^1 = 1$$

**Esperanza:**

$$E[X] = \int_0^1 x \cdot f(x) \, dx = \int_0^1 x \cdot 2x \, dx = \int_0^1 2x^2 \, dx = \left[ \frac{2x^3}{3} \right]_0^1 = \frac{2}{3} \approx 0.6667$$

**E[X^2]:**

$$E[X^2] = \int_0^1 x^2 \cdot 2x \, dx = \int_0^1 2x^3 \, dx = \left[ \frac{2x^4}{4} \right]_0^1 = \frac{1}{2} = 0.5$$

**Varianza:**

$$Var(X) = E[X^2] - (E[X])^2 = \frac{1}{2} - \left(\frac{2}{3}\right)^2 = \frac{1}{2} - \frac{4}{9} = \frac{9 - 8}{18} = \frac{1}{18} \approx 0.0556$$

```python
f = lambda x: 2 * x

# Verificacion de densidad
integral_total, _ = quad(f, 0, 1)
print(f"Integral de f(x) en [0,1] = {integral_total} (debe ser 1)")

# E[X]
E_X, _ = quad(lambda x: x * f(x), 0, 1)
print(f"E[X] = integral(x * 2x, 0, 1) = {E_X:.4f} = 2/3")

# E[X^2]
E_X2, _ = quad(lambda x: x**2 * f(x), 0, 1)
print(f"E[X^2] = integral(x^2 * 2x, 0, 1) = {E_X2:.4f} = 1/2")

# Var(X)
Var_X = E_X2 - E_X**2
print(f"Var(X) = {E_X2:.4f} - ({E_X:.4f})^2 = {Var_X:.4f} = 1/18")

# Grafica: PDF con E[X] y area sombreada +/- 1 std
x_plot = np.linspace(0, 1, 200)
y_plot = 2 * x_plot
std = np.sqrt(Var_X)

fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, color=C_PRIMARY, linewidth=2.5, label='f(x) = 2x')
ax.fill_between(x_plot, y_plot, alpha=0.2, color=C_PRIMARY)

# Sombrear region E[X] +/- std
mask = (x_plot >= E_X - std) & (x_plot <= E_X + std)
ax.fill_between(x_plot[mask], y_plot[mask], alpha=0.4, color=C_ORANGE, label=f'E[X] +/- std')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X:.4f}')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF f(x) = 2x en [0, 1]', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 5:** Sea X ~ Poisson(lambda=5). Calcular E[X], Var(X) y P(X >= 3).

**Resolucion:**

Para $X \sim \text{Poisson}(\lambda)$, la esperanza y la varianza son iguales:

$$E[X] = \lambda = 5$$

$$Var(X) = \lambda = 5$$

**P(X >= 3):** Usamos el complemento.

$$P(X \geq 3) = 1 - P(X \leq 2) = 1 - \sum_{k=0}^{2} \frac{e^{-\lambda} \lambda^k}{k!}$$

$$P(X=0) = \frac{e^{-5} \cdot 5^0}{0!} = e^{-5} = 0.00674$$

$$P(X=1) = \frac{e^{-5} \cdot 5^1}{1!} = 5e^{-5} = 0.03369$$

$$P(X=2) = \frac{e^{-5} \cdot 5^2}{2!} = \frac{25e^{-5}}{2} = 0.08422$$

$$P(X \geq 3) = 1 - (0.00674 + 0.03369 + 0.08422) = 1 - 0.12465 = 0.8753$$

```python
lam = 5
E_X = lam
Var_X = lam

print(f"E[X] = lambda = {E_X}")
print(f"Var(X) = lambda = {Var_X}")

# P(X >= 3) por complemento
p0 = stats.poisson.pmf(0, lam)
p1 = stats.poisson.pmf(1, lam)
p2 = stats.poisson.pmf(2, lam)
p_ge3 = 1 - (p0 + p1 + p2)
print(f"\nP(X=0) = {p0:.5f}")
print(f"P(X=1) = {p1:.5f}")
print(f"P(X=2) = {p2:.5f}")
print(f"P(X<=2) = {p0+p1+p2:.5f}")
print(f"P(X>=3) = 1 - {p0+p1+p2:.5f} = {p_ge3:.4f}")

# Verificacion con scipy
print(f"Verificacion: 1 - CDF(2) = {1 - stats.poisson.cdf(2, lam):.4f}")

# Grafica: PMF con area sombreada para X>=3
x_vals = np.arange(0, 16)
pmf_vals = stats.poisson.pmf(x_vals, lam)

fig, ax = plt.subplots()
colores = [C_DANGER if x < 3 else C_SUCCESS for x in x_vals]
ax.bar(x_vals, pmf_vals, color=colores, edgecolor=C_DARK, alpha=0.85)
ax.axvline(x=E_X, color=C_ORANGE, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.set_xlabel('k')
ax.set_ylabel('P(X = k)')
ax.set_title(f'Poisson(lambda=5) - P(X>=3) = {p_ge3:.4f} (verde)', color=C_DARK, fontweight='bold')
ax.set_xticks(x_vals)
ax.legend()
plt.tight_layout()
plt.show()
```

$$E[aX + bY] = a \cdot E[X] + b \cdot E[Y]$$

$$E[3X + 2Y] = 3 \times 10 + 2 \times 20 = 30 + 40 = 70$$

**Propiedad de la varianza** (requiere independencia para que no haya termino de covarianza):

$$Var(aX + bY) = a^2 \cdot Var(X) + b^2 \cdot Var(Y)$$

- $a = 3$, $b = 2$
- La constante se eleva al cuadrado porque la varianza mide dispersion cuadratica

```python
E_X, E_Y = 10, 20
Var_X, Var_Y = 4, 9
a, b = 3, 2

E_combinacion = a * E_X + b * E_Y
Var_combinacion = a**2 * Var_X + b**2 * Var_Y

print(f"E[3X + 2Y] = 3*{E_X} + 2*{E_Y} = {a*E_X} + {b*E_Y} = {E_combinacion}")
print(f"Var(3X + 2Y) = 3^2*{Var_X} + 2^2*{Var_Y} = {a**2}*{Var_X} + {b**2}*{Var_Y} = {a**2*Var_X} + {b**2*Var_Y} = {Var_combinacion}")

# Verificacion por simulacion
np.random.seed(42)
n_sim = 100_000
X_sim = np.random.normal(E_X, np.sqrt(Var_X), n_sim)
Y_sim = np.random.normal(E_Y, np.sqrt(Var_Y), n_sim)
Z_sim = a * X_sim + b * Y_sim

print(f"\nVerificacion por simulacion ({n_sim:,} muestras):")
print(f"E[3X+2Y] simulada = {np.mean(Z_sim):.2f} (teorica: {E_combinacion})")
print(f"Var(3X+2Y) simulada = {np.var(Z_sim):.2f} (teorica: {Var_combinacion})")

# Grafica: histograma de la simulacion
fig, ax = plt.subplots()
ax.hist(Z_sim, bins=60, density=True, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.7, label='Simulacion 3X+2Y')
ax.axvline(x=E_combinacion, color=C_DANGER, linewidth=2, linestyle='--', label=f'E = {E_combinacion}')
ax.axvline(x=E_combinacion - np.sqrt(Var_combinacion), color=C_ORANGE, linewidth=1.5, linestyle=':', label=f'+/- std = {np.sqrt(Var_combinacion):.2f}')
ax.axvline(x=E_combinacion + np.sqrt(Var_combinacion), color=C_ORANGE, linewidth=1.5, linestyle=':')
ax.set_xlabel('3X + 2Y')
ax.set_ylabel('Densidad')
ax.set_title('Simulacion de 3X + 2Y (X,Y independientes)', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Varianza:**

$$Var(X) = \frac{1}{\lambda^2} = \frac{1}{4} = 0.25$$

**P(X > 1):** Calculamos la integral de supervivencia.

$$P(X > 1) = \int_1^{\infty} \lambda e^{-\lambda x} \, dx = \int_1^{\infty} 2e^{-2x} \, dx$$

**Ejercicio 8:** Demostrar que Var(X) = E[X^2] - (E[X])^2 usando un ejemplo discreto: un dado de 3 caras con valores {1, 2, 3} equiprobables. Calcular la varianza por ambos metodos y verificar que dan el mismo resultado.

**Resolucion:**

Dado de 3 caras: $X \in \{1, 2, 3\}$, cada uno con $P = 1/3$.

**Metodo 1 - Definicion directa:**

$$Var(X) = E[(X - \mu)^2] = \sum_{i} (x_i - \mu)^2 \cdot P(x_i)$$

Primero: $\mu = E[X] = \frac{1+2+3}{3} = 2$

$$Var(X) = \frac{1}{3}(1-2)^2 + \frac{1}{3}(2-2)^2 + \frac{1}{3}(3-2)^2$$

$$= \frac{1}{3}(1 + 0 + 1) = \frac{2}{3} \approx 0.6667$$

**Metodo 2 - Formula alternativa:**

$$Var(X) = E[X^2] - (E[X])^2$$

$$E[X^2] = \frac{1}{3}(1^2 + 2^2 + 3^2) = \frac{1 + 4 + 9}{3} = \frac{14}{3} \approx 4.6667$$

$$Var(X) = \frac{14}{3} - 2^2 = \frac{14}{3} - 4 = \frac{14 - 12}{3} = \frac{2}{3} \approx 0.6667$$

Ambos metodos dan el mismo resultado: $\frac{2}{3}$.

```python
x = np.array([1, 2, 3])
p = np.array([1/3, 1/3, 1/3])

# Metodo 1: definicion directa
E_X = np.sum(x * p)
var_metodo1 = np.sum((x - E_X)**2 * p)

# Metodo 2: formula alternativa
E_X2 = np.sum(x**2 * p)
var_metodo2 = E_X2 - E_X**2

print("=== Metodo 1: Var(X) = E[(X - mu)^2] ===")
print(f"E[X] = mu = {E_X}")
for xi, pi in zip(x, p):
    print(f"  ({xi} - {E_X})^2 * {pi:.4f} = {(xi-E_X)**2 * pi:.4f}")
print(f"Var(X) = {var_metodo1:.4f}")

print(f"\n=== Metodo 2: Var(X) = E[X^2] - (E[X])^2 ===")
print(f"E[X^2] = {E_X2:.4f}")
print(f"(E[X])^2 = {E_X**2:.4f}")
print(f"Var(X) = {E_X2:.4f} - {E_X**2:.4f} = {var_metodo2:.4f}")

print(f"\nAmbos metodos iguales: {np.isclose(var_metodo1, var_metodo2)}")
print(f"Valor exacto: 2/3 = {2/3:.4f}")

# Grafica: descomposicion visual de la varianza
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Metodo 1: desviaciones al cuadrado
desviaciones = (x - E_X)**2
contribuciones = desviaciones * p
ax1.bar(x, contribuciones, color=[C_DANGER, C_SUCCESS, C_DANGER], edgecolor=C_DARK, alpha=0.85, width=0.5)
ax1.axhline(y=0, color=C_DARK, linewidth=0.5)
for xi, c in zip(x, contribuciones):
    ax1.text(xi, c + 0.02, f'{c:.4f}', ha='center', fontsize=10, color=C_DARK)
ax1.set_xlabel('x')
ax1.set_ylabel('(x - mu)^2 * P(x)')
ax1.set_title(f'Metodo 1: contribuciones a Var = {var_metodo1:.4f}', color=C_DARK, fontweight='bold')
ax1.set_xticks(x)

# Metodo 2: E[X^2] vs (E[X])^2
categorias = ['E[X^2]', '(E[X])^2', 'Var(X)']
valores = [E_X2, E_X**2, var_metodo2]
colores_bar = [C_PRIMARY, C_ORANGE, C_SUCCESS]
ax2.bar(categorias, valores, color=colores_bar, edgecolor=C_DARK, alpha=0.85, width=0.5)
for i, v in enumerate(valores):
    ax2.text(i, v + 0.05, f'{v:.4f}', ha='center', fontweight='bold', color=C_DARK)
ax2.set_title(f'Metodo 2: E[X^2] - (E[X])^2 = Var', color=C_DARK, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Ejercicio 9:** X tiene PMF: P(X=1)=0.1, P(X=2)=0.3, P(X=3)=0.4, P(X=4)=0.2. Calcular E[X], E[X^2], Var(X).

**Resolucion:**

Se tiene una variable aleatoria discreta $X$ con funcion de masa de probabilidad (PMF) dada por:

| $x_i$ | $P(X = x_i)$ |
|--------|---------------|
| 1      | 0.1           |
| 2      | 0.3           |
| 3      | 0.4           |
| 4      | 0.2           |

Donde:
- $X$: variable aleatoria discreta que toma valores en $\{1, 2, 3, 4\}$.
- $x_i$: cada uno de los posibles valores de $X$. Aqui $x_1 = 1$, $x_2 = 2$, $x_3 = 3$, $x_4 = 4$.
- $P(X = x_i)$: probabilidad de que $X$ tome el valor $x_i$. La suma de todas las probabilidades es $0.1 + 0.3 + 0.4 + 0.2 = 1$.

**Esperanza E[X]:**

La esperanza es el valor promedio ponderado por las probabilidades:

$$E[X] = \sum_{i=1}^{4} x_i \cdot P(X = x_i)$$

- $E[X]$: esperanza de $X$, tambien llamada valor esperado o media $\mu$.
- $\sum_{i=1}^{4}$: sumatoria sobre los 4 valores posibles de $X$.
- $x_i \cdot P(X = x_i)$: cada valor multiplicado por su probabilidad.

$$E[X] = 1 \times 0.1 + 2 \times 0.3 + 3 \times 0.4 + 4 \times 0.2$$
$$= 0.1 + 0.6 + 1.2 + 0.8 = 2.7$$

**Segundo momento E[X^2]:**

$$E[X^2] = \sum_{i=1}^{4} x_i^2 \cdot P(X = x_i)$$

- $E[X^2]$: esperanza del cuadrado de $X$, es decir el segundo momento. Se calcula aplicando LOTUS (Law of the Unconscious Statistician): no necesitamos la distribucion de $X^2$, solo la de $X$.
- $x_i^2$: el cuadrado de cada valor posible de $X$.

$$E[X^2] = 1^2 \times 0.1 + 2^2 \times 0.3 + 3^2 \times 0.4 + 4^2 \times 0.2$$
$$= 0.1 + 1.2 + 3.6 + 3.2 = 8.1$$

**Varianza Var(X):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $Var(X)$: varianza de $X$, denotada $\sigma^2$. Mide la dispersion de $X$ respecto a su media.
- $E[X^2]$: segundo momento de $X$, calculado arriba como $8.1$.
- $(E[X])^2$: cuadrado de la esperanza, es decir $\mu^2 = (2.7)^2 = 7.29$.

$$Var(X) = 8.1 - (2.7)^2 = 8.1 - 7.29 = 0.81$$

La desviacion estandar es $\sigma = \sqrt{Var(X)} = \sqrt{0.81} = 0.9$.

```python
x = np.array([1, 2, 3, 4])
p = np.array([0.1, 0.3, 0.4, 0.2])

# Verificacion de PMF valida
print(f"Suma de probabilidades = {np.sum(p)} (debe ser 1)")

# E[X]
E_X = np.sum(x * p)
print(f"\nE[X] = {' + '.join([f'{xi}*{pi}' for xi, pi in zip(x, p)])}")
print(f"E[X] = {' + '.join([f'{xi*pi:.1f}' for xi, pi in zip(x, p)])} = {E_X}")

# E[X^2]
E_X2 = np.sum(x**2 * p)
print(f"\nE[X^2] = {' + '.join([f'{xi}^2*{pi}' for xi, pi in zip(x, p)])}")
print(f"E[X^2] = {' + '.join([f'{xi**2*pi:.1f}' for xi, pi in zip(x, p)])} = {E_X2}")

# Var(X)
Var_X = E_X2 - E_X**2
std_X = np.sqrt(Var_X)
print(f"\nVar(X) = E[X^2] - (E[X])^2 = {E_X2} - ({E_X})^2 = {E_X2} - {E_X**2} = {Var_X}")
print(f"Desviacion estandar = sqrt({Var_X}) = {std_X}")

# Grafica: PMF con E[X] y banda de +/- 1 std
fig, ax = plt.subplots()
ax.bar(x, p, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, width=0.5, label='P(X = x)')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.axvspan(E_X - std_X, E_X + std_X, alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {std_X:.2f}')

for xi, pi in zip(x, p):
    ax.text(xi, pi + 0.015, f'{pi}', ha='center', fontweight='bold', color=C_DARK)

ax.set_xlabel('x')
ax.set_ylabel('P(X = x)')
ax.set_title('PMF discreta con E[X] y desviacion estandar', color=C_DARK, fontweight='bold')
ax.set_xticks(x)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 10:** Variable continua f(x)=3x^2 para x en [0,1]. Calcular E[X] y Var(X) con integrales.

**Resolucion:**

Se tiene una variable aleatoria continua $X$ con funcion de densidad de probabilidad (PDF):

$$f(x) = 3x^2, \quad x \in [0, 1]$$

Donde:
- $f(x)$: funcion de densidad de probabilidad. Asigna una "densidad" a cada punto $x$, de modo que la probabilidad de un intervalo es el area bajo la curva.
- $x$: variable continua que toma valores en el intervalo $[0, 1]$.
- $3x^2$: funcion cuadratica que crece con $x$, lo que significa que valores cercanos a 1 son mas probables que valores cercanos a 0.

**Verificacion de densidad valida:**

$$\int_0^1 3x^2 \, dx = \left[ x^3 \right]_0^1 = 1^3 - 0^3 = 1$$

- $\int_0^1$: integral definida en el soporte $[0, 1]$.
- $dx$: diferencial de integracion.
- El resultado es 1, confirmando que $f(x)$ es una densidad valida.

**Esperanza E[X]:**

$$E[X] = \int_0^1 x \cdot f(x) \, dx = \int_0^1 x \cdot 3x^2 \, dx = \int_0^1 3x^3 \, dx$$

- $E[X]$: esperanza de $X$, el "centro de masa" de la distribucion.
- $x \cdot f(x)$: el integrando es el valor $x$ ponderado por su densidad $f(x)$.
- $3x^3$: resultado de multiplicar $x$ por $3x^2$.

$$E[X] = \left[ \frac{3x^4}{4} \right]_0^1 = \frac{3}{4} = 0.75$$

**Segundo momento E[X^2]:**

$$E[X^2] = \int_0^1 x^2 \cdot f(x) \, dx = \int_0^1 x^2 \cdot 3x^2 \, dx = \int_0^1 3x^4 \, dx$$

- $E[X^2]$: segundo momento de $X$. Se usa para calcular la varianza.
- $x^2 \cdot f(x)$: el cuadrado de $x$ ponderado por la densidad.

$$E[X^2] = \left[ \frac{3x^5}{5} \right]_0^1 = \frac{3}{5} = 0.6$$

**Varianza Var(X):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $Var(X)$: varianza, mide la dispersion de $X$ respecto a su media $\mu = E[X] = 0.75$.
- $E[X^2] = 0.6$: segundo momento calculado arriba.
- $(E[X])^2 = (0.75)^2 = 0.5625$: cuadrado de la esperanza.

$$Var(X) = 0.6 - (0.75)^2 = 0.6 - 0.5625 = 0.0375$$

La desviacion estandar es $\sigma = \sqrt{0.0375} \approx 0.1936$.

```python
f = lambda x: 3 * x**2

# Verificacion de densidad
integral_total, _ = quad(f, 0, 1)
print(f"Integral de f(x) en [0,1] = {integral_total} (debe ser 1)")

# E[X]
E_X, _ = quad(lambda x: x * f(x), 0, 1)
print(f"\nE[X] = integral(x * 3x^2, 0, 1) = integral(3x^3, 0, 1) = {E_X:.4f} = 3/4")

# E[X^2]
E_X2, _ = quad(lambda x: x**2 * f(x), 0, 1)
print(f"E[X^2] = integral(x^2 * 3x^2, 0, 1) = integral(3x^4, 0, 1) = {E_X2:.4f} = 3/5")

# Var(X)
Var_X = E_X2 - E_X**2
std_X = np.sqrt(Var_X)
print(f"\nVar(X) = E[X^2] - (E[X])^2 = {E_X2:.4f} - ({E_X:.4f})^2 = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")
print(f"Desviacion estandar = sqrt({Var_X:.4f}) = {std_X:.4f}")

# Grafica: PDF con E[X] y region de +/- 1 std
x_plot = np.linspace(0, 1, 300)
y_plot = 3 * x_plot**2

fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, color=C_PRIMARY, linewidth=2.5, label='f(x) = 3x^2')
ax.fill_between(x_plot, y_plot, alpha=0.2, color=C_PRIMARY)

# Sombrear region E[X] +/- std
mask = (x_plot >= E_X - std_X) & (x_plot <= E_X + std_X)
ax.fill_between(x_plot[mask], y_plot[mask], alpha=0.4, color=C_ORANGE, label=f'E[X] +/- std')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X:.4f}')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('PDF f(x) = 3x^2 en [0, 1]', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 11:** X ~ Geometrica(p=0.2). Calcular E[X] y Var(X). Cuantos intentos esperas para el primer exito?

**Resolucion:**

La distribucion geometrica modela el numero de ensayos independientes de Bernoulli necesarios para obtener el primer exito.

$$P(X = k) = (1 - p)^{k-1} \cdot p, \quad k = 1, 2, 3, \ldots$$

Donde:
- $X$: variable aleatoria que cuenta el numero de ensayos hasta el primer exito. Toma valores en $\{1, 2, 3, \ldots\}$.
- $k$: numero de ensayos realizados (incluyendo el exitoso).
- $p$: probabilidad de exito en cada ensayo individual. Aqui $p = 0.2$.
- $(1 - p)$: probabilidad de fracaso en cada ensayo. Aqui $1 - p = 0.8$.
- $(1 - p)^{k-1}$: probabilidad de obtener $k-1$ fracasos consecutivos antes del exito.
- La multiplicacion por $p$ al final representa el exito en el ensayo $k$-esimo.

**Esperanza E[X]:**

$$E[X] = \frac{1}{p}$$

- $E[X]$: numero esperado de ensayos hasta el primer exito.
- $\frac{1}{p}$: inverso de la probabilidad de exito. A menor probabilidad de exito, mas ensayos se esperan.

$$E[X] = \frac{1}{0.2} = 5$$

Interpretacion: en promedio, se esperan 5 intentos para obtener el primer exito.

**Varianza Var(X):**

$$Var(X) = \frac{1 - p}{p^2}$$

- $Var(X)$: varianza de $X$, mide la dispersion del numero de ensayos respecto a la media.
- $1 - p$: probabilidad de fracaso. Aqui $1 - p = 0.8$.
- $p^2$: cuadrado de la probabilidad de exito. Aqui $p^2 = 0.04$.

$$Var(X) = \frac{1 - 0.2}{(0.2)^2} = \frac{0.8}{0.04} = 20$$

La desviacion estandar es $\sigma = \sqrt{20} \approx 4.47$, indicando alta variabilidad en el numero de intentos.

```python
p = 0.2

# Formulas teoricas
E_X = 1 / p
Var_X = (1 - p) / p**2
std_X = np.sqrt(Var_X)

print(f"E[X] = 1/p = 1/{p} = {E_X}")
print(f"Var(X) = (1-p)/p^2 = {1-p}/{p**2} = {Var_X}")
print(f"Desviacion estandar = sqrt({Var_X}) = {std_X:.4f}")
print(f"\nInterpretacion: se esperan {E_X:.0f} intentos para el primer exito")

# Verificacion con scipy (geom en scipy cuenta desde 1)
dist = stats.geom(p)
print(f"\nVerificacion scipy: E[X] = {dist.mean()}, Var(X) = {dist.var()}")

# Simulacion
np.random.seed(42)
n_sim = 50_000
simulaciones = stats.geom.rvs(p, size=n_sim)
print(f"\nSimulacion ({n_sim:,} muestras):")
print(f"Media simulada = {np.mean(simulaciones):.2f} (teorica: {E_X})")
print(f"Varianza simulada = {np.var(simulaciones):.2f} (teorica: {Var_X})")

# Grafica: PMF geometrica
k_vals = np.arange(1, 26)
pmf_vals = stats.geom.pmf(k_vals, p)

fig, ax = plt.subplots()
ax.bar(k_vals, pmf_vals, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X = k)')
ax.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X:.0f}')
ax.axvspan(E_X - std_X, E_X + std_X, alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {std_X:.2f}')

ax.set_xlabel('k (numero de intentos)')
ax.set_ylabel('P(X = k)')
ax.set_title(f'Geometrica(p={p}): intentos hasta el primer exito', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 12:** E[aX+b] = aE[X]+b. Verificar con X ~ Binomial(10, 0.3), a=2, b=5.

**Resolucion:**

Se quiere verificar la propiedad de linealidad de la esperanza:

$$E[aX + b] = a \cdot E[X] + b$$

Donde:
- $E[aX + b]$: esperanza de la transformacion lineal $aX + b$.
- $a$: constante multiplicativa (escalar). Aqui $a = 2$.
- $b$: constante aditiva. Aqui $b = 5$.
- $X$: variable aleatoria. Aqui $X \sim \text{Binomial}(n=10, p=0.3)$.
- $E[X]$: esperanza de $X$.
- Esta propiedad se cumple siempre, independientemente de la distribucion de $X$.

**Paso 1: Calcular E[X] para X ~ Binomial(n, p):**

$$E[X] = n \cdot p$$

- $n$: numero de ensayos. Aqui $n = 10$.
- $p$: probabilidad de exito en cada ensayo. Aqui $p = 0.3$.

$$E[X] = 10 \times 0.3 = 3$$

**Paso 2: Calcular E[aX + b] usando la propiedad:**

$$E[aX + b] = a \cdot E[X] + b = 2 \times 3 + 5 = 6 + 5 = 11$$

**Paso 3: Verificar directamente con la definicion:**

$$E[aX + b] = \sum_{k=0}^{n} (a \cdot k + b) \cdot P(X = k)$$

- $k$: cada posible valor de $X$, desde 0 hasta $n = 10$.
- $a \cdot k + b$: transformacion lineal aplicada al valor $k$. Aqui $2k + 5$.
- $P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$: probabilidad binomial de obtener exactamente $k$ exitos.
- $\binom{n}{k}$: coeficiente binomial, numero de formas de elegir $k$ exitos entre $n$ ensayos.

**Tambien verificamos la varianza de la transformacion:**

$$Var(aX + b) = a^2 \cdot Var(X)$$

- $Var(aX + b)$: varianza de la transformacion lineal.
- $a^2$: la constante multiplicativa se eleva al cuadrado porque la varianza mide dispersion cuadratica. Aqui $a^2 = 4$.
- $Var(X) = n \cdot p \cdot (1-p) = 10 \times 0.3 \times 0.7 = 2.1$.
- $b$ no aparece porque trasladar una distribucion no cambia su dispersion.

$$Var(2X + 5) = 2^2 \times 2.1 = 4 \times 2.1 = 8.4$$

```python
n, p = 10, 0.3
a, b = 2, 5

# E[X] y Var(X) de la Binomial
E_X = n * p
Var_X = n * p * (1 - p)

# Propiedad de linealidad
E_aXb_propiedad = a * E_X + b
Var_aXb_propiedad = a**2 * Var_X

print("=== Usando propiedades ===")
print(f"E[X] = n*p = {n}*{p} = {E_X}")
print(f"E[aX+b] = a*E[X]+b = {a}*{E_X}+{b} = {E_aXb_propiedad}")
print(f"Var(X) = n*p*(1-p) = {Var_X}")
print(f"Var(aX+b) = a^2*Var(X) = {a**2}*{Var_X} = {Var_aXb_propiedad}")

# Verificacion directa con la definicion
k_vals = np.arange(0, n + 1)
pmf_vals = stats.binom.pmf(k_vals, n, p)
transformados = a * k_vals + b

E_aXb_directo = np.sum(transformados * pmf_vals)
E_aXb2_directo = np.sum(transformados**2 * pmf_vals)
Var_aXb_directo = E_aXb2_directo - E_aXb_directo**2

print(f"\n=== Verificacion directa (definicion) ===")
print(f"E[aX+b] = sum((2k+5)*P(k)) = {E_aXb_directo}")
print(f"Var(aX+b) = E[(aX+b)^2] - (E[aX+b])^2 = {Var_aXb_directo:.4f}")
print(f"\nCoinciden: E -> {np.isclose(E_aXb_propiedad, E_aXb_directo)}, Var -> {np.isclose(Var_aXb_propiedad, Var_aXb_directo)}")

# Grafica: comparar PMF de X vs PMF de aX+b
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(k_vals, pmf_vals, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=k)')
ax1.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title('X ~ Binomial(10, 0.3)', color=C_DARK, fontweight='bold')
ax1.legend()

ax2.bar(transformados, pmf_vals, color=C_SUCCESS, edgecolor=C_DARK, alpha=0.85, width=1.5, label='P(2X+5=v)')
ax2.axvline(x=E_aXb_propiedad, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[2X+5] = {E_aXb_propiedad}')
ax2.set_xlabel('v = 2k + 5')
ax2.set_ylabel('P(2X+5 = v)')
ax2.set_title('Y = 2X + 5 (transformacion lineal)', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 13:** X ~ Uniforme discreta {1,2,...,n}. Derivar E[X]=(n+1)/2 y Var(X)=(n^2-1)/12 para n=6.

**Resolucion:**

La distribucion uniforme discreta asigna la misma probabilidad a cada valor del conjunto $\{1, 2, \ldots, n\}$.

$$P(X = k) = \frac{1}{n}, \quad k = 1, 2, \ldots, n$$

Donde:
- $X$: variable aleatoria con distribucion uniforme discreta sobre $\{1, 2, \ldots, n\}$.
- $k$: cada uno de los $n$ valores posibles.
- $\frac{1}{n}$: probabilidad de cada valor. Todos tienen la misma probabilidad.
- $n$: numero total de valores posibles. Aqui $n = 6$.

**Derivacion de E[X]:**

$$E[X] = \sum_{k=1}^{n} k \cdot \frac{1}{n} = \frac{1}{n} \sum_{k=1}^{n} k$$

- $E[X]$: esperanza de $X$, el valor promedio.
- $\sum_{k=1}^{n} k$: suma de los primeros $n$ enteros positivos.
- $\frac{1}{n}$: factor constante que sale de la sumatoria.

Usando la formula conocida $\sum_{k=1}^{n} k = \frac{n(n+1)}{2}$:

$$E[X] = \frac{1}{n} \cdot \frac{n(n+1)}{2} = \frac{n+1}{2}$$

Para $n = 6$: $E[X] = \frac{6+1}{2} = \frac{7}{2} = 3.5$

**Derivacion de Var(X):**

$$Var(X) = E[X^2] - (E[X])^2$$

- $Var(X)$: varianza de $X$, mide la dispersion respecto a la media.
- $E[X^2]$: segundo momento de $X$.
- $(E[X])^2$: cuadrado de la esperanza.

Primero calculamos $E[X^2]$:

$$E[X^2] = \sum_{k=1}^{n} k^2 \cdot \frac{1}{n} = \frac{1}{n} \sum_{k=1}^{n} k^2$$

- $\sum_{k=1}^{n} k^2$: suma de los cuadrados de los primeros $n$ enteros.

Usando la formula conocida $\sum_{k=1}^{n} k^2 = \frac{n(n+1)(2n+1)}{6}$:

$$E[X^2] = \frac{1}{n} \cdot \frac{n(n+1)(2n+1)}{6} = \frac{(n+1)(2n+1)}{6}$$

Entonces:

$$Var(X) = \frac{(n+1)(2n+1)}{6} - \left(\frac{n+1}{2}\right)^2 = \frac{(n+1)(2n+1)}{6} - \frac{(n+1)^2}{4}$$

$$= (n+1) \left[\frac{2n+1}{6} - \frac{n+1}{4}\right] = (n+1) \cdot \frac{2(2n+1) - 3(n+1)}{12}$$

$$= (n+1) \cdot \frac{4n + 2 - 3n - 3}{12} = (n+1) \cdot \frac{n - 1}{12} = \frac{n^2 - 1}{12}$$

Para $n = 6$: $Var(X) = \frac{36 - 1}{12} = \frac{35}{12} \approx 2.9167$

```python
n = 6
x = np.arange(1, n + 1)
p = np.ones(n) / n

# Formulas teoricas
E_X_teorica = (n + 1) / 2
Var_X_teorica = (n**2 - 1) / 12

# Calculo directo
E_X = np.sum(x * p)
E_X2 = np.sum(x**2 * p)
Var_X = E_X2 - E_X**2

print(f"n = {n}")
print(f"\n=== Formula teorica ===")
print(f"E[X] = (n+1)/2 = ({n}+1)/2 = {E_X_teorica}")
print(f"Var(X) = (n^2-1)/12 = ({n}^2-1)/12 = ({n**2}-1)/12 = {Var_X_teorica:.4f}")

print(f"\n=== Calculo directo ===")
print(f"E[X] = (1/n) * sum(k) = (1/{n}) * {int(np.sum(x))} = {E_X}")
print(f"E[X^2] = (1/n) * sum(k^2) = (1/{n}) * {int(np.sum(x**2))} = {E_X2:.4f}")
print(f"Var(X) = E[X^2] - (E[X])^2 = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")

print(f"\nCoinciden: E -> {np.isclose(E_X, E_X_teorica)}, Var -> {np.isclose(Var_X, Var_X_teorica)}")

# Grafica: PMF uniforme con formulas para distintos n
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# PMF para n=6
ax1.bar(x, p, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=k) = 1/6')
ax1.axvline(x=E_X, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax1.axvspan(E_X - np.sqrt(Var_X), E_X + np.sqrt(Var_X), alpha=0.15, color=C_ORANGE,
            label=f'+/- 1 std = {np.sqrt(Var_X):.2f}')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title(f'Uniforme discreta n={n}', color=C_DARK, fontweight='bold')
ax1.set_xticks(x)
ax1.legend(fontsize=9)

# E[X] y Var(X) para distintos n
n_vals = np.arange(2, 21)
e_vals = (n_vals + 1) / 2
var_vals = (n_vals**2 - 1) / 12

ax2.plot(n_vals, e_vals, color=C_PRIMARY, linewidth=2, marker='o', markersize=4, label='E[X] = (n+1)/2')
ax2.plot(n_vals, var_vals, color=C_DANGER, linewidth=2, marker='s', markersize=4, label='Var(X) = (n^2-1)/12')
ax2.axvline(x=6, color=C_DARK, linestyle=':', alpha=0.5, label='n=6')
ax2.set_xlabel('n')
ax2.set_ylabel('Valor')
ax2.set_title('E[X] y Var(X) segun n', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 14:** X ~ Normal(50, 10). Calcular P(40<X<60), P(X>65), E[X], Var(X).

**Resolucion:**

Se tiene $X \sim \mathcal{N}(\mu, \sigma)$ con $\mu = 50$ y $\sigma = 10$.

La funcion de densidad de la distribucion normal es:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)$$

Donde:
- $f(x)$: funcion de densidad de probabilidad (PDF) de la normal. Es la famosa "campana de Gauss".
- $\mu$: media de la distribucion. Aqui $\mu = 50$. Es el centro de la campana.
- $\sigma$: desviacion estandar. Aqui $\sigma = 10$. Controla el ancho de la campana.
- $\sigma^2$: varianza, $\sigma^2 = 100$.
- $\pi$: constante pi ($\approx 3.14159$).
- $\exp(\cdot)$: funcion exponencial $e^{(\cdot)}$.
- $(x - \mu)^2$: distancia cuadratica de $x$ a la media.
- $2\sigma^2$: factor de escala en el exponente.
- $\frac{1}{\sigma\sqrt{2\pi}}$: constante de normalizacion que asegura que el area total bajo la curva sea 1.

**Esperanza y Varianza:**

$$E[X] = \mu = 50$$

- $E[X]$: esperanza de $X$. Para la distribucion normal, la esperanza coincide con el parametro $\mu$.

$$Var(X) = \sigma^2 = 10^2 = 100$$

- $Var(X)$: varianza de $X$. Para la normal, es el cuadrado del parametro $\sigma$.

**P(40 < X < 60):**

Estandarizamos usando $Z = \frac{X - \mu}{\sigma}$:

- $Z$: variable normal estandar, $Z \sim \mathcal{N}(0, 1)$.
- $\frac{X - \mu}{\sigma}$: transformacion que convierte $X$ a la escala estandar (media 0, desviacion 1).

$$P(40 < X < 60) = P\left(\frac{40 - 50}{10} < Z < \frac{60 - 50}{10}\right) = P(-1 < Z < 1)$$

$$= \Phi(1) - \Phi(-1) \approx 0.8413 - 0.1587 = 0.6827$$

- $\Phi(\cdot)$: funcion de distribucion acumulada (CDF) de la normal estandar $\mathcal{N}(0,1)$.
- $\Phi(1) \approx 0.8413$: probabilidad de que $Z \leq 1$.
- $\Phi(-1) \approx 0.1587$: probabilidad de que $Z \leq -1$.

Esto confirma la regla empirica: aproximadamente el 68.27% de los datos caen dentro de $\pm 1\sigma$ de la media.

**P(X > 65):**

$$P(X > 65) = P\left(Z > \frac{65 - 50}{10}\right) = P(Z > 1.5) = 1 - \Phi(1.5)$$

$$= 1 - 0.9332 = 0.0668$$

- $\Phi(1.5) \approx 0.9332$: probabilidad acumulada hasta $z = 1.5$.
- $1 - \Phi(1.5)$: cola derecha, probabilidad de estar a mas de 1.5 desviaciones estandar por encima de la media.

```python
mu, sigma = 50, 10
dist = stats.norm(loc=mu, scale=sigma)

E_X = dist.mean()
Var_X = dist.var()

# Probabilidades
p_40_60 = dist.cdf(60) - dist.cdf(40)
p_gt65 = 1 - dist.cdf(65)

# Estandarizacion
z_40 = (40 - mu) / sigma
z_60 = (60 - mu) / sigma
z_65 = (65 - mu) / sigma

print(f"E[X] = mu = {E_X}")
print(f"Var(X) = sigma^2 = {sigma}^2 = {Var_X}")
print(f"sigma = {sigma}")

print(f"\nP(40 < X < 60):")
print(f"  Z_40 = (40 - {mu}) / {sigma} = {z_40}")
print(f"  Z_60 = (60 - {mu}) / {sigma} = {z_60}")
print(f"  P(-1 < Z < 1) = Phi(1) - Phi(-1) = {stats.norm.cdf(1):.4f} - {stats.norm.cdf(-1):.4f} = {p_40_60:.4f}")

print(f"\nP(X > 65):")
print(f"  Z_65 = (65 - {mu}) / {sigma} = {z_65}")
print(f"  P(Z > 1.5) = 1 - Phi(1.5) = 1 - {stats.norm.cdf(1.5):.4f} = {p_gt65:.4f}")

# Grafica: PDF normal con regiones sombreadas
x_plot = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)
y_plot = dist.pdf(x_plot)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_plot, y_plot, color=C_DARK, linewidth=2, label=f'N({mu}, {sigma})')

# Sombrear P(40 < X < 60)
mask1 = (x_plot >= 40) & (x_plot <= 60)
ax.fill_between(x_plot[mask1], y_plot[mask1], alpha=0.4, color=C_PRIMARY,
                label=f'P(40<X<60) = {p_40_60:.4f}')

# Sombrear P(X > 65)
mask2 = x_plot >= 65
ax.fill_between(x_plot[mask2], y_plot[mask2], alpha=0.5, color=C_DANGER,
                label=f'P(X>65) = {p_gt65:.4f}')

ax.axvline(x=mu, color=C_ORANGE, linewidth=2, linestyle='--', label=f'mu = {mu}')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(f'Normal(mu={mu}, sigma={sigma}): probabilidades sombreadas', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 15:** Dos dados: Y=X1+X2. Calcular E[Y] y Var(Y) usando propiedades de linealidad.

**Resolucion:**

Se lanzan dos dados justos de 6 caras independientes. Definimos:

- $X_1$: resultado del primer dado. $X_1 \in \{1, 2, 3, 4, 5, 6\}$, cada valor con probabilidad $\frac{1}{6}$.
- $X_2$: resultado del segundo dado. $X_2 \in \{1, 2, 3, 4, 5, 6\}$, cada valor con probabilidad $\frac{1}{6}$.
- $Y = X_1 + X_2$: suma de ambos dados. $Y \in \{2, 3, \ldots, 12\}$.
- $X_1$ y $X_2$ son **independientes**: el resultado de un dado no afecta al otro.

**E[Y] usando linealidad de la esperanza:**

$$E[Y] = E[X_1 + X_2] = E[X_1] + E[X_2]$$

- $E[Y]$: esperanza de la suma $Y$.
- $E[X_1 + X_2]$: esperanza de la suma de dos variables aleatorias.
- $E[X_1] + E[X_2]$: la esperanza de la suma siempre es la suma de las esperanzas (no requiere independencia).

Para un dado justo (uniforme discreta en $\{1, \ldots, 6\}$):

$$E[X_1] = E[X_2] = \frac{n+1}{2} = \frac{6+1}{2} = 3.5$$

- $\frac{n+1}{2}$: formula de la esperanza de la uniforme discreta con $n = 6$.

$$E[Y] = 3.5 + 3.5 = 7$$

**Var(Y) usando propiedades de la varianza:**

$$Var(Y) = Var(X_1 + X_2) = Var(X_1) + Var(X_2) + 2 \cdot Cov(X_1, X_2)$$

- $Var(Y)$: varianza de la suma.
- $Var(X_1) + Var(X_2)$: suma de las varianzas individuales.
- $Cov(X_1, X_2)$: covarianza entre $X_1$ y $X_2$. Mide la dependencia lineal.
- Como $X_1$ y $X_2$ son **independientes**, $Cov(X_1, X_2) = 0$.

$$Var(Y) = Var(X_1) + Var(X_2)$$

Para un dado justo:

$$Var(X_1) = Var(X_2) = \frac{n^2 - 1}{12} = \frac{36 - 1}{12} = \frac{35}{12} \approx 2.9167$$

- $\frac{n^2-1}{12}$: formula de la varianza de la uniforme discreta con $n = 6$.

$$Var(Y) = \frac{35}{12} + \frac{35}{12} = \frac{70}{12} = \frac{35}{6} \approx 5.8333$$

La desviacion estandar de $Y$ es $\sigma_Y = \sqrt{35/6} \approx 2.4152$.

```python
# Propiedades teoricas
E_X1 = 3.5
E_X2 = 3.5
Var_X1 = 35 / 12
Var_X2 = 35 / 12

E_Y = E_X1 + E_X2
Var_Y = Var_X1 + Var_X2  # independientes
std_Y = np.sqrt(Var_Y)

print(f"E[X1] = E[X2] = {E_X1}")
print(f"Var(X1) = Var(X2) = 35/12 = {Var_X1:.4f}")
print(f"\nE[Y] = E[X1] + E[X2] = {E_X1} + {E_X2} = {E_Y}")
print(f"Var(Y) = Var(X1) + Var(X2) = {Var_X1:.4f} + {Var_X2:.4f} = {Var_Y:.4f}")
print(f"Desviacion estandar de Y = {std_Y:.4f}")

# Verificacion: construir PMF exacta de Y = X1 + X2
y_vals = np.arange(2, 13)
pmf_Y = np.zeros(11)
for x1 in range(1, 7):
    for x2 in range(1, 7):
        y = x1 + x2
        pmf_Y[y - 2] += 1 / 36

E_Y_directo = np.sum(y_vals * pmf_Y)
E_Y2_directo = np.sum(y_vals**2 * pmf_Y)
Var_Y_directo = E_Y2_directo - E_Y_directo**2

print(f"\n=== Verificacion directa (36 combinaciones) ===")
print(f"E[Y] = {E_Y_directo}")
print(f"Var(Y) = {Var_Y_directo:.4f}")

# Grafica: PMF de Y = X1 + X2
fig, ax = plt.subplots()
ax.bar(y_vals, pmf_Y, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(Y = y)')
ax.axvline(x=E_Y, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[Y] = {E_Y}')
ax.axvspan(E_Y - std_Y, E_Y + std_Y, alpha=0.15, color=C_ORANGE, label=f'+/- 1 std = {std_Y:.2f}')

for yi, pi in zip(y_vals, pmf_Y):
    ax.text(yi, pi + 0.005, f'{pi:.3f}', ha='center', fontsize=9, color=C_DARK)

ax.set_xlabel('Y = X1 + X2')
ax.set_ylabel('P(Y = y)')
ax.set_title('PMF de la suma de dos dados justos', color=C_DARK, fontweight='bold')
ax.set_xticks(y_vals)
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 16:** X ~ Exponencial(lambda=0.5). Calcular E[X], Var(X), P(X>3) con integral.

**Resolucion:**

La distribucion exponencial modela tiempos de espera entre eventos en un proceso de Poisson.

$$f(x) = \lambda e^{-\lambda x}, \quad x \geq 0$$

Donde:
- $f(x)$: funcion de densidad de probabilidad (PDF) de la exponencial.
- $\lambda$: parametro de tasa (rate). Aqui $\lambda = 0.5$. Representa el numero promedio de eventos por unidad de tiempo.
- $e^{-\lambda x}$: decaimiento exponencial. La densidad decrece rapidamente.
- $x$: variable continua no negativa (tiempo de espera, por ejemplo).

**Esperanza E[X]:**

$$E[X] = \frac{1}{\lambda}$$

- $E[X]$: tiempo de espera promedio hasta el proximo evento.
- $\frac{1}{\lambda}$: inverso de la tasa. A mayor tasa, menor tiempo de espera promedio.

$$E[X] = \frac{1}{0.5} = 2$$

**Varianza Var(X):**

$$Var(X) = \frac{1}{\lambda^2}$$

- $Var(X)$: varianza del tiempo de espera.
- $\frac{1}{\lambda^2}$: cuadrado del inverso de la tasa.
- $\lambda^2 = (0.5)^2 = 0.25$.

$$Var(X) = \frac{1}{(0.5)^2} = \frac{1}{0.25} = 4$$

La desviacion estandar es $\sigma = \sqrt{4} = 2$, que coincide con la media (propiedad de la exponencial).

**P(X > 3) con integral:**

$$P(X > 3) = \int_3^{\infty} \lambda e^{-\lambda x} \, dx = \int_3^{\infty} 0.5 \cdot e^{-0.5x} \, dx$$

- $\int_3^{\infty}$: integral desde $x = 3$ hasta infinito (cola derecha).
- $\lambda e^{-\lambda x}$: la PDF evaluada en el integrando.

$$= \left[ -e^{-0.5x} \right]_3^{\infty} = 0 - (-e^{-1.5}) = e^{-1.5} \approx 0.2231$$

- $e^{-1.5}$: resultado de evaluar $-e^{-0.5x}$ en los limites. En $x \to \infty$, $e^{-0.5x} \to 0$. En $x = 3$, $e^{-0.5 \cdot 3} = e^{-1.5}$.

```python
lam = 0.5
dist = stats.expon(scale=1/lam)

E_X = 1 / lam
Var_X = 1 / lam**2
std_X = np.sqrt(Var_X)

print(f"E[X] = 1/lambda = 1/{lam} = {E_X}")
print(f"Var(X) = 1/lambda^2 = 1/{lam**2} = {Var_X}")
print(f"Desviacion estandar = {std_X}")

# P(X > 3) con integral
p_gt3_integral, _ = quad(lambda x: lam * np.exp(-lam * x), 3, np.inf)
p_gt3_formula = np.exp(-lam * 3)
p_gt3_scipy = 1 - dist.cdf(3)

print(f"\nP(X > 3):")
print(f"  Por integral numerica: {p_gt3_integral:.4f}")
print(f"  Por formula e^(-lambda*3) = e^(-{lam*3}) = {p_gt3_formula:.4f}")
print(f"  Por scipy: {p_gt3_scipy:.4f}")

# Grafica: PDF con area sombreada para P(X > 3)
x_plot = np.linspace(0, 10, 300)
y_plot = dist.pdf(x_plot)

fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, color=C_PRIMARY, linewidth=2.5, label=f'f(x) = {lam}*exp(-{lam}x)')
ax.fill_between(x_plot, y_plot, where=(x_plot <= 3), alpha=0.3, color=C_PRIMARY,
                label=f'P(X<=3) = {1-p_gt3_formula:.4f}')
ax.fill_between(x_plot, y_plot, where=(x_plot >= 3), alpha=0.4, color=C_DANGER,
                label=f'P(X>3) = {p_gt3_formula:.4f}')
ax.axvline(x=E_X, color=C_ORANGE, linewidth=2, linestyle='--', label=f'E[X] = {E_X}')
ax.axvline(x=3, color=C_DARK, linewidth=1, linestyle=':')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title(f'Exponencial(lambda={lam}): P(X > 3) sombreada en rojo', color=C_DARK, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 17:** Desigualdad de Chebyshev: P(|X-mu|>=k*sigma) <= 1/k^2. Verificar con X ~ Normal para k=1,2,3.

**Resolucion:**

La desigualdad de Chebyshev establece una cota superior para la probabilidad de que una variable aleatoria se aleje de su media mas de $k$ desviaciones estandar:

$$P(|X - \mu| \geq k\sigma) \leq \frac{1}{k^2}$$

Donde:
- $P(|X - \mu| \geq k\sigma)$: probabilidad de que $X$ este a una distancia de al menos $k\sigma$ de su media $\mu$.
- $X$: variable aleatoria con media $\mu$ y varianza $\sigma^2$ finitas.
- $\mu$: media (esperanza) de $X$, $\mu = E[X]$.
- $\sigma$: desviacion estandar de $X$, $\sigma = \sqrt{Var(X)}$.
- $k$: numero de desviaciones estandar. Debe ser $k > 0$.
- $k\sigma$: distancia medida en unidades de desviacion estandar.
- $|X - \mu|$: valor absoluto de la desviacion de $X$ respecto a su media.
- $\frac{1}{k^2}$: cota superior. Es una cota conservadora que vale para **cualquier** distribucion.

**Equivalentemente**, la probabilidad de estar dentro de $k$ desviaciones:

$$P(|X - \mu| < k\sigma) \geq 1 - \frac{1}{k^2}$$

**Verificacion con X ~ Normal(0, 1):**

Para la normal estandar $Z \sim \mathcal{N}(0, 1)$, donde $\mu = 0$ y $\sigma = 1$:

- $Z$: variable normal estandar.
- $\mathcal{N}(0, 1)$: distribucion normal con media 0 y desviacion estandar 1.

| $k$ | Cota Chebyshev $\frac{1}{k^2}$ | P real Normal $P(\|Z\| \geq k)$ | La cota se cumple? |
|-----|-------------------------------|--------------------------------|-------------------|
| 1   | $\frac{1}{1} = 1.0000$       | $2(1 - \Phi(1)) = 0.3173$     | Si (0.3173 <= 1)  |
| 2   | $\frac{1}{4} = 0.2500$       | $2(1 - \Phi(2)) = 0.0455$     | Si (0.0455 <= 0.25)|
| 3   | $\frac{1}{9} = 0.1111$       | $2(1 - \Phi(3)) = 0.0027$     | Si (0.0027 <= 0.11)|

- $\Phi(k)$: funcion de distribucion acumulada (CDF) de la normal estandar evaluada en $k$.
- $2(1 - \Phi(k))$: probabilidad en ambas colas (simetria de la normal).
- La cota de Chebyshev es siempre mayor que la probabilidad real, confirmando que es una cota **conservadora**.
- La normal concentra mucha mas probabilidad cerca de la media que lo que Chebyshev garantiza.

```python
k_vals = [1, 2, 3]

print(f"{'k':>3} | {'Chebyshev 1/k^2':>16} | {'P real Normal':>16} | {'Cumple?':>10}")
print("-" * 55)

chebyshev_cotas = []
p_reales = []

for k in k_vals:
    cota = 1 / k**2
    p_real = 2 * (1 - stats.norm.cdf(k))
    cumple = p_real <= cota
    chebyshev_cotas.append(cota)
    p_reales.append(p_real)
    print(f"{k:>3} | {cota:>16.4f} | {p_real:>16.4f} | {'Si' if cumple else 'No':>10}")

print(f"\nLa cota de Chebyshev siempre se cumple.")
print(f"Para la Normal, la probabilidad real es mucho menor que la cota.")

# Grafica: comparacion Chebyshev vs Normal real
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# Barras comparativas
x_pos = np.arange(len(k_vals))
width = 0.35
ax1.bar(x_pos - width/2, chebyshev_cotas, width, color=C_DANGER, edgecolor=C_DARK,
        alpha=0.85, label='Cota Chebyshev (1/k^2)')
ax1.bar(x_pos + width/2, p_reales, width, color=C_PRIMARY, edgecolor=C_DARK,
        alpha=0.85, label='P real Normal')

for i, (c, r) in enumerate(zip(chebyshev_cotas, p_reales)):
    ax1.text(i - width/2, c + 0.02, f'{c:.4f}', ha='center', fontsize=9, color=C_DARK)
    ax1.text(i + width/2, r + 0.02, f'{r:.4f}', ha='center', fontsize=9, color=C_DARK)

ax1.set_xlabel('k (desviaciones estandar)')
ax1.set_ylabel('Probabilidad')
ax1.set_title('Chebyshev vs Normal: P(|X-mu| >= k*sigma)', color=C_DARK, fontweight='bold')
ax1.set_xticks(x_pos)
ax1.set_xticklabels([f'k={k}' for k in k_vals])
ax1.legend()

# PDF normal con regiones sombreadas
z_plot = np.linspace(-4, 4, 400)
y_plot = stats.norm.pdf(z_plot)
ax2.plot(z_plot, y_plot, color=C_DARK, linewidth=2)
ax2.fill_between(z_plot, y_plot, alpha=0.15, color=C_PRIMARY)

colores_k = [C_SUCCESS, C_ORANGE, C_DANGER]
for k, color in zip(k_vals, colores_k):
    ax2.axvline(x=k, color=color, linewidth=1.5, linestyle='--', alpha=0.8)
    ax2.axvline(x=-k, color=color, linewidth=1.5, linestyle='--', alpha=0.8)
    mask = (z_plot >= k) | (z_plot <= -k)
    ax2.fill_between(z_plot[mask], y_plot[mask], alpha=0.2, color=color, label=f'k={k}')

ax2.set_xlabel('z')
ax2.set_ylabel('f(z)')
ax2.set_title('Normal estandar: regiones fuera de k*sigma', color=C_DARK, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

**Ejercicio 18:** Covarianza: Cov(X,Y)=E[XY]-E[X]E[Y]. Calcular con datos discretos conjuntos.

**Resolucion:**

Consideremos dos variables aleatorias discretas $X$ e $Y$ con la siguiente distribucion conjunta:

| | Y=0 | Y=1 | Y=2 |
|---|---|---|---|
| X=0 | 0.10 | 0.05 | 0.05 |
| X=1 | 0.05 | 0.20 | 0.10 |
| X=2 | 0.05 | 0.10 | 0.30 |

Donde:
- $X$: variable aleatoria discreta que toma valores en $\{0, 1, 2\}$.
- $Y$: variable aleatoria discreta que toma valores en $\{0, 1, 2\}$.
- Cada entrada $P(X = x_i, Y = y_j)$: probabilidad conjunta de que $X = x_i$ y $Y = y_j$ simultaneamente.
- La suma de todas las entradas debe ser 1.

**Distribuciones marginales:**

$$P(X = x_i) = \sum_{j} P(X = x_i, Y = y_j)$$

- $P(X = x_i)$: probabilidad marginal de $X$. Se obtiene sumando la fila correspondiente.

$P(X=0) = 0.10 + 0.05 + 0.05 = 0.20$, $P(X=1) = 0.35$, $P(X=2) = 0.45$

$$P(Y = y_j) = \sum_{i} P(X = x_i, Y = y_j)$$

- $P(Y = y_j)$: probabilidad marginal de $Y$. Se obtiene sumando la columna correspondiente.

$P(Y=0) = 0.10 + 0.05 + 0.05 = 0.20$, $P(Y=1) = 0.35$, $P(Y=2) = 0.45$

**Esperanzas:**

$$E[X] = \sum_{i} x_i \cdot P(X = x_i) = 0 \times 0.20 + 1 \times 0.35 + 2 \times 0.45 = 1.25$$

- $E[X]$: esperanza de $X$, calculada con la distribucion marginal de $X$.

$$E[Y] = \sum_{j} y_j \cdot P(Y = y_j) = 0 \times 0.20 + 1 \times 0.35 + 2 \times 0.45 = 1.25$$

- $E[Y]$: esperanza de $Y$, calculada con la distribucion marginal de $Y$.

**E[XY]:**

$$E[XY] = \sum_{i} \sum_{j} x_i \cdot y_j \cdot P(X = x_i, Y = y_j)$$

- $E[XY]$: esperanza del producto $XY$. Se calcula usando la distribucion conjunta.
- $x_i \cdot y_j$: producto de los valores de $X$ e $Y$.
- $\sum_i \sum_j$: doble sumatoria sobre todos los pares $(x_i, y_j)$.

$$E[XY] = 0 \cdot 0 \cdot 0.10 + 0 \cdot 1 \cdot 0.05 + \ldots + 2 \cdot 2 \cdot 0.30$$

**Covarianza:**

$$Cov(X, Y) = E[XY] - E[X] \cdot E[Y]$$

- $Cov(X, Y)$: covarianza entre $X$ e $Y$. Mide la relacion lineal entre ambas variables.
- $E[XY]$: esperanza del producto.
- $E[X] \cdot E[Y]$: producto de las esperanzas individuales.
- Si $Cov(X, Y) > 0$: $X$ e $Y$ tienden a crecer juntas.
- Si $Cov(X, Y) < 0$: cuando una crece, la otra tiende a decrecer.
- Si $Cov(X, Y) = 0$: no hay relacion lineal (pero puede haber relacion no lineal).

**Correlacion:**

$$\rho(X, Y) = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$

- $\rho(X, Y)$: coeficiente de correlacion de Pearson. Siempre esta entre $-1$ y $1$.
- $\sigma_X = \sqrt{Var(X)}$: desviacion estandar de $X$.
- $\sigma_Y = \sqrt{Var(Y)}$: desviacion estandar de $Y$.
- Normaliza la covarianza para que sea comparable entre distintas escalas.

```python
# Distribucion conjunta P(X=xi, Y=yj)
# Filas: X=0,1,2; Columnas: Y=0,1,2
P_conjunta = np.array([
    [0.10, 0.05, 0.05],
    [0.05, 0.20, 0.10],
    [0.05, 0.10, 0.30]
])

x_vals = np.array([0, 1, 2])
y_vals = np.array([0, 1, 2])

print("Distribucion conjunta:")
print(P_conjunta)
print(f"Suma total = {P_conjunta.sum():.2f} (debe ser 1)")

# Marginales
P_X = P_conjunta.sum(axis=1)  # sumar filas
P_Y = P_conjunta.sum(axis=0)  # sumar columnas
print(f"\nMarginales:")
print(f"P(X) = {P_X}")
print(f"P(Y) = {P_Y}")

# Esperanzas
E_X = np.sum(x_vals * P_X)
E_Y = np.sum(y_vals * P_Y)
print(f"\nE[X] = {E_X}")
print(f"E[Y] = {E_Y}")

# E[XY]
E_XY = 0
print(f"\nCalculo de E[XY]:")
for i, xi in enumerate(x_vals):
    for j, yj in enumerate(y_vals):
        contrib = xi * yj * P_conjunta[i, j]
        if contrib != 0:
            print(f"  {xi}*{yj}*{P_conjunta[i,j]} = {contrib:.4f}")
        E_XY += contrib
print(f"E[XY] = {E_XY:.4f}")

# Covarianza
Cov_XY = E_XY - E_X * E_Y
print(f"\nCov(X,Y) = E[XY] - E[X]*E[Y] = {E_XY:.4f} - {E_X}*{E_Y} = {E_XY:.4f} - {E_X*E_Y:.4f} = {Cov_XY:.4f}")

# Varianzas y correlacion
E_X2 = np.sum(x_vals**2 * P_X)
E_Y2 = np.sum(y_vals**2 * P_Y)
Var_X = E_X2 - E_X**2
Var_Y = E_Y2 - E_Y**2
rho = Cov_XY / (np.sqrt(Var_X) * np.sqrt(Var_Y))

print(f"\nVar(X) = {Var_X:.4f}")
print(f"Var(Y) = {Var_Y:.4f}")
print(f"Correlacion rho = Cov(X,Y) / (sigma_X * sigma_Y) = {Cov_XY:.4f} / ({np.sqrt(Var_X):.4f} * {np.sqrt(Var_Y):.4f}) = {rho:.4f}")
print(f"Cov > 0: X e Y tienden a crecer juntas")

# Grafica: distribucion conjunta como mapa de calor y scatter de muestras
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Mapa de calor
im = ax1.imshow(P_conjunta, cmap='Blues', aspect='auto', origin='lower')
for i in range(3):
    for j in range(3):
        ax1.text(j, i, f'{P_conjunta[i,j]:.2f}', ha='center', va='center',
                fontsize=14, fontweight='bold', color='black' if P_conjunta[i,j] < 0.2 else 'white')

ax1.set_xticks([0, 1, 2])
ax1.set_yticks([0, 1, 2])
ax1.set_xlabel('Y')
ax1.set_ylabel('X')
ax1.set_title('Distribucion conjunta P(X, Y)', color=C_DARK, fontweight='bold')
plt.colorbar(im, ax=ax1, label='Probabilidad')

# Simulacion y scatter
np.random.seed(42)
n_sim = 5000
pares = []
probs = P_conjunta.flatten()
indices = [(i, j) for i in range(3) for j in range(3)]
elegidos = np.random.choice(len(indices), size=n_sim, p=probs)
X_sim = np.array([x_vals[indices[e][0]] for e in elegidos])
Y_sim = np.array([y_vals[indices[e][1]] for e in elegidos])

# Agregar jitter para visualizacion
jitter = 0.1
ax2.scatter(X_sim + np.random.uniform(-jitter, jitter, n_sim),
           Y_sim + np.random.uniform(-jitter, jitter, n_sim),
           alpha=0.15, s=10, color=C_PRIMARY)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title(f'Muestras simuladas (rho = {rho:.4f})', color=C_DARK, fontweight='bold')
ax2.set_xticks([0, 1, 2])
ax2.set_yticks([0, 1, 2])

plt.tight_layout()
plt.show()
```

**Ejercicio 19:** X ~ Poisson(lambda=3). Calcular los primeros 4 momentos: E[X], E[X^2], E[X^3], E[X^4].

**Resolucion:**

Para $X \sim \text{Poisson}(\lambda)$ con $\lambda = 3$, la PMF es:

$$P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}, \quad k = 0, 1, 2, \ldots$$

Donde:
- $X$: variable aleatoria con distribucion de Poisson. Modela el numero de eventos en un intervalo fijo.
- $k$: numero de eventos observados (entero no negativo).
- $\lambda$: parametro de la Poisson. Aqui $\lambda = 3$. Representa la tasa promedio de eventos.
- $e^{-\lambda}$: factor exponencial de normalizacion. $e \approx 2.71828$.
- $\lambda^k$: $\lambda$ elevado a la potencia $k$.
- $k!$: factorial de $k$, es decir $k! = k \times (k-1) \times \ldots \times 1$.

Los **momentos** $E[X^n]$ se pueden obtener usando la funcion generadora de momentos (MGF) de la Poisson:

$$M_X(t) = E[e^{tX}] = \exp\left(\lambda(e^t - 1)\right)$$

- $M_X(t)$: MGF de la Poisson evaluada en $t$.
- $\exp(\cdot)$: funcion exponencial.
- $\lambda(e^t - 1)$: argumento de la exponencial.
- $t$: parametro real de la MGF.

Los momentos se obtienen derivando la MGF y evaluando en $t = 0$:

$$E[X^n] = M_X^{(n)}(0)$$

- $M_X^{(n)}(0)$: $n$-esima derivada de la MGF evaluada en $t = 0$.

**Primer momento E[X]:**

$$E[X] = \lambda = 3$$

- $E[X]$: esperanza de $X$. Para la Poisson, coincide con $\lambda$.

**Segundo momento E[X^2]:**

$$E[X^2] = \lambda^2 + \lambda = 9 + 3 = 12$$

- $E[X^2]$: segundo momento. Se obtiene de $Var(X) + (E[X])^2 = \lambda + \lambda^2$.
- $Var(X) = \lambda = 3$: la varianza de la Poisson es igual a $\lambda$.

**Tercer momento E[X^3]:**

$$E[X^3] = \lambda^3 + 3\lambda^2 + \lambda = 27 + 27 + 3 = 57$$

- $E[X^3]$: tercer momento. Se puede derivar de la MGF o usando la relacion con momentos centrales.
- $\lambda^3 = 27$, $3\lambda^2 = 27$, $\lambda = 3$.

**Cuarto momento E[X^4]:**

$$E[X^4] = \lambda^4 + 6\lambda^3 + 7\lambda^2 + \lambda = 81 + 162 + 63 + 3 = 309$$

- $E[X^4]$: cuarto momento.
- $\lambda^4 = 81$, $6\lambda^3 = 162$, $7\lambda^2 = 63$, $\lambda = 3$.

```python
lam = 3

# Formulas teoricas para momentos de Poisson
E_X1_teorico = lam                                    # lambda
E_X2_teorico = lam**2 + lam                           # lambda^2 + lambda
E_X3_teorico = lam**3 + 3*lam**2 + lam               # lambda^3 + 3*lambda^2 + lambda
E_X4_teorico = lam**4 + 6*lam**3 + 7*lam**2 + lam    # lambda^4 + 6*lambda^3 + 7*lambda^2 + lambda

print(f"lambda = {lam}")
print(f"\n=== Momentos teoricos (formulas) ===")
print(f"E[X]   = lambda = {E_X1_teorico}")
print(f"E[X^2] = lambda^2 + lambda = {lam**2} + {lam} = {E_X2_teorico}")
print(f"E[X^3] = lambda^3 + 3*lambda^2 + lambda = {lam**3} + {3*lam**2} + {lam} = {E_X3_teorico}")
print(f"E[X^4] = lambda^4 + 6*lambda^3 + 7*lambda^2 + lambda = {lam**4} + {6*lam**3} + {7*lam**2} + {lam} = {E_X4_teorico}")

# Verificacion numerica directa (sumando la PMF)
K = 50  # truncar la suma en k=50 (suficiente para lambda=3)
k_vals = np.arange(0, K + 1)
pmf_vals = stats.poisson.pmf(k_vals, lam)

E_X1_num = np.sum(k_vals * pmf_vals)
E_X2_num = np.sum(k_vals**2 * pmf_vals)
E_X3_num = np.sum(k_vals**3 * pmf_vals)
E_X4_num = np.sum(k_vals**4 * pmf_vals)

print(f"\n=== Verificacion numerica (suma directa hasta k={K}) ===")
print(f"E[X]   = {E_X1_num:.4f} (teorico: {E_X1_teorico})")
print(f"E[X^2] = {E_X2_num:.4f} (teorico: {E_X2_teorico})")
print(f"E[X^3] = {E_X3_num:.4f} (teorico: {E_X3_teorico})")
print(f"E[X^4] = {E_X4_num:.4f} (teorico: {E_X4_teorico})")

# Momentos centrales
Var_X = E_X2_teorico - E_X1_teorico**2
skew = (E_X3_num - 3*E_X1_num*Var_X - E_X1_num**3) / Var_X**1.5
kurt = (E_X4_num - 4*E_X3_num*E_X1_num + 6*E_X2_num*E_X1_num**2 - 3*E_X1_num**4) / Var_X**2

print(f"\nVar(X) = E[X^2] - (E[X])^2 = {E_X2_teorico} - {E_X1_teorico**2} = {Var_X}")
print(f"Skewness (asimetria) = {skew:.4f}")
print(f"Kurtosis (exceso) = {kurt:.4f}")

# Grafica: PMF con contribuciones de cada momento
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# PMF
k_plot = np.arange(0, 15)
pmf_plot = stats.poisson.pmf(k_plot, lam)
ax1.bar(k_plot, pmf_plot, color=C_PRIMARY, edgecolor=C_DARK, alpha=0.85, label='P(X=k)')
ax1.axvline(x=E_X1_teorico, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {E_X1_teorico}')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title(f'Poisson(lambda={lam})', color=C_DARK, fontweight='bold')
ax1.legend()

# Momentos como barras
momentos = [E_X1_teorico, E_X2_teorico, E_X3_teorico, E_X4_teorico]
etiquetas = ['E[X]', 'E[X^2]', 'E[X^3]', 'E[X^4]']
colores = [C_PRIMARY, C_SUCCESS, C_ORANGE, C_DANGER]
bars = ax2.bar(etiquetas, momentos, color=colores, edgecolor=C_DARK, alpha=0.85)
for bar, val in zip(bars, momentos):
    ax2.text(bar.get_x() + bar.get_width()/2, val + 5, f'{val}',
            ha='center', fontweight='bold', fontsize=12, color=C_DARK)
ax2.set_ylabel('Valor del momento')
ax2.set_title(f'Primeros 4 momentos de Poisson(lambda={lam})', color=C_DARK, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Ejercicio 20:** Ley de los grandes numeros: Simular N lanzamientos de moneda y mostrar como la media converge a 0.5.

**Resolucion:**

La **Ley de los Grandes Numeros (LGN)** establece que la media muestral converge a la esperanza poblacional a medida que el tamano de muestra crece:

$$\bar{X}_N = \frac{1}{N} \sum_{i=1}^{N} X_i \xrightarrow{N \to \infty} E[X] = \mu$$

Donde:
- $\bar{X}_N$: media muestral de las primeras $N$ observaciones. Es una variable aleatoria que depende de la muestra.
- $N$: numero total de observaciones (tamano de muestra).
- $X_i$: la $i$-esima observacion. Cada $X_i$ es una variable aleatoria independiente con la misma distribucion (iid).
- $\sum_{i=1}^{N} X_i$: suma de las $N$ observaciones.
- $\frac{1}{N}$: factor de promediado.
- $E[X] = \mu$: esperanza de la distribucion comun de los $X_i$. Es el valor al que converge la media muestral.
- $\xrightarrow{N \to \infty}$: convergencia cuando $N$ tiende a infinito (en probabilidad para la LGN debil, casi seguramente para la LGN fuerte).

**Aplicacion: lanzamiento de moneda justa**

Cada lanzamiento $X_i \sim \text{Bernoulli}(p = 0.5)$:

- $X_i = 1$ si sale cara, $X_i = 0$ si sale cruz.
- $P(X_i = 1) = p = 0.5$: probabilidad de cara.
- $P(X_i = 0) = 1 - p = 0.5$: probabilidad de cruz.
- $E[X_i] = p = 0.5$: esperanza de cada lanzamiento.
- $Var(X_i) = p(1-p) = 0.25$: varianza de cada lanzamiento.

La media muestral $\bar{X}_N$ representa la proporcion de caras en $N$ lanzamientos:

$$\bar{X}_N = \frac{\text{numero de caras}}{N}$$

**Propiedades de la media muestral:**

$$E[\bar{X}_N] = \mu = 0.5$$

- $E[\bar{X}_N]$: esperanza de la media muestral. Es igual a $\mu$ para cualquier $N$ (la media muestral es un estimador insesgado).

$$Var(\bar{X}_N) = \frac{Var(X)}{N} = \frac{p(1-p)}{N} = \frac{0.25}{N}$$

- $Var(\bar{X}_N)$: varianza de la media muestral. Decrece con $N$.
- $\frac{Var(X)}{N}$: la varianza se divide entre $N$ porque promediamos $N$ variables independientes.
- A medida que $N$ crece, $Var(\bar{X}_N) \to 0$, lo que significa que $\bar{X}_N$ se concentra cada vez mas alrededor de $\mu$.

```python
np.random.seed(42)

p = 0.5
N_max = 10_000

# Simular N_max lanzamientos
lanzamientos = np.random.binomial(1, p, size=N_max)

# Media acumulada: media muestral despues de cada lanzamiento
n_vals = np.arange(1, N_max + 1)
media_acumulada = np.cumsum(lanzamientos) / n_vals

# Banda teorica de +/- 2*std de la media muestral
std_media = np.sqrt(p * (1 - p) / n_vals)
banda_sup = p + 2 * std_media
banda_inf = p - 2 * std_media

print(f"Moneda justa: p = {p}")
print(f"E[X] = p = {p}")
print(f"Var(X) = p(1-p) = {p*(1-p)}")
print(f"\nMedia muestral en distintos N:")
for n in [10, 100, 1000, 5000, 10000]:
    media_n = media_acumulada[n - 1]
    var_media_n = p * (1 - p) / n
    print(f"  N={n:>5}: media = {media_n:.4f}, Var(media) = {var_media_n:.6f}, +/- 2*std = {2*np.sqrt(var_media_n):.4f}")

# Grafica: convergencia de la media muestral
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Panel 1: media acumulada con bandas
ax1.plot(n_vals, media_acumulada, color=C_PRIMARY, linewidth=0.8, alpha=0.8, label='Media muestral')
ax1.axhline(y=p, color=C_DANGER, linewidth=2, linestyle='--', label=f'E[X] = {p}')
ax1.fill_between(n_vals, banda_inf, banda_sup, alpha=0.15, color=C_ORANGE,
                 label='+/- 2*std de la media')
ax1.set_xlabel('N (numero de lanzamientos)')
ax1.set_ylabel('Media muestral')
ax1.set_title('Ley de los Grandes Numeros: convergencia de la media muestral', color=C_DARK, fontweight='bold')
ax1.set_xscale('log')
ax1.set_xlim(1, N_max)
ax1.legend(loc='upper right')

# Panel 2: error absoluto |media - 0.5|
error = np.abs(media_acumulada - p)
ax2.plot(n_vals, error, color=C_DANGER, linewidth=0.8, alpha=0.7, label='|media - 0.5|')
ax2.plot(n_vals, 2 * std_media, color=C_ORANGE, linewidth=2, linestyle='--',
         label='2*sqrt(0.25/N) (cota teorica)')
ax2.set_xlabel('N (numero de lanzamientos)')
ax2.set_ylabel('Error absoluto')
ax2.set_title('Error |media muestral - E[X]| decrece con N', color=C_DARK, fontweight='bold')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlim(1, N_max)
ax2.legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 22: Covarianza bajo transformaciones lineales

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sean $X$ e $Y$ variables aleatorias con $\text{Cov}(X, Y) = 5$. Calcule $\text{Cov}(2X + 1,\; 3Y - 2)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La covarianza tiene la propiedad de **bilinealidad**. Esto significa que las constantes multiplicativas salen como factores y las constantes aditivas desaparecen:

$$\text{Cov}(aX + b,\; cY + d) = ac \cdot \text{Cov}(X, Y)$$

Donde:
- $a = 2$: coeficiente que multiplica a $X$.
- $b = 1$: constante sumada a $X$; no afecta la covarianza porque desplazar una variable no cambia como covarĂ­an.
- $c = 3$: coeficiente que multiplica a $Y$.
- $d = -2$: constante sumada a $Y$; tampoco afecta la covarianza.
- $\text{Cov}(X, Y) = 5$: dato del problema.

Sustituyendo:

$$\text{Cov}(2X + 1,\; 3Y - 2) = (2)(3) \cdot \text{Cov}(X, Y) = 6 \cdot 5 = 30$$

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(22)

# Generar X, Y con Cov(X,Y) = 5
mean = [0, 0]
cov_matrix = [[4, 5], [5, 9]]  # Var(X)=4, Var(Y)=9, Cov=5
samples = np.random.multivariate_normal(mean, cov_matrix, 50_000)
X, Y = samples[:, 0], samples[:, 1]

U = 2*X + 1
V = 3*Y - 2

cov_original = np.cov(X, Y)[0, 1]
cov_transformada = np.cov(U, V)[0, 1]

print(f"Cov(X, Y) estimada:          {cov_original:.4f}  (teorica: 5)")
print(f"Cov(2X+1, 3Y-2) estimada:    {cov_transformada:.4f}  (teorica: 30)")
print(f"Razon estimada:               {cov_transformada / cov_original:.4f}  (teorica: 6)")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(X[:2000], Y[:2000], alpha=0.3, s=8, color='#3498db')
axes[0].set_title('Variables originales (X, Y)', fontsize=13, color='#2c3e50')
axes[0].set_xlabel('X'); axes[0].set_ylabel('Y')

axes[1].scatter(U[:2000], V[:2000], alpha=0.3, s=8, color='#e74c3c')
axes[1].set_title('Transformadas (2X+1, 3Y-2)', fontsize=13, color='#2c3e50')
axes[1].set_xlabel('2X+1'); axes[1].set_ylabel('3Y-2')

plt.tight_layout()
plt.show()
```

## Ejercicio 23: Esperanza del maximo de dos uniformes independientes

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sean $X$ e $Y$ variables aleatorias independientes con distribucion $U(0,1)$. Calcule $E[\max(X, Y)]$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Definimos $M = \max(X, Y)$. La CDF de $M$ es:

$$F_M(m) = P(M \le m) = P(X \le m,\; Y \le m) = P(X \le m) \cdot P(Y \le m)$$

Donde la factorizacion se debe a la **independencia** de $X$ e $Y$.

- $P(X \le m) = m$ para $m \in [0,1]$: CDF de la uniforme.
- $P(Y \le m) = m$: idem.

Entonces $F_M(m) = m^2$ y la densidad es $f_M(m) = 2m$.

Calculamos la esperanza:

$$E[M] = \int_0^1 m \cdot f_M(m)\, dm = \int_0^1 m \cdot 2m\, dm = 2\int_0^1 m^2\, dm = 2 \cdot \frac{1}{3} = \frac{2}{3}$$

Donde:
- $m$: valor que toma el maximo.
- $f_M(m) = 2m$: densidad del maximo de dos uniformes.
- $\frac{2}{3} \approx 0.6667$: resultado final.

</div>

```python
np.random.seed(23)

n_sim = 100_000
X = np.random.uniform(0, 1, n_sim)
Y = np.random.uniform(0, 1, n_sim)
M = np.maximum(X, Y)

print(f"E[max(X,Y)] simulado:  {M.mean():.4f}")
print(f"E[max(X,Y)] teorico:   {2/3:.4f}")

fig, ax = plt.subplots(figsize=(8, 5))
m_vals = np.linspace(0, 1, 200)
ax.hist(M, bins=60, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='Simulacion')
ax.plot(m_vals, 2*m_vals, color='#e74c3c', lw=2.5, label=r'$f_M(m) = 2m$ (teorica)')
ax.axvline(2/3, color='#2c3e50', ls='--', lw=1.5, label=f'E[M] = 2/3')
ax.set_xlabel('m', fontsize=12)
ax.set_ylabel('Densidad', fontsize=12)
ax.set_title('Distribucion de max(X, Y) con X, Y ~ U(0,1)', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 24: Esperanza y varianza de $Y = X^2$ con $X \sim N(0,1)$

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X \sim N(0,1)$. Defina $Y = X^2$. Calcule $E[Y]$ y $\text{Var}(Y)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

$Y = X^2$ sigue una distribucion **chi-cuadrado** con 1 grado de libertad: $Y \sim \chi^2(1)$.

**Esperanza:**

$$E[Y] = E[X^2] = \text{Var}(X) + (E[X])^2 = 1 + 0 = 1$$

Donde:
- $E[X] = 0$: media de la normal estandar.
- $\text{Var}(X) = 1$: varianza de la normal estandar.
- $E[X^2] = \text{Var}(X) + (E[X])^2$: identidad fundamental que relaciona el segundo momento con varianza y media.

**Varianza:**

$$\text{Var}(Y) = E[Y^2] - (E[Y])^2 = E[X^4] - (E[X^2])^2$$

Para $X \sim N(0,1)$, el cuarto momento es $E[X^4] = 3$ (conocido como **curtosis** de la normal). Entonces:

$$\text{Var}(Y) = 3 - 1^2 = 2$$

Donde:
- $E[X^4] = 3$: cuarto momento de la normal estandar (se obtiene por la funcion generadora de momentos o integracion directa).
- $(E[X^2])^2 = 1$: cuadrado del segundo momento.

</div>

```python
from scipy import stats

np.random.seed(24)

X = np.random.standard_normal(200_000)
Y = X**2

print(f"E[Y] = E[X^2] simulado:  {Y.mean():.4f}  (teorico: 1)")
print(f"Var(Y) simulada:          {Y.var():.4f}  (teorica: 2)")
print(f"E[X^4] simulado:          {np.mean(X**4):.4f}  (teorico: 3)")

fig, ax = plt.subplots(figsize=(8, 5))
y_vals = np.linspace(0, 8, 300)
ax.hist(Y, bins=80, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='Simulacion $X^2$')
ax.plot(y_vals, stats.chi2.pdf(y_vals, df=1), color='#e74c3c', lw=2.5, label=r'$\chi^2(1)$ teorica')
ax.axvline(1, color='#2c3e50', ls='--', lw=1.5, label='E[Y] = 1')
ax.set_xlim(0, 8)
ax.set_xlabel('y', fontsize=12)
ax.set_ylabel('Densidad', fontsize=12)
ax.set_title(r'$Y = X^2$ con $X \sim N(0,1)$', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 25: Varianza condicional en la normal bivariada

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $(X, Y)$ un vector normal bivariado con medias $\mu_X, \mu_Y$, varianzas $\sigma_X^2, \sigma_Y^2$ y correlacion $\rho$. Determine $\text{Var}(Y | X = x)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Para la distribucion normal bivariada, la distribucion condicional $Y | X = x$ es tambien normal:

$$Y | X = x \;\sim\; N\!\left(\mu_Y + \rho \frac{\sigma_Y}{\sigma_X}(x - \mu_X),\;\; \sigma_Y^2(1 - \rho^2)\right)$$

Por lo tanto:

$$\text{Var}(Y | X = x) = \sigma_Y^2(1 - \rho^2)$$

Donde:
- $\sigma_Y^2$: varianza marginal de $Y$.
- $\rho$: coeficiente de correlacion entre $X$ e $Y$, con $\rho \in [-1, 1]$.
- $1 - \rho^2$: factor de reduccion de varianza. Cuando $|\rho| = 1$, la varianza condicional es 0 (determinismo perfecto). Cuando $\rho = 0$, no hay reduccion.
- La varianza condicional **no depende de $x$**: es constante para todos los valores de $X$. Esta propiedad es exclusiva de la normal bivariada (homocedasticidad condicional).

</div>

```python
np.random.seed(25)

mu_X, mu_Y = 5, 10
sigma_X, sigma_Y = 2, 3
rho = 0.7

cov_XY = rho * sigma_X * sigma_Y
cov_matrix = [[sigma_X**2, cov_XY], [cov_XY, sigma_Y**2]]
samples = np.random.multivariate_normal([mu_X, mu_Y], cov_matrix, 100_000)
X, Y = samples[:, 0], samples[:, 1]

var_teorica = sigma_Y**2 * (1 - rho**2)

# Varianza condicional para diferentes valores de x
x_slices = [3, 5, 7]
fig, axes = plt.subplots(1, len(x_slices), figsize=(14, 4), sharey=True)

for i, x0 in enumerate(x_slices):
    mask = np.abs(X - x0) < 0.2
    Y_cond = Y[mask]
    var_emp = Y_cond.var()
    axes[i].hist(Y_cond, bins=40, density=True, alpha=0.6, color='#3498db', edgecolor='white')
    axes[i].set_title(f'Y | X={x0}\nVar emp={var_emp:.2f}', fontsize=11, color='#2c3e50')
    axes[i].set_xlabel('Y')

axes[0].set_ylabel('Densidad')
fig.suptitle(f'Var(Y|X=x) teorica = {var_teorica:.2f} (constante para todo x)', fontsize=13, color='#e74c3c', y=1.02)
plt.tight_layout()
plt.show()

print(f"Var(Y|X=x) teorica: sigma_Y^2 * (1 - rho^2) = {sigma_Y}^2 * (1 - {rho}^2) = {var_teorica:.2f}")
```

## Ejercicio 26: Esperanza condicional con tabla conjunta discreta

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sean $X \in \{1, 2, 3\}$ e $Y \in \{0, 1, 2\}$ con la siguiente tabla de probabilidad conjunta:

| | Y=0 | Y=1 | Y=2 |
|---|---|---|---|
| X=1 | 0.05 | 0.10 | 0.05 |
| X=2 | 0.10 | 0.20 | 0.10 |
| X=3 | 0.05 | 0.15 | 0.20 |

Calcule $E[X | Y = 1]$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Usamos la definicion de esperanza condicional discreta:

$$E[X | Y = y] = \sum_x x \cdot P(X = x | Y = y) = \sum_x x \cdot \frac{P(X = x, Y = y)}{P(Y = y)}$$

**Paso 1:** Calcular la probabilidad marginal $P(Y = 1)$:

$$P(Y = 1) = P(X=1, Y=1) + P(X=2, Y=1) + P(X=3, Y=1) = 0.10 + 0.20 + 0.15 = 0.45$$

**Paso 2:** Calcular las probabilidades condicionales $P(X = x | Y = 1)$:

- $P(X=1 | Y=1) = \frac{0.10}{0.45} = \frac{2}{9}$
- $P(X=2 | Y=1) = \frac{0.20}{0.45} = \frac{4}{9}$
- $P(X=3 | Y=1) = \frac{0.15}{0.45} = \frac{3}{9} = \frac{1}{3}$

**Paso 3:** Calcular la esperanza condicional:

$$E[X | Y=1] = 1 \cdot \frac{2}{9} + 2 \cdot \frac{4}{9} + 3 \cdot \frac{1}{3} = \frac{2}{9} + \frac{8}{9} + \frac{9}{9} = \frac{19}{9} \approx 2.111$$

</div>

## Ejercicio 27: Momentos compuestos con Poisson y Binomial

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $N \sim \text{Poi}(10)$ y, dado $N$, sea $X | N \sim \text{Bin}(N, 0.3)$. Calcule $E[X]$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Aplicamos la **ley de la esperanza total** (o ley de la torre):

$$E[X] = E\!\big[E[X | N]\big]$$

**Paso 1:** Calcular $E[X | N]$. Dado $N = n$, $X \sim \text{Bin}(n, 0.3)$, entonces:

$$E[X | N] = N \cdot p = 0.3N$$

Donde:
- $N$: numero de ensayos (aleatorio, sigue una Poisson).
- $p = 0.3$: probabilidad de exito en cada ensayo.

**Paso 2:** Tomar la esperanza sobre $N$:

$$E[X] = E[0.3N] = 0.3 \cdot E[N] = 0.3 \cdot 10 = 3$$

Donde:
- $E[N] = \lambda = 10$: esperanza de la Poisson con parametro $\lambda = 10$.

Este modelo se conoce como **distribucion compuesta**. $X$ marginalmente sigue una Poisson con parametro $\lambda p = 3$.

</div>

```python
np.random.seed(27)

n_sim = 200_000
lam, p = 10, 0.3

N = np.random.poisson(lam, n_sim)
X = np.array([np.random.binomial(n, p) for n in N])

print(f"E[X] simulado:  {X.mean():.4f}  (teorico: {lam*p})")
print(f"Var(X) simulada: {X.var():.4f}  (teorica lambda*p = {lam*p})")

fig, ax = plt.subplots(figsize=(8, 5))
x_range = np.arange(0, 15)
ax.hist(X, bins=np.arange(-0.5, 15.5, 1), density=True, alpha=0.6, color='#3498db',
        edgecolor='white', label='Simulacion compuesta')
ax.plot(x_range, stats.poisson.pmf(x_range, lam*p), 'o-', color='#e74c3c',
        lw=2, markersize=6, label=f'Poi({lam*p}) teorica')
ax.axvline(lam*p, color='#2c3e50', ls='--', lw=1.5, label=f'E[X] = {lam*p}')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('Probabilidad', fontsize=12)
ax.set_title('Distribucion compuesta: N~Poi(10), X|N~Bin(N,0.3)', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 28: Desigualdad de Markov con exponencial

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X \sim \text{Exp}(\lambda = 0.5)$. Verifique la desigualdad de Markov para $a = 10$: $P(X \ge 10) \le \frac{E[X]}{10}$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **desigualdad de Markov** establece que para una variable aleatoria no negativa $X$ y $a > 0$:

$$P(X \ge a) \le \frac{E[X]}{a}$$

Para $X \sim \text{Exp}(\lambda = 0.5)$:
- $E[X] = \frac{1}{\lambda} = \frac{1}{0.5} = 2$: esperanza de la exponencial.
- $\lambda = 0.5$: tasa del proceso (inverso del tiempo medio).

**Cota de Markov:**

$$P(X \ge 10) \le \frac{E[X]}{10} = \frac{2}{10} = 0.2$$

**Valor exacto:**

$$P(X \ge 10) = e^{-\lambda \cdot 10} = e^{-5} \approx 0.00674$$

La cota de Markov ($0.2$) es valida pero muy holgada comparada con el valor real ($0.00674$). La desigualdad es general y no usa informacion sobre la forma de la distribucion, por eso tiende a ser conservadora.

</div>

```python
np.random.seed(28)

lam = 0.5
a_vals = np.linspace(0.5, 15, 200)
E_X = 1 / lam

# Cota de Markov vs probabilidad exacta
markov_bound = E_X / a_vals
exact_prob = np.exp(-lam * a_vals)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(a_vals, markov_bound, color='#e74c3c', lw=2.5, label=r'Cota Markov: $E[X]/a$')
ax.plot(a_vals, exact_prob, color='#3498db', lw=2.5, label=r'$P(X \geq a) = e^{-\lambda a}$')
ax.axvline(10, color='#2c3e50', ls='--', alpha=0.5)
ax.annotate(f'a=10\nMarkov: {E_X/10:.3f}\nExacto: {np.exp(-5):.5f}',
            xy=(10, 0.2), xytext=(11.5, 0.5), fontsize=10,
            arrowprops=dict(arrowstyle='->', color='#2c3e50'),
            bbox=dict(boxstyle='round', fc='white', ec='#2c3e50'))
ax.set_xlabel('a', fontsize=12)
ax.set_ylabel('Probabilidad / Cota', fontsize=12)
ax.set_title('Desigualdad de Markov vs probabilidad exacta (Exp(0.5))', fontsize=13, color='#2c3e50')
ax.legend(fontsize=11)
ax.set_ylim(0, 1.5)
plt.tight_layout()
plt.show()

print(f"Cota de Markov en a=10: {E_X/10:.4f}")
print(f"Valor exacto P(X>=10): {np.exp(-5):.6f}")
print(f"La cota se cumple: {np.exp(-5) <= E_X/10}")
```

## Ejercicio 29: Ley de los Grandes Numeros simulada con dados

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Simule el lanzamiento de $n$ dados y muestre que la media muestral converge a $E[X] = 3.5$ cuando $n \to \infty$ (Ley de los Grandes Numeros).

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **Ley de los Grandes Numeros (LGN)** establece que si $X_1, X_2, \ldots, X_n$ son variables aleatorias i.i.d. con $E[X_i] = \mu$, entonces:

$$\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i \xrightarrow{P} \mu \quad \text{cuando } n \to \infty$$

Donde:
- $\bar{X}_n$: media muestral de $n$ observaciones.
- $\xrightarrow{P}$: convergencia en probabilidad.
- $\mu = E[X_i]$: esperanza poblacional.
- i.i.d.: independientes e identicamente distribuidas.

Para un dado justo, $X_i \in \{1,2,3,4,5,6\}$ con probabilidad $1/6$ cada uno:

$$\mu = E[X] = \frac{1+2+3+4+5+6}{6} = 3.5$$

La LGN garantiza que $\bar{X}_n \to 3.5$ conforme $n$ crece, independientemente de la realizacion particular.

</div>

```python
np.random.seed(29)

n_max = 10_000
dados = np.random.randint(1, 7, n_max)
medias_acumuladas = np.cumsum(dados) / np.arange(1, n_max + 1)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(range(1, n_max + 1), medias_acumuladas, color='#3498db', lw=1, alpha=0.8, label=r'$\bar{X}_n$')
ax.axhline(3.5, color='#e74c3c', ls='--', lw=2, label=r'$\mu = 3.5$')
ax.fill_between(range(1, n_max + 1), 3.5 - 0.1, 3.5 + 0.1, alpha=0.15, color='#2ecc71')
ax.set_xscale('log')
ax.set_xlabel('Numero de lanzamientos (n)', fontsize=12)
ax.set_ylabel(r'Media muestral $\bar{X}_n$', fontsize=12)
ax.set_title('Ley de los Grandes Numeros: media de n dados', fontsize=13, color='#2c3e50')
ax.legend(fontsize=11)
ax.set_ylim(1, 6)
plt.tight_layout()
plt.show()

for n in [10, 100, 1000, 10000]:
    print(f"n = {n:>5}: media = {medias_acumuladas[n-1]:.4f}")
```

## Ejercicio 30: Teorema Central del Limite con uniformes

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sean $X_1, \ldots, X_{30}$ i.i.d. $\sim U(0,1)$. Verifique con un histograma que $S_{30} = \sum_{i=1}^{30} X_i$ se aproxima a una $N(15, 2.5)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

El **Teorema Central del Limite (TCL)** establece que si $X_1, \ldots, X_n$ son i.i.d. con media $\mu$ y varianza $\sigma^2$, entonces:

$$S_n = \sum_{i=1}^n X_i \;\xrightarrow{d}\; N(n\mu,\; n\sigma^2) \quad \text{cuando } n \to \infty$$

Para $X_i \sim U(0,1)$:
- $\mu = E[X_i] = \frac{1}{2}$: media de la uniforme en $(0,1)$.
- $\sigma^2 = \text{Var}(X_i) = \frac{1}{12}$: varianza de la uniforme en $(0,1)$.
- $n = 30$: numero de variables sumadas.

Parametros de la normal aproximante:
- $E[S_{30}] = n\mu = 30 \cdot \frac{1}{2} = 15$
- $\text{Var}(S_{30}) = n\sigma^2 = 30 \cdot \frac{1}{12} = 2.5$
- $\text{DE}(S_{30}) = \sqrt{2.5} \approx 1.581$

Por lo tanto $S_{30} \approx N(15, 2.5)$ donde $2.5$ es la varianza.

</div>

```python
np.random.seed(30)

n, n_sim = 30, 100_000
sumas = np.sum(np.random.uniform(0, 1, (n_sim, n)), axis=1)

mu_S = n * 0.5
var_S = n * (1/12)

fig, ax = plt.subplots(figsize=(8, 5))
x_vals = np.linspace(mu_S - 4*np.sqrt(var_S), mu_S + 4*np.sqrt(var_S), 300)
ax.hist(sumas, bins=70, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='Simulacion $S_{30}$')
ax.plot(x_vals, stats.norm.pdf(x_vals, mu_S, np.sqrt(var_S)), color='#e74c3c', lw=2.5,
        label=f'N({mu_S}, {var_S})')
ax.axvline(mu_S, color='#2c3e50', ls='--', lw=1.5, label=f'E[S] = {mu_S}')
ax.set_xlabel('$S_{30}$', fontsize=12)
ax.set_ylabel('Densidad', fontsize=12)
ax.set_title('TCL: Suma de 30 uniformes U(0,1)', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()

print(f"Media simulada:    {sumas.mean():.4f}  (teorica: {mu_S})")
print(f"Varianza simulada: {sumas.var():.4f}  (teorica: {var_S:.4f})")
```

## Ejercicio 31: Funcion generadora de momentos de la Poisson

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X \sim \text{Poi}(\lambda)$. Demuestre que la MGF es $M_X(t) = \exp\!\big(\lambda(e^t - 1)\big)$ y use las derivadas para obtener $E[X]$ y $\text{Var}(X)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **funcion generadora de momentos (MGF)** se define como $M_X(t) = E[e^{tX}]$.

**Derivacion de la MGF:**

$$M_X(t) = \sum_{k=0}^{\infty} e^{tk} \cdot \frac{\lambda^k e^{-\lambda}}{k!} = e^{-\lambda} \sum_{k=0}^{\infty} \frac{(\lambda e^t)^k}{k!} = e^{-\lambda} \cdot e^{\lambda e^t} = \exp\!\big(\lambda(e^t - 1)\big)$$

Donde se uso que $\sum_{k=0}^{\infty} \frac{a^k}{k!} = e^a$ (serie de Taylor de la exponencial).

**Primera derivada** (para obtener $E[X]$):

$$M'_X(t) = \lambda e^t \cdot \exp\!\big(\lambda(e^t - 1)\big)$$

$$E[X] = M'_X(0) = \lambda \cdot e^0 \cdot \exp(0) = \lambda$$

**Segunda derivada** (para obtener $E[X^2]$):

$$M''_X(t) = (\lambda e^t + \lambda^2 e^{2t}) \cdot \exp\!\big(\lambda(e^t - 1)\big)$$

$$E[X^2] = M''_X(0) = \lambda + \lambda^2$$

**Varianza:**

$$\text{Var}(X) = E[X^2] - (E[X])^2 = (\lambda + \lambda^2) - \lambda^2 = \lambda$$

Resultado: tanto $E[X]$ como $\text{Var}(X)$ son iguales a $\lambda$, propiedad caracteristica de la Poisson.

</div>

```python
np.random.seed(31)

lam = 7
t_vals = np.linspace(-1, 1.5, 200)

# MGF teorica
mgf_teorica = np.exp(lam * (np.exp(t_vals) - 1))

# MGF empirica via simulacion
X = np.random.poisson(lam, 100_000)
mgf_empirica = [np.mean(np.exp(t * X)) for t in t_vals]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(t_vals, mgf_teorica, color='#e74c3c', lw=2.5, label='MGF teorica')
axes[0].plot(t_vals, mgf_empirica, '--', color='#3498db', lw=2, label='MGF empirica')
axes[0].set_xlabel('t', fontsize=12)
axes[0].set_ylabel('$M_X(t)$', fontsize=12)
axes[0].set_title(f'MGF de Poisson($\\lambda$={lam})', fontsize=13, color='#2c3e50')
axes[0].legend()

# Verificacion numerica
lambdas = [1, 3, 5, 10, 20]
medias, varianzas = [], []
for l in lambdas:
    s = np.random.poisson(l, 100_000)
    medias.append(s.mean())
    varianzas.append(s.var())

axes[1].plot(lambdas, medias, 'o-', color='#3498db', lw=2, markersize=8, label='E[X] simulada')
axes[1].plot(lambdas, varianzas, 's-', color='#e74c3c', lw=2, markersize=8, label='Var(X) simulada')
axes[1].plot(lambdas, lambdas, '--', color='#2c3e50', lw=1.5, label=r'$y = \lambda$ (teorica)')
axes[1].set_xlabel(r'$\lambda$', fontsize=12)
axes[1].set_ylabel('Valor', fontsize=12)
axes[1].set_title(r'E[X] = Var(X) = $\lambda$', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 32: Esperanza y varianza de una mezcla de normales

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X$ una mezcla: con probabilidad $0.3$ sigue $N(10, 4)$ y con probabilidad $0.7$ sigue $N(20, 9)$. Calcule $E[X]$ y $\text{Var}(X)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Sea $Z \in \{1, 2\}$ la variable indicadora de componente, con $P(Z=1) = 0.3$ y $P(Z=2) = 0.7$.

- $X | Z=1 \sim N(\mu_1, \sigma_1^2) = N(10, 4)$
- $X | Z=2 \sim N(\mu_2, \sigma_2^2) = N(20, 9)$

**Esperanza** (ley de esperanza total):

$$E[X] = P(Z=1) \cdot E[X|Z=1] + P(Z=2) \cdot E[X|Z=2] = 0.3 \cdot 10 + 0.7 \cdot 20 = 3 + 14 = 17$$

**Varianza** (ley de varianza total): $\text{Var}(X) = E[\text{Var}(X|Z)] + \text{Var}(E[X|Z])$

Primer termino (esperanza de la varianza condicional):

$$E[\text{Var}(X|Z)] = 0.3 \cdot 4 + 0.7 \cdot 9 = 1.2 + 6.3 = 7.5$$

Segundo termino (varianza de la esperanza condicional):

$$\text{Var}(E[X|Z]) = E[(E[X|Z])^2] - (E[E[X|Z]])^2 = (0.3 \cdot 100 + 0.7 \cdot 400) - 17^2 = 310 - 289 = 21$$

Donde:
- $E[(E[X|Z])^2] = 0.3 \cdot 10^2 + 0.7 \cdot 20^2 = 310$
- $(E[X])^2 = 17^2 = 289$

$$\text{Var}(X) = 7.5 + 21 = 28.5$$

</div>

```python
np.random.seed(32)

n_sim = 200_000
p1, mu1, var1 = 0.3, 10, 4
p2, mu2, var2 = 0.7, 20, 9

Z = np.random.choice([1, 2], size=n_sim, p=[p1, p2])
X = np.where(Z == 1,
             np.random.normal(mu1, np.sqrt(var1), n_sim),
             np.random.normal(mu2, np.sqrt(var2), n_sim))

e_teorica = p1*mu1 + p2*mu2
var_teorica = (p1*var1 + p2*var2) + (p1*mu1**2 + p2*mu2**2 - e_teorica**2)

print(f"E[X] simulada:    {X.mean():.4f}  (teorica: {e_teorica})")
print(f"Var(X) simulada:  {X.var():.4f}  (teorica: {var_teorica})")

fig, ax = plt.subplots(figsize=(9, 5))
x_vals = np.linspace(-2, 32, 500)
pdf_mezcla = p1 * stats.norm.pdf(x_vals, mu1, np.sqrt(var1)) + \
             p2 * stats.norm.pdf(x_vals, mu2, np.sqrt(var2))

ax.hist(X, bins=80, density=True, alpha=0.5, color='#3498db', edgecolor='white', label='Simulacion')
ax.plot(x_vals, pdf_mezcla, color='#e74c3c', lw=2.5, label='Densidad mezcla teorica')
ax.plot(x_vals, p1*stats.norm.pdf(x_vals, mu1, np.sqrt(var1)), '--', color='#2ecc71', lw=1.5, label=f'0.3 N(10,4)')
ax.plot(x_vals, p2*stats.norm.pdf(x_vals, mu2, np.sqrt(var2)), '--', color='#2c3e50', lw=1.5, label=f'0.7 N(20,9)')
ax.axvline(e_teorica, color='#e74c3c', ls=':', lw=2, label=f'E[X] = {e_teorica}')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('Densidad', fontsize=12)
ax.set_title('Mezcla de normales', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 33: Ley de la varianza total con ejemplo discreto

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $Y \in \{1, 2\}$ con $P(Y=1) = 0.4$ y $P(Y=2) = 0.6$. Dado $Y$, $X|Y=1 \sim N(5, 2)$ y $X|Y=2 \sim N(8, 3)$. Verifique la ley de la varianza total: $\text{Var}(X) = E[\text{Var}(X|Y)] + \text{Var}(E[X|Y])$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **ley de la varianza total** (o formula de Eve) establece:

$$\text{Var}(X) = \underbrace{E[\text{Var}(X|Y)]}_{\text{variabilidad dentro}} + \underbrace{\text{Var}(E[X|Y])}_{\text{variabilidad entre}}$$

**Primer termino** -- $E[\text{Var}(X|Y)]$ (promedio de las varianzas condicionales):

$$E[\text{Var}(X|Y)] = P(Y=1) \cdot \text{Var}(X|Y=1) + P(Y=2) \cdot \text{Var}(X|Y=2)$$
$$= 0.4 \cdot 2 + 0.6 \cdot 3 = 0.8 + 1.8 = 2.6$$

**Segundo termino** -- $\text{Var}(E[X|Y])$ (varianza de las medias condicionales):

$$E[X|Y=1] = 5, \quad E[X|Y=2] = 8$$
$$E[E[X|Y]] = 0.4 \cdot 5 + 0.6 \cdot 8 = 6.8$$
$$E[(E[X|Y])^2] = 0.4 \cdot 25 + 0.6 \cdot 64 = 10 + 38.4 = 48.4$$
$$\text{Var}(E[X|Y]) = 48.4 - 6.8^2 = 48.4 - 46.24 = 2.16$$

**Total:**

$$\text{Var}(X) = 2.6 + 2.16 = 4.76$$

</div>

```python
np.random.seed(33)

n_sim = 200_000
p1, mu1, var1 = 0.4, 5, 2
p2, mu2, var2 = 0.6, 8, 3

Y = np.random.choice([1, 2], size=n_sim, p=[p1, p2])
X = np.where(Y == 1,
             np.random.normal(mu1, np.sqrt(var1), n_sim),
             np.random.normal(mu2, np.sqrt(var2), n_sim))

e_var = p1*var1 + p2*var2           # E[Var(X|Y)]
var_e = p1*mu1**2 + p2*mu2**2 - (p1*mu1 + p2*mu2)**2  # Var(E[X|Y])
var_total = e_var + var_e

print(f"E[Var(X|Y)]   = {e_var:.2f}")
print(f"Var(E[X|Y])   = {var_e:.2f}")
print(f"Var total teo  = {var_total:.2f}")
print(f"Var(X) simulada = {X.var():.4f}")

fig, ax = plt.subplots(figsize=(7, 5))
componentes = ['E[Var(X|Y)]\n(dentro)', 'Var(E[X|Y])\n(entre)', 'Var(X)\n(total)']
valores = [e_var, var_e, var_total]
colores = ['#3498db', '#e74c3c', '#2c3e50']
bars = ax.bar(componentes, valores, color=colores, edgecolor='white', width=0.5)
for bar, val in zip(bars, valores):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
            f'{val:.2f}', ha='center', fontsize=12, color='#2c3e50')
ax.set_ylabel('Varianza', fontsize=12)
ax.set_title('Descomposicion de la varianza total', fontsize=13, color='#2c3e50')
plt.tight_layout()
plt.show()
```

## Ejercicio 34: Esperanza condicional truncada $E[X | X > 0]$ para $X \sim N(0,1)$

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X \sim N(0,1)$. Calcule $E[X | X > 0]$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La densidad condicional de $X$ dado $X > 0$ es:

$$f_{X|X>0}(x) = \frac{f_X(x)}{P(X > 0)} = \frac{\phi(x)}{1/2} = 2\phi(x), \quad x > 0$$

Donde:
- $\phi(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$: densidad de la normal estandar.
- $P(X > 0) = 1/2$: por simetria de la normal centrada en 0.

La esperanza condicional es:

$$E[X | X > 0] = \int_0^{\infty} x \cdot 2\phi(x)\, dx = \frac{2}{\sqrt{2\pi}} \int_0^{\infty} x\, e^{-x^2/2}\, dx$$

Resolviendo la integral con la sustitucion $u = x^2/2$, $du = x\,dx$:

$$\int_0^{\infty} x\, e^{-x^2/2}\, dx = \left[-e^{-x^2/2}\right]_0^{\infty} = 0 - (-1) = 1$$

Por lo tanto:

$$E[X | X > 0] = \frac{2}{\sqrt{2\pi}} = \sqrt{\frac{2}{\pi}} \approx 0.7979$$

</div>

```python
np.random.seed(34)

X = np.random.standard_normal(500_000)
X_pos = X[X > 0]

teorico = np.sqrt(2 / np.pi)
print(f"E[X | X > 0] simulado:  {X_pos.mean():.4f}")
print(f"E[X | X > 0] teorico:   {teorico:.4f}")

fig, ax = plt.subplots(figsize=(8, 5))
x_vals = np.linspace(-4, 4, 500)
ax.plot(x_vals, stats.norm.pdf(x_vals), color='#3498db', lw=2, label=r'$\phi(x)$ normal estandar')
ax.fill_between(x_vals[x_vals > 0], stats.norm.pdf(x_vals[x_vals > 0]),
                alpha=0.3, color='#e74c3c', label=r'Region $X > 0$')
ax.axvline(teorico, color='#2c3e50', ls='--', lw=2,
           label=f'E[X|X>0] = $\\sqrt{{2/\\pi}}$ = {teorico:.4f}')
ax.axvline(0, color='gray', ls=':', lw=1)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('Densidad', fontsize=12)
ax.set_title('Esperanza condicional truncada E[X | X > 0]', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 35: Transformacion de Box-Muller

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sean $U_1, U_2 \sim U(0,1)$ independientes. Muestre que $Z_1 = \sqrt{-2\ln(U_1)} \cos(2\pi U_2)$ sigue una distribucion $N(0,1)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **transformacion de Box-Muller** convierte dos uniformes independientes en dos normales estandar independientes:

$$Z_1 = \sqrt{-2\ln(U_1)} \cos(2\pi U_2)$$
$$Z_2 = \sqrt{-2\ln(U_1)} \sin(2\pi U_2)$$

Donde:
- $U_1, U_2 \sim U(0,1)$: uniformes independientes (entrada del algoritmo).
- $R = \sqrt{-2\ln(U_1)}$: radio en coordenadas polares. Dado que $-\ln(U_1) \sim \text{Exp}(1)$, entonces $R^2 = -2\ln(U_1) \sim \chi^2(2)$.
- $\Theta = 2\pi U_2$: angulo uniformemente distribuido en $[0, 2\pi)$.
- $\cos(2\pi U_2)$ y $\sin(2\pi U_2)$: proyecciones ortogonales que generan independencia entre $Z_1$ y $Z_2$.

La idea geometrica es que un vector $(Z_1, Z_2)$ normal bivariado estandar, en coordenadas polares, tiene radio $R$ con $R^2 \sim \chi^2(2)$ y angulo $\Theta$ uniforme en $[0, 2\pi)$, ambos independientes.

</div>

```python
np.random.seed(35)

n_sim = 200_000
U1 = np.random.uniform(0, 1, n_sim)
U2 = np.random.uniform(0, 1, n_sim)

Z1 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)
Z2 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi * U2)

print(f"Z1: media = {Z1.mean():.4f}, var = {Z1.var():.4f}")
print(f"Z2: media = {Z2.mean():.4f}, var = {Z2.var():.4f}")
print(f"Corr(Z1, Z2) = {np.corrcoef(Z1, Z2)[0,1]:.4f}")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

x_vals = np.linspace(-4, 4, 300)
axes[0].hist(Z1, bins=80, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='Box-Muller $Z_1$')
axes[0].plot(x_vals, stats.norm.pdf(x_vals), color='#e74c3c', lw=2.5, label='N(0,1) teorica')
axes[0].set_title('$Z_1$ generada por Box-Muller', fontsize=13, color='#2c3e50')
axes[0].legend()
axes[0].set_xlabel('z')

axes[1].scatter(Z1[:3000], Z2[:3000], alpha=0.2, s=5, color='#3498db')
axes[1].set_xlabel('$Z_1$', fontsize=12)
axes[1].set_ylabel('$Z_2$', fontsize=12)
axes[1].set_title('$(Z_1, Z_2)$ independientes', fontsize=13, color='#2c3e50')
axes[1].set_aspect('equal')

plt.tight_layout()
plt.show()
```

## Ejercicio 36: Desigualdad de Jensen

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Verifique la desigualdad de Jensen $E[X^2] \ge (E[X])^2$ por simulacion para $X \sim \text{Exp}(1)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **desigualdad de Jensen** establece que si $g$ es una funcion convexa, entonces:

$$E[g(X)] \ge g(E[X])$$

Aqui $g(x) = x^2$ es convexa (su segunda derivada $g''(x) = 2 > 0$), por lo que:

$$E[X^2] \ge (E[X])^2$$

Para $X \sim \text{Exp}(1)$:
- $E[X] = 1$: media de la exponencial con $\lambda = 1$.
- $(E[X])^2 = 1$: cuadrado de la media.
- $E[X^2] = \text{Var}(X) + (E[X])^2 = 1 + 1 = 2$: segundo momento.
- La diferencia $E[X^2] - (E[X])^2 = \text{Var}(X) = 1 > 0$: la brecha de Jensen es exactamente la varianza.

En general, la igualdad se da si y solo si $X$ es constante (varianza cero) o $g$ es lineal.

</div>

```python
np.random.seed(36)

n_vals = np.logspace(1, 5, 50).astype(int)
e_x2_list = []
ex_2_list = []

for n in n_vals:
    X = np.random.exponential(1, n)
    e_x2_list.append(np.mean(X**2))
    ex_2_list.append(np.mean(X)**2)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(n_vals, e_x2_list, 'o-', color='#e74c3c', markersize=3, label=r'$E[X^2]$')
axes[0].plot(n_vals, ex_2_list, 's-', color='#3498db', markersize=3, label=r'$(E[X])^2$')
axes[0].axhline(2, color='#e74c3c', ls='--', alpha=0.5, label='Teorico $E[X^2]=2$')
axes[0].axhline(1, color='#3498db', ls='--', alpha=0.5, label='Teorico $(E[X])^2=1$')
axes[0].set_xscale('log')
axes[0].set_xlabel('n (muestras)', fontsize=12)
axes[0].set_ylabel('Valor', fontsize=12)
axes[0].set_title('Jensen: $E[X^2] \\geq (E[X])^2$', fontsize=13, color='#2c3e50')
axes[0].legend()

# Visualizacion geometrica de la convexidad
x_vals = np.linspace(0, 4, 200)
axes[1].plot(x_vals, x_vals**2, color='#2c3e50', lw=2.5, label='$g(x) = x^2$ (convexa)')
X_sample = np.random.exponential(1, 1000)
ex = X_sample.mean()
ex2 = np.mean(X_sample**2)
axes[1].plot(ex, ex**2, 'o', color='#3498db', markersize=12, zorder=5, label=f'$(E[X], g(E[X]))$ = ({ex:.2f}, {ex**2:.2f})')
axes[1].plot(ex, ex2, 's', color='#e74c3c', markersize=12, zorder=5, label=f'$(E[X], E[g(X)])$ = ({ex:.2f}, {ex2:.2f})')
axes[1].annotate('Brecha de\nJensen', xy=(ex, (ex**2+ex2)/2), xytext=(ex+0.8, (ex**2+ex2)/2),
                 fontsize=10, arrowprops=dict(arrowstyle='->', color='#2c3e50'))
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('$g(x)$', fontsize=12)
axes[1].set_title('Interpretacion geometrica', fontsize=13, color='#2c3e50')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.show()
```

## Ejercicio 37: Metodo delta para $\text{Var}(g(X))$

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X$ con $E[X] = \mu$ y $\text{Var}(X) = \sigma^2$. Use el metodo delta para aproximar $\text{Var}(\ln X)$ cuando $g(x) = \ln(x)$. Verifique con $X \sim \text{Gamma}(10, 1)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

El **metodo delta** aproxima la varianza de una funcion de una variable aleatoria mediante una expansion de Taylor de primer orden:

$$g(X) \approx g(\mu) + g'(\mu)(X - \mu)$$

Tomando varianza en ambos lados:

$$\text{Var}(g(X)) \approx [g'(\mu)]^2 \cdot \sigma^2$$

Donde:
- $g(x) = \ln(x)$: la funcion aplicada.
- $g'(x) = 1/x$: derivada de $\ln(x)$.
- $g'(\mu) = 1/\mu$: derivada evaluada en la media.
- $\sigma^2 = \text{Var}(X)$: varianza original.

Para $X \sim \text{Gamma}(\alpha=10, \beta=1)$:
- $\mu = E[X] = \alpha/\beta = 10$
- $\sigma^2 = \text{Var}(X) = \alpha/\beta^2 = 10$

Aproximacion delta:

$$\text{Var}(\ln X) \approx \left(\frac{1}{10}\right)^2 \cdot 10 = \frac{10}{100} = 0.1$$

El valor exacto es $\psi_1(10) \approx 0.1051$ (funcion trigamma), asi que la aproximacion es razonable.

</div>

```python
from scipy.special import polygamma

np.random.seed(37)

alpha, beta = 10, 1
mu = alpha / beta
sigma2 = alpha / beta**2
var_delta = (1/mu)**2 * sigma2
var_exacta = polygamma(1, alpha)  # trigamma

X = np.random.gamma(alpha, 1/beta, 300_000)
Y = np.log(X)

print(f"Var(ln X) metodo delta:   {var_delta:.4f}")
print(f"Var(ln X) exacta (trigamma): {var_exacta:.4f}")
print(f"Var(ln X) simulada:       {Y.var():.4f}")

# Comparar para distintos alpha
alphas = np.arange(2, 51)
var_delta_vals = 1 / alphas  # (1/alpha)^2 * alpha = 1/alpha
var_exacta_vals = [polygamma(1, a) for a in alphas]
var_sim_vals = [np.log(np.random.gamma(a, 1, 100_000)).var() for a in alphas]

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(alphas, var_delta_vals, '-', color='#e74c3c', lw=2.5, label='Metodo delta: $\\sigma^2/\\mu^2$')
ax.plot(alphas, var_exacta_vals, '--', color='#3498db', lw=2, label='Exacta: $\\psi_1(\\alpha)$')
ax.plot(alphas, var_sim_vals, 'o', color='#2ecc71', markersize=4, alpha=0.7, label='Simulada')
ax.set_xlabel(r'$\alpha$ (forma de la Gamma)', fontsize=12)
ax.set_ylabel(r'Var($\ln X$)', fontsize=12)
ax.set_title('Metodo delta vs valor exacto para Var(ln X)', fontsize=13, color='#2c3e50')
ax.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 38: Convergencia en probabilidad

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Muestre que $P(|\bar{X}_n - \mu| > \varepsilon) \to 0$ cuando $n \to \infty$ para $X_i \sim \text{Exp}(1)$ y varios valores de $\varepsilon$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **convergencia en probabilidad** de $\bar{X}_n$ a $\mu$ significa:

$$\forall \varepsilon > 0: \quad P(|\bar{X}_n - \mu| > \varepsilon) \to 0 \quad \text{cuando } n \to \infty$$

Esto es consecuencia directa de la Ley Debil de los Grandes Numeros. Usando la **desigualdad de Chebyshev**:

$$P(|\bar{X}_n - \mu| > \varepsilon) \le \frac{\text{Var}(\bar{X}_n)}{\varepsilon^2} = \frac{\sigma^2}{n\varepsilon^2}$$

Donde:
- $\bar{X}_n = \frac{1}{n}\sum_{i=1}^n X_i$: media muestral.
- $\mu = E[X_i] = 1$: media de la Exp(1).
- $\sigma^2 = \text{Var}(X_i) = 1$: varianza de la Exp(1).
- $\text{Var}(\bar{X}_n) = \sigma^2/n$: varianza de la media muestral (decrece con $n$).
- $\varepsilon > 0$: tolerancia; mientras mas pequena, mas lenta la convergencia.

La cota $\frac{\sigma^2}{n\varepsilon^2} \to 0$ cuando $n \to \infty$, lo que prueba la convergencia.

</div>

```python
np.random.seed(38)

mu = 1.0
n_vals = np.logspace(1, 4, 40).astype(int)
epsilons = [0.05, 0.1, 0.2, 0.5]
n_rep = 5000

fig, ax = plt.subplots(figsize=(9, 5))
colors = ['#e74c3c', '#3498db', '#2ecc71', '#2c3e50']

for eps, color in zip(epsilons, colors):
    probs = []
    for n in n_vals:
        medias = np.mean(np.random.exponential(1, (n_rep, n)), axis=1)
        prob = np.mean(np.abs(medias - mu) > eps)
        probs.append(prob)
    ax.plot(n_vals, probs, 'o-', color=color, markersize=4, lw=1.5, label=f'$\\varepsilon$ = {eps}')
    # Cota Chebyshev
    ax.plot(n_vals, np.minimum(1/(n_vals * eps**2), 1), '--', color=color, alpha=0.4, lw=1)

ax.set_xscale('log')
ax.set_xlabel('n (tamano de muestra)', fontsize=12)
ax.set_ylabel(r'$P(|\bar{X}_n - \mu| > \varepsilon)$', fontsize=12)
ax.set_title('Convergencia en probabilidad (lineas punteadas: cota Chebyshev)', fontsize=13, color='#2c3e50')
ax.legend(fontsize=11)
ax.set_ylim(-0.05, 1.05)
plt.tight_layout()
plt.show()
```

## Ejercicio 39: MLE de la exponencial

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Dada una muestra $x_1, \ldots, x_n$ de una $\text{Exp}(\lambda)$, derive el estimador de maxima verosimilitud $\hat{\lambda}_{MLE}$ y verifique con simulacion.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La funcion de verosimilitud para $x_1, \ldots, x_n$ i.i.d. $\text{Exp}(\lambda)$ es:

$$L(\lambda) = \prod_{i=1}^n \lambda e^{-\lambda x_i} = \lambda^n \exp\!\left(-\lambda \sum_{i=1}^n x_i\right)$$

La log-verosimilitud:

$$\ell(\lambda) = n \ln \lambda - \lambda \sum_{i=1}^n x_i$$

Donde:
- $n \ln \lambda$: contribucion del factor $\lambda$ en cada densidad.
- $\lambda \sum x_i$: suma ponderada de las observaciones.

Derivando e igualando a cero:

$$\frac{d\ell}{d\lambda} = \frac{n}{\lambda} - \sum_{i=1}^n x_i = 0$$

$$\hat{\lambda}_{MLE} = \frac{n}{\sum_{i=1}^n x_i} = \frac{1}{\bar{x}}$$

Donde:
- $\bar{x} = \frac{1}{n}\sum x_i$: media muestral.
- $\hat{\lambda}_{MLE} = 1/\bar{x}$: inverso de la media muestral; estima la tasa del proceso.

La segunda derivada $\frac{d^2\ell}{d\lambda^2} = -n/\lambda^2 < 0$ confirma que es un maximo.

</div>

```python
np.random.seed(39)

lam_true = 2.0
n = 50
n_rep = 10_000

lambda_hats = []
for _ in range(n_rep):
    X = np.random.exponential(1/lam_true, n)
    lambda_hats.append(1 / X.mean())
lambda_hats = np.array(lambda_hats)

print(f"lambda verdadero:        {lam_true}")
print(f"E[lambda_hat] simulado:  {lambda_hats.mean():.4f}")
print(f"Var(lambda_hat) simulada: {lambda_hats.var():.4f}")
print(f"Var teorica (lambda^2/(n-2)): {lam_true**2/(n-2):.4f}")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].hist(lambda_hats, bins=60, density=True, alpha=0.6, color='#3498db', edgecolor='white')
axes[0].axvline(lam_true, color='#e74c3c', ls='--', lw=2, label=f'$\\lambda$ = {lam_true}')
axes[0].axvline(lambda_hats.mean(), color='#2c3e50', ls=':', lw=2, label=f'$E[\\hat{{\\lambda}}]$ = {lambda_hats.mean():.3f}')
axes[0].set_xlabel(r'$\hat{\lambda}_{MLE}$', fontsize=12)
axes[0].set_ylabel('Densidad', fontsize=12)
axes[0].set_title(f'Distribucion de $\\hat{{\\lambda}}_{{MLE}}$ (n={n})', fontsize=13, color='#2c3e50')
axes[0].legend()

# Log-verosimilitud para una muestra
X_sample = np.random.exponential(1/lam_true, n)
lam_range = np.linspace(0.5, 5, 300)
log_lik = n * np.log(lam_range) - lam_range * X_sample.sum()
lam_hat = 1 / X_sample.mean()

axes[1].plot(lam_range, log_lik, color='#3498db', lw=2.5)
axes[1].axvline(lam_hat, color='#e74c3c', ls='--', lw=2, label=f'$\\hat{{\\lambda}}$ = {lam_hat:.3f}')
axes[1].axvline(lam_true, color='#2c3e50', ls=':', lw=1.5, label=f'$\\lambda$ verdadero = {lam_true}')
axes[1].set_xlabel(r'$\lambda$', fontsize=12)
axes[1].set_ylabel(r'$\ell(\lambda)$', fontsize=12)
axes[1].set_title('Log-verosimilitud', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 40: Informacion de Fisher para la Bernoulli

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Calcule la informacion de Fisher $I(p)$ para una sola observacion $X \sim \text{Ber}(p)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **informacion de Fisher** mide cuanta informacion sobre $p$ contiene una observacion:

$$I(p) = -E\!\left[\frac{d^2}{dp^2} \ln f(X; p)\right]$$

Para $X \sim \text{Ber}(p)$: $f(x; p) = p^x (1-p)^{1-x}$, con $x \in \{0, 1\}$.

**Log-verosimilitud:**

$$\ln f(x; p) = x \ln p + (1-x) \ln(1-p)$$

**Primera derivada (score):**

$$\frac{d}{dp} \ln f = \frac{x}{p} - \frac{1-x}{1-p}$$

**Segunda derivada:**

$$\frac{d^2}{dp^2} \ln f = -\frac{x}{p^2} - \frac{1-x}{(1-p)^2}$$

**Esperanza de la segunda derivada negativa:**

$$I(p) = E\!\left[\frac{X}{p^2} + \frac{1-X}{(1-p)^2}\right] = \frac{E[X]}{p^2} + \frac{1 - E[X]}{(1-p)^2} = \frac{p}{p^2} + \frac{1-p}{(1-p)^2} = \frac{1}{p} + \frac{1}{1-p} = \frac{1}{p(1-p)}$$

Donde:
- $E[X] = p$: media de la Bernoulli.
- $I(p) = \frac{1}{p(1-p)}$: maxima informacion en $p = 0$ o $p = 1$ (certeza total); minima en $p = 0.5$ (maxima incertidumbre).

</div>

```python
np.random.seed(40)

p_vals = np.linspace(0.01, 0.99, 200)
fisher_info = 1 / (p_vals * (1 - p_vals))

# Verificacion numerica: varianza del score
fisher_sim = []
for p in np.linspace(0.05, 0.95, 30):
    X = np.random.binomial(1, p, 100_000)
    score = X/p - (1-X)/(1-p)
    fisher_sim.append((p, score.var()))
fisher_sim = np.array(fisher_sim)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(p_vals, fisher_info, color='#e74c3c', lw=2.5, label=r'$I(p) = 1/[p(1-p)]$')
axes[0].plot(fisher_sim[:, 0], fisher_sim[:, 1], 'o', color='#3498db', markersize=5, label='Var(score) simulada')
axes[0].set_xlabel('p', fontsize=12)
axes[0].set_ylabel('I(p)', fontsize=12)
axes[0].set_title('Informacion de Fisher - Bernoulli', fontsize=13, color='#2c3e50')
axes[0].legend()

# Cota de Cramer-Rao para n observaciones
n_obs = 20
crlb = p_vals * (1 - p_vals) / n_obs  # 1/(n*I(p))
axes[1].plot(p_vals, crlb, color='#e74c3c', lw=2.5, label=f'CRLB = p(1-p)/{n_obs}')
axes[1].fill_between(p_vals, 0, crlb, alpha=0.15, color='#e74c3c')
axes[1].set_xlabel('p', fontsize=12)
axes[1].set_ylabel(r'Var$(\hat{p})$ minima', fontsize=12)
axes[1].set_title(f'Cota de Cramer-Rao (n={n_obs})', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 41: Cota de Cramer-Rao para la media muestral

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X_1, \ldots, X_n \sim N(\mu, \sigma^2)$ con $\sigma^2$ conocida. Verifique que $\bar{X}$ alcanza la cota de Cramer-Rao: $\text{Var}(\bar{X}) = \frac{1}{n \cdot I(\mu)} = \frac{\sigma^2}{n}$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **cota de Cramer-Rao (CRLB)** establece que para cualquier estimador insesgado $\hat{\theta}$:

$$\text{Var}(\hat{\theta}) \ge \frac{1}{n \cdot I(\theta)}$$

Donde $I(\theta)$ es la informacion de Fisher por observacion.

Para $X \sim N(\mu, \sigma^2)$:

$$\ln f(x; \mu) = -\frac{1}{2}\ln(2\pi\sigma^2) - \frac{(x-\mu)^2}{2\sigma^2}$$

**Score:**

$$\frac{d}{d\mu} \ln f = \frac{x - \mu}{\sigma^2}$$

**Informacion de Fisher:**

$$I(\mu) = E\!\left[\left(\frac{X-\mu}{\sigma^2}\right)^2\right] = \frac{\text{Var}(X)}{\sigma^4} = \frac{\sigma^2}{\sigma^4} = \frac{1}{\sigma^2}$$

**CRLB:**

$$\frac{1}{n \cdot I(\mu)} = \frac{1}{n \cdot 1/\sigma^2} = \frac{\sigma^2}{n}$$

Como $\text{Var}(\bar{X}) = \sigma^2/n$ coincide exactamente con la CRLB, $\bar{X}$ es un **estimador eficiente** (alcanza la cota minima de varianza).

</div>

```python
np.random.seed(41)

mu_true, sigma2 = 5.0, 4.0
n_sizes = [5, 10, 20, 50, 100, 200, 500]
n_rep = 10_000

var_media = []
crlb_vals = []

for n in n_sizes:
    medias = [np.random.normal(mu_true, np.sqrt(sigma2), n).mean() for _ in range(n_rep)]
    var_media.append(np.var(medias))
    crlb_vals.append(sigma2 / n)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(n_sizes, var_media, 'o-', color='#3498db', lw=2, markersize=8, label=r'Var($\bar{X}$) simulada')
ax.plot(n_sizes, crlb_vals, 's--', color='#e74c3c', lw=2, markersize=8, label=r'CRLB = $\sigma^2/n$')
ax.set_xlabel('n (tamano de muestra)', fontsize=12)
ax.set_ylabel('Varianza', fontsize=12)
ax.set_title(r'$\bar{X}$ alcanza la cota de Cramer-Rao', fontsize=13, color='#2c3e50')
ax.legend(fontsize=11)
ax.set_xscale('log')
ax.set_yscale('log')
plt.tight_layout()
plt.show()

print(f"{'n':>5} | {'Var(X_bar) sim':>15} | {'CRLB':>10} | {'Ratio':>8}")
print("-" * 48)
for n, vs, cr in zip(n_sizes, var_media, crlb_vals):
    print(f"{n:>5} | {vs:>15.4f} | {cr:>10.4f} | {vs/cr:>8.4f}")
```

## Ejercicio 42: Bootstrap para la varianza de la mediana

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Dada una muestra de $n = 30$ observaciones de una $\text{Exp}(1)$, estime $\text{Var}(\text{mediana})$ usando bootstrap con $B = 5000$ remuestreos.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

El **bootstrap** es un metodo de remuestreo para estimar la distribucion de un estadistico cuando no hay formula analitica disponible.

**Algoritmo:**

1. Dada la muestra original $\mathbf{x} = (x_1, \ldots, x_n)$.
2. Para $b = 1, \ldots, B$:
   - Generar una muestra bootstrap $\mathbf{x}^{*b}$ de tamano $n$ muestreando **con reemplazo** de $\mathbf{x}$.
   - Calcular el estadistico: $\hat{\theta}^{*b} = \text{mediana}(\mathbf{x}^{*b})$.
3. Estimar la varianza: $\widehat{\text{Var}}_{\text{boot}}(\hat{\theta}) = \frac{1}{B-1}\sum_{b=1}^B (\hat{\theta}^{*b} - \bar{\hat{\theta}}^*)^2$

Donde:
- $B = 5000$: numero de remuestreos bootstrap (suficiente para estimar la varianza).
- $n = 30$: tamano de la muestra original.
- $\hat{\theta}^{*b}$: mediana de la $b$-esima muestra bootstrap.
- $\bar{\hat{\theta}}^* = \frac{1}{B}\sum_b \hat{\theta}^{*b}$: promedio de las medianas bootstrap.
- Con reemplazo: cada observacion puede ser seleccionada multiples veces o ninguna.

</div>

```python
np.random.seed(42)

n = 30
B = 5000
X = np.random.exponential(1, n)

# Bootstrap
medianas_boot = np.array([np.median(np.random.choice(X, n, replace=True)) for _ in range(B)])
var_boot = medianas_boot.var(ddof=1)

# Comparacion: varianza "real" por Monte Carlo
n_mc = 20_000
medianas_mc = np.array([np.median(np.random.exponential(1, n)) for _ in range(n_mc)])
var_mc = medianas_mc.var()

print(f"Mediana muestral original: {np.median(X):.4f}")
print(f"Var(mediana) bootstrap:    {var_boot:.4f}")
print(f"Var(mediana) Monte Carlo:  {var_mc:.4f}")
print(f"IC 95% bootstrap: [{np.percentile(medianas_boot, 2.5):.4f}, {np.percentile(medianas_boot, 97.5):.4f}]")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].hist(medianas_boot, bins=50, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='Bootstrap')
axes[0].axvline(np.median(X), color='#e74c3c', ls='--', lw=2, label=f'Mediana original = {np.median(X):.3f}')
axes[0].axvline(np.log(2), color='#2c3e50', ls=':', lw=2, label=f'Mediana teorica = ln(2) = {np.log(2):.3f}')
axes[0].set_xlabel('Mediana bootstrap', fontsize=12)
axes[0].set_ylabel('Densidad', fontsize=12)
axes[0].set_title(f'Bootstrap (B={B})', fontsize=13, color='#2c3e50')
axes[0].legend()

axes[1].hist(medianas_mc, bins=50, density=True, alpha=0.6, color='#2ecc71', edgecolor='white', label='Monte Carlo')
axes[1].axvline(np.log(2), color='#2c3e50', ls=':', lw=2, label=f'Mediana teorica = {np.log(2):.3f}')
axes[1].set_xlabel('Mediana', fontsize=12)
axes[1].set_ylabel('Densidad', fontsize=12)
axes[1].set_title(f'Monte Carlo ({n_mc} muestras)', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 43: Jackknife para el sesgo del estimador de varianza

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Use el metodo jackknife para estimar el sesgo del estimador $\hat{\sigma}^2 = \frac{1}{n}\sum(x_i - \bar{x})^2$ (varianza sesgada) con una muestra de $n = 20$ observaciones de $N(0, 4)$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

El **jackknife** estima el sesgo de un estimador eliminando una observacion a la vez.

**Algoritmo:**

1. Calcular $\hat{\theta}$ con la muestra completa: $\hat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^2$.
2. Para $i = 1, \ldots, n$: calcular $\hat{\theta}_{(-i)}$ eliminando la $i$-esima observacion.
3. Promedio jackknife: $\bar{\hat{\theta}}_{(\cdot)} = \frac{1}{n}\sum_{i=1}^n \hat{\theta}_{(-i)}$.
4. Sesgo jackknife: $\widehat{\text{bias}}_{\text{jack}} = (n-1)(\bar{\hat{\theta}}_{(\cdot)} - \hat{\theta})$.

Donde:
- $\hat{\theta}_{(-i)}$: estadistico calculado sin la observacion $i$.
- El factor $(n-1)$ corrige la escala del sesgo.

**Sesgo teorico** de $\hat{\sigma}^2 = \frac{1}{n}\sum(x_i - \bar{x})^2$:

$$\text{bias} = E[\hat{\sigma}^2] - \sigma^2 = \frac{n-1}{n}\sigma^2 - \sigma^2 = -\frac{\sigma^2}{n}$$

Para $n = 20$ y $\sigma^2 = 4$: sesgo teorico $= -4/20 = -0.2$.

</div>

```python
np.random.seed(43)

n = 20
sigma2_true = 4.0
X = np.random.normal(0, np.sqrt(sigma2_true), n)

# Estimador sesgado de varianza
theta_hat = np.var(X, ddof=0)  # divide por n

# Jackknife
theta_jack = np.array([np.var(np.delete(X, i), ddof=0) for i in range(n)])
theta_bar_jack = theta_jack.mean()
bias_jack = (n - 1) * (theta_bar_jack - theta_hat)
bias_teorico = -sigma2_true / n

# Estimador corregido
theta_corrected = theta_hat - bias_jack

print(f"Varianza sesgada (1/n):     {theta_hat:.4f}")
print(f"Sesgo jackknife:             {bias_jack:.4f}")
print(f"Sesgo teorico (-sigma^2/n): {bias_teorico:.4f}")
print(f"Estimador corregido:         {theta_corrected:.4f}")
print(f"Varianza insesgada (1/(n-1)): {np.var(X, ddof=1):.4f}")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].bar(range(n), theta_jack, color='#3498db', edgecolor='white', alpha=0.7)
axes[0].axhline(theta_hat, color='#e74c3c', ls='--', lw=2, label=f'$\\hat{{\\sigma}}^2$ completo = {theta_hat:.3f}')
axes[0].axhline(theta_bar_jack, color='#2c3e50', ls=':', lw=2, label=f'Promedio jackknife = {theta_bar_jack:.3f}')
axes[0].set_xlabel('Observacion eliminada', fontsize=12)
axes[0].set_ylabel(r'$\hat{\sigma}^2_{(-i)}$', fontsize=12)
axes[0].set_title('Estimaciones jackknife', fontsize=13, color='#2c3e50')
axes[0].legend(fontsize=9)

# Repetir jackknife muchas veces
biases_jack = []
for _ in range(3000):
    X_rep = np.random.normal(0, np.sqrt(sigma2_true), n)
    th = np.var(X_rep, ddof=0)
    th_j = np.array([np.var(np.delete(X_rep, i), ddof=0) for i in range(n)])
    biases_jack.append((n-1) * (th_j.mean() - th))

axes[1].hist(biases_jack, bins=50, density=True, alpha=0.6, color='#3498db', edgecolor='white')
axes[1].axvline(bias_teorico, color='#e74c3c', ls='--', lw=2, label=f'Sesgo teorico = {bias_teorico:.2f}')
axes[1].axvline(np.mean(biases_jack), color='#2c3e50', ls=':', lw=2, label=f'Media jackknife = {np.mean(biases_jack):.3f}')
axes[1].set_xlabel('Sesgo estimado', fontsize=12)
axes[1].set_title('Distribucion del sesgo jackknife', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 44: Estadisticos de orden de uniformes

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sean $X_1, \ldots, X_n \sim U(0,1)$ i.i.d. Calcule $E[X_{(1)}]$ (minimo) y $E[X_{(n)}]$ (maximo) y verifique con simulacion para $n = 10$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Para $X_1, \ldots, X_n \sim U(0,1)$, la densidad del $k$-esimo estadistico de orden $X_{(k)}$ es:

$$f_{X_{(k)}}(x) = \frac{n!}{(k-1)!(n-k)!} x^{k-1}(1-x)^{n-k}, \quad x \in [0,1]$$

Esto es una distribucion $\text{Beta}(k, n-k+1)$, por lo que:

$$E[X_{(k)}] = \frac{k}{n+1}$$

**Minimo** ($k = 1$):

$$E[X_{(1)}] = \frac{1}{n+1}$$

Para $n = 10$: $E[X_{(1)}] = \frac{1}{11} \approx 0.0909$.

**Maximo** ($k = n$):

$$E[X_{(n)}] = \frac{n}{n+1}$$

Para $n = 10$: $E[X_{(n)}] = \frac{10}{11} \approx 0.9091$.

Donde:
- $X_{(k)}$: $k$-esimo valor mas pequeno de la muestra.
- $\text{Beta}(k, n-k+1)$: distribucion del $k$-esimo estadistico de orden de uniformes.
- Los estadisticos de orden dividen $[0,1]$ en $n+1$ intervalos de esperanza igual $\frac{1}{n+1}$.

</div>

```python
np.random.seed(44)

n = 10
n_sim = 100_000

samples = np.random.uniform(0, 1, (n_sim, n))
mins = samples.min(axis=1)
maxs = samples.max(axis=1)

print(f"n = {n}")
print(f"E[X_(1)] simulado: {mins.mean():.4f}  (teorico: {1/(n+1):.4f})")
print(f"E[X_(n)] simulado: {maxs.mean():.4f}  (teorico: {n/(n+1):.4f})")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
x_vals = np.linspace(0, 1, 300)

# Minimo ~ Beta(1, n)
axes[0].hist(mins, bins=60, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='Simulacion')
axes[0].plot(x_vals, stats.beta.pdf(x_vals, 1, n), color='#e74c3c', lw=2.5, label=f'Beta(1, {n})')
axes[0].axvline(1/(n+1), color='#2c3e50', ls='--', lw=2, label=f'E = 1/{n+1} = {1/(n+1):.4f}')
axes[0].set_xlabel('$X_{(1)}$', fontsize=12)
axes[0].set_ylabel('Densidad', fontsize=12)
axes[0].set_title(f'Minimo de {n} uniformes', fontsize=13, color='#2c3e50')
axes[0].legend()

# Maximo ~ Beta(n, 1)
axes[1].hist(maxs, bins=60, density=True, alpha=0.6, color='#2ecc71', edgecolor='white', label='Simulacion')
axes[1].plot(x_vals, stats.beta.pdf(x_vals, n, 1), color='#e74c3c', lw=2.5, label=f'Beta({n}, 1)')
axes[1].axvline(n/(n+1), color='#2c3e50', ls='--', lw=2, label=f'E = {n}/{n+1} = {n/(n+1):.4f}')
axes[1].set_xlabel('$X_{(n)}$', fontsize=12)
axes[1].set_ylabel('Densidad', fontsize=12)
axes[1].set_title(f'Maximo de {n} uniformes', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 45: Densidad condicional en la normal bivariada

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Derive $f(Y|X=x)$ para un vector normal bivariado $(X,Y)$ con parametros $\mu_X=0, \mu_Y=0, \sigma_X=1, \sigma_Y=2, \rho=0.6$ y verifique con simulacion.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La densidad condicional se obtiene como:

$$f(y|x) = \frac{f(x,y)}{f_X(x)}$$

Para la normal bivariada, el resultado es:

$$Y | X = x \sim N\!\left(\mu_Y + \rho\frac{\sigma_Y}{\sigma_X}(x - \mu_X),\;\; \sigma_Y^2(1-\rho^2)\right)$$

Con los parametros dados ($\mu_X = 0, \mu_Y = 0, \sigma_X = 1, \sigma_Y = 2, \rho = 0.6$):

**Media condicional:**

$$E[Y|X=x] = 0 + 0.6 \cdot \frac{2}{1} \cdot (x - 0) = 1.2x$$

Donde:
- $\rho \frac{\sigma_Y}{\sigma_X} = 0.6 \cdot 2 = 1.2$: coeficiente de regresion de $Y$ sobre $X$.

**Varianza condicional:**

$$\text{Var}(Y|X=x) = 4 \cdot (1 - 0.36) = 4 \cdot 0.64 = 2.56$$

Donde:
- $\sigma_Y^2 = 4$: varianza marginal de $Y$.
- $1 - \rho^2 = 1 - 0.36 = 0.64$: factor de reduccion por la correlacion.

</div>

```python
np.random.seed(45)

mu_X, mu_Y = 0, 0
sigma_X, sigma_Y, rho = 1, 2, 0.6
cov_XY = rho * sigma_X * sigma_Y
cov_mat = [[sigma_X**2, cov_XY], [cov_XY, sigma_Y**2]]

samples = np.random.multivariate_normal([mu_X, mu_Y], cov_mat, 100_000)
X, Y = samples[:, 0], samples[:, 1]

# Condicional en X=1
x0 = 1.0
mu_cond = mu_Y + rho * sigma_Y / sigma_X * (x0 - mu_X)
var_cond = sigma_Y**2 * (1 - rho**2)

mask = np.abs(X - x0) < 0.15
Y_cond = Y[mask]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].scatter(X[:3000], Y[:3000], alpha=0.15, s=5, color='#3498db')
axes[0].axvline(x0, color='#e74c3c', ls='--', lw=2, label=f'X = {x0}')
x_line = np.linspace(-3, 3, 100)
axes[0].plot(x_line, 1.2 * x_line, color='#2c3e50', lw=2, label='E[Y|X=x] = 1.2x')
axes[0].set_xlabel('X', fontsize=12)
axes[0].set_ylabel('Y', fontsize=12)
axes[0].set_title('Normal bivariada con recta de regresion', fontsize=13, color='#2c3e50')
axes[0].legend()

y_vals = np.linspace(-6, 8, 300)
axes[1].hist(Y_cond, bins=40, density=True, alpha=0.6, color='#3498db', edgecolor='white', label=f'Y | X={x0} (empirica)')
axes[1].plot(y_vals, stats.norm.pdf(y_vals, mu_cond, np.sqrt(var_cond)),
             color='#e74c3c', lw=2.5, label=f'N({mu_cond:.1f}, {var_cond:.2f})')
axes[1].set_xlabel('Y', fontsize=12)
axes[1].set_ylabel('Densidad', fontsize=12)
axes[1].set_title(f'Densidad condicional f(Y|X={x0})', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"Media condicional teorica:    {mu_cond:.4f}")
print(f"Media condicional empirica:   {Y_cond.mean():.4f}")
print(f"Varianza condicional teorica: {var_cond:.4f}")
print(f"Varianza condicional empirica: {Y_cond.var():.4f}")
```

## Ejercicio 46: Descomposicion de Cholesky para normales correlacionadas

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Use la descomposicion de Cholesky para generar un vector $(X_1, X_2, X_3)$ normal multivariado con una matriz de correlacion dada.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Para generar $\mathbf{X} \sim N(\boldsymbol{\mu}, \boldsymbol{\Sigma})$, usamos la **descomposicion de Cholesky**:

$$\boldsymbol{\Sigma} = \mathbf{L}\mathbf{L}^T$$

donde $\mathbf{L}$ es una matriz triangular inferior. El algoritmo es:

1. Generar $\mathbf{Z} \sim N(\mathbf{0}, \mathbf{I})$: vector de normales estandar independientes.
2. Calcular $\mathbf{X} = \boldsymbol{\mu} + \mathbf{L}\mathbf{Z}$.

Entonces $\mathbf{X} \sim N(\boldsymbol{\mu}, \boldsymbol{\Sigma})$ porque:

$$\text{Cov}(\mathbf{X}) = \mathbf{L} \cdot \text{Cov}(\mathbf{Z}) \cdot \mathbf{L}^T = \mathbf{L}\mathbf{I}\mathbf{L}^T = \mathbf{L}\mathbf{L}^T = \boldsymbol{\Sigma}$$

Donde:
- $\mathbf{L}$: factor de Cholesky; triangular inferior, unica para matrices definidas positivas.
- $\mathbf{Z}$: vector de ruido blanco independiente.
- $\mathbf{L}\mathbf{Z}$: introduce las correlaciones deseadas al mezclar las componentes independientes.

</div>

```python
np.random.seed(46)

# Matriz de covarianza objetivo
Sigma = np.array([
    [1.0, 0.8, 0.3],
    [0.8, 1.0, 0.5],
    [0.3, 0.5, 1.0]
])
mu = np.array([2, 5, -1])

# Cholesky
L = np.linalg.cholesky(Sigma)
print("Matriz L (Cholesky):")
print(np.round(L, 4))

# Generar muestras
n_sim = 50_000
Z = np.random.standard_normal((3, n_sim))
X = mu[:, None] + L @ Z
X = X.T  # (n_sim, 3)

print(f"\nCovarianza objetivo:\n{Sigma}")
print(f"\nCovarianza estimada:\n{np.round(np.cov(X.T), 4)}")

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
pairs = [(0,1), (0,2), (1,2)]
labels = ['$X_1$', '$X_2$', '$X_3$']
colors = ['#3498db', '#e74c3c', '#2ecc71']

for ax, (i, j), color in zip(axes, pairs, colors):
    ax.scatter(X[:2000, i], X[:2000, j], alpha=0.2, s=5, color=color)
    r_teo = Sigma[i, j]
    r_emp = np.corrcoef(X[:, i], X[:, j])[0, 1]
    ax.set_xlabel(labels[i], fontsize=12)
    ax.set_ylabel(labels[j], fontsize=12)
    ax.set_title(f'{labels[i]} vs {labels[j]}\n$\\rho$={r_teo:.1f} (emp={r_emp:.3f})', fontsize=11, color='#2c3e50')

plt.suptitle('Normales correlacionadas via Cholesky', fontsize=13, color='#2c3e50', y=1.02)
plt.tight_layout()
plt.show()
```

## Ejercicio 47: Metropolis-Hastings simple

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Implemente el algoritmo Metropolis-Hastings para muestrear de $N(0,1)$ usando una propuesta uniforme $q(x'|x) = U(x - \delta, x + \delta)$ con $\delta = 1$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

El **algoritmo Metropolis-Hastings** genera una cadena de Markov cuya distribucion estacionaria es la distribucion objetivo $\pi(x)$.

**Algoritmo:**

1. Inicializar $x_0$ (valor arbitrario).
2. Para $t = 1, 2, \ldots$:
   - Proponer $x' \sim q(x'|x_t) = U(x_t - \delta, x_t + \delta)$.
   - Calcular la razon de aceptacion: $\alpha = \min\!\left(1,\; \frac{\pi(x')}{\pi(x_t)}\right)$.
   - Con probabilidad $\alpha$, aceptar: $x_{t+1} = x'$. Si no, $x_{t+1} = x_t$.

Donde:
- $\pi(x) \propto e^{-x^2/2}$: densidad objetivo (normal estandar; no necesitamos la constante de normalizacion).
- $q(x'|x)$: propuesta simetrica (random walk), por lo que la razon de Hastings $q(x|x')/q(x'|x) = 1$.
- $\delta = 1$: ancho de la propuesta. Si es muy grande, muchos rechazos; si es muy pequeno, exploracion lenta.
- $\alpha$: probabilidad de aceptacion; estados mas probables siempre se aceptan.

</div>

```python
np.random.seed(47)

def log_target(x):
    """Log de la densidad objetivo (N(0,1), sin constante)."""
    return -0.5 * x**2

def metropolis_hastings(n_iter, delta, x0=0.0):
    chain = [x0]
    accepted = 0
    for _ in range(n_iter):
        x_curr = chain[-1]
        x_prop = x_curr + np.random.uniform(-delta, delta)
        log_alpha = log_target(x_prop) - log_target(x_curr)
        if np.log(np.random.uniform()) < log_alpha:
            chain.append(x_prop)
            accepted += 1
        else:
            chain.append(x_curr)
    return np.array(chain), accepted / n_iter

n_iter = 50_000
chain, acc_rate = metropolis_hastings(n_iter, delta=1.0)
burn_in = 1000
samples_mh = chain[burn_in:]

print(f"Tasa de aceptacion: {acc_rate:.3f}")
print(f"Media MH:     {samples_mh.mean():.4f}  (teorica: 0)")
print(f"Varianza MH:  {samples_mh.var():.4f}  (teorica: 1)")

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(chain[:2000], color='#3498db', lw=0.5, alpha=0.8)
axes[0].axhline(0, color='#e74c3c', ls='--', lw=1)
axes[0].set_xlabel('Iteracion', fontsize=11)
axes[0].set_ylabel('x', fontsize=11)
axes[0].set_title('Traza de la cadena', fontsize=12, color='#2c3e50')

x_vals = np.linspace(-4, 4, 300)
axes[1].hist(samples_mh, bins=70, density=True, alpha=0.6, color='#3498db', edgecolor='white', label='MH')
axes[1].plot(x_vals, stats.norm.pdf(x_vals), color='#e74c3c', lw=2.5, label='N(0,1)')
axes[1].set_title('Histograma vs objetivo', fontsize=12, color='#2c3e50')
axes[1].legend()

# Comparar deltas
deltas = [0.1, 0.5, 1.0, 2.0, 5.0, 10.0]
acc_rates = []
for d in deltas:
    _, ar = metropolis_hastings(20_000, d)
    acc_rates.append(ar)

axes[2].plot(deltas, acc_rates, 'o-', color='#2c3e50', lw=2, markersize=7)
axes[2].axhline(0.44, color='#e74c3c', ls='--', lw=1.5, label='Optimo ~0.44')
axes[2].set_xlabel(r'$\delta$', fontsize=12)
axes[2].set_ylabel('Tasa aceptacion', fontsize=12)
axes[2].set_title('Efecto de $\\delta$', fontsize=12, color='#2c3e50')
axes[2].legend()

plt.tight_layout()
plt.show()
```

## Ejercicio 48: Actualizacion bayesiana Beta-Binomial

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Se tiene un prior $p \sim \text{Beta}(2, 2)$ para la probabilidad de exito. Se observan 7 exitos en 10 ensayos. Determine la distribucion posterior y calcule su esperanza.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Usamos la **conjugacion Beta-Binomial**. Si el prior es $p \sim \text{Beta}(\alpha, \beta)$ y observamos $k$ exitos en $n$ ensayos, entonces el posterior es:

$$p | \text{datos} \sim \text{Beta}(\alpha + k,\; \beta + n - k)$$

**Prior:** $p \sim \text{Beta}(2, 2)$

Donde:
- $\alpha = 2, \beta = 2$: hiperparametros del prior (simetrico, centrado en 0.5).
- La Beta(2,2) concentra masa cerca de 0.5, representando creencia moderada de equidad.

**Datos:** $k = 7$ exitos, $n = 10$ ensayos.

**Posterior:**

$$p | \text{datos} \sim \text{Beta}(2 + 7,\; 2 + 3) = \text{Beta}(9, 5)$$

**Esperanza posterior:**

$$E[p | \text{datos}] = \frac{\alpha + k}{\alpha + k + \beta + n - k} = \frac{9}{9 + 5} = \frac{9}{14} \approx 0.6429$$

Comparaciones:
- Prior: $E[p] = 2/4 = 0.5$
- MLE: $\hat{p} = 7/10 = 0.7$
- Posterior: $0.6429$ (compromiso entre prior y datos, mas cerca de los datos porque $n$ es moderado).

</div>

```python
np.random.seed(48)

alpha_prior, beta_prior = 2, 2
k, n = 7, 10
alpha_post = alpha_prior + k
beta_post = beta_prior + (n - k)

p_vals = np.linspace(0, 1, 300)
prior_pdf = stats.beta.pdf(p_vals, alpha_prior, beta_prior)
post_pdf = stats.beta.pdf(p_vals, alpha_post, beta_post)
likelihood = stats.binom.pmf(k, n, p_vals)
likelihood_norm = likelihood / likelihood.max() * post_pdf.max()

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(p_vals, prior_pdf, '--', color='#3498db', lw=2.5, label=f'Prior Beta({alpha_prior},{beta_prior})')
axes[0].plot(p_vals, likelihood_norm, ':', color='#2ecc71', lw=2, label='Verosimilitud (escalada)')
axes[0].plot(p_vals, post_pdf, '-', color='#e74c3c', lw=2.5, label=f'Posterior Beta({alpha_post},{beta_post})')
axes[0].axvline(alpha_post/(alpha_post+beta_post), color='#e74c3c', ls=':', alpha=0.5)
axes[0].axvline(k/n, color='#2ecc71', ls=':', alpha=0.5)
axes[0].set_xlabel('p', fontsize=12)
axes[0].set_ylabel('Densidad', fontsize=12)
axes[0].set_title('Actualizacion bayesiana', fontsize=13, color='#2c3e50')
axes[0].legend()

# Actualizacion secuencial
datos = [1, 1, 0, 1, 1, 1, 0, 1, 0, 1]  # 7 exitos
a, b = alpha_prior, beta_prior
medias_post = [a/(a+b)]

for xi in datos:
    a += xi
    b += (1 - xi)
    medias_post.append(a / (a + b))

axes[1].plot(range(len(medias_post)), medias_post, 'o-', color='#3498db', lw=2, markersize=6)
axes[1].axhline(k/n, color='#e74c3c', ls='--', lw=1.5, label=f'MLE = {k/n:.2f}')
axes[1].axhline(0.5, color='#2c3e50', ls=':', lw=1, label='Prior E[p] = 0.5')
axes[1].set_xlabel('Observaciones acumuladas', fontsize=12)
axes[1].set_ylabel('E[p | datos]', fontsize=12)
axes[1].set_title('Actualizacion secuencial de la media posterior', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"Prior:     Beta({alpha_prior}, {beta_prior}), E[p] = {alpha_prior/(alpha_prior+beta_prior):.4f}")
print(f"Posterior: Beta({alpha_post}, {beta_post}), E[p|datos] = {alpha_post/(alpha_post+beta_post):.4f}")
print(f"MLE:       p_hat = {k/n:.4f}")
```

## Ejercicio 49: Aproximacion de Taylor para $E[\sin(X)]$

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Sea $X \sim N(\mu, \sigma^2)$. Use una expansion de Taylor de segundo orden para aproximar $E[\sin(X)]$ y verifique con simulacion para $\mu = 1$, $\sigma^2 = 0.25$.

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

Expandimos $g(X) = \sin(X)$ en Taylor alrededor de $\mu$:

$$\sin(X) \approx \sin(\mu) + \cos(\mu)(X - \mu) - \frac{\sin(\mu)}{2}(X - \mu)^2$$

Tomando esperanza:

$$E[\sin(X)] \approx \sin(\mu) + \cos(\mu) \cdot \underbrace{E[X - \mu]}_{=0} - \frac{\sin(\mu)}{2} \cdot \underbrace{E[(X - \mu)^2]}_{=\sigma^2}$$

$$E[\sin(X)] \approx \sin(\mu) - \frac{\sigma^2}{2}\sin(\mu)$$

Donde:
- $\sin(\mu)$: valor de la funcion en la media (termino de orden 0).
- $\cos(\mu)$: primera derivada; desaparece porque $E[X - \mu] = 0$.
- $-\sin(\mu)/2$: segunda derivada de $\sin$ evaluada en $\mu$, dividida por 2 (factor de Taylor).
- $\sigma^2$: varianza de $X$; controla la magnitud de la correccion.

Para $\mu = 1$, $\sigma^2 = 0.25$:

$$E[\sin(X)] \approx \sin(1) - \frac{0.25}{2}\sin(1) = \sin(1)(1 - 0.125) = 0.8415 \cdot 0.875 \approx 0.7363$$

</div>

```python
np.random.seed(49)

mu, sigma2 = 1.0, 0.25

# Aproximacion de Taylor
approx_taylor = np.sin(mu) - (sigma2 / 2) * np.sin(mu)

# Valor exacto por simulacion
X = np.random.normal(mu, np.sqrt(sigma2), 500_000)
e_sin_sim = np.mean(np.sin(X))

# Valor exacto analitico para normal: E[sin(X)] = sin(mu) * exp(-sigma^2/2)
e_sin_exact = np.sin(mu) * np.exp(-sigma2 / 2)

print(f"sin(mu) (orden 0):             {np.sin(mu):.6f}")
print(f"Aproximacion Taylor (orden 2): {approx_taylor:.6f}")
print(f"Valor exacto (analitico):      {e_sin_exact:.6f}")
print(f"Valor simulado:                {e_sin_sim:.6f}")

# Comparar para distintas varianzas
sigma2_vals = np.linspace(0.01, 2, 100)
taylor_vals = np.sin(mu) * (1 - sigma2_vals / 2)
exact_vals = np.sin(mu) * np.exp(-sigma2_vals / 2)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(sigma2_vals, exact_vals, color='#e74c3c', lw=2.5, label='Exacto: $\\sin(\\mu)e^{-\\sigma^2/2}$')
axes[0].plot(sigma2_vals, taylor_vals, '--', color='#3498db', lw=2, label='Taylor: $\\sin(\\mu)(1-\\sigma^2/2)$')
axes[0].axhline(np.sin(mu), color='#2c3e50', ls=':', lw=1, label='$\\sin(\\mu)$ (orden 0)')
axes[0].axvline(sigma2, color='gray', ls=':', alpha=0.5)
axes[0].set_xlabel(r'$\sigma^2$', fontsize=12)
axes[0].set_ylabel(r'$E[\sin(X)]$', fontsize=12)
axes[0].set_title(f'Aproximacion de Taylor ($\\mu = {mu}$)', fontsize=13, color='#2c3e50')
axes[0].legend(fontsize=9)

axes[1].plot(sigma2_vals, np.abs(taylor_vals - exact_vals), color='#e74c3c', lw=2.5)
axes[1].set_xlabel(r'$\sigma^2$', fontsize=12)
axes[1].set_ylabel('Error absoluto', fontsize=12)
axes[1].set_title('Error de la aproximacion de Taylor', fontsize=13, color='#2c3e50')

plt.tight_layout()
plt.show()
```

## Ejercicio 50: Regresion a la media

<div style="border-left:4px solid #3498db; padding:10px 15px; background:#f0f7fc; margin:10px 0;">

**Enunciado:** Simule alturas de padres e hijos con correlacion $\rho = 0.5$ y muestre que los hijos de padres excepcionalmente altos tienden a ser mas bajos que sus padres (regresion a la media).

</div>

<div style="border-left:4px solid #2ecc71; padding:10px 15px; background:#f0faf4; margin:10px 0;">

**Resolucion:**

La **regresion a la media** es el fenomeno por el cual valores extremos tienden a ser seguidos por valores mas cercanos al promedio.

Modelamos alturas (en cm) de padres ($X$) e hijos ($Y$) como normal bivariada:

$$E[Y | X = x] = \mu_Y + \rho \frac{\sigma_Y}{\sigma_X}(x - \mu_X)$$

Si asumimos $\mu_X = \mu_Y = \mu$ y $\sigma_X = \sigma_Y = \sigma$ (misma distribucion marginal):

$$E[Y | X = x] = \mu + \rho(x - \mu)$$

Donde:
- $\rho = 0.5$: correlacion padre-hijo (menor que 1).
- $\mu$: altura media poblacional.
- $x - \mu$: desviacion del padre respecto a la media.
- $\rho(x - \mu)$: desviacion **esperada** del hijo; como $|\rho| < 1$, esta es **menor** que la del padre.

Si un padre mide $x = \mu + 3\sigma$ (excepcionalmente alto):

$$E[Y | X] = \mu + 0.5 \cdot 3\sigma = \mu + 1.5\sigma$$

El hijo esperado mide $1.5\sigma$ sobre la media, no $3\sigma$. Esto no es que los hijos "encojan", sino que valores extremos contienen un componente aleatorio que no se hereda completamente.

</div>

```python
np.random.seed(50)

mu = 175  # cm
sigma = 7  # cm
rho = 0.5

cov_mat = [[sigma**2, rho*sigma**2], [rho*sigma**2, sigma**2]]
samples = np.random.multivariate_normal([mu, mu], cov_mat, 5000)
padres, hijos = samples[:, 0], samples[:, 1]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Scatter con regresion a la media
axes[0].scatter(padres, hijos, alpha=0.15, s=8, color='#3498db')
x_line = np.linspace(150, 200, 100)
axes[0].plot(x_line, x_line, '--', color='#2c3e50', lw=1.5, label='Y = X (sin regresion)')
axes[0].plot(x_line, mu + rho*(x_line - mu), '-', color='#e74c3c', lw=2.5,
             label=f'E[Y|X] = {mu} + {rho}(X-{mu})')
axes[0].set_xlabel('Altura del padre (cm)', fontsize=12)
axes[0].set_ylabel('Altura del hijo (cm)', fontsize=12)
axes[0].set_title('Regresion a la media', fontsize=13, color='#2c3e50')
axes[0].legend(fontsize=9)
axes[0].set_aspect('equal')

# Analisis por grupos
percentiles = [90, 95, 99]
for p in percentiles:
    umbral = np.percentile(padres, p)
    mask = padres >= umbral
    media_padres_altos = padres[mask].mean()
    media_hijos_altos = hijos[mask].mean()
    print(f"Padres percentil >= {p}: media padre = {media_padres_altos:.1f} cm, "
          f"media hijo = {media_hijos_altos:.1f} cm, "
          f"diferencia = {media_hijos_altos - media_padres_altos:.1f} cm")

# Barras comparativas
grupos = ['Padres\nbajos\n(<p10)', 'Padres\nmedios\n(p45-p55)', 'Padres\naltos\n(>p90)']
masks = [
    padres <= np.percentile(padres, 10),
    (padres >= np.percentile(padres, 45)) & (padres <= np.percentile(padres, 55)),
    padres >= np.percentile(padres, 90)
]
medias_p = [padres[m].mean() for m in masks]
medias_h = [hijos[m].mean() for m in masks]

x_pos = np.arange(3)
axes[1].bar(x_pos - 0.15, medias_p, 0.3, color='#3498db', label='Padres', edgecolor='white')
axes[1].bar(x_pos + 0.15, medias_h, 0.3, color='#e74c3c', label='Hijos', edgecolor='white')
axes[1].axhline(mu, color='#2c3e50', ls='--', lw=1.5, label=f'Media poblacional = {mu}')
axes[1].set_xticks(x_pos)
axes[1].set_xticklabels(grupos, fontsize=10)
axes[1].set_ylabel('Altura media (cm)', fontsize=12)
axes[1].set_title('Hijos regresan hacia la media', fontsize=13, color='#2c3e50')
axes[1].legend()

plt.tight_layout()
plt.show()
```
