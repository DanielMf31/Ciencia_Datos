## Fase 4: Tests de Hipotesis

**Por que va despues de la Fase 3:** Los tests de hipotesis son la aplicacion directa de la inferencia estadistica. Usan los mismos conceptos (distribuciones de muestreo, niveles de confianza) pero ahora para tomar decisiones: "es este efecto real o es ruido?". En ML, esto se traduce en: "es la diferencia entre dos modelos significativa o es varianza del muestreo?".

**Prerequisitos:** Fases 1-3 completas. Especialmente: intervalos de confianza, distribuciones t y Chi-cuadrado.

### Temas que cubre:
- Hipotesis nula y alternativa
- Errores Tipo I y Tipo II, potencia del test
- p-valor: que es realmente y que NO es
- Tests parametricos: z-test, t-test (una muestra, dos muestras, pareado)
- Tests no parametricos: Mann-Whitney, Wilcoxon, Kruskal-Wallis
- Chi-cuadrado: bondad de ajuste e independencia
- ANOVA: comparar mas de dos grupos
- Correccion por comparaciones multiples (Bonferroni)
