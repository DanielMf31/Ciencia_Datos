# Guia de Support Picks -- Overwatch 2026

**Principio fundamental:** Un support no solo cura -- habilita al equipo. Tu eleccion de support puede ganar o perder la partida antes de que empiece la fight.

**Los 3 factores para elegir support:**
1. **Tu tanque:** define el estilo de juego (dive, poke, brawl)
2. **La composicion enemiga:** tienen dive? necesitas peel. Tienen poke? necesitas movilidad
3. **El mapa:** mapas abiertos favorecen poke (Ana, Bap). Mapas cerrados favorecen brawl (Lucio, Brig)

**Subroles 2026:**
- **Tactician** (supports): el exceso de carga ultimate se conserva para el siguiente -- ciclas ults mas rapido
- **Stalwart** (algunos tanques): reduccion de knockback y slows

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyArrowPatch, FancyBboxPatch

C_PRIMARY = '#3498db'
C_DANGER = '#e74c3c'
C_SUCCESS = '#2ecc71'
C_DARK = '#2c3e50'
C_ORANGE = '#f39c12'
C_PURPLE = '#9b59b6'

fig, ax = plt.subplots(1, 3, figsize=(18, 7), facecolor='#1a1a2e')

arquetipos = [
    {
        'nombre': 'DIVE',
        'color': C_PRIMARY,
        'tanques': ['Winston', 'D.Va', 'Wrecking Ball', 'Doomfist'],
        'objetivo': 'Backline\nenemigo',
    },
    {
        'nombre': 'POKE',
        'color': C_SUCCESS,
        'tanques': ['Sigma', 'Domina', 'Orisa'],
        'objetivo': 'Sightlines',
    },
    {
        'nombre': 'BRAWL',
        'color': C_DANGER,
        'tanques': ['Reinhardt', 'Junker Queen', 'Mauga', 'Ramattra'],
        'objetivo': 'Avanzar\njuntos',
    },
]

for i, arq in enumerate(arquetipos):
    a = ax[i]
    a.set_facecolor('#1a1a2e')
    a.set_xlim(0, 10)
    a.set_ylim(0, 10)
    a.axis('off')

    # Titulo
    a.text(5, 9.3, arq['nombre'], ha='center', va='center',
           fontsize=22, fontweight='bold', color=arq['color'],
           fontfamily='monospace')

    # Caja de tanques
    box = FancyBboxPatch((1, 4.5), 8, 4, boxstyle="round,pad=0.3",
                         facecolor=arq['color'], alpha=0.15,
                         edgecolor=arq['color'], linewidth=2)
    a.add_patch(box)

    tanques_text = '\n'.join(arq['tanques'])
    a.text(5, 6.5, tanques_text, ha='center', va='center',
           fontsize=13, color='white', fontfamily='monospace')

    # Flecha hacia objetivo
    a.annotate('', xy=(5, 1.8), xytext=(5, 4.3),
               arrowprops=dict(arrowstyle='->', color=arq['color'],
                              lw=3, mutation_scale=20))

    # Objetivo
    obj_box = FancyBboxPatch((1.5, 0.5), 7, 1.5, boxstyle="round,pad=0.2",
                             facecolor=arq['color'], alpha=0.3,
                             edgecolor=arq['color'], linewidth=2)
    a.add_patch(obj_box)
    a.text(5, 1.25, arq['objetivo'], ha='center', va='center',
           fontsize=12, color='white', fontweight='bold', fontfamily='monospace')

fig.suptitle('Los 3 Arquetipos de Composicion', fontsize=18,
             color='white', fontweight='bold', fontfamily='monospace', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

## 6. Conceptos Fundamentales para Support

---

### Peeling
Proteger a tu co-support. Si tu Ana esta siendo diveada, tu trabajo es salvarla (Brig bash, Lucio boop, Kiriko suzu). Un support muerto = la fight se pierde.

---

### Cooldown Tracking
Saber que el enemigo NO tiene Suzu/Immortality/Sleep es MAS importante que tener buena punteria. Si Kiriko acaba de usar Suzu, tienes 14 segundos para hacer un engage agresivo.

---

### Ult Economy
No usar Sound Barrier Y Transcendence en la misma fight -- guardar uno para la siguiente. Los supports que ciclan ults mas eficientemente ganan mas partidas. Con el subrol Tactician, el exceso de carga se conserva, asi que no pierdas tiempo sin usar habilidades.

---

### Positioning
NUNCA estes delante de tu tanque. Siempre detras, con linea de vision a todo tu equipo. Si mueres primero, tu equipo pierde la fight. La posicion ideal es: detras del tanque, con cover cerca, con linea de vision a todos.

---

### Target Priority
Tanque low HP > DPS low HP > DPS full HP. Nunca dejes morir al tanque por curar a un DPS full. El tanque es el que crea espacio -- sin tanque, no hay fight.

---

### Support Diff
La diferencia entre supports es a menudo la diferencia entre ganar y perder. Si tu Ana acierta anti-nades y la enemiga no, ganas. El support es el rol con mayor impacto por habilidad individual en 2026.
